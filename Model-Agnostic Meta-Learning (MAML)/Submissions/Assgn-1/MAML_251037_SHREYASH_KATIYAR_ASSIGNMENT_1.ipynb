## Part 4 : Matplotlib
### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title('Simple Line Graph')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)

plt.show()
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
import matplotlib.pyplot as plt

subjects = ['Maths', 'Physics', 'Chemistry', 'CS']
marks = [90, 75, 88, 95]

plt.bar(subjects, marks, color='skyblue')
plt.title('Marks in Subjects')
plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.ylim(0, 100)  # optional: to show full 0–100 range
plt.grid(axis='y', linestyle='--', alpha=0.5)  # optional

plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(100)
plt.hist(data, bins=10, edgecolor='black')
plt.title('Histogram of 100 Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
