```python
s = "Python Assignment"


uppercase_s = s.upper()


reversed_s = s[::-1]

vowels = "aeiouAEIOU"
vowel_count = 0

for char in s:
    if char in vowels:
        vowel_count += 1


print("Original:", s)
print("Uppercase:", uppercase_s)
print("Reversed:", reversed_s)
print("Number of vowels:", vowel_count)
```

### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):

    if n <= 1:
        return False


    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

test_numbers = [7, 15, 23]

for num in test_numbers:
    print(f"Is {num} prime? {is_prime(num)}")
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

#Using the "Bubble Sort" algorithm
sorted_numbers = list(numbers)
n = len(sorted_numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if sorted_numbers[j] > sorted_numbers[j + 1]:

            sorted_numbers[j], sorted_numbers[j + 1] = sorted_numbers[j + 1], sorted_numbers[j]


total_sum = 0
for num in numbers:
    total_sum += num


evens_only = [num for num in numbers if num % 2 == 0]


print("Sorted List:", sorted_numbers)
print("Sum:", total_sum)
print("Even numbers only:", evens_only)
```

### Q13. Line Plot

**Plot a simple line graph for:**

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
import matplotlib.pyplot as plt


subjects = ["Maths", "Physics", "Chemistry", "CS"]
marks = [90, 75, 88, 95]


plt.figure(figsize=(8, 5))
plt.bar(subjects, marks, color=['red', 'blue', 'green', 'pink'])


plt.title("Subject-wise Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")


plt.savefig('bar_chart.png')
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt


data = np.random.randn(100)


plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, color='yellow', edgecolor='blue')

plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")


plt.savefig('histogram.png')
```
