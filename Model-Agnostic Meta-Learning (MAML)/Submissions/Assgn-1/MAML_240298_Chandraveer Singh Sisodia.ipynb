### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
import math
def is_prime(n):
  sqrt = int(math.sqrt(n))#I have checked till root N as a composite number always has atleast one root less than equal to root N becase if both
  if(n<=1):               #are greater than root N then their product will exeed N,which is simply not possible
    return False
  for i in range(2,sqrt+1):
    if((n%i)==0):
      return False

  return True
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
n=len(numbers)
for i in range(n):
  for j in range(0,n-i-1):
    if(numbers[j]>numbers[j+1]):
      temp=numbers[j]
      numbers[j]=numbers[j+1]
      numbers[j+1]=temp

sum=0
for i in range(n):
  sum+=numbers[i]
print(sum)

i=0

while(i<len(numbers)):
  if((numbers[i]%2)!=0):
    numbers.remove(numbers[i])
  else:
    i=i+1
print(numbers)
```
