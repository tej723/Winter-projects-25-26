### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
  c=0
  for i in range(1,n+1):
    if(n%i==0):
      c+=1
  if(c==2):
    return True
  else:
    return False
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
for i in range(len(numbers)):
    for j in range(0, len(numbers) - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
print("Sorted", numbers)
total=sum(numbers)
print("Sum:",total)
even = [x for x in numbers if x%2==0]
print("List after removing odds:",even)
```
