### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
#Q5. Functions
def is_prime(n):
  if(n == 1):
      return False
  sqrt_n = int(n**0.5) #getting sqrt of n
  for i in range (2 , sqrt_n + 1): #Running loop till sqrt of n
    if(n % i == 0):
      return False
  return True
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
#Q6. Lists
numbers = [5, 2, 9, 1, 7, 3]
new_numbers = numbers.copy() #Creating copy of list to store sorted list

for i in range(len(new_numbers)):
  for j in range(0 , len(new_numbers)-i-1):
    if new_numbers[j] > new_numbers[j+1]:
     new_numbers[j], new_numbers[j+1] = new_numbers[j+1], new_numbers[j]

print(new_numbers,"\n")

sum = 0
for i in numbers:
  sum += i #adding elements
print("Sum of all elements is",sum,"\n")

for i in range(len(numbers)-1, -1, -1): #Traversing list from end and poping elements
  if(numbers[i] % 2 != 0):
    numbers.pop(i)

print(numbers)  #Printing remaining list
```

### Q9. NumPy Indexing

**Create an array of numbers 1–20.**
Extract:
- All even numbers
- Values from index 5 to 10
- Every 3rd element

```python
#Q9. NumPy Indexing

import numpy as np
arr = np.arange(1, 21)
print("An array of numbers 1-20:",arr,"\n")

even = []#Creating list to store even elements
for i in arr:
  if(i % 2 == 0):
    even.append(int(i))
print("All even numbers:",even,"\n")

values = []#Storing values from 5th index to 10th index
for i in range(5,11):
  values.append(int(arr[i]))
print("Values from index 5 to 10:",values,"\n")

every_3rd = arr[::3] #Using slicing to get every 3rd element
print("Every 3rd element:",every_3rd)


```

## Part 4 : Matplotlib
### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
#Q13. Line Plot
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y, marker='o', color='blue', linestyle='-')
plt.title("Simple line Plot")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.xticks(x) #So that x and y axis values are integer
plt.yticks(y)

plt.grid(True) #Grids are also shown
plt.show()
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
#Q14. Bar Chart

import matplotlib.pyplot as plt

subjects = ['Maths', 'Physics', 'Chemistry', 'CS']
marks = [90, 75, 88, 95]
plt.bar(subjects, marks, color='blue')
plt.ylim(0, 100)
plt.yticks(range(0, 101, 5))
plt.title("Bar chart of 4 subjects vs marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
#Q15. Histogram

import numpy as np
import matplotlib.pyplot as plt
ran_data = np.random.rand(100)
plt.hist(ran_data, bins=10, color='blue', edgecolor='black')

plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
