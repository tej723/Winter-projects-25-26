### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
# function of determining if prime or not
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
#testing
test_numbers = [7, 15, 23]
for num in test_numbers:
    print(f" {num}  {is_prime(num)}")
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python

numbers = [5, 2, 9, 1, 7, 3]
# sorting using bullble sort
sorted_numbers = numbers.copy()
n = len(sorted_numbers)
for i in range(n):
    for j in range(0, n-i-1):
        if sorted_numbers[j] > sorted_numbers[j+1]:
            sorted_numbers[j], sorted_numbers[j+1] = sorted_numbers[j+1], sorted_numbers[j]

print("Sorted List:", sorted_numbers)
# fing sum of all elements
total_sum = 0
for x in numbers:
    total_sum += x
print("Sum of elements:", total_sum)

#removing all the odds and printing the remaining list
evens_only = [x for x in numbers if x % 2 == 0]
print("List after removing odd numbers:", evens_only)
```

## Part 4 : Matplotlib
### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
subjects = ["Maths", "Physics", "Chemistry", "CS"]
marks = [90, 75, 88, 95]

plt.figure(figsize=(6, 4))
plt.bar(subjects, marks, color=['blue', 'orange', 'green', 'red'])
plt.title("Subject Marks Distribution")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
random_data = np.random.randn(100)

plt.figure(figsize=(6, 4))
plt.hist(random_data, bins=10, color='purple', edgecolor='black', alpha=0.7)
plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
