### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
list=[5, 2, 9, 1, 7, 3]

sum=0


for i in range(len(list)):
  sum+=list[i]

for i in range(len(list)):
  for j in range(i + 1, len(list)):
      if list[i] > list[j]:
          temp=list[i]
          list[i]=list[j]
          list[j]=temp
print(list)

edited_list=[x for x in list if x%2==0]


print("sum =", sum)
print(edited_list)
```

### Q13. Line Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
subject = ["Maths", "Physics", "Chemistry", "CS"]
marks = [90, 75, 88, 95]

plt.bar(subject, marks)
plt.title("Bar Chart")
plt.xlabel("Subject")
plt.ylabel("Marks")
plt.show()
```
