### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
  c=0
  for i in range(1,n+1):
    if(n%i==0):
      c+=1
  if(c==2):
    return True
  else:
    return False

print("7:",is_prime(7))
print("15:",is_prime(15))
print("23:",is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5,2,9,1,7,3]

sorted_numbers = []
temp = numbers.copy()

while temp:
  min_val=temp[0]
  for num in temp:
    if num < min_val:
      min_val = num
  sorted_numbers.append(min_val)
  temp.remove(min_val)

total_sum=sum(numbers)

even_numbers = [num for num in numbers if num % 2 ==0]

print("sorted list = ",sorted_numbers)
print("sum of elements = ",total_sum)
print("list after removing odd numbers = ",even_numbers)
```
