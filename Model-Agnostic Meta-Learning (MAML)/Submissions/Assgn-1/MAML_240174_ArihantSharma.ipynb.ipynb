```python
s = "Python Assignment"

u_s = s.upper()
print("Uppercase: " ,u_s)

rev_s = s[::-1]
print("Reversed: ",rev_s)

v_set = "aeiouAEIOU"
v_count = 0
for char in s:
    if char in v_set:
        v_count += 1
print("Number of vowels: ",v_count)
```

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**.5)+1):
        if n % i == 0:
            return False
    return True

print("Is 7 prime? ",  is_prime(7))
print("Is 15 prime? ",  is_prime(15))
print("Is 23 prime? ",  is_prime(23))
```

```python
numbers = [5, 2, 9, 1, 7, 3]
numbers_len = len(numbers)
sorted_list = list(numbers)
for i in range(numbers_len):
    for j in range(0, numbers_len - i - 1):
        if sorted_list[j] > sorted_list[j + 1]:
            sorted_list[j], sorted_list[j + 1] = sorted_list[j + 1], sorted_list[j]
print("Original list: ", numbers)
print("Sorted list (without sort()): ", sorted_list)

total_sum = 0
for item in numbers:
    total_sum += item
print("Sum of all elements: ", total_sum)

even_nums = []
for item in numbers:
    if item % 2 == 0:
        even_nums.append(item)
print("List after removing odd numbers: ", even_nums)
```

```python
import matplotlib.pyplot as plt

x_data = [1, 2, 3, 4, 5]
y_data = [2, 4, 6, 8, 10]

plt.plot(x_data, y_data, marker='o', linestyle='-')

plt.title("Simple Line Plot")
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")

plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

subjects = ["Maths", "Physics", "Chemistry", "CS"]
scores = [90, 75, 88, 95]

plt.figure(figsize=(8, 6))
plt.bar(subjects, scores, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])

plt.title("Marks in Different Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")

for i, score in enumerate(scores):
    plt.text(i, score + 1, str(score), ha='center', va='bottom')

plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
random_numbers = np.random.randn(100) * 10 + 50

plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=10, edgecolor='black', alpha=0.7)

plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
