### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
  if n%2 == 0:
    return True
  elif n%3 == 0:
    return True
  elif n%5 == 0:
    return True
  elif n%7 == 0:
    return True
  else:
    return False
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
def bubble_sort(lst):
    n = len(lst)
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j]
    return lst
bubble_sort(numbers)
print(numbers)
sum = 0
for i in numbers:
  sum = sum + i
print(sum)
for i in range(len(numbers)-1, -1, -1): #backward iteration
    if numbers[i] % 2 != 0:
        numbers.remove(numbers[i])
print(numbers)
```

## Part 4 : Matplotlib
### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, marker='o')
plt.title("y = 2x")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
subjects = ["Maths", "Physics", "Chemistry", "CS"]
marks = [90, 75, 88, 95]

plt.bar(subjects, marks, color=["skyblue", "orange", "green", "red"])
plt.title("Subjectwise - Marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 100)
#plt.grid(axis="y", linestyle="--", alpha=0.5)

plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
data = np.random.rand(100)

plt.hist(data, bins=10, edgecolor='black')
plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3)

plt.show()
```
