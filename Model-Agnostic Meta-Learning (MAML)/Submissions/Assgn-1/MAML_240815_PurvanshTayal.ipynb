### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
  count = 0
  for i in range(1,n):
    if n % i == 0:
      count += 1
  if count == 1:
    return True
  elif count > 1:
    return False
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
num = numbers
for i in range(len(numbers)):
  for j in range(len(numbers)-1):
    if num[j]>num[j+1]:
      num[j],num[j+1] = num[j+1],num[j]
print(num)
sum = 0
for i in numbers:
  sum += i
print(sum)
even = []
for i in numbers:
  if i % 2 == 0:
    even.append(i)
print(even)
```
