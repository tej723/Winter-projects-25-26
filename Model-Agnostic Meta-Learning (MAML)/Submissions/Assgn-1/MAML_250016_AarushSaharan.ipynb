### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
import math
def is_prime(n):
  i = 2
  while i<=math.sqrt(n):
    if n%i==0:
      return False
    i+=1
  return True
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q7. NumPy Array Creation

**Create the following using NumPy:**
- A 1D array from 10 to 50

- A random array of shape (2,3)

### Q8. Array Operations
Given:

```python
import numpy as np
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

Perform:
- Element-wise addition
- Dot product
- Multiply array a by a scalar (5)

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 1. Element-wise addition
add_result = a + b
print("Element-wise addition:", add_result)

# 2. Dot product
dot_result = np.dot(a, b)
print("Dot product:", dot_result)

# 3. Multiply array a by a scalar (5)
scalar_result = a * 5
print("Scalar multiplication:", scalar_result)
```

### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure()
plt.plot(x, y)
plt.title("Simple Line Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

### Q14. Bar Chart

### Q15. Histogram
