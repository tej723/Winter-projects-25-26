### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
# Function to check if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
# Testing the function
print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

#Sorting numbers using Merge Sort
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

sorted_numbers = merge_sort(numbers)
print("Sorted list using merge sort:", sorted_numbers)
```

```python
total_sum = 0
for num in numbers:
    total_sum += num

print("Sum of elements:", total_sum)
```

```python
even_numbers = [num for num in numbers if num % 2 == 0]

print("Even numbers only:", even_numbers)
```

### Q9. NumPy Indexing

- Every 3rd element
