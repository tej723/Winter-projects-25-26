### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n//2 + 1):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]
#sorting the array
n = len(numbers)
for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j] > numbers[j + 1]:
            numbers[j], numbers[j + 1] = numbers[j + 1], numbers[j]
#Total sum of elements in array
total_sum = 0
for num in numbers:
    total_sum += num
#Creating new array of only even numbers keeping the original order
even_numbers = []
for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)

print(numbers)
print(total_sum)
print(even_numbers)
```

**Plot a bar chart of 4 subjects vs marks:**

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate 100 random numbers
data = np.random.randint(-5,6,size=(100))

# 2. Plot a histogram
plt.hist(data, bins=10, color='blue', edgecolor='black')

plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.savefig("histogram_random_numbers.png")
```
