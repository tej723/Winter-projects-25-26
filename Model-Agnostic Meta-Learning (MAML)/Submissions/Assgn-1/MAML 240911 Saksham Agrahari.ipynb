### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):

    if n < 2:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(15))
print(is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

sorted_list = sorted(numbers)

total_sum = sum(numbers)

even_list = []
for n in numbers:
    if n % 2 == 0:
        even_list.append(n)

print("Sorted List:", sorted_list)
print("Sum:", total_sum)
print("Even Numbers Only:", even_list)
```
