### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
def sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
numbers = [5, 2, 9, 1, 7, 3]
sorted_numbers = sort(numbers)
print(sorted_numbers)
sum = 0
for i in numbers:
  sum+=i
print(sum)
numbers = [x for x in numbers if x % 2 == 0]
print(numbers)
```
