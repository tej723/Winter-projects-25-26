### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):

    if n <= 1:
        return False

    if n == 2:
        return True

    if n % 2 == 0:
        return False

    i = 3
    while i * i <= n:
        if n % i == 0:
            return False
        i += 2

    return True



test_numbers = [7, 15, 23]


for num in test_numbers:
    result = is_prime(num)
    print(f"Is {num} prime? {result}")
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

sorted_numbers = numbers[:]
n = len(sorted_numbers)

for i in range(n):
    for j in range(0, n - i - 1):
        if sorted_numbers[j] > sorted_numbers[j + 1]:
            sorted_numbers[j], sorted_numbers[j + 1] = sorted_numbers[j + 1], sorted_numbers[j]

print(f"Original list: {numbers}")
print(f"1. Sorted list (without sort()): {sorted_numbers}")



list_sum = sum(numbers)
print(f"2. Sum of all elements: {list_sum}")


even_numbers = [num for num in numbers if num % 2 == 0]

print(f"3. List after removing odd numbers: {even_numbers}")
```

## Part 4 : Matplotlib
### Q13. Line Plot

**Plot a simple line graph for:**

x = [1, 2, 3, 4, 5]

y = [2, 4, 6, 8, 10]

Add title and labels.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]


plt.figure(figsize=(8, 6))


plt.plot(x, y,
         marker='o',
         linestyle='-',
         color='blue',
         label='y = 2x')


plt.title("Simple Linear Relationship: y = 2x")
plt.xlabel("X-Axis Values")
plt.ylabel("Y-Axis Values")


plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)


plt.savefig('simple_line_plot.png')
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
random_data = np.random.randn(100)


plt.figure(figsize=(8, 6))
plt.hist(random_data,
         bins=10,
         color='skyblue',
         edgecolor='black',
         alpha=0.7)

plt.title("Histogram of 100 Random Numbers (Standard Normal Distribution)")
plt.xlabel("Value Range (Bins)")
plt.ylabel("Frequency (Count)")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('random_data_histogram.png')
```
