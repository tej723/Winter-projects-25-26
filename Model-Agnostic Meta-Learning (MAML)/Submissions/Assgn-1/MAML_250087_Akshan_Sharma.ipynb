### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
  for i in range(2, int(n**0.5)+1):
    if not n%i:
      return False
  return True
print(is_prime(7), is_prime(15), is_prime(23))
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
#reverse bubble sort
numbers = [5, 2, 9, 1, 7, 3]
new_nums = []
k = len(numbers)
for i in range(k):
  new_nums.append(min(numbers))
  del numbers[numbers.index(min(numbers))]
numbers = new_nums.copy()
print(numbers)

print(sum(numbers))

new_nums = []
for i in numbers:
  if not i%2:
    new_nums.append(i)
numbers = new_nums.copy()

print(numbers)
```
