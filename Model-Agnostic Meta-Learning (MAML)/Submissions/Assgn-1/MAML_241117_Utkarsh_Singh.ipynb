### Q4. Strings

**Given the string:**

```python
s = "Python Assignment"
```

Do the following:

1. Convert it to uppercase

2. Reverse the string

3. Count the number of vowels

```python
s = "Python Assignment"

# 1. Convert to uppercase
uppercase_s = s.upper()
print("Uppercase string:", uppercase_s)

# 2. Reverse the string
reversed_s = s[::-1]
print("Reversed string:", reversed_s)

# 3. Count the number of vowels
vowels = "aeiouAEIOU"
vowel_count = 0
for char in s:
    if char in vowels:
        vowel_count += 1
print("Number of vowels:", vowel_count)
```

### Q5. Functions

**Write a function is_prime(n) that returns:**

- True if the number is prime

- False otherwise

Test it with: 7, 15, 23.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 15 prime? {is_prime(15)}")
print(f"Is 23 prime? {is_prime(23)}")
```

### Q6. Lists

Given a list:

numbers = [5, 2, 9, 1, 7, 3]


Do the following:

- Sort the list without using sort()

- Find the sum of all elements

- Remove all odd numbers and print the remaining list

```python
numbers = [5, 2, 9, 1, 7, 3]

# Sort the list without using sort() (using bubble sort for example)
n = len(numbers)
for i in range(n):
    for j in range(0, n-i-1):
        if numbers[j] > numbers[j+1]:
            numbers[j], numbers[j+1] = numbers[j+1], numbers[j]
print("Sorted list (without using sort()):", numbers)

# Find the sum of all elements
sum_elements = 0
for num in numbers:
    sum_elements += num
print("Sum of all elements:", sum_elements)

# Remove all odd numbers and print the remaining list
even_numbers = []
for num in numbers:
    if num % 2 == 0:
        even_numbers.append(num)
print("List after removing odd numbers:", even_numbers)
```

### Q9. NumPy Indexing

**Create an array of numbers 1–20.**
Extract:
- All even numbers
- Values from index 5 to 10
- Every 3rd element

```python
import numpy as np

# Create an array of numbers 1-20
arr = np.arange(1, 21)
print("Original Array:", arr)

# Extract all even numbers
even_numbers = arr[arr % 2 == 0]
print("Even numbers:", even_numbers)

# Extract values from index 5 to 10
values_5_to_10 = arr[5:11] # Index 10 is included, so slice up to 11
print("Values from index 5 to 10:", values_5_to_10)

# Extract every 3rd element
every_3rd_element = arr[::3]
print("Every 3rd element:", every_3rd_element)
```

### Q14. Bar Chart

**Plot a bar chart of 4 subjects vs marks:**
- Maths – 90
- Physics – 75
- Chemistry – 88
- CS – 95

```python
import matplotlib.pyplot as plt

subjects = ['Maths', 'Physics', 'Chemistry', 'CS']
marks = [90, 75, 88, 95]

plt.figure(figsize=(8, 6))
plt.bar(subjects, marks, color=['blue', 'green', 'red', 'purple'])
plt.title("Marks in Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.ylim(0, 100) # Assuming marks are out of 100
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Q15. Histogram

Generate 100 random numbers using NumPy.

Plot a histogram using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random numbers
random_data = np.random.randn(100)

# Plot a histogram
plt.figure(figsize=(10, 6))
plt.hist(random_data, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of 100 Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', alpha=0.75)
plt.show()
```
