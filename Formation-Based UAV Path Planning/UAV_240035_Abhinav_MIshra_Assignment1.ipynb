
### Question 1: DFS




```python
import numpy as np

n, m = map(int, input().split())

edges = np.zeros((m, 2))
visited = np.zeros(n, dtype=bool)
cycle = 0

for i in range(m):

    edges[i][0], edges[i][1] = map(int, input().split())

adj = [[] for _ in range(n + 1)]

for i in range(m):
    u, v = int(edges[i][0]), int(edges[i][1])
    adj[u].append(v)
    adj[v].append(u)

def cycles(u, p):

    visited[u-1] = True

    for v in adj[u]:
        if v == p:
            continue

        if visited[v-1]:
            return True

        if not visited[v-1]:
            if cycles(v, u):
                return True
    return False

for i in range(1, n + 1):
    if not visited[i-1]:
        if cycles(i, -1):
            cycle = 1
            break

if(cycle):
    print("YES")
else:
    print("NO")
```

### Question 2: BFS

```python
import numpy as np

n, m = map(int, input().split())

edges = np.zeros((m, 2))
visited = np.zeros(n + 1, dtype=bool)
colors = np.zeros(n + 1, dtype=int)
bipartite = 1

for i in range(m):
    edges[i][0], edges[i][1] = map(int, input().split())

adj = [[] for _ in range(n + 1)]
for i in range(m):
    u, v = int(edges[i][0]), int(edges[i][1])
    adj[u].append(v)
    adj[v].append(u)

queue_buffer = np.zeros(n + 1, dtype=int)

def check_bipartite(start_node):
    head = 0
    tail = 0

    queue_buffer[tail] = start_node
    tail += 1

    visited[start_node] = True
    colors[start_node] = 0

    while head < tail:
        u = queue_buffer[head]
        head += 1

        current_color = colors[u]
        next_color = 1 - current_color

        for v in adj[u]:
            if not visited[v]:
                visited[v] = True
                colors[v] = next_color

                queue_buffer[tail] = v
                tail += 1

            elif colors[v] == current_color:
                return False

    return True

for i in range(1, n + 1):
    if not visited[i]:
        if not check_bipartite(i):
            bipartite = 0
            break

if bipartite:
    print("YES")
    print(*(colors[1:]))
else:
    print("NO")
```
