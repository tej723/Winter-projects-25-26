Task 1: Grid as Graph



*   Each cell (x,y) is a node
*   Each node can move to 8 neighbour cells


*   Valid moves: Inside the grid, not in any obstacle
*   Edge Weight: 1 for horizontal and root 2 for diagonal





Graph making and other fucntions needed

```python
import heapq
import math

N = 200
START = (0, 0)
GOAL = (199, 199)

obstacles = [((40, 40), (70, 70)), ((90, 120), (130, 160)), ((150, 50), (180, 90)), ((60, 150), (100, 190))]

def is_obstacle(x, y):
    for (x1, y1), (x2, y2) in obstacles:
        if x1 <= x <= x2 and y1 <= y <= y2:
            return True
    return False


def in_bounds(x, y):
    return 0 <= x < N and 0 <= y < N


def get_neighbors(x, y):
    nbrs = []

    for dx, dy in [(1,0), (-1,0), (0,1), (0,-1)]:
        nx, ny = x + dx, y + dy
        if in_bounds(nx, ny) and not is_obstacle(nx, ny):
            nbrs.append(((nx, ny), 1.0))

    for dx, dy in [(1,1), (1,-1), (-1,1), (-1,-1)]:
        nx, ny = x + dx, y + dy
        if in_bounds(nx, ny) and not is_obstacle(nx, ny):
            if not is_obstacle(x + dx, y) and not is_obstacle(x, y + dy):
                nbrs.append(((nx, ny), math.sqrt(2)))

    return nbrs
```

TASK 2: Disjstra's Algorithm

```python
def dijkstra():
    pq = [(0.0, START)]
    dist = {START: 0.0}
    parent = {START: None}
    visited = set()
    expanded = 0

    while pq:
        cost, node = heapq.heappop(pq)

        if node in visited:
            continue
        visited.add(node)
        expanded += 1

        if node == GOAL:
            break

        for nxt, w in get_neighbors(*node):
            new_cost = cost + w
            if nxt not in dist or new_cost < dist[nxt]:
                dist[nxt] = new_cost
                parent[nxt] = node
                heapq.heappush(pq, (new_cost, nxt))

    path = []
    cur = GOAL
    while cur is not None:
        path.append(cur)
        cur = parent[cur]
    path.reverse()

    return dist[GOAL], expanded, path

d_cost, d_exp, d_path = dijkstra()
print("Path length:", len(d_path))
print("Cost:", d_cost)
print("Expanded nodes:", d_exp)
```

TASK 3: A* Algo

```python
def heuristic(a, b):
    return math.hypot(a[0] - b[0], a[1] - b[1])

def astar():
    pq = [(heuristic(START, GOAL), 0.0, START)]
    g = {START: 0.0}
    parent = {START: None}
    visited = set()
    expanded = 0

    while pq:
        _, cost, node = heapq.heappop(pq)

        if node in visited:
            continue
        visited.add(node)
        expanded += 1

        if node == GOAL:
            break

        for nxt, w in get_neighbors(*node):
            new_g = cost + w
            if nxt not in g or new_g < g[nxt]:
                g[nxt] = new_g
                parent[nxt] = node
                f = new_g + heuristic(nxt, GOAL)
                heapq.heappush(pq, (f, new_g, nxt))

    path = []
    cur = GOAL
    while cur is not None:
        path.append(cur)
        cur = parent[cur]
    path.reverse()

    return g[GOAL], expanded, path

a_cost, a_exp, a_path = astar()
print("Path length:", len(a_path))
print("Cost:", a_cost)
print("Expanded nodes:", a_exp)

# Yes is does not overestimates as the diagonal movements are allowed with cost root 2 and matches the straight line distance in grid.
```
