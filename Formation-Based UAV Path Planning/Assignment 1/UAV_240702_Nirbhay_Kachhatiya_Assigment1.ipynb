## **Problem 1: Cycle Detection Using DFS**

```python
class graph1:
  def __init__(self,vertices):
    self.V = vertices
    self.adj = []
    for i in range(self.V +1):
      self.adj.append([])

  def addEdge(self,u,v):
    self.adj[u].append(v)
    self.adj[v].append(u)

  def isCycle(self,v,parent,visited):
    if(visited[v]):
      return False
    visited[v]=1;
    ans=False
    for i in self.adj[v]:
      if (visited[i]==0):
        ans = ans or self.isCycle(i,v,visited)
      elif (i!=parent) :
        return True
    return ans

  def findCycleDFS(self):
    visited = [0]*(self.V + 1)
    for i in range(self.V+1):
      if(self.isCycle(i,-1,visited)):
        print("Yes , Cycle is present")
        return
    print("No , Cycle is not present")
```

## **Problem 2: Bipartite Graph Check Using BFS**

```python
class graph2:
  def __init__(self,vertices):
    self.V = vertices
    self.adj = []
    for i in range(self.V +1):
      self.adj.append([])

  def addEdge(self,u,v):
    self.adj[u].append(v)
    self.adj[v].append(u)

  def isBipartiteBFS(self):
    color = [-1]*(self.V + 1)
    for i in range(1,self.V + 1):
      if(color[i]==-1):
        queue=[]
        queue.append(i)
        color[i]=0
        while(queue):
          front=queue.pop(0)
          for j in self.adj[front]:
            if(color[j]==-1):
              color[j] = 1 - color[front]
              queue.append(j)
            elif color[j]==color[front]:
              print("No, graph is not bipartite")
              return

    print("Yes , graph is bipartite")
    for i in range(1,self.V + 1):
      print("Vertex : {} , Color : {}".format(i,color[i]))
```

## **Problem 3: Shortest Path Using Dijkstraâ€™s Algorithm**

```python
import heapq

class graph3:
  def __init__(self,vertices):
    self.V = vertices
    self.adj = []
    for i in range(self.V +1):
      self.adj.append([])

  def addEdge(self,u,v,weight):
    self.adj[u].append((v,weight))

  def dijkstra(self):
    distances = [float('inf')]*(self.V + 1)
    distances[1]=0;
    visited = [0]*(self.V + 1)
    pq = [(0,1)]
    while pq:
      current_distance, current_node = heapq.heappop(pq)
      if(visited[current_node]==1):
        continue
      visited[current_node]=1
      for neighbour,weight in self.adj[current_node] :
        new_dist = current_distance + weight
        if new_dist < distances[neighbour]:
          distances[neighbour] = new_dist
          heapq.heappush(pq,(new_dist,neighbour))
    if(distances[self.V]==float('inf')):
      print(-1)
      return
    print(distances[self.V])
```

## **Problem 4: Shortest Path With a Single Hyperloop Jump Between Waypoints**

```python
import heapq

class graph4:
  def __init__(self,vertices):
    self.V = vertices
    self.adj = []
    for i in range(self.V +1):
      self.adj.append([])

  def addEdge(self,u,v,weight):
    self.adj[u].append((v,weight))

  def dijkstra(self,start):
    distances = [float('inf')]*(self.V + 1)
    distances[start]=0;
    visited = [0]*(self.V + 1)
    pq = [(0,start)]
    while pq:
      current_distance, current_node = heapq.heappop(pq)
      if(visited[current_node]==1):
        continue
      visited[current_node]=1
      for neighbour,weight in self.adj[current_node] :
        new_dist = current_distance + weight
        if new_dist < distances[neighbour]:
          distances[neighbour] = new_dist
          heapq.heappush(pq,(new_dist,neighbour))
    return distances

  def shortestPath(self,waypoints):
    dist_from_1 = self.dijkstra(1)
    min_from_1 = float('inf')
    min_from_n = float('inf')
    for i in waypoints:
      min_from_1 = min(min_from_1,dist_from_1[i])
      min_from_n = min(min_from_n,self.dijkstra(i)[self.V])
    ans = min(dist_from_1[self.V] , min_from_1+min_from_n)
    if(ans==float('inf')):
      return -1
    return ans
```
