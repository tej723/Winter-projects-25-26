## Problem 1 Solution

```python
def solve(n,m,graph):
    visited=[0]*(n+1)

    def dfs(u,parent):
        visited[u]=1
        for v in graph[u]:
            if not visited[v]:
                if dfs(v,u):
                    return True
            elif v!=parent:
                return True
        return False

    for i in range(n):
        if not visited[i]:
            if dfs(i,-1):
                print("YES")
                return
    print("NO")
```

## Problem 2 Solution

```python
def solve2(n,m,graph):
    color=[-1]*(n+1)

    def bfs(start):
        queue=[]
        queue.append(start)
        color[start]=0

        while queue:
            current=queue.pop(0)
            for v in graph[current]:
                if color[v]==-1:
                    color[v]=1-color[current]
                    queue.append(v)
                elif color[v]==color[current]:
                    return False
        return True

    for i in range(1, n+1):
        if color[i]==-1:
            if not bfs(i):
                print("NO")
                return

    print("YES")
```

## Problem 3 solution:


```python
def solve3(n,m,graph):
    INF=10**7
    dist=[INF]*(n+1)
    sptSet=[False]*(n+1)

    dist[1]=0

    def minDistance():
        mn=INF
        idx=-1
        for i in range(1,n+1):
            if not sptSet[i] and dist[i]<mn:
                mn=dist[i]
                idx=i
        return idx

    for _ in range(n):
        u=minDistance()
        if u==-1:
            break

        sptSet[u]=True

        for v in range(1, n + 1):
            if graph[u][v]>0 and not sptSet[v]:
                if dist[v]>dist[u]+graph[u][v]:
                    dist[v]=dist[u]+graph[u][v]

    if dist[n]==INF:
        print(-1)
    else:
        print(dist[n])
```

## Problem 4 Solution

```python
import heapq

def solve4(n,m,k,graph,waypoints):
    rev_graph=[[] for _ in range(n+1)]
    for u in range(1, n+1):
      for v,w in graph[u]:
        rev_graph[v].append((u,w))

    INF=10**18

    def dijkstra(start, g):
        dist=[INF]*(n+1)
        dist[start]=0
        pq=[(0,start)]

        while pq:
            d,u=heapq.heappop(pq)
            if d>dist[u]:
                continue
            for v,w in g[u]:
                if dist[v]>dist[u]+w:
                    dist[v]=dist[u]+w
                    heapq.heappush(pq,(dist[v],v))
        return dist

    dist_from_start=dijkstra(1,graph)
    dist_to_end=dijkstra(n,rev_graph)

    ans=dist_from_start[n]

    best_entry=min(dist_from_start[w] for w in waypoints)
    best_exit=min(dist_to_end[w] for w in waypoints)

    ans=min(ans,best_entry+best_exit)

    if ans<INF:
      print(ans)
    else:
      print(-1)
```
