```python
yes=False

n, m =map(int, input().split())
adj= []
visited=[]
for x in range(n+1):
    adj.append([])
    visited.append(False)
for i in range(m):
    u, v = map(int, input().split())
    adj[u].append(v)
    adj[v].append(u)

for start in range(1, n + 1):
    if visited[start]:
        continue

    stack =[(start, -1)]
    while stack:
        node, parent = stack.pop()
        if visited[node]:
            continue
        visited[node] = True

        for w in adj[node]:
            if not visited[w]:
                stack.append((w, node))
            elif w!=parent:
                yes=True
                break
        if yes:
            break
    if yes: 
        break
if yes:
    print("YES")
else:
    print("NO")
```

```python
n, m = map(int, input().split())

adj = [[] for i in range(n + 1)]
for i in range(m):
    u, v = map(int, input().split())
    adj[u].append(v)
    adj[v].append(u)

color = [-1] * (n + 1)  
#-1 meaning no color assigned yet

bipartite = True
for start in range(1, n + 1):
    if color[start] != -1:
        continue
    stack = [start]
    color[start] = 0
    while stack:
        node = stack.pop()
        for w in adj[node]:
            if color[w] == -1:
                if color[node]==0:
                    color[w] = 1
                if color[node]==1:
                    color[w] = 0
                stack.append(w)
            elif color[w] ==color[node]:
                bipartite = False
                break

        if not bipartite:
            break

    if not bipartite:
        break

if not bipartite:
    print("NO")
else:
    print("YES")
    print(*color[1:])
```

```python
n, m = map(int, input().split())

adj = [[] for i in range(n + 1)]
for _ in range(m):
    u, v, w = map(int, input().split())
    adj[u].append((v, w))

INF = 10**7
dist = []
for i in range(n + 1):
    dist.append(INF)
used = []
for i in range(n + 1):
    used.append(False)
    
dist[1] = 0

for _ in range(n):
    u = -1
    b = INF
    for i in range(1, n + 1):
        if not used[i] and dist[i]<b:
            b= dist[i]
            u = i
    if u == -1:
        break
    used[u] = True

    for v,w in adj[u]:
        if dist[v] > dist[u] + w:
            dist[v] = dist[u] + w

if dist[n]!=INF:
    print(dist[n])
else:
    print(-1)
```
