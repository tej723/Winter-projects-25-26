#Solution 1

```python
def cycle_detector(n, edges):
    adj = [[] for _ in range(n + 1)]

    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    visited = [False] * (n + 1)

    def has_cycle(u, p):
        visited[u] = True
        for v in adj[u]:
            if v == p: continue
            if visited[v]: return True
            if has_cycle(v, u): return True
        return False

    for i in range(1, n + 1):
        if not visited[i]:
            if has_cycle(i, -1):
                return "YES"
    return "NO"

test_cases_1 = [
    ("Sample 1", 4, [(1,2), (2,3), (3,1), (3,4)]),
    ("Sample 2", 5, [(1,2), (2,3), (3,4), (4,5)]),
    ("Test 1", 3, [(1,2), (2,3), (3,1)]),
    ("Test 2", 6, [(1,2), (3,4), (5,6)]),
    ("Test 3", 7, [(1,2), (2,3), (3,4), (4,2), (5,6), (6,7)]),
    ("Test 4", 1, []),
    ("Test 5", 4, [(1,2), (3,4)])
]

for name, n, edges in test_cases_1:
    print(f"{name}: {cycle_detector(n, edges)}")
```

#Solution 2

```python
from collections import deque

def grph_bipartite(n, edges):
    adj = [[] for _ in range(n + 1)]
    for u, v in edges:
        adj[u].append(v)
        adj[v].append(u)

    color = [-1] * (n + 1)
    for i in range(1, n + 1):
        if color[i] == -1:
            queue = deque([i])
            color[i] = 0
            while queue:
                u = queue.popleft()
                for v in adj[u]:
                    if color[v] == -1:
                        color[v] = 1 - color[u]
                        queue.append(v)
                    elif color[v] == color[u]:
                        return "NO"

    res_colors = " ".join(map(str, color[1:]))
    return f"YES\n{res_colors}"

test_cases_2 = [
    ("Sample 1", 3, [(1,2), (2,3)]),
    ("Sample 2", 3, [(1,2), (2,3), (3,1)]),
    ("Add. Test 1", 4, [(1,2), (3,4)]),
    ("Add. Test 2", 5, [(1,2), (2,3), (3,4), (4,5)]),
    ("Add. Test 3", 4, [(1,2), (2,3), (3,4), (4,1)]),
    ("Add. Test 4", 4, [(1,2), (2,3), (3,4), (4,1), (1,3)]),
    ("AdD. Test 5", 1, [])

]

for name, n, edges in test_cases_2:
    print(f"{name}: {grph_bipartite(n, edges)}\n")
```

#Solution 3

```python
import heapq

def shortest_path(n, edges):
    adj = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        adj[u].append((v, w))

    dist = [float('inf')] * (n + 1)
    dist[1] = 0
    pq = [(0, 1)]

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]: continue
        for v, w in adj[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    return dist[n] if dist[n] != float('inf') else -1

test_cases_3 = [
    ("Sample 1", 5, [(1,2,3), (2,3,4), (1,4,2), (4,3,2), (3,5,1), (4,5,10)]),
    ("Sample 2", 4, [(1,2,5), (3,4,7)]),
    ("Add. Test 1", 3, [(1,2,5), (2,3,1), (1,3,10)]),
    ("Add. Test 2", 4, [(1,2,1), (2,3,1), (3,4,1), (1,4,10)]),
    ("ADd. Test 3", 6, [(1,2,4), (2,3,4), (5,6,1)]),
    ("Add. Test 4", 5, [(1,2,2), (1,3,4), (2,4,7), (3,4,1), (4,5,3), (2,5,20), (3,5,10)]),
    ("ADd. Test 5", 1, [])
]

for name, n, edges in test_cases_3:
    print(f"{name}: {shortest_path(n, edges)}")
```

```python
def shortest_path_hyperloop(n, m, k, edges, waypoints):
    adj = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        adj[u].append((v, w))

    def dijkstra(start_node):
        dists = [float('inf')] * (n + 1)
        dists[start_node] = 0
        pq = [(0, start_node)]
        while pq:
            d, u = heapq.heappop(pq)
            if d > dists[u]: continue
            for v, w in adj[u]:
                if dists[u] + w < dists[v]:
                    dists[v] = dists[u] + w
                    heapq.heappush(pq, (dists[v], v))
        return dists

    d1 = dijkstra(1)
    ans = d1[n]

    rev_adj = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        rev_adj[v].append((u, w))

    def dijkstra_rev(start_node):
        dists = [float('inf')] * (n + 1)
        dists[start_node] = 0
        pq = [(0, start_node)]
        while pq:
            d, u = heapq.heappop(pq)
            if d > dists[u]: continue
            for v, w in rev_adj[u]:
                if dists[u] + w < dists[v]:
                    dists[v] = dists[u] + w
                    heapq.heappush(pq, (dists[v], v))
        return dists

    dn = dijkstra_rev(n)

    min_to_wp = min([d1[wp] for wp in waypoints])
    min_from_wp = min([dn[wp] for wp in waypoints])

    ans = min(ans, min_to_wp + min_from_wp)
    return ans if ans != float('inf') else -1

test_cases_4 = [
    ("Sample 1", 6, 7, 2, [(1,2,5), (2,3,2), (3,6,5), (1,4,3), (4,5,10), (5,6,1), (2,5,4)], [2, 4]),
    ("Sample 2", 5, 4, 3, [(1,2,4), (2,3,4), (3,5,4), (4,5,100)], [2, 4, 5]),
    ("Test 1", 5, 5, 2, [(1,2,2), (2,3,2), (3,4,2), (4,5,2), (1,5,50)], [2, 4]),
    ("Test 2", 6, 7, 3, [(1,2,1), (2,3,1), (3,6,10), (1,4,5), (4,5,1), (5,6,1), (2,5,20)], [2, 4, 6]),
    ("Test 3", 4, 2, 2, [(1,2,5), (3,4,5)], [1, 3]),
    ("Test 4", 7, 9, 3, [(1,2,3), (2,3,3), (3,7,3), (1,4,10), (4,5,2), (5,6,2), (6,7,2), (2,6,20), (2,4,1)], [3, 5, 7]),
    ("test 5", 3, 1, 1, [(1,2,10)], [2])
]

for name, n, m, k, edges, wps in test_cases_4:
    print(f"{name}: {shortest_path_hyperloop(n, m, k, edges, wps)}")
```
