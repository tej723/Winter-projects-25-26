```python
from collections import deque
import heapq
```

```python
def detect_cycle():
    vertices, edges = map(int, input().split())
    adj = {}
    for i in range(1, vertices + 1):
        adj[i] = []

    for _ in range(edges):
        a, b = map(int, input().split())
        adj[a].append(b)
        adj[b].append(a)

    seen = set()

    def explore(current, prev):
        seen.add(current)
        for next_node in adj[current]:
            if next_node not in seen:
                if explore(next_node, current):
                    return True
            elif next_node != prev:
                return True
        return False

    result = False
    for node in range(1, vertices + 1):
        if node not in seen:
            if explore(node, -1):
                result = True
                break

    print("YES" if result else "NO")
```

```python
def check_bipartite():
    num_nodes, num_edges = map(int, input().split())
    adjacency = [[] for _ in range(num_nodes + 1)]

    for _ in range(num_edges):
        x, y = map(int, input().split())
        adjacency[x].append(y)
        adjacency[y].append(x)

    colors = {}

    def bfs_coloring(source):
        q = deque()
        q.append(source)
        colors[source] = 0

        while q:
            curr = q.popleft()
            current_color = colors[curr]

            for adjacent in adjacency[curr]:
                if adjacent not in colors:
                    colors[adjacent] = 1 - current_color
                    q.append(adjacent)
                elif colors[adjacent] == current_color:
                    return False
        return True

    is_bipartite = True
    for v in range(1, num_nodes + 1):
        if v not in colors:
            if not bfs_coloring(v):
                is_bipartite = False
                break

    if is_bipartite:
        print("YES")
        result = []
        for i in range(1, num_nodes + 1):
            result.append(str(colors.get(i, 0)))
        print(' '.join(result))
    else:
        print("NO")
```

```python
def shortest_path_dijkstra():
    node_count, edge_count = map(int, input().split())
    edges_list = [[] for _ in range(node_count + 1)]

    for _ in range(edge_count):
        src, dest, cost = map(int, input().split())
        edges_list[src].append((dest, cost))

    INF = float('inf')
    distances = [INF] * (node_count + 1)
    distances[1] = 0

    heap = [(0, 1)]

    while heap:
        current_dist, u = heapq.heappop(heap)

        if current_dist > distances[u]:
            continue

        for v, weight in edges_list[u]:
            new_dist = distances[u] + weight
            if new_dist < distances[v]:
                distances[v] = new_dist
                heapq.heappush(heap, (new_dist, v))

    answer = distances[node_count]
    print(-1 if answer == INF else answer)
```

```python
def hyperloop_shortest_path():
    total_nodes, total_edges, waypoint_count = map(int, input().split())

    graph_adj = [[] for _ in range(total_nodes + 1)]
    for _ in range(total_edges):
        u, v, w = map(int, input().split())
        graph_adj[u].append((v, w))

    special_nodes = list(map(int, input().split()))

    def run_dijkstra(start_node):
        INF = float('inf')
        distance = [INF] * (total_nodes + 1)
        distance[start_node] = 0
        priority_queue = [(0, start_node)]

        while priority_queue:
            curr_dist, node = heapq.heappop(priority_queue)

            if curr_dist > distance[node]:
                continue

            for neighbor, edge_weight in graph_adj[node]:
                if distance[node] + edge_weight < distance[neighbor]:
                    distance[neighbor] = distance[node] + edge_weight
                    heapq.heappush(priority_queue, (distance[neighbor], neighbor))

        return distance

    from_start = run_dijkstra(1)

    reverse_adj = [[] for _ in range(total_nodes + 1)]
    for node in range(1, total_nodes + 1):
        for neighbor, weight in graph_adj[node]:
            reverse_adj[neighbor].append((node, weight))

    temp_graph = graph_adj
    graph_adj = reverse_adj
    from_end = run_dijkstra(total_nodes)
    graph_adj = temp_graph

    optimal = from_start[total_nodes]

    for waypoint_a in special_nodes:
        for waypoint_b in special_nodes:
            if from_start[waypoint_a] != float('inf') and from_end[waypoint_b] != float('inf'):
                candidate = from_start[waypoint_a] + from_end[waypoint_b]
                if candidate < optimal:
                    optimal = candidate

    print(-1 if optimal == float('inf') else optimal)
```
