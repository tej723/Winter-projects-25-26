```python
def has_cycle(n, edges):
    graph = [[] for _ in range(n + 1)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    visited = [False] * (n + 1)
    def dfs(node, parent):
        visited[node] = True
        for neighbor in graph[node]:
            if not visited[neighbor]:
                if dfs(neighbor, node):
                    return True
            elif neighbor != parent:
                return True
        return False
    #Since it is a graph and not a treee we need to check all connected components
    for i in range(1, n + 1):
        if not visited[i]:
            if dfs(i, -1):
                return True
    return False
#taking the input(first input the testcases and then input all the graphs)
t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    edges = []
    for _ in range(m):
        u, v = map(int, input().split())
        edges.append((u, v))

    if has_cycle(n, edges):
        print("YES")
    else:
        print("NO")
```

```python
from collections import deque
def is_bipartite(n, edges):
    graph = [[] for _ in range(n + 1)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    color = [-1] * (n + 1)
    for start in range(1, n + 1):
        if color[start] == -1:
            queue = deque([start])
            color[start] = 0
            while queue:
                u = queue.popleft()
                for v in graph[u]:
                    if color[v] == -1:
                        color[v] = 1 - color[u]
                        queue.append(v)
                    elif color[v] == color[u]:
                        return False
    return True
#input, if we have a cycle of odd length then the graph is not bipartite else the graph is bipartite, the same testcases format as in above code
t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    edges = []
    for _ in range(m):
        u, v = map(int, input().split())
        edges.append((u, v))

    if is_bipartite(n, edges):
        print("YES")
    else:
        print("NO")
```

```python
import heapq

def shortest_path(n, edges):
    graph = [[] for _ in range(n + 1)]
    for u, v, w in edges:
        graph[u].append((v, w))

    INF = 10**18
    dist = [INF] * (n + 1)
    dist[1] = 0
#we are storing (distance, node) in the priority queue
    pq = [(0, 1)]
    while pq:
        curr_dist, u = heapq.heappop(pq)
        if curr_dist > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[v] > curr_dist + w:
                dist[v] = curr_dist + w
                heapq.heappush(pq, (dist[v], v))
    # if node n is unreachable
    if dist[n] == INF:
        return -1
    return dist[n]
  #taking the input in testcases format
t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    edges = []
    for _ in range(m):
        u, v, w = map(int, input().split())
        edges.append((u, v, w))

    print(shortest_path(n, edges))
```

```python
import heapq

INF = 10**18
def shortest_path_with_hyperloop(n, edges, waypoints):
    S = n + 1
    graph = [[] for _ in range(n + 2)]
    for u, v, w in edges:
        graph[u].append((v, w))
    for x in waypoints:
        graph[x].append((S, 0))
        graph[S].append((x, 0))
    dist = [[INF, INF] for _ in range(n + 2)]
    dist[1][0] = 0
    pq = [(0, 1, 0)]
    while pq:
        d, u, used = heapq.heappop(pq)
        if d > dist[u][used]:
            continue
        for v, w in graph[u]:
            if u == S:
                if dist[v][1] > d + w:
                    dist[v][1] = d + w
                    heapq.heappush(pq, (dist[v][1], v, 1))
            else:
                if dist[v][used] > d + w:
                    dist[v][used] = d + w
                    heapq.heappush(pq, (dist[v][used], v, used))
    ans = min(dist[n][0], dist[n][1])
    return -1 if ans == INF else ans
t = int(input())
for _ in range(t):
    n, m, k = map(int, input().split())
    edges = []
    for _ in range(m):
        u, v, w = map(int, input().split())
        edges.append((u, v, w))
    waypoints = list(map(int, input().split()))
    print(shortest_path_with_hyperloop(n, edges, waypoints))
```
