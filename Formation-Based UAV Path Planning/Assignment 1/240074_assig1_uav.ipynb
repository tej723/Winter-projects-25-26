```python
def problem1():
    n, m = map(int, input().split())
    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)

    visited = [False] * (n + 1)

    def dfs(node, parent):
        visited[node] = True
        for neighbor in graph[node]:
            if not visited[neighbor]:
                if dfs(neighbor, node):
                    return True
            elif neighbor != parent:
                return True
        return False

    for i in range(1, n + 1):
        if not visited[i]:
            if dfs(i, -1):
                print("YES")
                return

    print("NO")
```

```python
def problem2():
    n, m = map(int, input().split())
    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        u, v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)

    color = [-1] * (n + 1)

    def bfs(start):
        queue = [start]
        color[start] = 0
        head = 0

        while head < len(queue):
            node = queue[head]
            head += 1

            for neighbor in graph[node]:
                if color[neighbor] == -1:
                    color[neighbor] = 1 - color[node]
                    queue.append(neighbor)
                elif color[neighbor] == color[node]:
                    return False
        return True

    for i in range(1, n + 1):
        if color[i] == -1:
            if not bfs(i):
                print("NO")
                return

    print("YES")
    print(' '.join(map(str, color[1:n+1])))
```

```python
def problem3():
    n, m = map(int, input().split())
    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        u, v, w = map(int, input().split())
        graph[u].append((v, w))

    import heapq

    dist = [float('inf')] * (n + 1)
    dist[1] = 0
    pq = [(0, 1)]

    while pq:
        d, node = heapq.heappop(pq)

        if d > dist[node]:
            continue

        for neighbor, weight in graph[node]:
            if dist[node] + weight < dist[neighbor]:
                dist[neighbor] = dist[node] + weight
                heapq.heappush(pq, (dist[neighbor], neighbor))

    if dist[n] == float('inf'):
        print(-1)
    else:
        print(dist[n])
```

```python
def problem4():
    n, m, k = map(int, input().split())
    graph = [[] for _ in range(n + 1)]

    for _ in range(m):
        u, v, w = map(int, input().split())
        graph[u].append((v, w))

    waypoints = list(map(int, input().split()))

    import heapq

    def dijkstra(start):
        dist = [float('inf')] * (n + 1)
        dist[start] = 0
        pq = [(0, start)]

        while pq:
            d, node = heapq.heappop(pq)

            if d > dist[node]:
                continue

            for neighbor, weight in graph[node]:
                if dist[node] + weight < dist[neighbor]:
                    dist[neighbor] = dist[node] + weight
                    heapq.heappush(pq, (dist[neighbor], neighbor))

        return dist

    dist_from_1 = dijkstra(1)
    dist_to_n = [float('inf')] * (n + 1)

    reverse_graph = [[] for _ in range(n + 1)]
    for u in range(1, n + 1):
        for v, w in graph[u]:
            reverse_graph[v].append((u, w))

    dist_from_n_reverse = dijkstra(n)
    for u in range(1, n + 1):
        dist_to_n[u] = dist_from_n_reverse[u]

    min_dist = dist_from_1[n]

    for i in range(k):
        for j in range(k):
            wp1 = waypoints[i]
            wp2 = waypoints[j]

            if dist_from_1[wp1] != float('inf') and dist_to_n[wp2] != float('inf'):
                min_dist = min(min_dist, dist_from_1[wp1] + dist_to_n[wp2])

    if min_dist == float('inf'):
        print(-1)
    else:
        print(min_dist)
```
