```python
# Defining the input parsing function
def parse_tests(s: str):
    lines = [l.strip() for l in s.strip().splitlines() if l.strip()]
    i = 0
    tests = []
    while i < len(lines):
        n, m = map(int, lines[i].split())
        i += 1
        adj = {u: [] for u in range(n)}
        for _ in range(m):
            u, v = map(int, lines[i].split()) 
            i += 1
            u -= 1; v -= 1
            adj[u].append(v); adj[v].append(u)
        tests.append(adj)
    return tests

# example usage with the provided cases
data = """
4 4
1 2
2 3
3 1
3 4

5 4
1 2
2 3
3 4
4 5

3 3
1 2
2 3
3 1

6 3
1 2
3 4
5 6

7 6
1 2
2 3
3 4
4 2
5 6
6 7

1 0

4 2
1 2
3 4
"""
tests_problem1 = parse_tests(data)

print(tests_problem1)
```

```python
# Implementing DFS-based cycle detection
def DFS_cycle_finder(adjacency_list,start):
    n = len(adjacency_list)

    visited = [False] * n
    parent = [-1] * n
    curr_stack = []

    cycle_flag = False

    for s in range(n):

        if not visited[start]:
            curr_stack.append(start)
            visited[start] = True
        elif not visited[s]:
            curr_stack.append(s)
            visited[s] = True

        while len(curr_stack) > 0:
            node = curr_stack.pop()

            for neighbor in adjacency_list[node]:
                if not visited[neighbor]:
                    visited[neighbor] = True
                    curr_stack.append(neighbor)
                    parent[neighbor] = node
                elif parent[node] == neighbor:
                    continue
                else:
                    cycle_flag = True
                    return cycle_flag

    return cycle_flag

def calling_DFS_cycle_finder(adjacency_list,start):
    flag = DFS_cycle_finder(adjacency_list,start)

    if(flag):
        print("YES")
    else:
        print("NO")

# Running the cycle detection on each test case and printing results
for test in tests_problem1:
    calling_DFS_cycle_finder(test,0)
    
```

```python
# Implementing BFS-based bipartite checker
def BFS_bipartite_checker(adjacency_list,start):
    n = len(adjacency_list)

    color = [-1] * n
    curr_queue = []

    bipartite_flag = True

    for v in range(n):
        if color[start] == -1:
            curr_queue.append(start)
            color[start] = 1
        elif color[v] == -1:
            curr_queue.append(v)
            color[v] = 1

        while len(curr_queue) > 0:
            node = curr_queue.pop(0)

            for neighbour in adjacency_list[node]:
                if color[neighbour] == -1:
                    if color[node] == 1:
                        color[neighbour] = 2
                    else:
                        color[neighbour] = 1
                    curr_queue.append(neighbour)
                elif color[neighbour] == color[node]:
                    bipartite_flag = False
                    return bipartite_flag

    return bipartite_flag

def calling_BFS_bipartite_checker(adjacency_list,start):
    flag = BFS_bipartite_checker(adjacency_list,start)

    if(flag):
        print("YES")
    else:
        print("NO")

# Running the bipartite check on each test case and printing results
for test in tests_problem2:
    calling_BFS_bipartite_checker(test,0)
```

```python
# Taking input for Problem 3
def parse_tests_directed_weighted_graph(s: str):
    lines = [l.strip() for l in s.strip().splitlines() if l.strip()]
    i = 0
    tests = []
    while i < len(lines):
        n, m = map(int, lines[i].split())
        i += 1
        adj = {u: [] for u in range(n)}
        for _ in range(m):
            u, v, w = map(int, lines[i].split()) 
            i += 1
            u -= 1; v -= 1
            adj[u].append((v, w))
        tests.append(adj)
    return tests

str_input_directed_weighted = """
5 6
1 2 3
2 3 4
1 4 2
4 3 2
3 5 1
4 5 10

4 2
1 2 5
3 4 7

3 3
1 2 5
2 3 1
1 3 10

4 4
1 2 1
2 3 1
3 4 1
1 4 10

6 3
1 2 4
2 3 4
5 6 1

5 7
1 2 2
1 3 4
2 4 7
3 4 1
4 5 3
2 5 20
3 5 10

1 0
"""

tests_problem3 = parse_tests_directed_weighted_graph(str_input_directed_weighted)

print(tests_problem3)
```

```python
import heapq

# Implementing Dijkstra's algorithm for shortest path
def dijjkstra_shortest_path(adjacency_list,start):
    n = len(adjacency_list)

    distances = [-1] * n
    distances[start] = 0

    min_heap = [(0, start)]  # (distance, node)

    while len(min_heap) > 0:
        curr_distance, node = heapq.heappop(min_heap)

        if curr_distance > distances[node]:
            continue

        for neighbor, weight in adjacency_list[node]:
            distance = curr_distance + weight

            if distances[neighbor] == -1 or distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(min_heap, (distance, neighbor))

    return distances

# Running Dijkstra's algorithm on each test case and printing results
for test in tests_problem3:
    result = dijjkstra_shortest_path(test,0)
    print(result[len(result)-1])
    
```

```python
# Taking input for Problem 4
def parse_tests_directed_weighted_hyperloops(s: str):
    lines = [l.strip() for l in s.strip().splitlines() if l.strip()]
    i = 0
    tests = []
    while i < len(lines):
        n, m, k = map(int, lines[i].split())
        i += 1
        adj = {u: [] for u in range(n)}
        for _ in range(m):
            u, v, w = map(int, lines[i].split()) 
            i += 1
            u -= 1; v -= 1
            adj[u].append((v, w))

        hyperloop = []
        for _ in range(k):
            node = int(lines[i]) - 1
            hyperloop.append(node)
            i += 1

        tests.append((adj, hyperloop))

    return tests

str_input_directed_weighted_hyperloops = """
6 7 2
1 2 5
2 3 2
3 6 5
1 4 3
4 5 10
5 6 1
2 5 4
2
4

5 4 3
1 2 4
2 3 4
3 5 4
4 5 100
2
4
5

5 5 2
1 2 2
2 3 2
3 4 2
4 5 2
1 5 50
2
4

6 7 3
1 2 1
2 3 1
3 6 10
1 4 5
4 5 1
5 6 1
2 5 20
2
4
6

4 2 2
1 2 5
3 4 5
1
3

7 9 3
1 2 3
2 3 3
3 7 3
1 4 10
4 5 2
5 6 2
6 7 2
2 6 20
3 4 1
3
5
7

3 1 1
1 2 10
2
"""

tests_problem4 = parse_tests_directed_weighted_hyperloops(str_input_directed_weighted_hyperloops)

print(tests_problem4)
```

```python
# Implementing the hyperlinks shortest path algorithm
def hyperlinks(adjacency_list, hyperloop, start):
    n = len(adjacency_list)
    list_1 = dijjkstra_shortest_path(adjacency_list, start)

    # build reversed graph so distances are to the target (n-1)
    rev_adj = {u: [] for u in range(n)}
    for u, neighs in adjacency_list.items():
        for v, w in neighs:
            rev_adj[v].append((u, w))
    list_2 = dijjkstra_shortest_path(rev_adj, n - 1)

    min_distance = list_1[n - 1]
    if min_distance == -1:
        min_distance = float('inf')

    # Debugging
    # print("Distances from start:", list_1)
    # print("Distances to target:", list_2)
    # print("Hyperloop nodes:", hyperloop)
    
    u = float('inf')
    for i in hyperloop:
        if list_1[i] != -1:
            u = min(u, list_1[i])
    v = float('inf')
    for i in hyperloop:
        if list_2[i] != -1:
            v = min(v, list_2[i])
    
    if u != float('inf') and v != float('inf'):
        min_distance = min(min_distance, u + v)

    if min_distance == float('inf'):
        return -1
    return min_distance

# Running the hyperlinks shortest path on each test case and printing results
for test, hyperloops in tests_problem4:
    result = hyperlinks(test, hyperloops, 0)
    print(result)
```
