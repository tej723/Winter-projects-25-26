```python
import tensorflow as tf

(X_train_full, Y_train_full), (X_test_full, Y_test_full) = tf.keras.datasets.mnist.load_data()

print("MNIST dataset loaded successfully.")
print(f"X_train_full shape: {X_train_full.shape}, Y_train_full shape: {Y_train_full.shape}")
print(f"X_test_full shape: {X_test_full.shape}, Y_test_full shape: {Y_test_full.shape}")
```

**Reasoning**:
I will reshape and normalize the image data, implement one-hot encoding for labels, and then filter the dataset to include 1000 images per digit for both training and testing sets, ensuring the correct final dimensions for X and Y.



```python
image_size = X_train_full.shape[1] * X_train_full.shape[2]
num_classes = 10

# Reshape and normalize X_train_full
X_train_reshaped = X_train_full.reshape(X_train_full.shape[0], image_size).astype('float32') / 255.0
# Reshape and normalize X_test_full
X_test_reshaped = X_test_full.reshape(X_test_full.shape[0], image_size).astype('float32') / 255.0

# One-hot encoding function
def one_hot_encode(labels, num_classes):
    encoded_labels = np.zeros((labels.shape[0], num_classes))
    for i, label in enumerate(labels):
        encoded_labels[i, label] = 1
    return encoded_labels.T # Transpose to get (num_classes, num_samples)

# Apply one-hot encoding to Y_train_full and Y_test_full
Y_train_encoded = one_hot_encode(Y_train_full, num_classes)
Y_test_encoded = one_hot_encode(Y_test_full, num_classes)

# Initialize lists to store filtered data
X_train_filtered_list = []
Y_train_filtered_list = []
X_test_filtered_list = []
Y_test_filtered_list = []

# Filter 1000 images per digit for training and testing
images_per_digit = 1000

for digit in range(num_classes):
    # Training data
    train_indices = np.where(Y_train_full == digit)[0]
    if len(train_indices) > images_per_digit:
        train_indices = train_indices[:images_per_digit]
    X_train_filtered_list.append(X_train_reshaped[train_indices])
    Y_train_filtered_list.append(Y_train_encoded[:, train_indices])

    # Testing data
    test_indices = np.where(Y_test_full == digit)[0]
    if len(test_indices) > images_per_digit:
        test_indices = test_indices[:images_per_digit]
    X_test_filtered_list.append(X_test_reshaped[test_indices])
    Y_test_filtered_list.append(Y_test_encoded[:, test_indices])

# Concatenate filtered data
X_train = np.vstack(X_train_filtered_list).T # Transpose to get (features, samples)
Y_train = np.hstack(Y_train_filtered_list) # Keep (classes, samples)
X_test = np.vstack(X_test_filtered_list).T   # Transpose to get (features, samples)
Y_test = np.hstack(Y_test_filtered_list)   # Keep (classes, samples)

print("Preprocessing complete.")
print(f"X_train shape: {X_train.shape}, Y_train shape: {Y_train.shape}")
print(f"X_test shape: {X_test.shape}, Y_test shape: {Y_test.shape}")
```

## Implement Backward Propagation

### Subtask:
Complete the `backward_prop` function in cell `regulation-internship` by implementing the backpropagation equations provided in the notebook, using the derivative of the ReLU activation for the hidden layer.


**Reasoning**:
I need to complete the `backward_prop` function as per the instructions, implementing all the backpropagation equations provided. This involves calculating `dz2`, `dw2`, `db2`, `dz1`, `dw1`, and `db1` using the specified formulas and the `derivative_relu` function.



```python
def backward_prop(x, y, parameters, forward_cache):

    m = x.shape[1]

    # Fetching our parameters
    w1 = parameters['w1']
    b1 = parameters['b1']
    w2 = parameters['w2']
    b2 = parameters['b2']

    # Fetching our forward_cache
    a1 = forward_cache['a1']
    a2 = forward_cache['a2']
    z1 = forward_cache['z1'] # Need z1 for derivative_relu


    ## Complete the Code below ##
    dz2 = a2 - y
    dw2 = (1/m) * np.dot(dz2, a1.T)
    db2 = (1/m) * np.sum(dz2, axis=1, keepdims=True)

    dz1 = np.dot(w2.T, dz2) * derivative_relu(z1)
    dw1 = (1/m) * np.dot(dz1, x.T)
    db1 = (1/m) * np.sum(dz1, axis=1, keepdims=True)
    ## Your Code ends ##

    ## Returning the Gradients
    gradients = {
        "dw1" : dw1,
        "db1" : db1,
        "dw2" : dw2,
        "db2" : db2
    }

    return gradients
```

## Implement Update Parameters

### Subtask:
Fill in the `update_parameters` function in cell `convertible-samba` to update the weights and biases using gradient descent: `W = W - learning_rate * dW` and `B = B - learning_rate * dB`.


**Reasoning**:
I need to complete the `update_parameters` function in cell `convertible-samba` by implementing the gradient descent update rules for `w1`, `b1`, `w2`, and `b2` using the `learning_rate` and respective gradients.



```python
def update_parameters(parameters, gradients, learning_rate):

    # Fetching our parameters
    w1 = parameters['w1']
    b1 = parameters['b1']
    w2 = parameters['w2']
    b2 = parameters['b2']

    # Fetching our gradients
    dw1 = gradients['dw1']
    db1 = gradients['db1']
    dw2 = gradients['dw2']
    db2 = gradients['db2']

    ## Complete the Code below ##
    w1 = w1 - learning_rate * dw1
    b1 = b1 - learning_rate * db1
    w2 = w2 - learning_rate * dw2
    b2 = b2 - learning_rate * db2
    ## Your code ends ##

    # Returning the updated parameters
    Parameters = {
        "w1" : w1,
        "b1" : b1,
        "w2" : w2,
        "b2" : b2
    }

    return Parameters
```
