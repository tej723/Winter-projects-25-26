```python
import numpy as np
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def sigmderi(a):
    return a * (1 - a)

input_data = np.array([[0, 0],[0, 1],[1, 0],[1, 1]])
target_output = np.array([[0],[1],[1],[0]])

np.random.seed(50)

w1 = np.random.uniform(size=(2, 2))
b1 = np.random.uniform(size=(1, 2))

w2 = np.random.uniform(size=(2, 1))
b2 = np.random.uniform(size=(1, 1))

lr = 0.5
epochs = 10000

for epoch in range(epochs):

    H_input = np.dot(input_data, w1) + b1
    H_output = sigmoid(H_input)

    O_input = np.dot(H_output, w2) + b2
    predicted_O = sigmoid(O_input)

    error = target_output - predicted_O
    mse = np.mean(np.square(error))

    delta_O = error * sigmderi(predicted_O)

    error_H = delta_O.dot(w2.T)
    delta_H = error_H * sigmderi(H_output)

    w2 += H_output.T.dot(delta_O) * lr
    b2 += np.sum(delta_O, axis=0, keepdims=True) * lr

    w1 += input_data.T.dot(delta_H) * lr
    b1 += np.sum(delta_H, axis=0, keepdims=True) * lr

    if epoch % 2000 == 0:
        print(f"Iteration {epoch+1:5d} MSE: {mse:.8f}")

print(f"Training Complete")
print(f"Final Loss Value: {mse:.8f}")

finalpred = sigmoid(np.dot(sigmoid(np.dot(input_data, w1) + b1), w2) + b2)
finalres = (finalpred > 0.5).astype(int)

for input_val, output_val in zip(input_data, finalres):
    print(f"Input={input_val} Predicted Output={output_val[0]}")
```
