```python
import numpy as np
class LinearSVM:
    def __init__(self,learning_rate,n_iters,lambda_param):
        self.lr=learning_rate
        self.lambda_param=lambda_param
        self.n_iters=n_iters
        self.w=0
        self.b=0

    def fit(self,x,y):
        row,col=x.shape
        y=np.where(y<=0,-1,1)
        self.w=np.zeros(col)
        self.b=0
        for i in range(self.n_iters):
            for rowe,xi in enumerate(x):
                condition=y[rowe]*(np.dot(self.w,xi)-self.b)>=1
                if condition:
                    dw=2*self.lambda_param*self.w
                    self.w-=self.lr*dw
                else:
                    dw=2*self.lambda_param*self.w-y[rowe]*xi
                    db=-y[rowe]

                    self.w-=self.lr*dw
                    self.b-=self.lr*db

    def predict(self,x):
        linear_output=np.dot(x, self.w)-self.b
        return np.sign(linear_output)
```
