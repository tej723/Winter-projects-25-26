```python
import numpy as np
```

```python
x = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])
```

```python
np.random.seed(42)
```

```python
w1 = np.random.rand(2,2)
w2 = np.random.rand(2,1)
b1 = np.zeros((1,2))
b2 = np.zeros((1,1))
```

```python
lr = 0.01
iters = 20000
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def tanh_derivative(x):
    return 1-np.power(tanh(x),2)
```

```python
for i in range(iters):
    l1 = tanh(np.dot(x,w1)+b1)
    y_ = tanh(np.dot(l1,w2)+b2)

    error = y_-y
    d2 = error*tanh_derivative(y_)

    dw2 = np.dot(l1.T,d2)
    db2 = np.sum(d2,axis=0,keepdims=True)

    d1 = np.dot(d2,w2.T)*tanh_derivative(l1)
    dw1 = np.dot(x.T,d1)
    db1 = np.sum(d1,axis=0,keepdims=True)

    w1 -= lr*dw1
    b1 -= lr*db1
    w2 -= lr*dw2
    b2 -= lr*db2
```
