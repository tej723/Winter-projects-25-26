```python
import numpy as np
```

```python
class LinearSVM:
    #.Initialization
    def __init__(self,learning_rate=0.001,lambda_params=0.01,n_iters=100000):
        self.learning_rate = learning_rate
        self.lambda_params = lambda_params
        self.n_iters = n_iters
        self.w = None
        self.b = None

    #.Training
    def fit(self,X,y):
        samples,features = X.shape

        self.w = np.zeros(features)
        self.b = 0

        np.where(y <= 0, -1, 1)

        for i in range(self.n_iters):
          for i in range(len(X)):
            x_i = X[i]
            y_i = y[i]
            margin = y_i * (np.dot(self.w, x_i) - self.b)
            if margin >= 1:
                w -= self.learning_rate * (2 * self.lambda_param * self.w)
            else:
                w -= self.learning_rate * (2 * self.lambda_param * self.w - y_i * x_i)
                b -= self.learning_rate * y_i

    #.Prediction
    def predict(self,X):
        linear_output = np.dot(X,self.w) - self.b
```
