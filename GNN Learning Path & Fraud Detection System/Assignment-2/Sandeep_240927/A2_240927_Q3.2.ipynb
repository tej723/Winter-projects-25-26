```python
import numpy as np
```

```python
def sigmoid(n):
    return 1/(1+np.exp(-n))

def sigmoid_derivative(n):
    return n*(1-n)

class XOR:
    def __init__(self,learning_rate=0.1, n_iters=10000):
        self.learning_rate = learning_rate
        self.n_iters = n_iters


        self.w1 = np.random.rand(2,2)#input to hidden
        self.b1 = np.zeros((1,2))

        self.w2 = np.random.rand(2,1)#hidden to output
        self.b2 = np.zeros((1,1))

    def forward(self,X):
        self.z1 = np.dot(X,self.w1)+self.b1
        self.a1 = sigmoid(self.z1)

        self.z2 = np.dot(self.a1,self.w2)+self.b2
        self.a2 = sigmoid(self.z2)

        return self.a2

    def train(self,X,y):
        for i in range(self.n_iters):
            output = self.forward(X)

            error = y-output
            d_output = error*sigmoid_derivative(output)

            error_hidden = d_output.dot(self.w2.T)
            d_hidden = error_hidden*sigmoid_derivative(self.a1)


            self.w2 += self.a1.T.dot(d_output)*self.learning_rate
            self.b2 += np.sum(d_output,axis=0,keepdims=True)*self.learning_rate
            self.w1 += X.T.dot(d_hidden)*self.learning_rate
            self.b1 += np.sum(d_hidden,axis=0,keepdims=True)*self.learning_rate

    def predict(self,X):
      return np.round(self.forward(X))

```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([
    [0],
    [1],
    [1],
    [0]
])

model = XOR()
model.train(X,y)
print(model.predict(X))
```
