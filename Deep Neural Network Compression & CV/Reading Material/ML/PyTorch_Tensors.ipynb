## Statistics

```python
e = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
e
```

```python
# sum
torch.sum(e)
# sum along columns
torch.sum(e, dim=0)
# sum along rows
torch.sum(e, dim=1)
```

```python
# mean
torch.mean(e)
# mean along col
torch.mean(e, dim=0)
```

```python
# median
torch.median(e)
```

```python
# max and min
torch.max(e)
torch.min(e)
```

```python
# standard deviation
torch.std(e)
```

```python
# variance
torch.var(e)
```

```python
# argmax
torch.argmax(e)
```

```python
# argmin
torch.argmin(e)
```

## Matrix Operations

```python
m = torch.randint(size=(2,3), low=0, high=10)
n = torch.randint(size=(3,3), low=0, high=10)
print(m)
print(n)
```

```python
# matrix multiplcation
torch.matmul(m, n)
```

```python
vector1 = torch.tensor([1, 2])
vector2 = torch.tensor([3, 4])

# dot product --> done between 1-D vectors
torch.dot(vector1, vector2)
```

```python
# transpose
torch.transpose(m, 0, 1)
```

```python
# determinant (works for floating point tensor as an input)
n = n.to(torch.float32) # converting n to floating point from int.
torch.det(n)
```

```python
# inverse
torch.inverse(n)
```

## Special Functions( Activation Functions also)

```python
z = torch.randint(size=(2,3), low=0, high=10, dtype=torch.float32)
z
```

```python
# log
torch.log(z)
```

```python
# exp
torch.exp(z)
```

```python
# sqrt
torch.sqrt(z)
```

```python
# sigmoid
torch.sigmoid(z)
```

```python
# softmax
torch.softmax(z, dim=0)
```

```python
# relu
torch.relu(z)
```

## Inplace Operations

```python
m = torch.randn(2,3, dtype=torch.float32)
m
```

```python
n = torch.randn(2,3, dtype=torch.float32)
n
```

```python
m+n
# A completely new tensor is created occupying memory
```

```python
m.add_(n)
# Inplace operation
```

```python
m
```

```python
torch.relu_(m)
```

```python
m
```
