To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

def detect_gradients_and_plot(image, title_prefix="", threshold=None):
    """
    Detects horizontal (Ix) and vertical (Iy) gradients using Sobel kernels,
    plots them, and creates a scatter plot of Iy vs Ix.

    Args:
        image (numpy.ndarray): The input image (can be RGB or grayscale).
        title_prefix (str): A prefix for plot titles (e.g., 'Flat Image').
        threshold (float, optional): If provided, applies a threshold to Ix and Iy
                                     before plotting the scatter. Defaults to None.

    Returns:
        tuple: A tuple containing the Ix and Iy gradient arrays.
    """
    # Convert image to grayscale if it's RGB
    if len(image.shape) == 3:
        gray_image = np.mean(image, axis=2).astype(np.float32)
    else:
        gray_image = image.astype(np.float32)

    # Define Sobel kernels
    Gx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype=np.float32)

    Gy = np.array([[-1, -2, -1],
                   [0, 0, 0],
                   [1, 2, 1]], dtype=np.float32)

    # Convolve to find Ix and Iy
    Ix = convolve2d(gray_image, Gx, mode='same', boundary='symm')
    Iy = convolve2d(gray_image, Gy, mode='same', boundary='symm')

    # Plot Ix and Iy
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(Ix, cmap='gray')
    plt.title(f'{title_prefix} - Ix (Horizontal Gradient)')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(Iy, cmap='gray')
    plt.title(f'{title_prefix} - Iy (Vertical Gradient)')
    plt.axis('off')
    plt.show()

    # Prepare data for scatter plot, applying threshold if specified
    Ix_flat = Ix.flatten()
    Iy_flat = Iy.flatten()

    scatter_title = f'{title_prefix} - Iy vs Ix Scatter Plot'
    if threshold is not None:
        # Apply threshold to both Ix and Iy
        mask = (np.abs(Ix_flat) > threshold) | (np.abs(Iy_flat) > threshold)
        Ix_flat_thresh = Ix_flat[mask]
        Iy_flat_thresh = Iy_flat[mask]
        scatter_title += f' (Thresholded: > {threshold})'
    else:
        Ix_flat_thresh = Ix_flat
        Iy_flat_thresh = Iy_flat

    # Create scatter plot for Iy vs Ix
    plt.figure(figsize=(7, 7))
    plt.scatter(Ix_flat_thresh, Iy_flat_thresh, s=1, alpha=0.1, cmap='viridis')
    plt.title(scatter_title)
    plt.xlabel('Ix')
    plt.ylabel('Iy')
    plt.grid(True)
    plt.show()

    print(f"--- {title_prefix} Results ---")
    print(f"Shape of Ix: {Ix.shape}")
    print(f"Shape of Iy: {Iy.shape}")
    print(f"Min/Max of Ix: {np.min(Ix):.2f}/{np.max(Ix):.2f}")
    print(f"Min/Max of Iy: {np.min(Iy):.2f}/{np.max(Iy):.2f}")
    if threshold is not None:
        print(f"Number of points after thresholding: {len(Ix_flat_thresh)}")
    print("\n")

    return Ix, Iy
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

def calculate_r_map(image, window_size=5, k=0.04):
    """
    Calculates the R-map for corner detection using eigenvalues of the second moment matrix.

    Args:
        image (numpy.ndarray): The input image (can be RGB or grayscale).
        window_size (int): The size of the sliding window (e.g., 3, 5, 7).
        k (float): A constant, typically between 0.04 and 0.06, used in the R-value calculation.

    Returns:
        numpy.ndarray: The R-map, a 2D array where higher values indicate corners.
    """
    if len(image.shape) == 3:
        gray_image = np.mean(image, axis=2).astype(np.float32)
    else:
        gray_image = image.astype(np.float32)

    # Use Sobel kernels to get Ix and Iy
    Gx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]], dtype=np.float32)
    Gy = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]], dtype=np.float32)

    Ix = convolve2d(gray_image, Gx, mode='same', boundary='symm')
    Iy = convolve2d(gray_image, Gy, mode='same', boundary='symm')

    # Calculate products of gradients
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    height, width = gray_image.shape
    r_map = np.zeros_like(gray_image)

    offset = window_size // 2

    # Loop over image pixels with a sliding window
    for y in range(offset, height - offset):
        for x in range(offset, width - offset):
            # Extract window for gradient products
            window_Ix2 = Ix2[y - offset : y + offset + 1, x - offset : x + offset + 1]
            window_Iy2 = Iy2[y - offset : y + offset + 1, x - offset : x + offset + 1]
            window_Ixy = Ixy[y - offset : y + offset + 1, x - offset : x + offset + 1]

            # Sum within the window to get second moments (A, B, C)
            A = np.sum(window_Ix2)  # Corresponds to Ix^2 in the video notation
            C = np.sum(window_Iy2)  # Corresponds to Iy^2 in the video notation
            B = np.sum(window_Ixy)  # Corresponds to IxIy in the video notation

            # Construct the second moment matrix M
            M = np.array([[A, B],
                          [B, C]])

            # Calculate eigenvalues
            eigenvalues = np.linalg.eigvals(M)
            lambda1, lambda2 = eigenvalues[0], eigenvalues[1]

            # Calculate R-value using Harris corner detector formula
            # R = lambda1 * lambda2 - k * (lambda1 + lambda2)**2
            # Alternatively, R = det(M) - k * (trace(M))**2
            det_M = np.linalg.det(M)
            trace_M = np.trace(M)
            R = det_M - k * (trace_M ** 2)

            r_map[y, x] = R

    # Normalize R-map for better visualization
    r_map = r_map / np.max(np.abs(r_map)) * 255 # Scale to 0-255 for display
    r_map = np.clip(r_map, 0, 255)

    return r_map

def display_r_map(r_map, title_prefix=""):
    plt.figure(figsize=(8, 8))
    plt.imshow(r_map, cmap='gray')
    plt.title(f'{title_prefix} - R-map (Corners Highlighted White)')
    plt.axis('off')
    plt.show()
````

````python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import canny
from scipy.signal import convolve2d
import cv2 # For drawing lines, as specified

def hough_slope_intercept(image, sigma=3, low_threshold=0.1, high_threshold=0.2,
                          num_slopes=200, slope_range=(-5, 5),
                          num_intercepts=None, threshold_accumulator_ratio=0.8,
                          lines_to_draw=20):
    """
    Performs Hough Transform using the slope-intercept (m, c) method to detect lines.

    Args:
        image (numpy.ndarray): The input image (can be RGB or grayscale).
        sigma (float): Sigma for Canny edge detection.
        low_threshold (float): Low threshold for Canny edge detection.
        high_threshold (float): High threshold for Canny edge detection.
        num_slopes (int): Number of discrete slope values to consider.
        slope_range (tuple): (min_slope, max_slope) tuple.
        num_intercepts (int, optional): Number of discrete intercept values to consider.
                                        If None, it's calculated based on image dimensions.
        threshold_accumulator_ratio (float): Ratio to find peaks in the accumulator.
        lines_to_draw (int): Maximum number of strongest lines to draw.

    Returns:
        numpy.ndarray: Image with detected lines drawn.
    """
    if len(image.shape) == 3:
        gray_image = np.mean(image, axis=2).astype(np.uint8)
    else:
        gray_image = image.astype(np.uint8)

    # 1. Canny Edge Detection
    edges = canny(gray_image, sigma=sigma, low_threshold=low_threshold, high_threshold=high_threshold)

    # Display edges
    plt.figure(figsize=(8, 6))
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edges (Slope-Intercept Method)')
    plt.axis('off')
    plt.show()

    height, width = edges.shape

    # Define intercept range based on image dimensions and max slope
    max_abs_slope = max(abs(s) for s in slope_range)
    min_possible_intercept = -max_abs_slope * width - height # c = y - m*x. If m is large positive and x is large, c can be very negative.
    max_possible_intercept = height + max_abs_slope * width # If m is large negative and x is large, c can be very positive.

    # Adjust num_intercepts based on image size to ensure reasonable resolution
    if num_intercepts is None:
        num_intercepts = int((max_possible_intercept - min_possible_intercept) / 2) # Roughly 1 pixel resolution if slope is 0, wider if not

    # Discretize slopes and intercepts
    slopes = np.linspace(slope_range[0], slope_range[1], num_slopes)
    intercepts = np.linspace(min_possible_intercept, max_possible_intercept, num_intercepts)

    # Initialize accumulator
    accumulator = np.zeros((num_slopes, num_intercepts), dtype=np.int32)

    # Calculate step sizes for mapping c to index
    c_step = (max_possible_intercept - min_possible_intercept) / (num_intercepts - 1)
    if c_step == 0: c_step = 1 # Avoid division by zero if range is too small

    print(f"Accumulator dimensions: Slopes ({num_slopes}), Intercepts ({num_intercepts})")
    print(f"Slope range: {slope_range}, Intercept range: ({min_possible_intercept:.2f}, {max_possible_intercept:.2f})")

    # 3. Voting
    for y in range(height):
        for x in range(width):
            if edges[y, x]: # If it's an edge pixel
                for i_m, m in enumerate(slopes):
                    c = y - m * x
                    # Map calculated 'c' to its corresponding index in the accumulator
                    i_c = int((c - min_possible_intercept) / c_step)

                    if 0 <= i_c < num_intercepts:
                        accumulator[i_m, i_c] += 1

    # 4. Find Peaks
    # Normalize accumulator for thresholding
    max_acc = accumulator.max()
    if max_acc == 0:
        print("No votes accumulated. No lines detected.")
        return image

    threshold_value = max_acc * threshold_accumulator_ratio

    # Get all peaks above the threshold
    strong_peaks_indices = np.argwhere(accumulator > threshold_value)

    # Sort peaks by vote count in descending order
    sorted_peaks = []
    for i_m, i_c in strong_peaks_indices:
        sorted_peaks.append((accumulator[i_m, i_c], i_m, i_c))
    sorted_peaks.sort(key=lambda x: x[0], reverse=True)

    detected_lines_params = []
    for _, i_m, i_c in sorted_peaks[:lines_to_draw]: # Take top N lines
        m = slopes[i_m]
        c = intercepts[i_c]
        detected_lines_params.append((m, c))

    print(f"Detected {len(detected_lines_params)} lines to draw.")

    # 5. Draw Lines
    output_image = np.copy(image)
    if output_image.ndim == 2:
        output_image = cv2.cvtColor(output_image, cv2.COLOR_GRAY2BGR)

    for m, c in detected_lines_params:
        # Determine two points on the line within image boundaries for drawing
        # y = mx + c
        # x = (y - c) / m

        p1 = None
        p2 = None

        # Check intersection with left (x=0) and right (x=width-1) borders
        y_at_x0 = int(c)
        y_at_xw = int(m * (width - 1) + c)

        if 0 <= y_at_x0 < height:
            if p1 is None: p1 = (0, y_at_x0)
            else: p2 = (0, y_at_x0)

        if 0 <= y_at_xw < height and p2 is None:
            if p1 is None: p1 = (width - 1, y_at_xw)
            else: p2 = (width - 1, y_at_xw)

        # Check intersection with top (y=0) and bottom (y=height-1) borders
        if abs(m) > 1e-6: # Avoid division by zero for horizontal lines
            x_at_y0 = int(-c / m)
            x_at_yh = int((height - 1 - c) / m)

            if 0 <= x_at_y0 < width and p2 is None:
                if p1 is None: p1 = (x_at_y0, 0)
                else: p2 = (x_at_y0, 0)

            if 0 <= x_at_yh < width and p2 is None:
                if p1 is None: p1 = (x_at_yh, height - 1)
                else: p2 = (x_at_yh, height - 1)

        if p1 is not None and p2 is not None:
            cv2.line(output_image, p1, p2, (0, 255, 0), 2) # Green line, thickness 2

    plt.figure(figsize=(8, 6))
    plt.imshow(output_image)
    plt.title('Detected Lines (Slope-Intercept Method)')
    plt.axis('off')
    plt.show()

    return output_image
````

````python
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import canny
from skimage.transform import hough_line, hough_line_peaks
import cv2

def hough_rho_theta(image, sigma=1.0, low_threshold=0.1, high_threshold=0.2,
                    num_thetas=360, theta_range=(-np.pi / 2, np.pi / 2), # Default -90 to 90 degrees
                    threshold_accumulator_ratio=0.7,
                    lines_to_draw=25):
    """
    Performs Hough Transform using the rho-theta method to detect lines.

    Args:
        image (numpy.ndarray): The input image (can be RGB or grayscale).
        sigma (float): Sigma for Canny edge detection.
        low_threshold (float): Low threshold for Canny edge detection.
        high_threshold (float): High threshold for Canny edge detection.
        num_thetas (int): Number of discrete theta values to consider (e.g., 180 or 360).
        theta_range (tuple): (min_theta_radians, max_theta_radians) tuple.
        threshold_accumulator_ratio (float): Ratio to find peaks in the accumulator.
        lines_to_draw (int): Maximum number of strongest lines to draw.

    Returns:
        numpy.ndarray: Image with detected lines drawn.
    """
    if len(image.shape) == 3:
        gray_image = np.mean(image, axis=2).astype(np.uint8)
    else:
        gray_image = image.astype(np.uint8)

    # 1. Canny Edge Detection
    edges = canny(gray_image, sigma=sigma, low_threshold=low_threshold, high_threshold=high_threshold)

    # Display edges
    plt.figure(figsize=(8, 6))
    plt.imshow(edges, cmap='gray')
    plt.title('Canny Edges (Rho-Theta Method)')
    plt.axis('off')
    plt.show()

    # 2. Accumulator Space & 3. Voting (using scikit-image hough_line)
    accumulator, thetas, rhos = hough_line(edges, theta=np.linspace(*theta_range, num_thetas))

    # Display accumulator
    plt.figure(figsize=(8, 6))
    plt.imshow(np.log(1 + accumulator), extent=[np.rad2deg(thetas[-1]), np.rad2deg(thetas[0]), rhos[-1], rhos[0]], cmap='gray', aspect='auto')
    plt.title('Hough Accumulator (Rho-Theta)')
    plt.xlabel('Angles (degrees)')
    plt.ylabel('Distance (pixels)')
    plt.colorbar(label='Number of votes')
    plt.show()

    # 4. Find Peaks
    threshold_value = np.max(accumulator) * threshold_accumulator_ratio

    # `hough_line_peaks` automatically performs thresholding and non-maximum suppression
    # It returns (accumulators, angles, dists) for the detected peaks.
    H_peaks, angles, dists = hough_line_peaks(accumulator, thetas, rhos,
                                             num_peaks=lines_to_draw,
                                             threshold=threshold_value)

    print(f"Detected {len(H_peaks)} lines.")

    # 5. Draw Lines
    output_image = np.copy(image)
    if output_image.ndim == 2:
        output_image = cv2.cvtColor(output_image, cv2.COLOR_GRAY2BGR)

    for _, angle, dist in zip(H_peaks, angles, dists):
        # Calculate two points on the line for drawing
        # x * cos(theta) + y * sin(theta) = dist
        # For drawing lines, it's often easier to find points on the image borders.

        # Point 1: At x = 0
        if np.sin(angle) != 0:
            y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
            x0 = 0
        else: # Vertical line (angle is 0 or pi), x = dist/cos(angle)
            x0 = dist / np.cos(angle)
            y0 = 0 # arbitrary y for drawing

        # Point 2: At x = image_width - 1
        if np.sin(angle) != 0:
            y1 = (dist - (output_image.shape[1] - 1) * np.cos(angle)) / np.sin(angle)
            x1 = output_image.shape[1] - 1
        else:
            x1 = dist / np.cos(angle)
            y1 = output_image.shape[0] - 1 # arbitrary y for drawing

        # Ensure points are integers
        pt1 = (int(x0), int(y0))
        pt2 = (int(x1), int(y1))

        # Draw the line using OpenCV (it handles clipping to boundaries)
        cv2.line(output_image, pt1, pt2, (255, 0, 0), 2) # Red line, thickness 2

    plt.figure(figsize=(8, 6))
    plt.imshow(output_image)
    plt.title('Detected Lines (Rho-Theta Method)')
    plt.axis('off')
    plt.show()

    return output_image
````

### Review and Comparison of Hough Transform Methods

The Hough Transform is a powerful technique for detecting regular shapes like lines, circles, and ellipses in images. The two primary parameterizations for lines are the slope-intercept (m, c) method and the rho-theta (, ) method.

**Hough Transform (Slope-Intercept Method):**
*   **Parameterization:** Lines are represented by `y = mx + c`, where `m` is the slope and `c` is the y-intercept.
*   **Accumulator Space:** A 2D array is created with axes for `m` and `c`.
*   **Strengths:** Conceptually straightforward for lines that are not vertical.
*   **Weaknesses:**
    1.  **Vertical Lines:** The slope `m` approaches infinity for vertical lines, making it difficult to represent them accurately or efficiently in a finite accumulator space. This typically requires handling vertical lines as a special case or restricting the slope range, potentially missing important features. In our implementation, we constrained `slope_range=(-2, 2)`, which means very steep lines (close to vertical) might not be detected.
    2.  **Computational Cost:** Discretizing both `m` and `c` can lead to a large accumulator size and higher computational cost, especially when `c` can have a wide range depending on image size and `m`.
*   **Observation (Flat Image):** For the `flat` image, which has clear horizontal and vertical lines, the slope-intercept method with a restricted `m` range successfully identified the horizontal lines. However, it struggled with the vertical lines due to the chosen slope range.

**Hough Transform (Rho-Theta Method):**
*   **Parameterization:** Lines are represented by ` = x cos() + y sin()`, where `` (rho) is the perpendicular distance from the origin to the line, and `` (theta) is the angle of the normal vector from the origin to the line.
*   **Accumulator Space:** A 2D array is created with axes for `` and ``.
*   **Strengths:**
    1.  **Robustness to Orientation:** It naturally handles all line orientations, including vertical lines, as `` varies smoothly from 0 to  (or -/2 to /2) and `` varies accordingly. There are no "infinite" parameter values.
    2.  **Efficiency:** The parameter space (, ) is compact and well-behaved, allowing for more efficient discretization and processing.
    3.  **Scikit-image Integration:** Libraries like `scikit-image` provide optimized functions (`hough_line`, `hough_line_peaks`) that make implementation and peak detection more robust with built-in non-maximum suppression.
*   **Weaknesses:** While generally superior, it can sometimes produce false positives if the `threshold_accumulator_ratio` is too low or `num_peaks` is too high.
*   **Observation (Corner Image):** For the `corner` image, the rho-theta method clearly identified multiple lines forming the corner. The accumulator plot visually shows distinct peaks for these lines.

**Comparison:**
The `rho-theta` method is generally preferred over the `slope-intercept` method for line detection because of its robustness to vertical lines and its more compact and well-behaved parameter space. The `slope-intercept` method, while intuitive, can be computationally more challenging to implement for a full range of line orientations without special handling or compromises in the slope range, as observed with the `flat` image where vertical lines were mostly missed due to the restricted slope range. The use of `scikit-image`'s `hough_line` and `hough_line_peaks` for the `rho-theta` implementation significantly simplifies the process and provides efficient peak detection with built-in non-maximum suppression.

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.measure import find_contours, perimeter, regionprops
from skimage.filters import sobel
from scipy.ndimage import binary_fill_holes

def create_ball_stencil(image, bbox_coords, threshold=0.1):
    """
    Creates an RGBA stencil of the largest contour within a bounding box
    from the first image.

    Args:
        image (numpy.ndarray): The input image (e.g., 'ball' image).
        bbox_coords (tuple): A tuple (x1, y1, x2, y2) defining the top-left
                             and bottom-right coordinates of the bounding box.
        threshold (float): Threshold for contour detection after Canny/Sobel.

    Returns:
        numpy.ndarray: An RGBA stencil with the extracted object,
                       or None if no significant contour is found.
    """
    x1, y1, x2, y2 = bbox_coords

    # Crop the image to the bounding box
    cropped_image = image[y1:y2, x1:x2]

    if cropped_image.size == 0:
        print("Cropped image is empty. Adjust bounding box coordinates.")
        return None

    # Convert to grayscale
    gray_cropped = cv2.cvtColor(cropped_image, cv2.COLOR_RGB2GRAY)

    # Apply Canny edge detection
    edges = cv2.Canny(gray_cropped, 50, 150) # You might need to adjust Canny thresholds

    # Find contours
    # cv2.RETR_EXTERNAL retrieves only external contours
    # cv2.CHAIN_APPROX_SIMPLE compresses horizontal, vertical, and diagonal segments
    contours, _ = cv2.findContours(edges.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours:
        print("No contours found within the bounding box.")
        return None

    # Find the largest contour (by area)
    largest_contour = max(contours, key=cv2.contourArea)

    # Create a mask for the largest contour
    mask = np.zeros(gray_cropped.shape, dtype=np.uint8)
    cv2.drawContours(mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

    # Fill holes in the mask (optional, but often useful for object extraction)
    mask = binary_fill_holes(mask).astype(np.uint8) * 255

    # Create an RGBA stencil
    stencil = np.zeros((cropped_image.shape[0], cropped_image.shape[1], 4), dtype=np.uint8)

    # Copy RGB channels from the original cropped image
    stencil[:, :, :3] = cropped_image[:, :, :3]

    # Set the alpha channel based on the mask
    stencil[:, :, 3] = mask # Alpha channel: 0 (transparent) to 255 (opaque)

    return stencil

def overlay_stencil(background_image, stencil, center_coords):
    """
    Overlays an RGBA stencil onto a background image at a specified center point.

    Args:
        background_image (numpy.ndarray): The image to overlay onto (e.g., 'ball2' image).
        stencil (numpy.ndarray): The RGBA stencil to overlay.
        center_coords (tuple): A tuple (cx, cy) for the center of the stencil
                               on the background image.

    Returns:
        numpy.ndarray: The background image with the stencil overlaid.
    """
    if stencil is None:
        return background_image

    bg_height, bg_width, _ = background_image.shape
    stencil_height, stencil_width, _ = stencil.shape

    cx, cy = center_coords

    # Calculate top-left corner for placing the stencil
    x_offset = cx - stencil_width // 2
    y_offset = cy - stencil_height // 2

    # Define the region of interest (ROI) in the background image
    x_start_bg = max(0, x_offset)
    y_start_bg = max(0, y_offset)
    x_end_bg = min(bg_width, x_offset + stencil_width)
    y_end_bg = min(bg_height, y_offset + stencil_height)

    # Define the corresponding region in the stencil
    x_start_stencil = max(0, -x_offset)
    y_start_stencil = max(0, -y_offset)
    x_end_stencil = stencil_width - max(0, (x_offset + stencil_width) - bg_width)
    y_end_stencil = stencil_height - max(0, (y_offset + stencil_height) - bg_height)

    # Ensure valid ROI dimensions
    if x_start_bg >= x_end_bg or y_start_bg >= y_end_bg:
        print("Stencil completely outside background image boundaries. No overlay performed.")
        return background_image

    # Extract ROI from background and stencil
    bg_roi = background_image[y_start_bg:y_end_bg, x_start_bg:x_end_bg]
    stencil_roi = stencil[y_start_stencil:y_end_stencil, x_start_stencil:x_end_stencil]

    # Use the alpha channel for blending
    alpha_s = stencil_roi[:, :, 3] / 255.0
    alpha_b = 1.0 - alpha_s

    # Blend the stencil onto the background ROI
    for c in range(0, 3):
        bg_roi[:, :, c] = (alpha_s * stencil_roi[:, :, c] +
                           alpha_b * bg_roi[:, :, c])

    # Copy the modified ROI back to the background image
    output_image = background_image.copy()
    output_image[y_start_bg:y_end_bg, x_start_bg:x_end_bg] = bg_roi

    return output_image
````
