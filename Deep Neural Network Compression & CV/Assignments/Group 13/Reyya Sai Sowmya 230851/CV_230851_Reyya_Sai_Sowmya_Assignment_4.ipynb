To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def sobel_analysis(img_gray, title="Image"):
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[-1, -2, -1],
                        [ 0,  0,  0],
                        [ 1,  2,  1]], dtype=np.float32)

    Ix = cv2.filter2D(img_gray, -1, sobel_x)
    Iy = cv2.filter2D(img_gray, -1, sobel_y)

    plt.figure(figsize=(12, 8))

    plt.subplot(2, 2, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title(f"{title} (Grayscale)")
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix (Sobel X)")
    plt.axis('off')

    plt.subplot(2, 2, 3)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy (Sobel Y)")
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.scatter(Ix.flatten(), Iy.flatten(), s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Scatter Plot: Iy vs Ix")

    plt.tight_layout()
    plt.show()
    return Ix, Iy
````

Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def harris_r_map(img_gray, window_size=5, k=0.04):

    Ix = cv2.Sobel(img_gray, cv2.CV_32F, 1, 0, ksize=3)
    Iy = cv2.Sobel(img_gray, cv2.CV_32F, 0, 1, ksize=3)

    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy
    a = cv2.boxFilter(Ix2, -1, (window_size, window_size))
    b = cv2.boxFilter(Ixy, -1, (window_size, window_size))
    c = cv2.boxFilter(Iy2, -1, (window_size, window_size))
    det_M = a * c - b * b
    trace_M = a + c
    R_map = det_M - k * (trace_M ** 2)

    return R_map
````

````python
def show_r_map(R_map, title="R-map"):
    R_norm = cv2.normalize(R_map, None, 0, 255, cv2.NORM_MINMAX)
    R_norm = R_norm.astype(np.uint8)

    plt.figure(figsize=(5, 5))
    plt.imshow(R_norm, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
````

````python
def hough_mc(image_gray,
             m_range=(-2, 2),
             num_m=400,
             num_c=400,
             threshold_ratio=0.9):
    edges = cv2.Canny(image_gray.astype(np.uint8), 100, 200)

    h, w = image_gray.shape
    m_vals = np.linspace(m_range[0], m_range[1], num_m)
    c_vals = np.linspace(-h, h, num_c)

    accumulator = np.zeros((num_m, num_c), dtype=np.float32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for i, m in enumerate(m_vals):
            c = y - m * x
            c_idx = np.searchsorted(c_vals, c)
            if 0 <= c_idx < num_c:
                accumulator[i, c_idx] += 1
    accumulator /= accumulator.max()

    strong_votes = np.where(accumulator > threshold_ratio)

    img_color = cv2.cvtColor(image_gray.astype(np.uint8), cv2.COLOR_GRAY2BGR)

    for m_idx, c_idx in zip(strong_votes[0], strong_votes[1]):
        m = m_vals[m_idx]
        c = c_vals[c_idx]

        x1, x2 = 0, w
        y1 = int(c)
        y2 = int(m * w + c)

        cv2.line(img_color, (x1, y1), (x2, y2), (0, 0, 255), 1)

    return accumulator, img_color
````

````python
def hough_rho_theta(image_gray,
                    num_theta=180,
                    num_rho=400,
                    threshold_ratio=0.5):
    edges = cv2.Canny(image_gray.astype(np.uint8), 100, 200)
    h, w = image_gray.shape
    diag_len = int(np.sqrt(h*h + w*w))

    thetas = np.linspace(-np.pi/2, np.pi/2, num_theta)
    rhos = np.linspace(-diag_len, diag_len, num_rho)

    accumulator = np.zeros((num_rho, num_theta), dtype=np.float32)

    ys, xs = np.nonzero(edges)
    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            rho_idx = np.searchsorted(rhos, rho)
            if 0 <= rho_idx < num_rho:
                accumulator[rho_idx, t_idx] += 1

    accumulator /= accumulator.max()
    strong = np.where(accumulator > threshold_ratio)
    img_color = cv2.cvtColor(image_gray.astype(np.uint8), cv2.COLOR_GRAY2BGR)

    for rho_idx, theta_idx in zip(strong[0], strong[1]):
        rho = rhos[rho_idx]
        theta = thetas[theta_idx]

        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_color, (x1, y1), (x2, y2), (0, 255, 0), 1)

    return accumulator, img_color
````

````python
def create_ball_stencil(img, x1, y1, x2, y2):
    roi = img[y1:y2, x1:x2].copy()

    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    gray_blur = cv2.GaussianBlur(gray, (5, 5), 0)
    Ix = cv2.Sobel(gray_blur, cv2.CV_32F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray_blur, cv2.CV_32F, 0, 1, ksize=3)
    grad_mag = np.sqrt(Ix**2 + Iy**2)
    grad_mag = cv2.normalize(grad_mag, None, 0, 255, cv2.NORM_MINMAX)
    grad_mag = grad_mag.astype(np.uint8)

    _, edge_map = cv2.threshold(grad_mag, 40, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(edge_map, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    largest_contour = max(contours, key=cv2.contourArea)
    alpha = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(alpha, [largest_contour], -1, 255, thickness=-1)
    rgba = np.zeros((roi.shape[0], roi.shape[1], 4), dtype=np.uint8)
    rgba[..., :3] = roi
    rgba[..., 3] = alpha

    return rgba
````

````python
def overlay_stencil(background, stencil_rgba, center_x, center_y):
    bg = background.copy()
    h_s, w_s, _ = stencil_rgba.shape
    x_start = center_x - w_s // 2
    y_start = center_y - h_s // 2

    for y in range(h_s):
        for x in range(w_s):
            bg_y = y_start + y
            bg_x = x_start + x

            if bg_y < 0 or bg_x < 0 or bg_y >= bg.shape[0] or bg_x >= bg.shape[1]:
                continue

            alpha = stencil_rgba[y, x, 3] / 255.0
            if alpha > 0:
                bg[bg_y, bg_x] = (
                    alpha * stencil_rgba[y, x, :3] +
                    (1 - alpha) * bg[bg_y, bg_x]
                )

    return bg.astype(np.uint8)
````
