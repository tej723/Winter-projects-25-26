To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Convolution Function
def convolve(img, kernel):
    k_h, k_w = kernel.shape
    pad_h, pad_w = k_h // 2, k_w // 2
    
    img_padded = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
    output = np.zeros(img.shape, dtype = np.float32)
    
    for i in range(k_h):
        for j in range(k_w):
            region = img_padded[i : i + img.shape[0], j : j + img.shape[1]]
            output += region * kernel[i, j]
            
    return output

# Sobel Kernels
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
sobel_y = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

#B&W images
flat_gray = 0.299 * flat[:, :, 0] +0.587 * flat[:, :, 1] + 0.114 * flat[:, :, 2]
edge_gray = 0.299 * edge[:, :, 0] +0.587 * edge[:, :, 1] + 0.114 * edge[:, :, 2]
corner_gray = 0.299 * corner[:, :, 0] +0.587 * corner[:, :, 1] + 0.114 * corner[:, :, 2]
corner2_gray = 0.299 * corner2[:, :, 0] +0.587 * corner2[:, :, 1] + 0.114 * corner2[:, :, 2] 

# Sobel Edge Detection Function
def sobel_edge_detection(img, plot=False):
    # Compute I_x and I_y
    i_x = convolve(img, sobel_x)
    i_y = convolve(img, sobel_y)
    
    if plot:
        # Display I_x and I_y
        plt.figure(figsize=(2*6, 6))

        plt.subplot(1, 2, 1)
        plt.imshow(i_x, cmap = 'gray')
        plt.title("I_x", fontsize = 7)
        plt.axis("off")

        plt.subplot(1, 2, 2)
        plt.imshow(i_y, cmap = 'gray')
        plt.title("I_y", fontsize = 7)
        plt.axis("off")
        
        plt.show()

        plt.scatter(i_x.flatten(), i_y.flatten(), s=0.1)
        plt.title("Scatter Plot of I_y vs I_x", fontsize = 7)
        plt.show()
    
    return i_x, i_y

# Execute Sobel Edge Detection on all images
print("Sobel Edge Detection on flat_gray:")
sobel_edge_detection(flat_gray, plot=True)
print("\n")

print("Sobel Edge Detection on edge_gray:")
sobel_edge_detection(edge_gray, plot=True)
print("\n")

print("Sobel Edge Detection on corner_gray:")
sobel_edge_detection(corner_gray, plot=True)
print("\n")
````

````python
# Corner detection function using a window of size 3x3
def corner_detector(img):
    i_x, i_y = sobel_edge_detection(img)
    m1, m2, m3 = i_x**2, 2*i_x * i_y, i_y**2 # Second moments

    # Pad the second moment matrices (to handle borders)
    m1 = np.pad(m1, 1, mode='reflect')
    m2 = np.pad(m2, 1, mode='reflect')
    m3 = np.pad(m3, 1, mode='reflect')

    # Initialize R matrix
    R = np.zeros(img.shape)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # Sum the moments over the 3x3 window centered at (i, j)
            a = np.sum(m1[i:i+3, j:j+3])
            b = np.sum(m2[i:i+3, j:j+3])
            c = np.sum(m3[i:i+3, j:j+3])

            # Calculate the eigenvalues
            lambda1 = 0.5 * (a + c + np.sqrt(b**2 + (a - c)**2))
            lambda2 = 0.5 * (a + c - np.sqrt(b**2 + (a - c)**2))

            # Calculate R using Eigenvalues
            R[i][j] = (lambda1 * lambda2) - 0.04 * ((lambda1 + lambda2)**2)
    
    return R

# Display the R matrix
plt.imagesize = (6, 6)

plt.imshow(corner_detector(flat_gray), cmap='gray')
plt.title("Corner Response: flat")
plt.axis("off")
plt.show()
print('Maximum value of R:', np.max(corner_detector(flat_gray)))
print('No corners detected as expected')
print("\n")

plt.imshow(corner_detector(edge_gray), cmap='gray', vmin = 0)
plt.title("Corner Response: edge")
plt.axis("off")
plt.show()
print('Maximum value of R:', np.max(corner_detector(edge_gray)))
print("Few unexpected weak corners detected along the edges. Tried relative thresholding but it did not work since there aren't any other strong corners in the image")
print("\n")

plt.imshow(corner_detector(corner_gray), cmap='gray', vmin = 0)
plt.title("Corner Response: corner")
plt.axis("off")
plt.show()
print('Maximum value of R:', np.max(corner_detector(corner_gray)))
print('Single strong corner detected as expected')
````

````python
# rho limits
rho_min, rho_max = -1*np.sqrt(height**2 + width**2), np.sqrt(height**2 + width**2)
rho_step = 1
rho = np.arange(rho_min, rho_max, rho_step)

# theta limits: theta is measured from +x axis in the anti-clockwise direction
# +x axis points the left and +y axis points downwards in image coordinates
theta_min, theta_max = 0, np.pi
theta_step = np.pi / 180
theta = np.arange(theta_min, theta_max, theta_step)

def create_accumulator_rho_theta(img):
    # Finding Canny Edge Map
    img_canny = cv2.Canny(img, 50, 150)
    
    # Initializing Accumulator
    accumulator = np.zeros((len(rho), len(theta)), dtype=np.int32)
    
    # Getting Edge Pixel Coordinates where canny edge map is non-zero
    y_idx, x_idx = np.where(img_canny != 0)
    num_edge_pixels = len(y_idx)
  
    # Looping through each canny edge pixel and update the accumulator
    for i in range(num_edge_pixels):
        x = x_idx[i]
        y = y_idx[i]
        
        # For each canny edge pixel, looping through all angles contained in theta
        for theta_idx, theta_val in enumerate(theta):
            # rho_val corresponding to theta_val and current (x, y)
            rho_val = y * np.cos(theta_val) - x * np.sin(theta_val)
            
            # Finding the index of rho_val in rho
            rho_idx = int(round((rho_val - rho_min) / rho_step))
            
            # Boundary Check & Voting
            if 0 <= rho_idx < len(rho):
                accumulator[rho_idx, theta_idx] += 1

    return accumulator

# Calculating Accumulator
accumulator = create_accumulator_rho_theta(img_gray)

# Normalizing Accumulator
accumulator_norm = accumulator / np.max(accumulator)

# Using a 40% threshold to detect peak values in the accumulator
rho_peaks, theta_peaks = np.where(accumulator > 0.4 * np.max(accumulator))

# Drawing Gray lines around detected lines
result_img = img_gray.copy()

for i in range(len(rho_peaks)):
    rho_i = rho[rho_peaks[i]]
    theta_i = theta[theta_peaks[i]]

    x0 = -rho_i * np.sin(theta_i)
    y0 = rho_i * np.cos(theta_i)
    
    # Finding two points on the line segment
    x1 = int(x0 + 1000 * np.cos(theta_i))
    y1 = int(y0 + 1000 * np.sin(theta_i))
    x2 = int(x0 - 1000 * np.cos(theta_i))
    y2 = int(y0 - 1000 * np.sin(theta_i))
    
    # Drawing a gray line with parameters rho_i and theta_i using cv2
    cv2.line(result_img, (x1, y1), (x2, y2), (122, 122, 122), 2)

# Display
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.imshow(result_img, cmap='gray')
plt.title("Detected Lines using (rho, theta)")
plt.axis('off')

plt.subplot(1, 2, 1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.show()
print("Time taken during testing for me: 1.6s")
print("(rho, theta) parameterization is faster(x4.25) and can also handle lines with high slopes")
````
