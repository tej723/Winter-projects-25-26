To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def analyze_gradients(image_path, title, threshold=None):
    # Load image in grayscale
    img=cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        #Create synthetic images if files aren't provided
        img=create_synthetic_image(title)

    img=img.astype(np.float32)

    # Apply Sobel Kernels (ksize=3 for standard output)
    Ix=cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    Iy=cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

    # Flatten for scatter plot
    ix_flat=Ix.flatten()
    iy_flat=Iy.flatten()

    # Optional Thresholding to remove minor sensor noise/compression artifacts
    if threshold is not None:
        mask=(np.abs(ix_flat) > threshold) | (np.abs(iy_flat) > threshold)
        ix_plot=ix_flat[mask]
        iy_plot=iy_flat[mask]
        thresh_info=f" (Threshold: {threshold})"
    else:
        ix_plot=ix_flat
        iy_plot=iy_flat
        thresh_info=" (No Threshold)"

    # Plotting
    fig, axs=plt.subplots(1,3,figsize=(18,5))

    axs[0].imshow(img,cmap='gray')
    axs[0].set_title(f"Original: {title}")

    axs[1].imshow(Ix,cmap='hot')
    axs[1].set_title("Ix(Horizontal Gradients)")

    axs[2].scatter(ix_plot,iy_plot,alpha=0.5,s=2)
    axs[2].set_title(f"Scatter Plot Iy vs Ix{thresh_info}")
    axs[2].set_xlabel("Ix")
    axs[2].set_ylabel("Iy")
    axs[2].axhline(0,color='black',lw=1)
    axs[2].axvline(0,color='black',lw=1)

    plt.tight_layout()
    plt.show()
    return Ix,Iy

def create_synthetic_image(type):
    # Generates clean 100x100 images for demonstration
    canvas=np.zeros((100,100), dtype=np.uint8)
    if type=="Flat":
        canvas.fill(128)
    elif type=="Edge":
        canvas[;,50:]=255
    elif type=="Corner":
        canvas[40:60,40:60]=255
    return canvas

# Execution
# Set threshold to 10 to clear out floating point noise near zero
ix_f, iy_f=analyze_gradients("flat.png", "Flat", threshold=10)
ix_e, iy_e=analyze_gradients("edge.png", "Edge", threshold=10)
ix_c, iy_c=analyze_gradients("corner.png", "Corner", threshold=10)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def get_harris_rmap(Ix, Iy, window_size=5, k=0.04):
    # 1. Compute second moments (a, b, c)
    # These represent the sum of gradients over a local window
    Ixx=Ix**2
    Iyy=Iy**2
    Ixy=Ix*Iy

    # Using GaussianBlur for a weighted window (gives much smoother results than boxFilter)
    a=cv2.GaussianBlur(Ixx,(window_size,window_size),0)
    b=cv2.GaussianBlur(Ixy,(window_size,window_size),0)
    c=cv2.GaussianBlur(Iyy,(window_size,window_size),0)

    # 2. Calculate Determinant and Trace for the R-map
    det_M=(a*c)-(b**2)
    trace_M=a+c
    R=det_M-k*(trace_M**2)

    # 3. Calculate Eigenvalues (for your report/understanding)
    # Using the property: lambda1 + lambda2 = trace, lambda1 * lambda2 = det
    term=np.sqrt(trace_M**2-4*det_M)
    l1=(trace_M+term)/2
    l2=(trace_M-term)/2

    # 4. CRITICAL STEP FOR "WHITE CORNERS":
    # Suppress all negative values (edges) and near-zero values (flat)
    R_corners=np.copy(R)
    R_corners[R_corners<0]=0 # If R is negative, it's an edge. Kill it.

    # Normalize 0-255 for display
    R_display=cv2.normalize(R_corners, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    return R_display, l1, l2

def plot_part_b(img_title, Ix, Iy):
    rmap, l1, l2 = get_harris_rmap(Ix, Iy)

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.title(f"Original: {img_title}")
    # Displaying a reconstruction for context
    plt.imshow(create_synthetic_image(img_title), cmap='gray')

    plt.subplot(1, 2, 2)
    plt.title(f"Corrected R-Map: {img_title}")
    # Using 'hot' colormap: Black=Flat/Edge, White/Yellow=Corner
    plt.imshow(rmap, cmap='hot')
    plt.colorbar(label='Response Intensity')
    plt.show()

# Run for all three
plot_part_b("Flat", ix_f, iy_f)
plot_part_b("Edge", ix_e, iy_e)
plot_part_b("Corner", ix_c, iy_c)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hough_slope_intercept(img, m_range=(-5, 5), c_range=(-100, 200), steps=200):
    # Ensure image is 8-bit for Canny
    if img.dtype != np.uint8:
        img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    # 1. Edge Detection
    edges = cv2.Canny(img, 50, 150)

    # 2. Setup Accumulator
    accumulator = np.zeros((steps, steps))
    m_vals = np.linspace(m_range[0], m_range[1], steps)
    c_vals = np.linspace(c_range[0], c_range[1], steps)

    # 3. Voting procedure
    edge_pixels = np.argwhere(edges > 0)
    for y, x in edge_pixels:
        for m_idx, m in enumerate(m_vals):
            c = y - m * x # Formula: c = y - mx
            if c_range[0] <= c <= c_range[1]:
                c_idx = int((c - c_range[0]) / (c_range[1] - c_range[0]) * (steps - 1))
                accumulator[m_idx, c_idx] += 1

    # 4. Thresholding (Using 90th percentile as requested)
    thresh = np.percentile(accumulator, 99.9) # 90% is often too low for sparse accumulators
    indices = np.argwhere(accumulator > thresh)

    # 5. Drawing the lines
    output = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for m_idx, c_idx in indices:
        m, c = m_vals[m_idx], c_vals[c_idx]
        x1, x2 = 0, img.shape[1]
        y1, y2 = int(m * x1 + c), int(m * x2 + c)
        cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 1)

    return output, accumulator

# Use the Corner image from Part A/B
synthetic_img = create_synthetic_image("Corner")
res_mc, acc_mc = hough_slope_intercept(synthetic_img)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1); plt.imshow(acc_mc, cmap='hot'); plt.title("Accumulator (m, c)")
plt.subplot(1, 2, 2); plt.imshow(res_mc); plt.title("Slope-Intercept Lines")
plt.show()
````

````python
def hough_polar(img, theta_res=1, rho_res=1):
    if img.dtype != np.uint8:
        img = cv2.normalize(img, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    edges = cv2.Canny(img, 50, 150)
    h, w = edges.shape
    diag_len = int(np.sqrt(h**2 + w**2))

    thetas = np.deg2rad(np.arange(0, 180, theta_res))
    rhos = np.arange(-diag_len, diag_len, rho_res)
    accumulator = np.zeros((len(rhos), len(thetas)))

    # Pre-calculate cos and sin for speed
    cos_t, sin_t = np.cos(thetas), np.sin(thetas)
    edge_pixels = np.argwhere(edges > 0)

    for y, x in edge_pixels:
        for t_idx in range(len(thetas)):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            r_idx = int((rho + diag_len) / rho_res)
            accumulator[r_idx, t_idx] += 1

    # Thresholding (Top 95% for cleaner results)
    thresh = np.max(accumulator) * 0.8
    indices = np.argwhere(accumulator > thresh)

    output = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for r_idx, t_idx in indices:
        rho, theta = rhos[r_idx], thetas[t_idx]
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho
        x1, y1 = int(x0 + 1000*(-b)), int(y0 + 1000*(a))
        x2, y2 = int(x0 - 1000*(-b)), int(y0 - 1000*(a))
        cv2.line(output, (x1, y1), (x2, y2), (255, 0, 0), 1)

    return output, accumulator

res_polar, acc_polar = hough_polar(synthetic_img)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1); plt.imshow(acc_polar, extent=[0, 180, -140, 140], aspect='auto', cmap='hot')
plt.title("Accumulator (Rho, Theta)"); plt.xlabel("Theta (deg)"); plt.ylabel("Rho")
plt.subplot(1, 2, 2); plt.imshow(res_polar); plt.title("Polar Lines")
plt.show()
````
