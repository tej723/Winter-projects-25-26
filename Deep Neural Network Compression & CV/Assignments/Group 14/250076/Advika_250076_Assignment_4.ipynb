To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def compute_harris_rmap(Ix, Iy, k=0.04, window_size=3):
    # a = Ix^2, b = Iy^2, c = Ix*Iy
    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy


    a = cv2.boxFilter(Ixx, -1, (window_size, window_size))
    b = cv2.boxFilter(Iyy, -1, (window_size, window_size))
    c = cv2.boxFilter(Ixy, -1, (window_size, window_size))

    det = (a * b) - (c**2)
    trace = a + b
    R = det - k * (trace**2)


    R_norm = cv2.normalize(R, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    plt.figure(figsize=(6, 5))
    plt.imshow(R_norm, cmap='magma')
    plt.title("Harris R-Map")
    plt.colorbar()
    plt.show()

    return R

R_flat = compute_harris_rmap(Ix_f, Iy_f)
R_edge = compute_harris_rmap(Ix_e, Iy_e)
R_corner = compute_harris_rmap(Ix_c, Iy_c)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def hough_transform_comparison(image_input):
    # --- 1. PRE-PROCESSING ---
    if len(image_input.shape) == 3:
        gray = cv2.cvtColor(image_input, cv2.COLOR_RGB2GRAY)
    else:
        gray = image_input


    gray = (gray).astype(np.uint8)
    edges = cv2.Canny(gray, 50, 150)
    h, w = edges.shape
    y_idxs, x_idxs = np.nonzero(edges)

    # METHOD A: SLOPE-INTERCEPT (y = mx + c)
    m_range = np.linspace(-10, 10, 250)
    c_range = np.linspace(-h, h, 250)
    acc_mc = np.zeros((len(m_range), len(c_range)))

    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for m_idx, m in enumerate(m_range):
            c = y - m * x
            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                acc_mc[m_idx, c_idx] += 1


    img_mc = image_input.copy()

    thresh_mc = np.max(acc_mc) * 0.6
    m_p, c_p = np.where(acc_mc > thresh_mc)
    for mi, ci in zip(m_p, c_p):
        m, c = m_range[mi], c_range[ci]
        pt1 = (0, int(c))
        pt2 = (w, int(m * w + c))
        cv2.line(img_mc, pt1, pt2, (255, 0, 0), 2) # Blue lines

    # METHOD B: RHO-THETA
    diag_len = int(np.sqrt(h**2 + w**2))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    thetas = np.deg2rad(np.arange(0, 180))
    acc_rt = np.zeros((len(rhos), len(thetas)))

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for t_idx in range(len(thetas)):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            r_idx = int(rho + diag_len)
            acc_rt[r_idx, t_idx] += 1


    img_rt = image_input.copy()
    thresh_rt = np.max(acc_rt) * 0.5 # Usually need a lower thresh for polar
    r_p, t_p = np.where(acc_rt > thresh_rt)
    for ri, ti in zip(r_p, t_p):
        r, t = rhos[ri], thetas[ti]
        a, b = np.cos(t), np.sin(t)
        x0, y0 = a*r, b*r
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        cv2.line(img_rt, pt1, pt2, (0, 255, 0), 2) # Green lines


    fig, axes = plt.subplots(2, 3, figsize=(18, 10))


    axes[0, 0].imshow(image_input); axes[0, 0].set_title("Original Image")
    axes[0, 1].imshow(edges, cmap='gray'); axes[0, 1].set_title("Canny Edges")
    axes[0, 2].axis('off') # Spacer


    axes[1, 0].imshow(acc_mc, extent=[c_range[0], c_range[-1], m_range[-1], m_range[0]], aspect='auto')
    axes[1, 0].set_title("Accumulator (m, c)"); axes[1, 0].set_ylabel("Slope (m)"); axes[1, 0].set_xlabel("Intercept (c)")

    axes[1, 1].imshow(img_mc); axes[1, 1].set_title("Lines via m,c (Blue)")

    axes[1, 2].imshow(img_rt); axes[1, 2].set_title("Lines via Rho-Theta (Green)")

    plt.tight_layout()
    plt.show()

    print(f"DEBUG: Max votes (m,c): {np.max(acc_mc)}, Max votes (rho,theta): {np.max(acc_rt)}")


hough_transform_comparison(corner2)
````
