To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
from google.colab import files
uploaded = files.upload()

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load uploaded image
img = cv2.imread(list(uploaded.keys())[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
def partA_auto_fixed(img, threshold=None):

    import numpy as np
    import cv2
    import matplotlib.pyplot as plt

    # Grayscale
    gray = (0.299*img[...,0] +
            0.587*img[...,1] +
            0.114*img[...,2]).astype(np.float32)

    # Gaussian smoothing
    gray = cv2.GaussianBlur(gray, (5,5), 1.0)

    h, w = gray.shape

    #  Sobel kernels
    kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], dtype=np.float32)
    ky = np.array([[-1,-2,-1],[0,0,0],[1,2,1]], dtype=np.float32)

    Ix_full = cv2.filter2D(gray, -1, kx)
    Iy_full = cv2.filter2D(gray, -1, ky)

    win = 30
    records = []

    for y in range(0, h-win, win):
        for x in range(0, w-win, win):
            Ix_p = Ix_full[y:y+win, x:x+win]
            Iy_p = Iy_full[y:y+win, x:x+win]

            a = np.mean(Ix_p**2)
            c = np.mean(Iy_p**2)

            records.append((y, x, a, c))

    records = np.array(records, dtype=float)


    energy = records[:,2] + records[:,3]
    diff   = np.abs(records[:,2] - records[:,3])

    flat_box   = records[np.argmin(energy)]
    edge_box = records[np.argmax(energy * diff)]
    corner_box = records[np.argmax(energy - diff)]

    boxes = {
        "Flat Region"  : (int(flat_box[0]), int(flat_box[1])),
        "Edge Region"  : (int(edge_box[0]), int(edge_box[1])),
        "Corner Region": (int(corner_box[0]), int(corner_box[1]))
    }

    for name, (y,x) in boxes.items():

        patch = gray[y:y+win, x:x+win]

        Ix = cv2.filter2D(patch, -1, kx)
        Iy = cv2.filter2D(patch, -1, ky)

        if threshold is not None:
            Ix[np.abs(Ix) < threshold] = 0
            Iy[np.abs(Iy) < threshold] = 0

        Ix_vis = np.abs(Ix)
        Iy_vis = np.abs(Iy)
        Ix_vis /= Ix_vis.max() if Ix_vis.max()!=0 else 1
        Iy_vis /= Iy_vis.max() if Iy_vis.max()!=0 else 1

        plt.figure(figsize=(12,3))

        plt.subplot(1,3,1)
        plt.imshow(Ix_vis, cmap="gray")
        plt.title(name + " : Ix")
        plt.axis("off")

        plt.subplot(1,3,2)
        plt.imshow(Iy_vis, cmap="gray")
        plt.title(name + " : Iy")
        plt.axis("off")

        plt.subplot(1,3,3)
        plt.scatter(Ix.flatten(), Iy.flatten(), s=3)
        plt.xlabel("Ix")
        plt.ylabel("Iy")
        plt.title("Iy vs Ix")

        plt.tight_layout()
        plt.show()
results = partA_auto_fixed(img, threshold=10)
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def harris_r_map(img, window_size=5, k=0.04):


    # Convert to grayscale
    gray = (0.299*img[...,0] +
            0.587*img[...,1] +
            0.114*img[...,2]).astype(np.float32)

    # Gaussian smoothing
    gray = cv2.GaussianBlur(gray, (5,5), 1.0)

    # Sobel gradients
    Ix = cv2.Sobel(gray, cv2.CV_32F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_32F, 0, 1, ksize=3)

    # Second moment terms
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    #  Windowed averaging
    Sx2 = cv2.GaussianBlur(Ix2, (window_size, window_size), 1.0)
    Sy2 = cv2.GaussianBlur(Iy2, (window_size, window_size), 1.0)
    Sxy = cv2.GaussianBlur(Ixy, (window_size, window_size), 1.0)

    #  Harris R response
    detM = Sx2 * Sy2 - Sxy**2
    traceM = Sx2 + Sy2
    R = detM - k * (traceM**2)

    return R

def show_r_map(R, title="Harris R Map"):
    R_norm = R.copy()
    R_norm[R_norm < 0] = 0
    R_norm /= R_norm.max() if R_norm.max() != 0 else 1

    plt.figure(figsize=(6,6))
    plt.imshow(R_norm, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()

R = harris_r_map(img, window_size=7, k=0.04)
show_r_map(R, "Harris R-map (Corners in White)")
````

````python
from google.colab import files
uploaded = files.upload()

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread(list(uploaded.keys())[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()


def hough_rho_theta(img,
                    canny_low=100,
                    canny_high=200,
                    theta_res=1,
                    rho_res=1,
                    vote_thresh_ratio=0.6):


    # Grayscale
    gray = (0.299*img[...,0] +
            0.587*img[...,1] +
            0.114*img[...,2]).astype(np.uint8)

    # Canny Edge Detection
    edges = cv2.Canny(gray, canny_low, canny_high)

    # Parameter Space
    h, w = gray.shape
    diag_len = int(np.sqrt(h*h + w*w))

    rhos = np.arange(-diag_len, diag_len+1, rho_res)
    thetas = np.deg2rad(np.arange(-90, 90, theta_res))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.uint64)

    # Voting
    y_idxs, x_idxs = np.nonzero(edges)

    for x, y in zip(x_idxs, y_idxs):
        for t_idx, theta in enumerate(thetas):
            rho = int(x*np.cos(theta) + y*np.sin(theta))
            rho_idx = rho + diag_len
            accumulator[rho_idx, t_idx] += 1

    #  Thresholding
    acc_norm = accumulator / accumulator.max()
    detected = np.where(acc_norm >= vote_thresh_ratio)

    # Draw detected lines
    output = img.copy()

    for r_idx, t_idx in zip(*detected):
        rho = rhos[r_idx]
        theta = thetas[t_idx]

        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a*rho, b*rho

        x1, y1 = int(x0 + 1000*(-b)), int(y0 + 1000*(a))
        x2, y2 = int(x0 - 1000*(-b)), int(y0 - 1000*(a))

        cv2.line(output, (x1,y1), (x2,y2), (255,0,0), 2)

    return edges, acc_norm, output


#  FUNCTION CALL
edges, acc, final_img = hough_rho_theta(
    img,
    canny_low=100,
    canny_high=200,
    theta_res=1,
    rho_res=1,
    vote_thresh_ratio=0.6
)


plt.figure(figsize=(16,5))

plt.subplot(1,3,1)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Map")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(acc, cmap="hot")
plt.title("Hough Accumulator (-)")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(final_img)
plt.title("Detected Lines using Hough Transform")
plt.axis("off")

plt.tight_layout()
plt.show()
````
