To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Using the images provided above
# Writing the function
def corner_detection(img, title="Image"):
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  # calculting Ix and Iy
  Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
  Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

  # calculating the magnitude
  magnitude = np.sqrt(Ix**2 + Iy**2)

  # plotting all results
  plt.figure(figsize=(15, 5))
  plt.subplot(1, 3, 1)
  plt.imshow(gray, cmap='gray')
  plt.title(f'{title} (Grayscale)')
  plt.axis("off")
  plt.subplot(1, 3, 2)
  plt.imshow(magnitude, cmap='gray')
  plt.title("Gradient magnitude")
  plt.axis("off")
  plt.subplot(1, 3, 3)
  plt.scatter(Ix.flatten(), Iy.flatten(), alpha=0.1, s=1)
  plt.title("Scatter plot: Iy vs Ix")
  plt.xlabel("Ix")
  plt.ylabel("Iy")
  plt.show()

  return Ix, Iy

Ix_flat, Iy_flat = corner_detection(flat, 'flat')
Ix_edge, Iy_edge = corner_detection(edge, 'edge')
Ix_corner, Iy_corner = corner_detection(corner, 'corner')
````

Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def ellipse_fitting_R_maps_corner_detection(img, Ix, Iy, title='Image', window_size=3, k=0.04):
    height, width = Ix.shape
    offset = window_size // 2

    # Initializing the R-map
    R_map = np.zeros((height, width))

    # calculating pixel moments
    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy

    # looping over each pixel with window
    for i in range(offset, height - offset):
        for j in range(offset, width - offset):

            # calculating second moments (a, b, c)
            a = np.sum(Ixx[i-offset : i+offset+1, j-offset : j+offset+1])
            b = np.sum(Ixy[i-offset : i+offset+1, j-offset : j+offset+1])
            c = np.sum(Iyy[i-offset : i+offset+1, j-offset : j+offset+1])

            # forming matrix to find eigenvalues
            M = np.array([[a, b], [b, c]])
            lambdas = np.linalg.eigvals(M)
            l1, l2 = lambdas[0], lambdas[1]

            # Calculating value of R
            R = (l1 * l2) - k * (l1 + l2)**2

            # forming the final R-map
            R_map[i, j] = R

    # plotting the Results
    plt.figure(figsize=(6, 6))
    plt.imshow(R_map, cmap='gray')
    plt.title(f'R map for {title} image')
    plt.colorbar(label='R value')
    plt.axis('off')
    plt.show()

    return R_map

'''Corners in a corner image would be visible in white because at the corner point(where both edges meet) both lambdas will be very high resulting in high value of R hence a bright white spot.'''
R_flat = ellipse_fitting_R_maps_corner_detection(flat, Ix_flat, Iy_flat, "flat")
R_edge = ellipse_fitting_R_maps_corner_detection(edge, Ix_edge, Iy_edge, "edge")
R_corner = ellipse_fitting_R_maps_corner_detection(corner, Ix_corner, Iy_corner, "corner")
````

````python
def hough_slope_intercept(img, threshold_p=0.9):
    # Generating Canny edge map
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 50, 150)
    y_idxs, x_idxs = np.nonzero(edges)

    # Settingup Accumulator with defined limits
    m_steps, c_steps = 200, 200
    m_vals = np.linspace(-10, 10, m_steps)
    c_vals = np.linspace(-img.shape[0], img.shape[0], c_steps)
    accumulator = np.zeros((m_steps, c_steps))

    # Voting procedure
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for m_idx, m in enumerate(m_vals):
            c = y - m * x
            if c_vals[0] <= c <= c_vals[-1]:
                c_idx = np.argmin(np.abs(c_vals - c))
                accumulator[m_idx, c_idx] += 1

    # Thresholding :-
    thresh = np.max(accumulator) * threshold_p
    m_inds, c_inds = np.where(accumulator > thresh)

    # Marking lines on the image
    output = img.copy()
    for i in range(len(m_inds)):
        m, c = m_vals[m_inds[i]], c_vals[c_inds[i]]
        x1, y1 = 0, int(c)
        x2, y2 = img.shape[1], int(m * img.shape[1] + c)
        cv2.line(output, (x1, y1), (x2, y2), (255, 0, 0), 2)

    # plotting the results
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 3, 1);
    plt.imshow(edges, cmap='gray');
    plt.title("Canny Edges")
    plt.axis("off")
    plt.subplot(1, 3, 2);
    plt.imshow(accumulator, aspect='auto', extent=[c_vals[0], c_vals[-1], m_vals[0], m_vals[-1]])
    plt.title("m-c Accumulator");
    plt.xlabel("c");
    plt.ylabel("m")
    plt.subplot(1, 3, 3);
    plt.imshow(output);
    plt.title("Final Output (Lines Marked)")
    plt.axis("off")
    plt.show()

    return accumulator
acumulator = hough_slope_intercept(corner2)
````

````python
def hough_rho_theta(img, threshold_p=0.9):
    # Generating Canny edge map
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 50, 150)
    y_idxs, x_idxs = np.nonzero(edges)

    # Settingup Accumulator with defined limits
    h, w = edges.shape
    diag_len = int(np.sqrt(h**2 + w**2))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    thetas = np.deg2rad(np.arange(0, 180))
    accumulator = np.zeros((len(rhos), len(thetas)))

    # Voting procedure
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for t_idx, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            r_idx = int(rho + diag_len)
            if 0 <= r_idx < len(rhos):
                accumulator[r_idx, t_idx] += 1

    # Thresholding :-
    thresh = np.max(accumulator) * threshold_p
    r_inds, t_inds = np.where(accumulator > thresh)

    # Marking lines on the image
    output = img.copy()
    for i in range(len(r_inds)):
        rho, theta = rhos[r_inds[i]], thetas[t_inds[i]]
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        cv2.line(output, pt1, pt2, (0, 255, 0), 2)

    # plotting the results
    plt.figure(figsize=(18, 6))
    plt.subplot(1, 3, 1);
    plt.imshow(edges, cmap='gray');
    plt.title("Canny Edges")
    plt.subplot(1, 3, 2);
    plt.imshow(accumulator, aspect='auto', cmap='hot', extent=[0, 180, -diag_len, diag_len])
    plt.title("Rho-Theta Accumulator");
    plt.xlabel("Theta (rad)");
    plt.ylabel("Rho (pixels)")
    plt.subplot(1, 3, 3);
    plt.imshow(output);
    plt.title("Final Output (Lines Marked)")
    plt.show()

    return accumulator
accumulator1 = hough_rho_theta(corner2)
````
