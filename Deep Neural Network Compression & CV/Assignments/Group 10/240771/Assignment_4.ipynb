To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import requests
from PIL import Image
from io import BytesIO
import cv2 as cv

gray_flat = cv.cvtColor(flat, cv.COLOR_BGR2GRAY)
gray_edge = cv.cvtColor(edge, cv.COLOR_BGR2GRAY)
gray_corner = cv.cvtColor(corner, cv.COLOR_BGR2GRAY)
gray_corner2 = cv.cvtColor(corner2, cv.COLOR_BGR2GRAY)

def convolve(img, kernel):
    img = img.astype(np.float32)
    kernel = kernel.astype(np.float32)

    kH, kW = kernel.shape
    pad_h = kH // 2
    pad_w = kW // 2

    padded = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
    # reflect padding mirrors image values at the boundary.
    # This avoids artificial dark borders that zero-padding would introduce
    # and preserves edge continuity during convolution.

    H, W = img.shape
    output = np.zeros((H, W), dtype=np.float32)

    for i in range(H):
        for j in range(W):
            region = padded[i:i+kH, j:j+kW]
            output[i, j] = np.sum(region * kernel)

    # In NumPy, * performs element-wise multiplication.
    # This is valid here because 'region' and 'kernel' have the same shape.
    # NumPy does NOT perform convolution automatically with *.

    return output

sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [1, 2, 1]
], dtype=np.float32)

def calc_G(img):
  # Gx: horizontal intensity changes (vertical edges)
  Gx = convolve(img, sobel_x)
  # Gy: vertical intensity changes (horizontal edges)
  Gy = convolve(img, sobel_y)
  return Gx, Gy

def plot_G(img):
  Gx,Gy = calc_G(img)

  # Create the scatter plot
  plt.scatter(Gx, Gy, s=5)

  # Add title and labels
  plt.title("Gy vs Gx Scatter Plot")
  plt.xlabel("Gx")
  plt.ylabel("Gy")

  # Display the plot
  plt.show()

  # Add title and labels
  plt.imshow(Gx)
  plt.title("Gx Plot")
  plt.xlabel("Gx")

  # Display the plot
  plt.show()

  # Add title and labels
  plt.imshow(Gy)
  plt.title("Gy Plot")
  plt.xlabel("Gy")

  # Display the plot
  plt.show()

plot_G(gray_flat)
plot_G(gray_edge)
plot_G(gray_corner)
````

````python
def gaussian_kernel(ksize, sigma):
    ax = np.arange(-(ksize // 2), ksize // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)

    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel /= np.sum(kernel)

    return kernel

# sigma = 1
# ksize = 7
def plotR(img):
  Gx, Gy = calc_G(img)
  pre_a = Gx**2
  pre_b = Gx*Gy
  pre_c = Gy**2

  # Second moments summarize local gradient distribution. Eigenvalues of the structure tensor indicate intensity variation along orthogonal directions.
  a = convolve(pre_a, gaussian_kernel(7, 1))
  b = convolve(pre_b, gaussian_kernel(7, 1))
  c = convolve(pre_c, gaussian_kernel(7, 1))

  # The Harris response combines these into a scalar measure that is high for corners, low for edges, and near zero for flat regions.
  lambda1 = (a+c)/2 + ((((a-c)/2)**2) + b**2)**(1/2)
  lambda2 = (a+c)/2 - ((((a-c)/2)**2) + b**2)**(1/2)

  #The R-map is the spatial distribution of the Harris corner response, with one response value computed per pixel.
  R = lambda1*lambda2 - 0.05*(lambda1+lambda2)**2

  plt.title("R Map")
  plt.imshow(R, cmap='gray')
  plt.axis('off')
  plt.show()

plotR(gray_flat)
plotR(gray_edge)
plotR(gray_corner)
plotR(gray_corner2)
````

````python
url = "https://picsum.photos/200"
response = requests.get(url)

img = Image.open(BytesIO(response.content)).convert("L")
gray_img = np.array(img.convert("L"))

plt.title("Original Image")
plt.imshow(gray_img, cmap="gray")
plt.axis("off")
plt.show()

Gx, Gy = calc_G(gray_img)

# Gradient magnitude combines both directions
gradient_magnitude = np.sqrt(Gx ** 2 + Gy ** 2)

# Normalize gradient magnitude to range 0255 for visualization and saving
grad_norm = gradient_magnitude / gradient_magnitude.max()
grad_norm = (grad_norm * 255).astype(np.uint8)

# Thresholding:
# The normalized gradient magnitude contains both strong edges and weak responses due to noise.
# By applying a threshold, only pixels with gradient values greater than or equal to the threshold
# are retained as edges, while all weaker responses are suppressed.
threshold = 50
edge_map = np.zeros_like(grad_norm)
edge_map[grad_norm >= threshold] = 255

plt.title("Edge map")
plt.imshow(edge_map, cmap="gray")
plt.axis("off")
plt.show()

# Initialize the accumulator
m_min, m_max, m_step = -2, 2, 0.1
c_min, c_max, c_step = -200, 200, 1

m_vals = np.arange(m_min, m_max, m_step)
c_vals = np.arange(c_min, c_max, c_step)

acc = np.zeros((len(m_vals), len(c_vals)))

#looping through edge pixels
H, W = edge_map.shape

for y in range(H):
    for x in range(W):
        if edge_map[y, x] == 255:
            for mi, m in enumerate(m_vals):
                c = y - m * x
                if c_min <= c < c_max:
                    ci = int((c - c_min) / c_step)
                    if 0 <= ci < len(c_vals):
                      acc[mi, ci] += 1
                else:
                  continue
        else:
          continue

#normalize accumulator
acc_max = acc.max()
if acc.max() != 0:
    acc_normalized = acc / acc.max()
else:
    acc_normalized = acc

#threshold the accumulator
threshold = 0.7
lines = np.where(acc_normalized >= threshold)

#recover the accumulator and draw them
img_color = cv.cvtColor(gray_img, cv.COLOR_GRAY2BGR)

for mi, ci in zip(lines[0], lines[1]):
    m = m_vals[mi]
    c = c_vals[ci]

    x1, y1 = 0, int(c)
    x2, y2 = W, int(m * W + c)
    if 0 <= y1 < H and 0 <= y2 < H:
      cv.line(img_color, (x1, y1), (x2, y2), (0, 0, 255), 1)

plt.figure()
plt.title("Detected Lines (Slope-Intercept Hough)")
plt.imshow(img_color)
plt.axis("off")
plt.show()
````
