To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import io
import base64

def get_images():   # Setting up the image for questions

    # 1. Flat image referenced as flat
    flat = np.full((100, 100), 128, dtype=np.uint8)

    # 2. Edge image referenced as edge
    edge = np.zeros((100, 100), dtype=np.uint8)
    edge[:, 50:] = 255   #dimensions and pixels

    # 3. Corner image referenced as corner
    corner = np.zeros((100, 100), dtype=np.uint8)
    corner[50:, 50:] = 255   #dimensions and pixels

    # 4. Multiple Corner image with first corner as coner and 2nd corner as corner 2
    corner2 = np.zeros((100, 100), dtype=np.uint8)
    corner2[30:70, 30:70] = 255 #dimensions and pixels


    return flat, edge, corner, corner2

# Assigning random images to the references/variables
flat, edge, corner, corner2= get_images()

def analyze_gradients(img, name):
    # Defining Sobel kernels for horizontal (Ix) and vertical (Iy) gradients
    # Kx highlights vertical lines & Ky highlights horizontal lines
    Kx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    Ky = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

    # Use cv2.filter2D to combine the image with the kernels
    Ix = cv2.filter2D(img.astype(float), -1, Kx)  # type casting done to float to handle negative values
    Iy = cv2.filter2D(img.astype(float), -1, Ky)

    # Creating the visualization plots
    plt.figure(figsize=(15, 4))

    # Plotting the Original Grayscale image
    plt.subplot(1, 3, 1); plt.imshow(img, cmap='gray'); plt.title(f"{name} Original")

    # Plotting the The Gradient Map
    plt.subplot(1, 3, 2); plt.imshow(np.abs(Ix) + np.abs(Iy), cmap='gray'); plt.title(f"{name} Gradients")

    # Plotting the Scatter Plot of Iy vs Ix
    plt.subplot(1, 3, 3)
    plt.scatter(Ix.flatten(), Iy.flatten(), alpha=0.2, s=1) # Flatten converts the 2D image into a 1D pixel values
    plt.axhline(0, color='red', lw=1); plt.axvline(0, color='red', lw=1) # Reference lines
    plt.title(f"Iy vs Ix: {name}")
    plt.xlabel("Ix"); plt.ylabel("Iy")

    plt.show()
    return Ix, Iy  # returning values of Ix and Iy

# Running  it for flat, edge, and corner image
flat_Ix, flat_Iy = analyze_gradients(flat, "Flat")
edge_Ix, edge_Iy = analyze_gradients(edge, "Edge")
corner_Ix, corner_Iy = analyze_gradients(corner, "Corner")
````

````python
def compute_r_map(img, Ix, Iy, window_size=3, k=0.04):
    # Computing the products of gradients at every pixel
    Ixx = Ix * Ix
    Iyy = Iy * Iy
    Ixy = Ix * Iy

    # Summing these products over a local window using box filter(this will represent fitting of ellipse in the window)

    # Calculating values of a,b,c and eigenvalues
    a =Sumxx = cv2.boxFilter(Ixx, -1, (window_size, window_size))
    b=Sumxy = cv2.boxFilter(Ixy, -1, (window_size, window_size))
    c =Sumyy = cv2.boxFilter(Iyy, -1, (window_size, window_size))

    # eigen values are calculated using characteristic equation for a matrix(M-lambda*I=0)
#  Calculating eigenvalues (lambda1 and lambda2) using the quadratic formula
    term_sqrt = np.sqrt((a - c)**2 + 4 * (cv2.boxFilter(Ixy, -1, (window_size, window_size))**2))   # Term under the square root: (a-c)^2 + 4*b^2
    lambda1 = (a + c + term_sqrt) / 2
    lambda2 = (a + c - term_sqrt) / 2

    # Calculating the determinant and trace of the matrix M = [[Sxx, Sxy], [Sxy, Syy]]
    determinant_M = (cv2.boxFilter(Ixx, -1, (window_size, window_size)) * cv2.boxFilter(Iyy, -1, (window_size, window_size))) - (cv2.boxFilter(Ixy, -1, (window_size, window_size))**2)
    trace_M = cv2.boxFilter(Ixx, -1, (window_size, window_size)) + cv2.boxFilter(Iyy, -1, (window_size, window_size))

    # Calculating the  Harris Response R (R > 0- Corner; R < 0-Edge; |R| small- Flat region)
    R = determinant_M - k * (trace_M**2)

    # Displaying the R-map
    plt.figure(figsize=(5, 5))
    plt.imshow(R, cmap='hot')
    plt.title("Harris R-Map (Corners in White/Yellow)")
    plt.show()
    return R

# Running  it for Corner and Multiple Corner images
print("R-Map for Corner:")
r_map1 = compute_r_map(corner, corner_Ix, corner_Iy)
print("R-Map for Multiple Corners:")
# Helper to get gradients for corner2 since we didn't save them in Part A
c2_Ix = cv2.filter2D(corner2.astype(float), -1, np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]))
c2_Iy = cv2.filter2D(corner2.astype(float), -1, np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]]))
r_map2 = compute_r_map(corner2, c2_Ix, c2_Iy)
````

````python
# Your PART C codes follow here. You can create more cells under this. Write clean and commented code.
# calculating hough transform using slope intercept method
def hough_slope_intercept(img, threshold_percent=0.9):
    # Edge Detection: Getting the edge map using Canny
    edges = cv2.Canny(img, 50, 150)
    h, w = edges.shape

    # Defining Accumulator limits
    # m is for slope and c is for intercept on y axis
    m_range = np.linspace(-100, 100, 10000) # by fixing m between any two real numbers we avoid infinite slope problem
    c_range = np.linspace(-h, h, 10000)
    accumulator = np.zeros((len(m_range), len(c_range)))

    # Voting Procedure
    y_idxs, x_idxs = np.nonzero(edges) # A-Find all edge pixels
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for m_idx, m in enumerate(m_range):
            # B-Calculate c = y - mx
            c = y - m * x
            # C-Check if calculated c is within our accumulator range
            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                accumulator[m_idx, c_idx] += 1

    # Normalization and Thresholding
    acc_max = np.max(accumulator)
    # Getting coordinates where votes are higher than the threshold(90%)
    line_indices = np.argwhere(accumulator > (acc_max * threshold_percent))

    # Drawing the detected lines back on the image
    output_img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    for m_idx, c_idx in line_indices:
        m = m_range[m_idx]
        c = c_range[c_idx]
        # Getting two points to draw the line at x=0 and x=width
        p1 = (0, int(c))
        p2 = (w, int(m * w + c))
        cv2.line(output_img, p1, p2, (0, 0, 255), 1)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1); plt.imshow(accumulator, extent=[c_range[0], c_range[-1], m_range[-1], m_range[0]], aspect='auto')
    plt.title("Accumulator (m, c)"); plt.xlabel("c"); plt.ylabel("m")
    plt.subplot(1, 2, 2); plt.imshow(output_img)
    plt.title("Detected Lines (Slope-Intercept)")
    plt.show()

# Running on Multiple Corners
hough_slope_intercept(corner2)
````

````python
# The rho-Theta method uses rho = x cos theta + y sin theta to map points to curves in parameter space

def hough_transform(img):
    # Using Canny to get binary edge map
    edges = cv2.Canny(img, 50, 150)

    # Rho ranges from -diagonal to +diagonal,Theta ranges from -90 to 90 degrees
    diag_len = int(np.sqrt(img.shape[0]**2 + img.shape[1]**2))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    thetas = np.deg2rad(np.arange(-90, 90))

    # Initializing the accumulator with zeros
    accumulator = np.zeros((len(rhos), len(thetas)), dtype=int)

    # Voting: For every edge pixel, calculate all possible (rho, theta) pairs
    y_idxs, x_idxs = np.nonzero(edges)
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for t_idx, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            # Find the closest rho index in our accumulator
            rho_idx = np.argmin(np.abs(rhos - rho))
            accumulator[rho_idx, t_idx] += 1

    # Displaying the Accumulator
    plt.figure(figsize=(10, 5))
    plt.imshow(accumulator, extent=[-90, 90, diag_len, -diag_len], cmap='jet', aspect='auto')
    plt.title("Hough Accumulator (Rho-Theta)")
    plt.xlabel("Theta (degrees)"); plt.ylabel("Rho (pixels)")
    plt.show()

# Running on the Multiple Corner image
hough_transform(corner2)

# using rho-theta method the code is executed very fastly in about 1 sec while for slope intercept method it takes about 28 sec to execute for m between -100 to 100
# with 100 points .Thus the rho,theta method is very fast.
````
