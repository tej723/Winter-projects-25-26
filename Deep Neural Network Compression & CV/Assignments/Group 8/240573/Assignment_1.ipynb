````python
# B1. Compute the FFT of the grayscale image ( use NumPy )


# take grayscale image
gray = gray_manual.astype(float)

# FFT
F = np.fft.fft2(gray)
Fshift = np.fft.fftshift(F)


#B2. Compute magnitude, phase and display Magnitude Spectrum


# magnitude and phase
mag = np.abs(Fshift)
phase = np.angle(Fshift)

# log magnitude for display
mag_show = np.log1p(mag)
mag_show = mag_show / mag_show.max()

# normalize phase just for display
phase_show = (phase + np.pi) / (2*np.pi)

# show everything
plt.figure(figsize=(12,5))
plt.subplot(1,3,1); plt.imshow(gray, cmap='gray'); plt.title("Gray Modiji"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(mag_show, cmap='gray'); plt.title("Magnitude of Modiji"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(phase_show, cmap='twilight'); plt.title("Phase of Modiji"); plt.axis('off')
plt.show()
````

````python
# Defining all the required functions


# Circular Mask
def circular_mask(h, w, r):
    cy, cx = h//2, w//2
    Y, X = np.ogrid[:h, :w]
    D2 = (Y - cy)**2 + (X - cx)**2
    return (D2 <= r*r).astype(np.float32)

# Square Mask
def square_mask(h, w, side):
    cy, cx = h//2, w//2
    half = side//2
    m = np.zeros((h, w), dtype=np.float32)
    y0, y1 = max(0, cy-half), min(h, cy+half)
    x0, x1 = max(0, cx-half), min(w, cx+half)
    m[y0:y1, x0:x1] = 1.0
    return m

# Reconstructor
def recon(Fs):
    img = np.fft.ifft2(np.fft.ifftshift(Fs))
    img = np.real(img)
    img = img - img.min()
    if img.max() != 0:
        img = img / img.max()
    return img

# Normalization function
def norm01(x):
    y = x - x.min()
    if y.max() != 0:
        return y / y.max()
    return y
````

````python
# Create 3 LPF masks


radii = [80, 30, 10]
max_r = min(h, w)//2 - 1

# Checking bound on radii
radii = [min(r, max_r) for r in radii]

# Container for images
imgs_lpf = []

# Creating and applying LPF masks
for r in radii:

    m = circular_mask(h, w, r)
    Fs = Fshift * m
    imgs_lpf.append(recon(Fs))

# Display
plt.figure(figsize=(12,4))
plt.subplot(1, len(radii)+1, 1)
plt.imshow(norm01(gray), cmap='gray'); plt.title('orignal Modiji'); plt.axis('off')
for i, r in enumerate(radii):
    plt.subplot(1, len(radii)+1, i+2)
    plt.imshow(imgs_lpf[i], cmap='gray')
    plt.title(f'LPC r={r}')
    plt.axis('off')
plt.suptitle('Applying Circular LPFs')
plt.show()
````

Why blur level increases as we decrease the radius ?

--> This is because we have constructed the magnitude only graph of Modiji and shifted it so that low frequencies are at its center . These low frequencies capture the stable details of the images . High frequencies are responsible for sharp details like edge and all . By applying low pass filters we remove these details . The radius of low pass filter is responsiblle for how much frequencies are getting passed . Less radius means lesser frequinces that are small enough are getting passed . Hence the details that are stored in high frequencies get lost and hence as the radius decreases blurring increases .

````python
# C2. Create HPF mask

# Creating HPF mask
r_hpf = min(30, max_r)
lp = circular_mask(h, w, r_hpf)
hp = 1.0 - lp

# Applying it
img_hpf = recon(Fshift * hp)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(norm01(gray), cmap='gray'); plt.title('Orignal Modiji'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(img_hpf, cmap='gray'); plt.title(f'HPF r={r_hpf}'); plt.axis('off')
plt.suptitle('C2  High-Pass Filter')
plt.show()
````

````python
# C3. Create a square LPF mask

# Creating mask
side = 50
side = min(side, min(h, w))
sq = square_mask(h, w, side)

# Applying it
img_sq = recon(Fshift * sq)

# Display
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(norm01(gray), cmap='gray'); plt.title('orig'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(img_sq, cmap='gray'); plt.title(f'square LPF side={side}'); plt.axis('off')
plt.suptitle('C3  Square LPF')
plt.show()
````

````python
# D1. Phase-only reconstructions
eps = 1e-69

# From Angle
phase = np.angle(Fshift)
spec_phaseA = np.exp(1j * phase)
img_phaseA = recon(spec_phaseA)

# Dividing by magnitude
mag = np.abs(Fshift)
spec_phaseB = Fshift / (mag + eps)
img_phaseB = recon(spec_phaseB)

# Display
plt.figure(figsize=(12,4))
plt.subplot(1,4,1); plt.imshow(norm01(gray), cmap='gray'); plt.title('Original Modiji'); plt.axis('off')
plt.subplot(1,4,2); plt.imshow(img_phaseA, cmap='gray'); plt.title('Angle Modiji'); plt.axis('off')
plt.subplot(1,4,3); plt.imshow(img_phaseB, cmap='gray'); plt.title('F/|F| Modiji '); plt.axis('off')
phase_vis = (phase + np.pi) / (2*np.pi)
plt.subplot(1,4,4); plt.imshow(phase_vis, cmap='twilight'); plt.title('Phase of Modiji'); plt.axis('off')
plt.suptitle('Phase only Modiji')
plt.show()
````

* Whats the difference between those two ?
--> np.angle() just gives the angle in radians and not the complex number which is necessary for reconstruction where as dividing by magnitude gives the complex number corresponding to the phase


* How to do reconstruction from the angle part ?
--> Just find the corresponding complex number e^(i * angle)

````python
# D2. Magnitude-only reconstruction

eps = 1e-69

# Magnitude
mag = np.abs(Fshift)

# Build spectrum with zero phase
spec_mag = mag * np.exp(1j * 0.0)
img_mag = recon(spec_mag)

# Log magnitude for display
mag_vis = np.log1p(mag)
if mag_vis.max() != 0:
    mag_vis = mag_vis / mag_vis.max()

# Display
plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(norm01(gray), cmap='gray'); plt.title('Original Modiji'); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(img_mag, cmap='gray'); plt.title('Magnitude only Modiji'); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(mag_vis, cmap='gray'); plt.title('Log |F| of Modiji'); plt.axis('off')
plt.suptitle('Magnitude only Modiji')
plt.show()

````

* Why to use log in magnitude spectrum ?
--> It is just for visualization . The frequencies are present in very diverse amount and the difference between their occurences can differ alot . Hence to change the scale for visaulisation purpose we use the log !

* Difference between magnitude only reconstruction and magnitude spectrum ?
--> First one is zero phase reconstruction and the second one is visualization of |F|
