````python
# Histogram functions and plotting


# Gray histogram function
def gray_histogram(gray):
    assert gray.ndim == 2
    vals = gray.ravel()
    hist = np.bincount(vals, minlength=256)
    return hist


# RGB histogram function
def rgb_histogram(img_rgb):
    r = img_rgb[..., 0].ravel()
    g = img_rgb[..., 1].ravel()
    b = img_rgb[..., 2].ravel()
    hr = np.bincount(r, minlength=256)
    hg = np.bincount(g, minlength=256)
    hb = np.bincount(b, minlength=256)
    return hr, hg, hb
````

````python
# Plotting

# Grawscale bananalisa
rgb_f = rgb.astype(np.float32)
gray = (0.299*rgb_f[...,0] + 0.587*rgb_f[...,1] + 0.114*rgb_f[...,2]).round().astype(np.uint8)

# Histpgrams
h_gray = gray_histogram(gray)
hr, hg, hb = rgb_histogram(rgb)

# Plotting gray bananalisa histogram
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(h_gray, color='k')
plt.title('Grayscale bananalisa histogram')
plt.xlim([0,255])
plt.xlabel('Intensity'); plt.ylabel('Count')

# Plotting RGB bananalisa histogram
plt.subplot(1,2,2)
x = np.arange(256)
plt.plot(x, hr, color='r', label='R')
plt.plot(x, hg, color='g', label='G')
plt.plot(x, hb, color='b', label='B')
plt.title('RGB bananalisa histograms')
plt.xlim([0,255]); plt.xlabel('Intensity'); plt.ylabel('Count'); plt.legend()
plt.tight_layout()
````

````python
# Normalise RGB

def rgb_to_hsv(image) :
  rgb1 = image.astype(np.float32) / 255.0
  r, g, b = rgb1[...,0], rgb1[...,1], rgb1[...,2]

  # Finding max,min,difference between them for RGB

  mx = np.maximum.reduce([r,g,b])
  mn = np.minimum.reduce([r,g,b])
  diff = mx - mn

  # Computing hue

  h = np.zeros_like(mx)

  # r max
  mask = (mx == r) & (diff != 0)
  h[mask] = (60 * ((g[mask] - b[mask]) / diff[mask])) % 360
  # g max
  mask = (mx == g) & (diff != 0)
  h[mask] = 60 * ((b[mask] - r[mask]) / diff[mask] + 2)
  # b max
  mask = (mx == b) & (diff != 0)
  h[mask] = 60 * ((r[mask] - g[mask]) / diff[mask] + 4)

  # Computing saturation

  s = np.zeros_like(mx)
  mask = mx != 0
  s[mask] = diff[mask] / mx[mask]

  # Computing value

  v = mx


  # Converting to OpenCV ranges

  H = (h / 2).round()
  S = (s * 255).round()
  V = (v * 255).round()


  # Storing

  hsv = np.stack([H, S, V], axis=-1).astype(np.uint8)
  return hsv
````

````python
# Filter functions

# Gamma correction
def gamma_correction(img, gamma):
    inv = 1.0 / gamma
    lut = (np.arange(256) / 255.0) ** inv * 255.0
    lut = np.clip(lut, 0, 255).astype(np.uint8)
    out = cv2.LUT(img, lut)
    return out

# out = alpha*(img-128) + 128 + beta
def contrast_brightness(img, alpha=1.0, beta=0.0):
    out = img.astype(np.float32)
    out = alpha * (out - 128.0) + 128.0 + beta
    out = np.clip(out, 0, 255).astype(np.uint8)
    return out

# Filter 1: Sunny Boost
def sunny_boost(img_rgb, sat_gain=1.25, contrast=1.12, bright=8, gamma=1.0):
    """
    - Increase warmth (shift red up)
    - Increase saturation
    - Slight contrast and brightness boost
    Good for outdoor / beach photos.
    """
    out = img_rgb.copy()

    # 1) contrast + brightness
    out = contrast_brightness(out, alpha=contrast, beta=bright)

    # 2) warm tint: raise R channel a bit, reduce B a bit
    out = out.astype(np.float32)
    out[..., 0] = np.clip(out[..., 0] * 1.06, 0, 255)  # R
    out[..., 2] = np.clip(out[..., 2] * 0.97, 0, 255)  # B
    out = out.astype(np.uint8)

    # 3) saturation boost via HSV (use our rgb_to_hsv + cv2 to convert back)
    hsv = rgb_to_hsv(out)
    hsv = hsv.astype(np.uint8)

    # scale S channel
    S = hsv[...,1].astype(np.float32) * sat_gain
    hsv[...,1] = np.clip(S, 0, 255).astype(np.uint8)

    # back to RGB via cv2
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)
    rgb_out = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

    # 4) optional gamma
    if gamma != 1.0:
        rgb_out = gamma_correction(rgb_out, gamma)

    return rgb_out

# Filter 2: Moody Cool
def moody_cool(img_rgb, hue_shift_deg=10, contrast=0.9, gamma=1.15, shadow_blue=15):
    """
    - Cool blue tint in shadows, slight desaturation, gamma brighten midtones
    - Good for moody portraits, urban scenes.
    """
    out = img_rgb.copy().astype(np.float32)

    # 1) mild contrast reduction
    out = contrast_brightness(out.astype(np.uint8), alpha=contrast, beta=0).astype(np.float32)

    # 2) convert to HSV to reduce saturation a bit and rotate hue slightly
    hsv = rgb_to_hsv(out.astype(np.uint8)).astype(np.int32)

    # hue shift: convert back to degrees then shift
    H = hsv[...,0].astype(np.float32) * 2.0   # deg 0..360
    H = (H + hue_shift_deg) % 360.0
    hsv[...,0] = np.round(H / 2.0).astype(np.uint8)

    # reduce saturation for moody look
    hsv[...,1] = np.clip(hsv[...,1].astype(np.float32) * 0.85, 0, 255).astype(np.uint8)

    # back to RGB
    bgr = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2BGR)
    rgb_back = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB).astype(np.float32)

    # 3) split-tone-ish: boost blue in shadows
    # compute mask where pixel is dark (V small)
    v = hsv[...,2].astype(np.float32)
    shadow_mask = (v < 100).astype(np.float32)[..., None]  # narrow range
    rgb_back[..., 2] = np.clip(rgb_back[..., 2] + shadow_blue * shadow_mask[...,0], 0, 255)

    # 4) gamma for midtones
    rgb_out = gamma_correction(rgb_back.astype(np.uint8), gamma)

    return rgb_out
````

Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image.
````python
# White patch balance


def white_patch_balance(img_rgb):
    """
    img_rgb: HxWx3 uint8
    returns balanced uint8 image (0..255)
    """

    img = img_rgb.astype(np.float32)

    # max of every channel
    max_r = img[...,0].max()
    max_g = img[...,1].max()
    max_b = img[...,2].max()

    # avoid division by zero
    fr = 255.0 / max_r if max_r > 0 else 1.0
    fg = 255.0 / max_g if max_g > 0 else 1.0
    fb = 255.0 / max_b if max_b > 0 else 1.0

    # scale channels
    out = img.copy()
    out[...,0] = np.clip(out[...,0] * fr, 0, 255)
    out[...,1] = np.clip(out[...,1] * fg, 0, 255)
    out[...,2] = np.clip(out[...,2] * fb, 0, 255)
    return out.astype(np.uint8)

# Apply and Display
wb = white_patch_balance(rgb)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1); plt.imshow(rgb); plt.title('Original Bananalisa'); plt.axis('off')
plt.subplot(1,2,2); plt.imshow(wb); plt.title('White-patch Balanced Bananalisa'); plt.axis('off')
````

````python
# White_balance slider


def white_balance(img_rgb, value):
    """
    img_rgb: HxWx3 uint8
    value: float in [0,1] where 0 = coldest, 0.5 = original, 1 = warmest
    Strategy:
      - interpolate R and B channel gains: at 0 => R*0.85,B*1.15 (cold)
                                       at 0.5 => 1.0,1.0
                                       at 1 => R*1.15,B*0.85 (warm)
      - slight saturation adjustment tied to value (warm increases sat slightly)
    """
    v = float(value)

    # Linear interpolation
    r_gain = 1.0 + (v - 0.5) * 0.6
    b_gain = 1.0 - (v - 0.5) * 0.6
    s_gain = 1.0 + (v - 0.5) * 0.15
    out = img_rgb.astype(np.float32)
    out[...,0] = np.clip(out[...,0] * r_gain, 0, 255)
    out[...,2] = np.clip(out[...,2] * b_gain, 0, 255)


    # adjust saturation
    hsv = rgb_to_hsv(out.astype(np.uint8)).astype(np.int32)
    hsv[...,1] = np.clip(hsv[...,1].astype(np.float32) * s_gain, 0, 255).astype(np.uint8)
    bgr = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2BGR)
    out_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    return out_rgb

# Apply and Display
cold = white_balance(rgb, 0.0)
orig = white_balance(rgb, 0.5)
warm = white_balance(rgb, 1.0)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.imshow(cold); plt.title('value=0.0 (cold)'); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(orig); plt.title('value=0.5 (original)'); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(warm); plt.title('value=1.0 (warm)'); plt.axis('off')
````
