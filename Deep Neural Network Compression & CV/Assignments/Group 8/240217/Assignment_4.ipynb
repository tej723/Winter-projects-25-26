To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# RGB image to grayscale
def to_grayscale(img_np):
    # convert to float for calculation
    img_float = img_np.astype(np.float32)

    # apply weighted sum formula
    gray = (0.299 * img_float[:, :, 0] +
            0.587 * img_float[:, :, 1] +
            0.114 * img_float[:, :, 2])

    # convert back to uint8 (0-255 range)
    return gray.astype(np.uint8)


# function to convolve
def convolve2d(image, kernel):

    # image and kernel sizes
    H, W = image.shape
    kH, kW = kernel.shape

    # Amount of padding needed
    pad_h = kH // 2
    pad_w = kW // 2

    # Pad image with zeros on all sides
    padded = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')

    # Output image
    output = np.zeros((H, W), dtype=np.float32)


    kernel = np.flipud(np.fliplr(kernel))
    for i in range(H):
        for j in range(W):
            region = padded[i:i+kH, j:j+kW]
            output[i, j] = np.sum(region * kernel)

    return output



def sobel_corner_analysis(gray_img, threshold=None, title="Image"):

    # Sobel kernels
    SOBEL_X = np.array([
        [-1, 0, 1],
        [-2, 0, 2],
        [-1, 0, 1]
    ])

    SOBEL_Y = np.array([
        [-1, -2, -1],
        [ 0,  0,  0],
        [ 1,  2,  1]
    ])

    # Computing gradients using convolution
    Ix = convolve2d(gray_img, SOBEL_X)
    Iy = convolve2d(gray_img, SOBEL_Y)

    # Preparing scatter data
    if threshold is not None:
        magnitude = np.sqrt(Ix**2 + Iy**2)
        mask = magnitude > threshold
        Ix_scatter = Ix[mask]
        Iy_scatter = Iy[mask]
    else:
        Ix_scatter = Ix.flatten()
        Iy_scatter = Iy.flatten()

    # Display
    plt.figure(figsize=(15, 4))

    plt.subplot(1, 3, 1)
    plt.imshow(Ix, cmap='gray')
    plt.title(f"{title} - Ix")
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.imshow(Iy, cmap='gray')
    plt.title(f"{title} - Iy")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.scatter(Ix_scatter, Iy_scatter, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix Scatter")

    plt.tight_layout()
    plt.show()

    return Ix, Iy
````

````python
# Fn. to compute R map
def compute_R_map(Ix, Iy, window_size=5, k=0.04):

    H, W = Ix.shape
    offset = window_size // 2

    R_map = np.zeros((H, W), dtype=np.float32)

    # Loop over pixels
    for i in range(offset, H - offset):
        for j in range(offset, W - offset):

            # Extract local window
            Ix_win = Ix[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy_win = Iy[i-offset:i+offset+1, j-offset:j+offset+1]

            # Compute second moments
            a = np.sum(Ix_win ** 2)
            b = np.sum(Iy_win ** 2)
            c = np.sum(Ix_win * Iy_win)

            # Determinant and trace
            det_M = a * b - c ** 2
            trace_M = a + b

            # Harris response
            R = det_M - k * (trace_M ** 2)

            R_map[i, j] = R

    return R_map
````

````python
#Display
def show_R_map(R_map, title="R Map"):

    # Normalize for display
    R_norm = R_map - R_map.min()
    R_norm = R_norm / (R_norm.max() + 1e-8)

    plt.figure(figsize=(5, 5))
    plt.imshow(R_norm, cmap='gray')
    plt.title(title)
    plt.axis("off")
    plt.show()
````

````python
# Hough Line Transform using slopeintercept form y = mx + c

def hough_slope_intercept(
    img,
    m_min=-2, m_max=2,
    m_bins=150,
    c_bins=200,
    acc_threshold_ratio=0.9
):

    # Convert to grayscale
    gray = to_grayscale(img_np)

    # Canny edge detection
    edges = cv2.Canny(gray, 80, 200)

    H, W = edges.shape

    # slope and intercept ranges
    m_values = np.linspace(m_min, m_max, m_bins)

    c_max = int(np.sqrt(H**2 + W**2))
    c_values = np.linspace(-c_max, c_max, c_bins)

    # Accumulator
    accumulator = np.zeros((m_bins, c_bins), dtype=np.uint64)

    # Edge coordinates
    edge_pixels = np.argwhere(edges > 0)

    for y, x in edge_pixels:
        for m_idx, m in enumerate(m_values):
            c = y - m * x
            c_idx = np.searchsorted(c_values, c)

            # if inside accumulator
            if 0 <= c_idx < c_bins:
                accumulator[m_idx, c_idx] += 1

    # Normalize
    acc_norm = accumulator / accumulator.max()

    # Threshold accumulator
    strong_params = np.argwhere(acc_norm >= acc_threshold_ratio)

    # Draw detected lines
    output_img = img.copy()

    for m_idx, c_idx in strong_params:
        m = m_values[m_idx]
        c = c_values[c_idx]

        x0, x1 = 0, W
        y0 = int(m * x0 + c)
        y1 = int(m * x1 + c)

        if 0 <= y0 < H and 0 <= y1 < H:
            cv2.line(output_img, (x0, y0), (x1, y1), (255, 0, 0), 2)

    return output_img, acc_norm
````

````python
output_si, acc_si = hough_slope_intercept(
    img_np,
    m_min=-2, m_max=2,
    m_bins=150,
    c_bins=200,
    acc_threshold_ratio=0.9
)


plt.imshow(output_si)
plt.title("Hough Lines  Slope Intercept")
plt.axis("off")
plt.show()
````

````python
# Hough Line Transform using rhotheta form

def hough_rho_theta(
    img,
    rho_bins=180,
    theta_bins=180,
    acc_threshold_ratio=0.6
):

    gray = to_grayscale(img_np)
    edges = cv2.Canny(gray, 80, 200)

    H, W = edges.shape
    diag_len = int(np.sqrt(H**2 + W**2))

    rhos = np.linspace(-diag_len, diag_len, rho_bins)
    thetas = np.linspace(0, np.pi, theta_bins)

    accumulator = np.zeros((rho_bins, theta_bins), dtype=np.uint64)

    edge_pixels = np.argwhere(edges > 0)

    # Voting
    for y, x in edge_pixels:
        for t_idx, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            r_idx = np.searchsorted(rhos, rho)

            if 0 <= r_idx < rho_bins:
                accumulator[r_idx, t_idx] += 1

    # Normalize
    acc_norm = accumulator / accumulator.max()

    # Threshold
    strong_params = np.argwhere(acc_norm >= acc_threshold_ratio)

    output_img = img.copy()

    # Draw detected lines
    for r_idx, t_idx in strong_params:
        rho = rhos[r_idx]
        theta = thetas[t_idx]

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(output_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return output_img, acc_norm
````
