````python
# Your PART A codes follow here. You can create more cells under this. Write clean and commented code.
# what * operation does with python numpy arrays is to multiply the elements of the arrays
# that appears at somewaht same position in both arrays
# np.pad function is used to pad an array in different ways
# arguments are as follows np.pad(array,pad_values,mode_of_padding)
def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 0)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value


def convolve2d(img, kernel):
    kh, kw = kernel.shape

    padded = np.pad(img,((kh//2, kh//2), (kw//2, kw//2)),pad_with)

    H, W = img.shape
    output = np.zeros((H, W), dtype=np.float32)

    for i in range(H):
        for j in range(W):
            region = padded[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)

    return output


def convolve_rgb(img, kernel):
    channels = []
    for c in range(3):
        channels.append(convolve2d(img[:, :, c], kernel))

    output = np.stack(channels, axis=2)
    return np.clip(output, 0, 255).astype(np.uint8)


def average_blur_gray(img, ksize):
    kernel = np.full((ksize, ksize), 1.0/(ksize*ksize), dtype=np.float32)
    return convolve2d(img, kernel)


def average_blur_rgb(img, ksize):
    kernel = np.full((ksize, ksize), 1.0/(ksize*ksize), dtype=np.float32)
    return convolve_rgb(img, kernel)


def gaussian_kernel(ksize, sigma):
    ax = np.arange(-(ksize // 2), ksize // 2 + 1)
    xx, yy = np.meshgrid(ax, ax)

    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    kernel /= np.sum(kernel)

    return kernel.astype(np.float32)

gaussian_blurred_image = convolve_rgb(img_part5,gaussian_kernel(10,10))
avg_blurred_image = average_blur_rgb(img_part5,10)
avg_blurred_gray_image = average_blur_gray(cv2.cvtColor(img_part5,cv2.COLOR_RGB2GRAY),10)
gaussian_blurred_gray_image = convolve2d(cv2.cvtColor(img_part5,cv2.COLOR_RGB2GRAY),gaussian_kernel(10,10)).astype(np.uint8)

plt.figure(figsize=(12,5))

plt.subplot(2,3,1)
plt.imshow(img_part5)
plt.title("original image")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(gaussian_blurred_image)
plt.title("gaussian blurred image")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(avg_blurred_image)
plt.title("avg blurred image")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(avg_blurred_gray_image,cmap="gray")
plt.title("avg blurred gray image")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(gaussian_blurred_gray_image,cmap="gray")
plt.title("gaussian blurred gray image")
plt.axis("off")

plt.show()

````

#### Task - Sobel Edge Detection (NumPy Only)
Implement Sobel X and Sobel Y filters manually using convolution. Apply on the grayscale image.

Compute:
1. Gx
2. Gy
3. Gradient magnitude

Normalize to 0255 and save the edge map. Threshold if you wish to, if you do then make the thresholding part clear using code comments.

````python
# Your PART B codes follow here. You can create more cells under this. Write clean and commented code.
SOBEL_X = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

SOBEL_Y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)


def sobel_gray(img_gray):
    gx = convolve2d(img_gray, SOBEL_X)
    gy = convolve2d(img_gray, SOBEL_Y)

    magnitude = np.sqrt(gx**2 + gy**2)

    return gx, gy, magnitude

def normalize_to_uint8(img):
    img = np.abs(img)
    img = img / img.max() * 255
    return np.clip(img, 0, 255).astype(np.uint8)


def sobel_rgb(img):
    gx_channels = []
    gy_channels = []
    mag_channels = []

    for c in range(3):
        gx, gy, mag = sobel_gray(img[:, :, c])
        gx_channels.append(gx)
        gy_channels.append(gy)
        mag_channels.append(mag)

    gx = np.stack(gx_channels, axis=2)
    gy = np.stack(gy_channels, axis=2)
    mag = np.stack(mag_channels, axis=2)

    return (
        normalize_to_uint8(gx),
        normalize_to_uint8(gy),
        normalize_to_uint8(mag)
    )

gray = cv2.cvtColor(img_part5, cv2.COLOR_RGB2GRAY)

gx, gy, mag = sobel_gray(gray)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(normalize_to_uint8(gx), cmap='gray')
plt.title("Sobel X")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(normalize_to_uint8(gy), cmap='gray')
plt.title("Sobel Y")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(normalize_to_uint8(mag), cmap='gray')
plt.title("Gradient Magnitude")
plt.axis("off")

plt.show()
````

````python
def unsharp_mask_gray(img, ksize=5, sigma=1.0, alpha=1.0):
    blur = convolve2d(img, gaussian_kernel(ksize, sigma))
    mask = img - blur
    sharp = img + alpha * mask
    return np.clip(sharp, 0, 255).astype(np.uint8)

def unsharp_mask_rgb(img, ksize=5, sigma=1.0, alpha=1.0):
    channels = []
    for c in range(3):
        sharp = unsharp_mask_gray(
            img[:, :, c],
            ksize=ksize,
            sigma=sigma,
            alpha=alpha
        )
        channels.append(sharp)
    return np.stack(channels, axis=2)
sharp_a1 = unsharp_mask_gray(gray, alpha=1.0)
sharp_a2 = unsharp_mask_gray(gray, alpha=20.0)
blur_a  = unsharp_mask_gray(gray, alpha=-5)

plt.figure(figsize = (12,12))

plt.subplot(1,3,1)
plt.imshow(sharp_a1, cmap='gray')
plt.title("sharped image with masks")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(sharp_a2, cmap='gray')
plt.title("more sharped image with masks")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(blur_a, cmap='gray')
plt.title("blured image with masks")
plt.axis("off")

plt.show()
````

Compute the magnitude spectrum of the original image, gaussian blurred image and the unsharp masking image. Now, play with the alpha value in sharpening and gaussian kernel and notice how the magntude spectrum changes.
Then, in code comments, explain your observations of dark and bright areas in the magnitude spectrum. In nutshell, verify how convolution is actually filtering frequencies depending on the kernel.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
def magnitude_spectrum(img):
    img_f = img.astype(np.float32)

    # 2D FFT
    F = np.fft.fft2(img_f)

    # Shift zero-frequency component to center
    F_shift = np.fft.fftshift(F)

    # Magnitude spectrum
    magnitude = np.log(1 + np.abs(F_shift))

    return magnitude

````

````python
blur = convolve2d(gray, gaussian_kernel(10, sigma=20))

alpha = 1.0
mask = gray - blur
sharp = gray + alpha * mask

spec_orig  = magnitude_spectrum(gray)
spec_blur  = magnitude_spectrum(blur)
spec_sharp = magnitude_spectrum(sharp)

plt.figure(figsize=(15,5))

plt.subplot(2,3,1)
plt.imshow(spec_orig, cmap='gray')
plt.title("Original image spectrum")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(spec_blur, cmap='gray')
plt.title("Gaussian blurred spectrum")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(spec_sharp, cmap='gray')
plt.title("Unsharp masked spectrum")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(gray, cmap='gray')
plt.title("Original image")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(blur, cmap='gray')
plt.title("Gaussian blurred image")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(sharp, cmap='gray')
plt.title("Unsharp masked image")
plt.axis("off")

plt.show()
````
