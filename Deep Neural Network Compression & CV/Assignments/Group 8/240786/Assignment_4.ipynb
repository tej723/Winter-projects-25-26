To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def r_map(Ix, Iy, window_size=3, k=0.04):
    h, w = Ix.shape
    offset = window_size // 2

    r_map = np.zeros((h, w))

    for i in range(offset, h - offset):
        for j in range(offset, w - offset):

            # Window extraction
            Ix_win = Ix[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy_win = Iy[i-offset:i+offset+1, j-offset:j+offset+1]

            # Second moments
            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            # Harris R value
            r = (a * c - b ** 2) - k * (a + c) ** 2

            r_map[i, j] = r

    return r_map

r_flat   = r_map(Ix_flat, Iy_flat)
r_edge   = r_map(Ix_edge, Iy_edge)
r_corner = r_map(Ix_corner, Iy_corner)

def show_r_map(r, title):
    r_disp = r.copy()
    r_disp[r_disp < 0] = 0
    r_disp = r_disp / r_disp.max()

    plt.figure(figsize=(4,4))
    plt.imshow(r_disp, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
show_r_map(r_flat, "r-map (Flat Image)")
show_r_map(r_edge, "r-map (Edge Image)")
show_r_map(r_corner, "r-map (Corner Image)")

````

````python
from google.colab import files
import cv2
import matplotlib.pyplot as plt

uploaded = files.upload()

img_name = list(uploaded.keys())[0]

img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)

# Display image
plt.figure(figsize=(5,5))
plt.imshow(img, cmap='gray')
plt.title("Input Image")
plt.axis('off')
plt.show()
````

````python
def canny_edges(img):
    edges = cv2.Canny(img, 100, 200)
    return edges


def hough_slope_intercept(img, m_bins=180, c_bins=300):
    edges = canny_edges(img)
    h, w = edges.shape

    # parameter space limits
    m_min, m_max = -1, 1
    c_min, c_max = -h, h

    m_vals = np.linspace(m_min, m_max, m_bins)
    c_vals = np.linspace(c_min, c_max, c_bins)

    accumulator = np.zeros((m_bins, c_bins))

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for i, m in enumerate(m_vals):
            c = y - m * x
            if c_min <= c <= c_max:
                j = int((c - c_min) / (c_max - c_min) * (c_bins - 1))
                accumulator[i, j] += 1

    return accumulator, m_vals, c_vals


def extract_lines_mc(accumulator, m_vals, c_vals, threshold_ratio=0.9):
    acc_norm = accumulator / accumulator.max()
    indices = np.where(acc_norm > threshold_ratio)

    lines = []
    for i, j in zip(indices[0], indices[1]):
        lines.append((m_vals[i], c_vals[j]))

    return lines


def draw_lines_mc(img, lines):
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    h, w = img.shape

    for m, c in lines:
        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img_color, (x1, y1), (x2, y2), (0, 0, 255), 1)

    return img_color


# slope intercept hough
acc_mc, m_vals, c_vals = hough_slope_intercept(img)
lines_mc = extract_lines_mc(acc_mc, m_vals, c_vals, threshold_ratio=0.9)
img_lines_mc = draw_lines_mc(img, lines_mc)

plt.figure(figsize=(6,6))
plt.imshow(img_lines_mc)
plt.title("Hough Lines (Slope-Intercept)")
plt.axis("off")
plt.show()


# rough theta
def hough_rho_theta(img, rho_bins=300, theta_bins=180):
    edges = canny_edges(img)
    h, w = img.shape
    diag = int(np.sqrt(h*h + w*w))

    rhos = np.linspace(-diag, diag, rho_bins)
    thetas = np.linspace(0, np.pi, theta_bins)

    accumulator = np.zeros((rho_bins, theta_bins))

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for j, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            i = int((rho + diag) / (2 * diag) * (rho_bins - 1))
            accumulator[i, j] += 1

    return accumulator, rhos, thetas


def extract_lines_rt(accumulator, rhos, thetas, threshold_ratio=0.9):
    acc_norm = accumulator / accumulator.max()
    indices = np.where(acc_norm > threshold_ratio)

    lines = []
    for i, j in zip(indices[0], indices[1]):
        lines.append((rhos[i], thetas[j]))

    return lines


def draw_lines_rt(img, lines):
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    for rho, theta in lines:
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a*rho, b*rho

        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(img_color, (x1, y1), (x2, y2), (255, 0, 0), 1)

    return img_color


# rho thera hough
acc_rt, rhos, thetas = hough_rho_theta(img)
lines_rt = extract_lines_rt(acc_rt, rhos, thetas, threshold_ratio=0.9)
img_lines_rt = draw_lines_rt(img, lines_rt)

plt.figure(figsize=(6,6))
plt.imshow(img_lines_rt)
plt.title("Hough Lines (Rho-Theta)")
plt.axis("off")
plt.show()




````
