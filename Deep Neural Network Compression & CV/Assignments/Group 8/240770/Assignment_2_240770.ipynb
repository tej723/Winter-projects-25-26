#### RGB to HSV ( only Numpy ofc )
Write a function <code>rgb_to_hsv</code> that takes a 0-255 ( not normalised ) RGB image and returns an HSV image ( also not normalised ).

**Note :** Research for the formulas, maybe ask some AI to explain you the algo and formulas ( NOT to give you the code, seriously youre not gonna learn anything if you do that )

The output should be in a format that is compatible with OpenCV cvtColor. Read the documentation to find out any Hue scaling etc.
Finally, use your function and test it by using cv2.cvtColor to convert your function output into RGB again and display it. Your image will appear the same after reconversion into RGB.

````python
#B.
def rgb_to_hsv(i):
  r = i[:,:,0]/255.0
  g = i[:,:,1]/255.0
  b = i[:,:,2]/255.0
  cmin = np.minimum(r,np.minimum(g,b))
  cmax = np.maximum(np.maximum(r, g), b)
  v = cmax
  h = np.zeros_like(cmax)
  s = np.zeros_like(cmax)
  delta = cmax - cmin
  #saturation = delta/cmax when cmax !=0 else zero
  mask = cmax != 0
  s[mask] = delta[mask] / cmax[mask]
  #hue

  mask_delta = delta != 0
  #r is max h = g-b/delta%6
  idx = (cmax==r)&mask_delta
  h[idx] = ((g[idx] - b[idx]) / delta[idx]) % 6
  #green is max h = b-r/delta +2
  idx = (cmax == g)&mask_delta
  h[idx] = (b[idx] - r[idx]) / delta[idx] + 2
  #blue is max h = r - g/delta + 4
  idx = (cmax == b)&mask_delta
  h[idx] = (r[idx] - g[idx]) / delta[idx] + 4

  h *=60 #converting h to degrees
  H = (h / 2).astype(np.uint8)
  S  = (s * 255).astype(np.uint8)     #scaled into 8 bit so for open cv scaling
  V = (v * 255).astype(np.uint8)
  hsv =np.dstack([H,S,V])
  return hsv


````

````python
#test
img_hsv = rgb_to_hsv(img)
img_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

plt.figure(figsize = (12,4))

plt.subplot(1,2,1)
plt.title("RGB that was converted from HSV")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Orignal RGB")
plt.imshow(img)
plt.axis("off")
plt.show()
````

````python
#filter 2
def dawn(img):

    contrast = 1.1
    sat_boost = 1.15
    darken = 0.8
    r = img[..., 0]
    g = img[..., 1]
    b = img[..., 2]

   #saturation boosting only red and green channels
    avg = (r+ g + b) / 3.0

    r= (avg +(r - avg) * sat_boost)*1.1    #here I am giving more weight to red so that I can get more orange in picture
    g = avg + (g - avg) * sat_boost
    out = np.stack([r, g, 0.9*b], axis=-1)   #reducing blue light to enhance warmer colors

  ##darkened the entire image and slightly increased the contrast the make colors pop more
    out *= darken
    out = (out - 127.5)*contrast + 127.5
    return clamp_uint8(out)
dawn= dawn(img)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("original image")
plt.axis("off")
plt.subplot(1,2,2)
plt.title("Dawn")
plt.imshow(dawn)
plt.axis("off")
plt.show()
##this filter suits this image specifically because of the red car and the golden grass which get enhanced due to the red and green saturation boosts
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
#D
def white_patch(img):

    img = img_rgb.astype(np.float32)

    #max value in each channel
    maxr = np.max(img[:,:, 0])
    maxg = np.max(img[:,:, 1])
    maxb = np.max(img[:,:, 2])



    # applying scaling to each channel considering max value to be brightest
    i = img.copy()
    i[:,:, 0] *= 255/maxr
    i[:,:, 1] *= 255/maxg
    i[:,:, 2] *= 255/maxb


    return clamp_uint8(i)

white_balance=white_patch(img)

plt.figure(figsize=(10, 4))
plt.subplot(1,2,1)
plt.imshow(img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(white_balance)
plt.title("while balanced image")
plt.axis("off")

plt.show()
##the image is inherantly very warm so we cant properly see the result only subtle difference is observable
````
