To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def corner_detection(image, title_prefix=""):
    # Convert image to grayscale
    gray_image = np.dot(image[...,:3], [0.2989, 0.5870, 0.1140])
    # Normalize to 0-255 range and convert to uint8 for consistent processing and display
    gray_image = (gray_image - np.min(gray_image)) / (np.max(gray_image) - np.min(gray_image)) * 255
    gray_image = gray_image.astype(np.uint8)

    # Define Sobel kernels
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])

    sobel_y = np.array([[-1, -2, -1],
                        [0, 0, 0],
                        [1, 2, 1]])


    Ix = np.zeros_like(gray_image, dtype=float)
    Iy = np.zeros_like(gray_image, dtype=float)

    img_rows, img_cols = gray_image.shape
    kernel_size = sobel_x.shape[0]
    pad = kernel_size // 2

    # Pad the image to handle borders during convolution
    padded_image = np.pad(gray_image, pad, mode='edge')

    for i in range(img_rows):
        for j in range(img_cols):
            window = padded_image[i:i+kernel_size, j:j+kernel_size]
            Ix[i, j] = np.sum(window * sobel_x)
            Iy[i, j] = np.sum(window * sobel_y)

    # Plot Ix and Iy
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(gray_image, cmap='gray')
    plt.title(f'{title_prefix} Grayscale Image')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(Ix, cmap='gray')
    plt.title(f'{title_prefix} Ix (Gx)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(Iy, cmap='gray')
    plt.title(f'{title_prefix} Iy (Gy)')
    plt.axis('off')
    plt.show()

    # Create a scatter plot for Iy vs Ix
    plt.figure(figsize=(5, 5))
    # Flatten the 2D arrays to 1D for the scatter plot
    # Optionally, apply a threshold to reduce the number of points for better visualization if the image is too large
    plt.scatter(Ix.flatten(), Iy.flatten(), s=1, alpha=0.5)
    plt.xlabel('Ix (Gx)')
    plt.ylabel('Iy (Gy)')
    plt.title(f'{title_prefix} Iy vs Ix Scatter Plot')
    plt.grid(True)
    plt.show()

    return Ix, Iy
````

````python
def generate_r_map(img, window_size=3, k=0.04):
    # Load and convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY).astype(np.float32)

    # 1. Compute Gradients Ix and Iy
    Ix = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    # 2. Compute Products of Gradients
    Ixx = Ix**2
    Ixy = Ix*Iy
    Iyy = Iy**2

    # 3. Sum of products over a window (using a box filter or Gaussian)
    # This represents the summing part of a, b, and c formulas
    a = cv2.boxFilter(Ixx, -1, (window_size, window_size), normalize=False)
    b = cv2.boxFilter(Ixy, -1, (window_size, window_size), normalize=False)
    c = cv2.boxFilter(Iyy, -1, (window_size, window_size), normalize=False)

    # 4. Calculate Determinant and Trace for R
    # det(M) = ac - b^2, trace(M) = a + c
    det_M = (a * c) - (b**2)
    trace_M = a + c

    # 5. Calculate R-map
    R = det_M - k * (trace_M**2)

    return R

generate_r_map(ball)
# Process your 3 images
image_files = [flat, edge, ball]
plt.figure(figsize=(15, 5))

for i, img_path in enumerate(image_files):
    r_map = generate_r_map(img_path)

    # Normalize for display so corners appear white
    r_map_display = cv2.normalize(r_map, None, 0, 255, cv2.NORM_MINMAX)

    plt.subplot(1, 3, i+1)
    plt.imshow(r_map_display, cmap='gray')
    if(i==1):
        plt.title("Edge Image")

    elif(i==2):
        plt.title("Ball Image")

    else:
      plt.title("Flat Image")

    plt.axis('off')

plt.show()

````

````python

def hough_slope_intercept(img_input, title_prefix="", m_limit=5, c_limit_factor=1, m_resolution=200, c_resolution=200, threshold_percentage=0.85):
    # Make a copy of the image to draw lines on
    img = img_input.copy()

    # Convert to grayscale and find Canny edges
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)

    h, w = edges.shape

    # Define ranges for slope (m) and intercept (c)
    m_range = np.linspace(-m_limit, m_limit, m_resolution)
    c_range = np.linspace(-h * c_limit_factor, h * c_limit_factor, c_resolution)

    # Initialize accumulator
    accumulator = np.zeros((len(m_range), len(c_range)))

    # Loop through edge pixels to vote
    edge_pixels = np.argwhere(edges > 0)
    for y, x in edge_pixels:
        for m_idx, m in enumerate(m_range):
            c = y - m * x
            # Check if intercept is within defined limits
            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                accumulator[m_idx, c_idx] += 1

    # Normalize accumulator for display and thresholding
    acc_norm = cv2.normalize(accumulator, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)
    threshold = threshold_percentage * np.max(acc_norm)


    line_params = np.argwhere(acc_norm > threshold)

    # Draw the detected lines
    for m_idx, c_idx in line_params:
        m = m_range[m_idx]
        c = c_range[c_idx]

        # Lines can be vertical or horizontal, handle edge cases for drawing
        if np.isclose(m, 0, atol=1e-5): # Horizontal line
            y1, y2 = int(c), int(c)
            x1, x2 = 0, w
        elif np.isinf(m): # Vertical line (approx. infinite slope)
            x1, x2 = int(-c/m), int(-c/m) # Since c = y - mx => x = (y-c)/m
            y1, y2 = 0, h
        else:
            # Calculate y-coordinates for x=0 and x=width
            y1 = int(m * 0 + c)
            y2 = int(m * w + c)
            x1, x2 = 0, w

            # Filter points that are outside image bounds
            # Clip coordinates to image boundaries for drawing
            if (y1 < 0 and y2 < 0) or (y1 > h and y2 > h):
                continue
            if (x1 < 0 and x2 < 0) or (x1 > w and x2 > w):
                continue

        cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 2) # Green lines

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(img_input, cv2.COLOR_BGR2RGB))
    plt.title(f'{title_prefix} Original Image')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f'{title_prefix} Detected Lines (Slope-Intercept)')
    plt.axis('off')
    plt.show()

    return acc_norm

# --- Execution for Slope-Intercept Method ---
print("Processing for Hough Slope-Intercept Method:")
# Note: cv2 functions expect BGR order, so convert numpy arrays accordingly if they were loaded as RGB
ball_bgr = cv2.cvtColor(ball, cv2.COLOR_RGB2BGR)
acc_si = hough_slope_intercept(ball_bgr, "Ball Image", m_limit=2, c_limit_factor=0.5, m_resolution=100, c_resolution=100, threshold_percentage=0.9)
````

````python
def hough_rho_theta(img):

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)

    h, w = edges.shape
    diag_len = int(np.sqrt(h**2 + w**2))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    thetas = np.deg2rad(np.arange(0, 180))

    accumulator = np.zeros((len(rhos), len(thetas)))

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    edge_pixels = np.argwhere(edges > 0)
    for y, x in edge_pixels:
        for t_idx in range(len(thetas)):
            # Calculate rho
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            rho_idx = np.argmin(np.abs(rhos - rho))
            accumulator[rho_idx, t_idx] += 1

    # Thresholding
    acc_norm = cv2.normalize(accumulator, None, 0, 255, cv2.NORM_MINMAX)
    threshold = 0.5 * np.max(acc_norm) # Typical threshold for rho-theta

    line_indices = np.argwhere(acc_norm > threshold)
    for r_idx, t_idx in line_indices:
        rho = rhos[r_idx]
        theta = thetas[t_idx]
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a*rho, b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 2)

    return img, acc_norm

# Execution
result_rt, acc_rt = hough_rho_theta(ball)
plt.imshow(cv2.cvtColor(result_rt, cv2.COLOR_BGR2RGB))
plt.title("Rho-Theta Method")
plt.show()
````
