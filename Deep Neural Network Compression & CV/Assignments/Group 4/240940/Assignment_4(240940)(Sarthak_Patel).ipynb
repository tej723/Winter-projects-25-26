To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# defining the Sobel matrices
def sobel_gradients(image):
  Gx=np.array([[-1,0,1],[-2,0,2],[-1,0,1]],dtype=np.float32)
  Gy=np.array([[-1,-2,-1],[0,0,0],[1,2,1]],dtype=np.float32)
  Ix=ndimage.convolve(image,Gx)
  Iy=ndimage.convolve(image,Gy)
  return Ix,Iy
````

````python
# RGB to Grayscale function
def rgb_to_grayscale(img_rgb):
  R = img_rgb[:, :, 0].astype(np.float32)
  G = img_rgb[:, :, 1].astype(np.float32)
  B = img_rgb[:, :, 2].astype(np.float32)
  gray_numpy = (0.299 * R + 0.587 * G + 0.114 * B).astype(np.uint8)
  return gray_numpy
````

````python
# Function calculating the R_MAP along with the related values like a,b,c and
def calculate_r_map(Ix, Iy, window_size=5, k=0.04):
    height, width = Ix.shape
    R_map = np.zeros((height, width), dtype=np.float32)

    # Products of gradients
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    offset = window_size // 2

    for y in range(offset, height - offset):
        for x in range(offset, width - offset):

            window_Ix2 = Ix2[y - offset:y + offset + 1, x - offset:x + offset + 1]
            window_Iy2 = Iy2[y - offset:y + offset + 1, x - offset:x + offset + 1]
            window_Ixy = Ixy[y - offset:y + offset + 1, x - offset:x + offset + 1]

            # Summing over the window to get second moments
            a = np.sum(window_Ix2)
            b = np.sum(window_Iy2)
            c = np.sum(window_Ixy)

            # Constructing the M matrix
            M = np.array([[a, c],
                          [c, b]])

            # Calculating R value
            det_M = np.linalg.det(M)
            trace_M = np.trace(M)
            R = det_M - k * (trace_M**2)

            R_map[y, x] = R

    return R_map

def analyze_r_map(image_name, Ix, Iy):
    R_map = calculate_r_map(Ix, Iy)

    plt.figure(figsize=(8, 8))
    plt.imshow(R_map, cmap='gray')
    plt.title(f'{image_name} - R Map for Corner Detection')
    plt.axis('off')
    plt.colorbar(label='R Value')
    plt.show()


    threshold_value = R_map.max() * 0.01
    corners_image = (R_map > threshold_value).astype(np.uint8) * 255

    plt.figure(figsize=(8, 8))
    plt.imshow(corners_image, cmap='gray')
    plt.title(f'{image_name} - Detected Corners (Thresholded R Map)')
    plt.axis('off')
    plt.show()

# Analyzeing thw R Maps for the three images
analyze_r_map('Flat Image', Ix_flat, Iy_flat)
analyze_r_map('Edge Imsge', Ix_edge, Iy_edge)
analyze_r_map('Corner Image', Ix_corner, Iy_corner)
````

````python
# Slope intercept form
import time

def hough_slope_intercept(image):
    # Canny edge detection and with standard thresholds
    edges = cv.Canny(image, 50, 150)

    # Getting coordinates of all edge pixels
    y_idxs, x_idxs = np.nonzero(edges)

    height, width = edges.shape

    # Defining the Parameter Limits
    # Slope : We limit this to -10 to 10 to avoid infinite slope issues
    # Resolution: 200 steps
    m_vals = np.linspace(-10, 10, 200)

    # c = y - mx.
    c_max = int(height + 2 * width)
    c_min = int(-2 * width)
    num_c_bins = 500 # Resolution for intercept
    c_vals = np.linspace(c_min, c_max, num_c_bins)

    # Creating Accumulator
    accumulator = np.zeros((num_c_bins, len(m_vals)), dtype=np.uint64)

    start_time = time.time()

    # Voting Procedure
    # Looping through every edge pixel
    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        # Looping through every slope in our range
        for m_idx, m in enumerate(m_vals):

            c = y - (m * x)


            # Formula: (value - min) / step_size
            c_step = (c_max - c_min) / num_c_bins
            c_bin = int((c - c_min) / c_step)

            if 0 <= c_bin < num_c_bins:
                accumulator[c_bin, m_idx] += 1

    print("Voting Time:", {time.time() - start_time})

    # Normalization and Thresholding
    max_vote = np.max(accumulator)
    normalized_acc = accumulator / max_vote

    # Threshold: 0.6
    threshold = 0.6

    c_bins_selected, m_bins_selected = np.where(normalized_acc > threshold)

    print("Lines found:" ,{len(c_bins_selected)})

    # Drawing the lienws
    output_img = cv.cvtColor(image, cv.COLOR_GRAY2BGR)

    for k in range(len(c_bins_selected)):
        # Retrieving m and c from the bin indices
        m_idx = m_bins_selected[k]
        c_idx = c_bins_selected[k]

        m = m_vals[m_idx]
        c = c_vals[c_idx]

        x1 = 0
        y1 = int(c)
        x2 = width
        y2 = int(m * width + c)

        cv.line(output_img, (x1, y1), (x2, y2), (0, 0, 255), 2)

    return edges, output_img, accumulator

si_edges, si_result, si_acc = hough_slope_intercept(rgb_to_grayscale(img_rgb))


plt.figure(figsize=(15, 6))
plt.subplot(131), plt.imshow(si_edges, cmap='gray'), plt.title("Canny Edges")
plt.subplot(132), plt.imshow(si_acc, cmap='jet'), plt.title("Accumulator (M vs C)")
plt.subplot(133), plt.imshow(cv.cvtColor(si_result, cv.COLOR_BGR2RGB)), plt.title("Detected Lines (Slope-Int)")
plt.show()
````

````python
# Rho Theta method.
import math

def hough_rho_theta(image):
    #Canny Edge Detection
    edges = cv.Canny(image, 50, 150)
    y_idxs, x_idxs = np.nonzero(edges)

    height, width = edges.shape
    diag_len = int(np.ceil(np.sqrt(width**2 + height**2)))

    # Define parameter limits (Rho and Theta)
    # Theta: 0 to 180 degrees (in radians)
    thetas = np.deg2rad(np.arange(-90, 90, 1)) # 1 degree resolution

    # Rho: -Diagonal to +Diagonal
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)

    cos_thetas = np.cos(thetas)
    sin_thetas = np.sin(thetas)
    num_thetas = len(thetas)
    num_rhos = len(rhos)

    # Accumulator: Rows=Rho, Cols=Theta
    accumulator = np.zeros((num_rhos, num_thetas), dtype=np.uint64)

    start_time = time.time()

    # Voting Procedure
    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        # Iterating through all Theta angles
        for t_idx in range(num_thetas):
            # rho = x*cos(theta) + y*sin(theta)
            rho = round(x * cos_thetas[t_idx] + y * sin_thetas[t_idx])

            # Mapping Rho to bin index
            rho_idx = int(rho + diag_len)

            # checking the limits.
            if 0 <= rho_idx < num_rhos:
                accumulator[rho_idx, t_idx] += 1

    print("time taken ",{time.time() - start_time} )

    # Normalization and Thresholding
    max_vote = np.max(accumulator)
    normalized_acc = accumulator / max_vote

    # Threshold=0.6
    threshold = 0.6
    rho_bins_sel, theta_bins_sel = np.where(normalized_acc > threshold)

    print("Lines found:", {len(rho_bins_sel)})

    # Draw Lines
    output_img = cv.cvtColor(image, cv.COLOR_GRAY2BGR)

    for k in range(len(rho_bins_sel)):
        r_idx = rho_bins_sel[k]
        t_idx = theta_bins_sel[k]

        rho = rhos[r_idx]
        theta = thetas[t_idx]

        # x0 = rho*cos, y0 = rho*sin
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        # Extrapolate points to edge of image (1000 is arbitrary large number)
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))

        cv.line(output_img, pt1, pt2, (0, 255, 0), 2)

    return edges, output_img, accumulator

# Execute
rt_edges, rt_result, rt_acc = hough_rho_theta(rgb_to_grayscale(img_rgb))

# Display
plt.figure(figsize=(15, 6))
plt.subplot(131), plt.imshow(rt_edges, cmap='gray'), plt.title("Canny Edges")
plt.subplot(132), plt.imshow(rt_acc, cmap='jet'), plt.title("Accumulator (Rho vs Theta)")
plt.subplot(133), plt.imshow(cv.cvtColor(rt_result, cv.COLOR_BGR2RGB)), plt.title("Detected Lines (Rho-Theta)")
plt.show()
````

````python
# creating the ball stencil
def create_ball_stencil(image,x1,y1,x2,y2):
  area=image[y1:y2,x1:x2] # making the area of our concern containing the ball
  gray=rgb_to_grayscale(area) # Converting to grayscale
  edges=detect_canny_edges(gray,50,150) # edge detection
  contours, _ = cv.findContours(edges, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE) # This retrieves the outermost contour
  max_contour = max(contours, key=cv.contourArea)  # This finds the biggest contour from the list based on the area. Hence the ball is the largest shape in the cropped box.
  mask = np.zeros_like(gray) # creates a black mask of all 0s of same width and height as the gray image.
  cv.drawContours(mask, [max_contour], -1, 255, thickness=cv.FILLED)
  b,g,r=cv.split(area) # splits the original image into r g and b channels
  rgba=cv.merge((b,g,r,mask))
  return rgba
````

````python
# Function overlaying the created stencil onto the ball2 image
def overlay_stencil(background,stencil,center_x,center_y):
  result=background.copy()
  background_float=background.astype(np.float32)
  stencil_float=stencil.astype(np.float32)

  bg_h, bg_w = background_float.shape[:2]
  s_h, s_w = stencil_float.shape[:2]

  # Calculating top-left corner of the stencil on the background
  x_offset = center_x - s_w // 2
  y_offset = center_y - s_h // 2

  # Determine the region of overlap on the background
  d_x1 = max(0, x_offset)
  d_y1 = max(0, y_offset)
  d_x2 = min(bg_w, x_offset + s_w)
  d_y2 = min(bg_h, y_offset + s_h)

  # Calculate corresponding coordinates on the stencil
  s_x1 = max(0, -x_offset)
  s_y1 = max(0, -y_offset)
  s_x2 = s_w - max(0, (x_offset + s_w) - bg_w)
  s_y2 = s_h - max(0, (y_offset + s_h) - bg_h)

  # Ensure the clipped regions have valid dimensions
  if d_x2 <= d_x1 or d_y2 <= d_y1 or s_x2 <= s_x1 or s_y2 <= s_y1:
      print("No overlap or invalid overlap region. Skipping overlay.")
      return result.astype(np.uint8)

  # Getting the actual regions for blending
  background_roi = background_float[d_y1:d_y2, d_x1:d_x2]
  stencil_roi = stencil_float[s_y1:s_y2, s_x1:s_x2]

  # Extracting BGR and Alpha channels from the clipped stencil
  stencil_bgr = stencil_roi[:, :, :3]
  stencil_alpha = stencil_roi[:, :, 3] / 255.0

  alpha_mask = cv.merge([stencil_alpha, stencil_alpha, stencil_alpha])

  # Blending
  blended_roi = (stencil_bgr * alpha_mask) + (background_roi * (1 - alpha_mask))

  result[d_y1:d_y2, d_x1:d_x2] = blended_roi.astype(np.uint8)
  return result.astype(np.uint8)
````
