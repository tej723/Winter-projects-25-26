To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np
def convolve(img,kernel):
  img=img.astype(np.float64)
  kh,kw=kernel.shape
  pad_h=kh//2
  pad_w=kw//2
  padded_img=np.pad(img,((pad_h,pad_h),(pad_w,pad_w)),mode='reflect')
  H,W=img.shape
  output=np.zeros((H,W))
  for i in range(H):
      for j in range(W):
          region=padded_img[i:i+kh,j:j+kw]
          output[i,j]=np.sum(region*kernel)
  return output
````

````python
sobel_x=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
sobel_y=np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
def to_grayscale(img):
    if img.ndim==3:
        R=img[:,:,0]
        G=img[:,:,1]
        B=img[:,:,2]
        gray_img=0.299*R+0.587*G+0.114*B
        return gray_img
    return img
def corner_detection(img,title="",threshold=None):
  gray_img=to_grayscale(img).astype(np.float64)
  Gx=convolve(gray_img,sobel_x)
  Gy=convolve(gray_img,sobel_y)
  Gx_flat=Gx.flatten()
  Gy_flat=Gy.flatten()
  if threshold is not None:
        mask=(Gx_flat**2+Gy_flat**2)>threshold
        Gx_flat=Gx_flat[mask]
        Gy_flat=Gy_flat[mask]
  plt.figure(figsize=(14,4))
  plt.subplot(1,4,1)
  plt.imshow(gray_img,cmap='gray')
  plt.title(f"{title} grayscale")
  plt.axis("off")
  plt.subplot(1,4,2)
  plt.title("Sobel Gx")
  plt.imshow(np.abs(Gx),cmap='gray')
  plt.axis("off")
  plt.subplot(1,4,3)
  plt.title("Sobel Gy")
  plt.imshow(np.abs(Gy),cmap='gray')
  plt.subplot(1,4,4)
  plt.scatter(Gx_flat,Gy_flat,s=1)
  plt.xlabel("Gx")
  plt.ylabel("Gy")
  plt.title("Gy vs Gx")
  plt.show()
  return Gx_flat,Gy_flat
Gx_flat,Gy_flat=corner_detection(flat,"Flat image",threshold=100)
Gx_edge,Gy_edge=corner_detection(edge,"Edge image",threshold=100)
Gx_corner,Gy_corner=corner_detection(corner,"Corner image",threshold=100)
````

````python
def R_map(img,window_size=3,k=0.04):
    gray_img=to_grayscale(img).astype(np.float64)
    Gx=convolve(gray_img,sobel_x)
    Gy=convolve(gray_img,sobel_y)
    Gx2=Gx**2
    Gy2=Gy**2
    Gxy=Gx*Gy
    H,W=gray_img.shape
    offset=window_size//2
    R=np.zeros((H,W))
    for i in range(offset,H-offset):
        for j in range(offset,W-offset):
            a=np.sum(Gx2[i-offset:i+offset+1,j-offset:j+offset+1])
            b=np.sum(Gxy[i-offset:i+offset+1,j-offset:j+offset+1])
            c=np.sum(Gy2[i-offset:i+offset+1,j-offset:j+offset+1])
            det=a*c-b*b  #Harris response
            trace=a+c
            R[i,j]=det-k*(trace**2)
    return R
def show_R_map(R,title="R map"):
    plt.figure(figsize=(5,5))
    plt.imshow(R,cmap='gray')
    plt.title(title)
    plt.axis("off")
    plt.show()
R_flat=R_map(flat)
R_edge=R_map(edge)
R_corner=R_map(corner)
show_R_map(R_flat,"Flat image R map")
show_R_map(R_edge,"Edge image R map")
show_R_map(R_corner,"Corner image R map")
````

````python
#slope-intercept method
def get_edges(img,low=100,high=200):
    gray_img=to_grayscale(img)
    gray_img=gray_img.astype(np.uint8)
    edges=cv2.Canny(gray_img,low,high)
    return edges #canny edge map
edges=get_edges(img)
plt.imshow(edges,cmap='gray')
plt.title("Canny edge map")
plt.axis("off")
plt.show()
````

````python
def hough_slope_intercept(img,m_range=(-0.7,0.7),num_m=200,c_range=(-300,300),num_c=400,threshold_percent=99.8):
    edges=get_edges(img)
    H,W=edges.shape
    m_vals=np.linspace(m_range[0],m_range[1],num_m)
    c_vals=np.linspace(c_range[0],c_range[1],num_c)
    acc=np.zeros((num_m,num_c),dtype=np.int32)
    ys,xs=np.where(edges>0)  #voting
    for x, y in zip(xs,ys):
        for i, m in enumerate(m_vals):
            c=y-m*x  #line for each pixel
            j=int((c-c_range[0])/(c_range[1]-c_range[0])*num_c)
            if 0<=j<num_c:
                acc[i,j]+=1
    thresh=np.percentile(acc,threshold_percent)  #threshold accumulator
    lines=np.argwhere(acc>=thresh)
    lines=lines[:20]
    out=img.copy()
    for i,j in lines:
        m=m_vals[i]
        c=c_vals[j]
        x0,x1=0,W
        y0=int(c)
        y1=int(m*x1+c)
        cv2.line(out,(x0,y0),(x1,y1),(0,0,255),1)
    return acc,out
acc_mc,out_mc=hough_slope_intercept(img,m_range=(-0.7,0.7),threshold_percent=99.5)
plt.figure(figsize=(6,6))
plt.imshow(cv2.cvtColor(out_mc,cv2.COLOR_BGR2RGB))
plt.title("Slope-intercept hough lines")
plt.axis("off")
plt.show()
````
