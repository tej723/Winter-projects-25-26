To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def convolve(img,kernel):
  img = img.astype(np.float32)
  kernel = kernel.astype(np.float32)
  img_h,img_w=img.shape
  kernel_h,kernel_w=kernel.shape
  pad_h=kernel_h//2
  pad_w=kernel_w//2
  out=np.zeros(img.shape)
  padded_img=np.pad(img,((pad_h,pad_h),(pad_w,pad_w)),mode='reflect')
  for x in range(img_h):
    for y in range(img_w):
      patch=padded_img[x:x+kernel_h,y:y+kernel_w]
      out[x,y]=np.sum(patch*kernel)
  return out
````

````python
def map(img, window_size=5, k=0.04):
    Ix = convolve(img, SOBEL_X)
    Iy = convolve(img, SOBEL_Y)
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy
    w = np.ones((window_size, window_size))
    a = convolve(Ix2, w)
    b = convolve(Iy2, w)
    c = convolve(Ixy, w)
    detM = a * b - c * c
    traceM = a + b
    R = detM - k * (traceM ** 2)
    return R
````

````python
def show_map(img, title):
    R = map(img)

    plt.figure(figsize=(10, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Input Image")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(R, cmap='gray')
    plt.title("Harris R Map")
    plt.axis("off")

    plt.tight_layout()
    plt.show()

    return R
````

````python
def to_uint8(img):
    img = img - img.min()
    img = img / (img.max() + 1e-6)
    img = (img * 255).astype(np.uint8)
    return img
````

````python
import cv2
def hough_slope_intercept(img_gray,
                          m_min=-2, m_max=2, m_bins=300,
                          c_min=-300, c_max=300, c_bins=300,
                          acc_thresh_ratio=0.9):
    img_u8 = to_uint8(img_gray)
    edges = cv2.Canny(img_u8, 100, 200)
    h, w = img_gray.shape
    m_vals = np.linspace(m_min, m_max, m_bins)
    c_vals = np.linspace(c_min, c_max, c_bins)
    accumulator = np.zeros((m_bins, c_bins), dtype=np.float32)
    ys, xs = np.where(edges > 0)
    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int((c - c_min) / (c_max - c_min) * (c_bins - 1))
            if 0 <= ci < c_bins:
                accumulator[mi, ci] += 1

    acc_max = accumulator.max()
    if acc_max > 0:
        accumulator /= acc_max

    indices = np.where(accumulator > acc_thresh_ratio)

    lines = [(m_vals[mi], c_vals[ci]) for mi, ci in zip(indices[0], indices[1])]
    img_lines = cv2.cvtColor(img_u8, cv2.COLOR_GRAY2BGR)

    for m, c in lines:
        y1 = int(c)
        y2 = int(m * w + c)
        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img_lines, (0, y1), (w, y2), (0, 0, 255), 1)

    return accumulator, img_lines
````

````python
def hough_rho_theta(img_gray,
                    rho_res=1,
                    theta_res=1,
                    acc_thresh_ratio=0.9):

    # Convert ONCE for OpenCV
    img_u8 = to_uint8(img_gray)

    # Canny edge detection
    edges = cv2.Canny(img_u8, 100, 200)

    h, w = img_gray.shape
    diag_len = int(np.sqrt(h*h + w*w))

    rhos = np.arange(-diag_len, diag_len, rho_res)
    thetas = np.deg2rad(np.arange(-90, 90, theta_res))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    # Voting
    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int((rho + diag_len) / rho_res)
            if 0 <= ri < len(rhos):
                accumulator[ri, ti] += 1

    # Safe normalization
    acc_max = accumulator.max()
    if acc_max > 0:
        accumulator /= acc_max

    # Threshold
    indices = np.where(accumulator > acc_thresh_ratio)
    lines = [(rhos[ri], thetas[ti]) for ri, ti in zip(indices[0], indices[1])]

    # Draw lines
    img_lines = cv2.cvtColor(img_u8, cv2.COLOR_GRAY2BGR)

    for rho, theta in lines:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_lines, (x1, y1), (x2, y2), (0, 255, 0), 1)

    return accumulator, img_lines
````
