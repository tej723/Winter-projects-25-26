To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
#The sobel kernels
SOBEL_X = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

SOBEL_Y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)
````

````python
#Convolution
def convolve2d(img, kernel):
    h, w = img.shape
    kh, kw = kernel.shape
    pad_h, pad_w = kh//2, kw//2

    padded = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')
    out = np.zeros_like(img)

    for i in range(h):
        for j in range(w):
            region = padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * kernel)
    return out
````

````python
flat = Image.open("/content/Unknown.png")
edge = Image.open("/content/Unknown-2.png")
corner = Image.open("/content/Unknown-3.png")


def gaussian_kernel(size=5, sigma=1.0):
    ax = np.linspace(-(size // 2), size // 2, size)
    xx, yy = np.meshgrid(ax, ax)
    kernel = np.exp(-(xx**2 + yy**2) / (2 * sigma**2))
    return kernel / kernel.sum()



def sobel_and_scatter(img, title=""):
    # Convert to grayscale
    gray = load_grayscale(img)

    # Smooth image (important!)
    G = gaussian_kernel(size=5, sigma=1.0)
    gray = convolve2d(gray, G)

    # Sobel gradients
    Ix = convolve2d(gray, SOBEL_X)
    Iy = convolve2d(gray, SOBEL_Y)

    # Gradient magnitude
    mag = np.sqrt(Ix**2 + Iy**2)

    # Adaptive threshold (top 10% strongest gradients)
    thresh = np.percentile(mag, 90)
    mask = mag > thresh

    # Display gradients
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix")
    plt.axis("off")

    plt.subplot(1, 2, 2)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy")
    plt.axis("off")

    plt.show()

    # Scatter plot
    plt.figure(figsize=(5, 5))
    plt.scatter(Ix[mask], Iy[mask], s=1, alpha=0.3)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix ({title})")
    plt.axhline(0, color='gray')
    plt.axvline(0, color='gray')
    plt.show()

    return Ix, Iy


Ix_flat, Iy_flat = sobel_and_scatter(flat, title="Flat")
Ix_edge, Iy_edge = sobel_and_scatter(edge, title="Edge")
Ix_corner, Iy_corner = sobel_and_scatter(corner, title="Corner")

````

````python
def harris_r_map(img, window=3, k=0.04):
    gray = load_grayscale(img)

    Ix = convolve2d(gray, SOBEL_X)
    Iy = convolve2d(gray, SOBEL_Y)

    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix*Iy

    pad = window // 2
    h, w = gray.shape
    R = np.zeros((h, w), dtype=np.float32)

    for i in range(pad, h-pad):
        for j in range(pad, w-pad):
            Sxx = np.sum(Ixx[i-pad:i+pad+1, j-pad:j+pad+1])
            Syy = np.sum(Iyy[i-pad:i+pad+1, j-pad:j+pad+1])
            Sxy = np.sum(Ixy[i-pad:i+pad+1, j-pad:j+pad+1])

            det = (Sxx * Syy) - (Sxy ** 2)
            trace = Sxx + Syy
            R[i, j] = det - k * (trace ** 2)

    return R
````

````python
def show_r_map(R, title="R Map"):
    R_norm = (R - R.min()) / (R.max() - R.min() + 1e-8)
    plt.figure(figsize=(5,5))
    plt.imshow(R_norm, cmap='gray')
    plt.title(title)
    plt.axis("off")
    plt.show()
````

Canny Edge detection

````python
edges = cv2.Canny(gray, threshold1=50, threshold2=150)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Map")
plt.axis("off")
plt.show()
````

````python
# m and b limits
m_min, m_max = -1, 1
b_min, b_max = -gray.shape[0], gray.shape[0]

num_m = 200
num_b = 400

ms = np.linspace(m_min, m_max, num_m)
bs = np.linspace(b_min, b_max, num_b)

accumulator = np.zeros((num_m, num_b), dtype=np.uint64)

````

````python
gx = cv2.Sobel(gray, cv2.CV_32F, 1, 0, ksize=3)
gy = cv2.Sobel(gray, cv2.CV_32F, 0, 1, ksize=3)

angle = np.arctan2(gy, gx)

ys, xs = np.where(edges > 0)

for x, y in zip(xs, ys):

    # Skipping near-vertical gradients (infinite slope issue)
    if abs(np.cos(angle[y, x])) < 0.2:
        continue

    # Voting in (m, b) space
    for mi, m in enumerate(ms):
        b = y - m * x
        bi = int((b - b_min) / (b_max - b_min) * num_b)

        if 0 <= bi < num_b:
            accumulator[mi, bi] += 1
````

````python

acc_norm = accumulator / accumulator.max()
threshold = 0.7
lines_idx = np.where(acc_norm > threshold)


print(f"Detected {len(lines_idx[0])} strong lines")
````

````python
output_si = img.copy()

for mi, bi in zip(lines_idx[0], lines_idx[1]):
    m = ms[mi]
    b = bs[bi]

    x1, x2 = 0, gray.shape[1]
    y1 = int(m * x1 + b)
    y2 = int(m * x2 + b)

    if 0 <= y1 < gray.shape[0] and 0 <= y2 < gray.shape[0]:
        cv2.line(output_si, (x1, y1), (x2, y2), (255, 0, 0), 2)

plt.imshow(output_si)
plt.title("SlopeIntercept Hough")
plt.axis("off")
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

ball_img = np.array(Image.open("/content/Unknown-4.png"))
pool_img = np.array(Image.open("/content/Unknown-5.png"))

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(ball_img)
plt.title("Ball Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(pool_img)
plt.title("Pool Image")
plt.axis("off")
plt.show()
````

````python
def extract_ball_stencil(img, x1, y1, x2, y2):
    roi = img[y1:y2, x1:x2]

    gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
    gray_blur = cv2.GaussianBlur(gray, (7, 7), 0)

    # Threshold
    _, binary = cv2.threshold(
        gray_blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU
    )

    # Ensuring ball is foreground
    if np.mean(binary) > 127:
        binary = cv2.bitwise_not(binary)

    kernel = np.ones((3, 3), np.uint8)
    binary = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel)
    binary = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)

    contours, _ = cv2.findContours(
        binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )

    if len(contours) == 0:
        raise ValueError("No contours found")

    largest = max(contours, key=cv2.contourArea)


    M = cv2.moments(largest)
    cx = int(M["m10"] / M["m00"])
    cy = int(M["m01"] / M["m00"])

    # Create alpha mask
    alpha = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(alpha, [largest], -1, 255, -1)


    h, w = roi.shape[:2]
    size = max(h, w)

    centered_rgba = np.zeros((size, size, 4), dtype=np.uint8)

    offset_x = size // 2 - cx
    offset_y = size // 2 - cy

    for i in range(h):
        for j in range(w):
            ni = i + offset_y
            nj = j + offset_x
            if 0 <= ni < size and 0 <= nj < size:
                centered_rgba[ni, nj, :3] = roi[i, j]
                centered_rgba[ni, nj, 3] = alpha[i, j]

    return centered_rgba
````

````python
def expand_box(x1, y1, x2, y2, margin, img_shape):
    h, w = img_shape[:2]

    x1 = max(0, x1 - margin)
    y1 = max(0, y1 - margin)
    x2 = min(w, x2 + margin)
    y2 = min(h, y2 + margin)

    return x1, y1, x2, y2
# original box (rough)
x1, y1 = 50, 60
x2, y2 = 200, 210

x1, y1, x2, y2 = expand_box(
    x1, y1, x2, y2,
    margin=40,
    img_shape=ball_img.shape
)

stencil = extract_ball_stencil(ball_img, x1, y1, x2, y2)

plt.imshow(stencil)
plt.title("Extracted RGBA Stencil")
plt.axis("off")
plt.show()
````

````python
def paste_stencil(base_img, stencil, x_center, y_center):
    """
    base_img : RGB image
    stencil  : RGBA image
    (x_center, y_center) : center placement
    """

    h, w, _ = stencil.shape
    x1 = x_center - w // 2
    y1 = y_center - h // 2

    result = base_img.copy()

    for i in range(h):
        for j in range(w):
            if stencil[i, j, 3] > 0:
                if 0 <= y1+i < result.shape[0] and 0 <= x1+j < result.shape[1]:
                    result[y1+i, x1+j] = stencil[i, j, :3]

    return result
````
