To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import io, base64

# Helpers
def show(img, title="", cmap="gray"):
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

def rgb_to_gray(img_rgb):
    """
    Convert RGB uint8 image to grayscale float32 using standard weights.
    Output range: 0..255 (float)
    """
    img_rgb = img_rgb.astype(np.float32)
    gray = 0.299*img_rgb[:,:,0] + 0.587*img_rgb[:,:,1] + 0.114*img_rgb[:,:,2]
    return gray

def conv2d_same(img, kernel):
    """
    2D convolution (same output size) using only NumPy.
    img: 2D float array
    kernel: 2D float array
    """
    kh, kw = kernel.shape
    pad_h, pad_w = kh//2, kw//2
    padded = np.pad(img, ((pad_h,pad_h),(pad_w,pad_w)), mode="reflect")
    out = np.zeros_like(img, dtype=np.float32)

    # flip kernel for convolution
    k = np.flipud(np.fliplr(kernel)).astype(np.float32)

    H, W = img.shape
    for i in range(H):
        for j in range(W):
            region = padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * k)
    return out

# Sobel kernels
SOBEL_X = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]], dtype=np.float32)

SOBEL_Y = np.array([[-1,-2,-1],
                    [ 0, 0, 0],
                    [ 1, 2, 1]], dtype=np.float32)

# Function
def partA_corner_prep(img_rgb, threshold=None, scatter_max_points=40000):
    """
    Takes an RGB image (H,W,3) uint8, converts to grayscale,
    computes Ix, Iy using Sobel, plots Ix, Iy, and scatter Iy vs Ix.

    If threshold is given:
      uses magnitude sqrt(Ix^2 + Iy^2) >= threshold to keep points for scatter.
      prints threshold used.

    Returns: Ix, Iy (float32)
    """
    gray = rgb_to_gray(img_rgb)  # grayscale only (hint followed)

    Ix = conv2d_same(gray, SOBEL_X)
    Iy = conv2d_same(gray, SOBEL_Y)

    # Visualize gradients (scaled for viewing)
    show(gray, "Grayscale (input)")
    show(Ix, "Ix (Sobel X)")
    show(Iy, "Iy (Sobel Y)")

    # Prepare scatter points
    x = Ix.flatten()
    y = Iy.flatten()

    if threshold is not None:
        mag = np.sqrt(Ix**2 + Iy**2).flatten()
        mask = mag >= threshold
        x = x[mask]
        y = y[mask]
        print(f" Threshold used (gradient magnitude): {threshold}")

    # Downsample points if too many (keeps plot fast + readable)
    if x.size > scatter_max_points:
        idx = np.random.choice(x.size, scatter_max_points, replace=False)
        x = x[idx]
        y = y[idx]

    plt.figure(figsize=(6,6))
    plt.scatter(x, y, s=2)   # Iy vs Ix
    plt.title("Scatter Plot: Iy vs Ix")
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.grid(True)
    plt.show()

    return Ix, Iy

# Load 3 images (Flat / Edge / Corner)
# Create synthetic test images
def make_flat_image(H=200, W=200, value=128):
    img = np.ones((H,W), dtype=np.uint8) * value
    return np.stack([img,img,img], axis=2)

def make_edge_image(H=200, W=200):
    img = np.zeros((H,W), dtype=np.uint8)
    img[:, :W//2] = 30
    img[:, W//2:] = 220
    return np.stack([img,img,img], axis=2)

def make_corner_image(H=200, W=200):
    img = np.zeros((H,W), dtype=np.uint8)
    img[:H//2, :W//2] = 220  # bright top-left block -> corners at its borders
    img[H//2:, W//2:] = 80
    return np.stack([img,img,img], axis=2)

flat_img   = make_flat_image()
edge_img   = make_edge_image()
corner_img = make_corner_image()

print("FLAT IMAGE")
Ix_flat, Iy_flat = partA_corner_prep(flat_img, threshold=50)

print("EDGE IMAGE")
Ix_edge, Iy_edge = partA_corner_prep(edge_img, threshold=50)

print("CORNER IMAGE")
Ix_corner, Iy_corner = partA_corner_prep(corner_img, threshold=50)
````

````python
import numpy as np
import matplotlib.pyplot as plt

def normalize_for_display(img):
    """Normalize a 2D array for display (0..255)"""
    img = img.astype(np.float32)
    mn, mx = img.min(), img.max()
    if mx - mn < 1e-8:
        return np.zeros_like(img, dtype=np.uint8)
    out = (img - mn) / (mx - mn)
    out = (out * 255).astype(np.uint8)
    return out

def box_filter_sum(img, window_size=5):
    """
    Fast windowed SUM filter using integral image.
    Returns sum in window (window_size x window_size) at each pixel (same size output).
    """
    k = window_size
    pad = k // 2
    # pad so output size matches input
    padded = np.pad(img, ((pad,pad),(pad,pad)), mode="reflect").astype(np.float32)

    # integral image
    integral = np.cumsum(np.cumsum(padded, axis=0), axis=1)

    H, W = img.shape
    out = np.zeros((H, W), dtype=np.float32)

    # compute sum over kxk using integral image
    for i in range(H):
        for j in range(W):
            r1, c1 = i, j
            r2, c2 = i + k - 1, j + k - 1

            A = integral[r2, c2]
            B = integral[r1-1, c2] if r1 > 0 else 0
            C = integral[r2, c1-1] if c1 > 0 else 0
            D = integral[r1-1, c1-1] if (r1 > 0 and c1 > 0) else 0

            out[i, j] = A - B - C + D

    return out

def harris_r_map_from_IxIy(Ix, Iy, window_size=5, k=0.04):
    """
    Compute Harris R-map using Ix and Iy.
    a = sum(Ix^2), b = sum(IxIy), c = sum(Iy^2) in local window.
    R = det(M) - k * trace(M)^2
    """
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    # second-moment sums in window
    a = box_filter_sum(Ix2, window_size)
    b = box_filter_sum(Ixy, window_size)
    c = box_filter_sum(Iy2, window_size)

    detM = a * c - b * b
    traceM = a + c
    R = detM - k * (traceM ** 2)

    return R

def partB_rmap(img_rgb, window_size=5, k=0.04, threshold=None):
    """
    For a given RGB image:
    1) compute grayscale
    2) compute Ix, Iy (Sobel)
    3) compute R-map (Harris)
    4) display R-map (corners should be bright)
    Optionally threshold R-map to visualize strong corners.
    Returns: R (float32)
    """
    gray = rgb_to_gray(img_rgb)
    Ix = conv2d_same(gray, SOBEL_X)
    Iy = conv2d_same(gray, SOBEL_Y)

    R = harris_r_map_from_IxIy(Ix, Iy, window_size=window_size, k=k)

    # Display raw R-map (normalized)
    R_disp = normalize_for_display(R)
    plt.figure(figsize=(5,5))
    plt.imshow(R_disp, cmap="gray")
    plt.title(f"R-map (window={window_size}, k={k})")
    plt.axis("off")
    plt.show()

    # If they want threshold mention clearly
    if threshold is not None:
        print(f" R threshold used: {threshold}")
        mask = (R > threshold)
        plt.figure(figsize=(5,5))
        plt.imshow(mask, cmap="gray")
        plt.title("Corner Mask after Threshold on R")
        plt.axis("off")
        plt.show()

    return R

print("Part B: FLAT IMAGE R-map")
R_flat = partB_rmap(flat_img, window_size=7, k=0.04)

print("Part B: EDGE IMAGE R-map")
R_edge = partB_rmap(edge_img, window_size=7, k=0.04)

print("Part B: CORNER IMAGE R-map")
R_corner = partB_rmap(corner_img, window_size=7, k=0.04)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Helpers
def show_bgr(img_bgr, title=""):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(7,7))
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()

def show_gray(img, title=""):
    plt.figure(figsize=(7,7))
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()

# Load Image
img_path = "/content/uk_image.jpg"
img_bgr = cv2.imread(img_path)

if img_bgr is None:
    raise FileNotFoundError(
        "Image not found. Upload your image to Colab files and name it 'uk_image.jpg' "
        "or update img_path."
    )

show_bgr(img_bgr, "Original Image")

# Canny Edge Map
def get_edges(img_bgr, low=50, high=150, blur_ksize=5):
    gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    if blur_ksize is not None and blur_ksize > 0:
        gray_blur = cv2.GaussianBlur(gray, (blur_ksize, blur_ksize), 0)
    else:
        gray_blur = gray
    edges = cv2.Canny(gray_blur, low, high)
    return gray, edges

gray, edges = get_edges(img_bgr, low=50, high=150, blur_ksize=5)
show_gray(gray, "Grayscale")
show_gray(edges, "Canny Edge Map")


# METHOD 1: HOUGH (m, c)
def hough_mc(edges, m_min=-2.0, m_max=2.0, m_steps=220,
             c_min=None, c_max=None, c_steps=260):
    """
    Hough Transform in slope-intercept space: y = m x + c
    Returns accumulator, m_values, c_values
    """
    H, W = edges.shape

    if c_min is None: c_min = -H
    if c_max is None: c_max = 2*H

    m_vals = np.linspace(m_min, m_max, m_steps).astype(np.float32)
    c_vals = np.linspace(c_min, c_max, c_steps).astype(np.float32)

    acc = np.zeros((m_steps, c_steps), dtype=np.uint32)

    ys, xs = np.where(edges > 0)

    # voting
    for x, y in zip(xs, ys):
        c_for_all_m = y - m_vals * x
        c_idx = np.round((c_for_all_m - c_min) * (c_steps - 1) / (c_max - c_min)).astype(np.int32)

        valid = (c_idx >= 0) & (c_idx < c_steps)
        acc[np.where(valid)[0], c_idx[valid]] += 1

    return acc, m_vals, c_vals

def draw_lines_mc(img_bgr, m_vals, c_vals, peaks):
    """
    Draw lines y = m x + c for peak indices in accumulator
    """
    H, W = img_bgr.shape[:2]
    out = img_bgr.copy()

    for mi, ci in peaks:
        m = float(m_vals[mi])
        c = float(c_vals[ci])

        # line endpoints using x=0 and x=W-1
        y0 = int(m * 0 + c)
        y1 = int(m * (W - 1) + c)

        # draw if not totally crazy
        if -H <= y0 <= 2*H and -H <= y1 <= 2*H:
            cv2.line(out, (0, y0), (W - 1, y1), (0, 255, 0), 2)

    return out

# Run Hough (m,c)
acc_mc, m_vals, c_vals = hough_mc(
    edges,
    m_min=-1.5, m_max=1.5, m_steps=220,       # keep small for speed
    c_min=-gray.shape[0], c_max=2*gray.shape[0], c_steps=260
)

acc_mc_norm = acc_mc.astype(np.float32) / (acc_mc.max() + 1e-8)
show_gray(acc_mc_norm, "Accumulator (m,c) - normalized")

# Choose threshold (example: 90% of max)
th_mc = 0.90
peaks_mc = np.argwhere(acc_mc_norm >= th_mc)

print("---- (m,c) RESULTS ----")
print("Max votes:", int(acc_mc.max()))
print("Threshold (normalized):", th_mc)
print("Lines detected (peaks):", len(peaks_mc))

out_mc = draw_lines_mc(img_bgr, m_vals, c_vals, peaks_mc)
show_bgr(out_mc, "Final Output: Lines using (m,c) Hough")


# METHOD 2: HOUGH (rho, theta)  (FAST)
def hough_rho_theta(edges, theta_steps=180, rho_steps=None):
    """
    Hough Transform in rho-theta:
       rho = x cos(theta) + y sin(theta)
    Returns accumulator, rhos, thetas
    """
    H, W = edges.shape
    diag = int(np.ceil(np.sqrt(H*H + W*W)))

    if rho_steps is None:
        rho_steps = 2*diag  # rho in [-diag, +diag]

    thetas = np.deg2rad(np.linspace(0, 180, theta_steps, endpoint=False)).astype(np.float32)
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    rhos = np.linspace(-diag, diag, rho_steps).astype(np.float32)

    acc = np.zeros((rho_steps, theta_steps), dtype=np.uint32)

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        rho_vals = x*cos_t + y*sin_t
        rho_idx = np.round((rho_vals + diag) * (rho_steps - 1) / (2*diag)).astype(np.int32)
        acc[rho_idx, np.arange(theta_steps)] += 1

    return acc, rhos, thetas

def draw_lines_rho_theta(img_bgr, rhos, thetas, peaks):
    """
    Draw lines corresponding to (rho, theta)
    """
    out = img_bgr.copy()

    for ri, ti in peaks:
        rho = float(rhos[ri])
        theta = float(thetas[ti])

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        # long segment
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(out, (x1, y1), (x2, y2), (255, 0, 0), 2)

    return out

# Run Hough (rho,theta)
acc_rt, rhos, thetas = hough_rho_theta(edges, theta_steps=180, rho_steps=None)

acc_rt_norm = acc_rt.astype(np.float32) / (acc_rt.max() + 1e-8)
show_gray(acc_rt_norm, "Accumulator (rho,theta) - normalized")

# Choose threshold (rho-theta usually needs lower than 0.90)
th_rt = 0.55
peaks_rt = np.argwhere(acc_rt_norm >= th_rt)

print("---- (rho,theta) RESULTS ----")
print("Max votes:", int(acc_rt.max()))
print("Threshold (normalized):", th_rt)
print("Lines detected (peaks):", len(peaks_rt))

out_rt = draw_lines_rho_theta(img_bgr, rhos, thetas, peaks_rt)
show_bgr(out_rt, "Final Output: Lines using (rho,theta) Hough")
````
