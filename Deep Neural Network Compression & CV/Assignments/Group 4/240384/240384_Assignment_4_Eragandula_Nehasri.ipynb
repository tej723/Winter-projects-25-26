To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Flat region: smooth sky
flat_img = img[50:300, 200:700]
# Edge region: skyline boundary
edge_img = img[550:700, 200:700]
# Corner region: buildings
corner_img = img[700:950, 600:1100]
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.imshow(flat_img)
plt.title("Flat Region")
plt.axis("off")
plt.subplot(1,3,2)
plt.imshow(edge_img)
plt.title("Edge Region")
plt.axis("off")
plt.subplot(1,3,3)
plt.imshow(corner_img)
plt.title("Corner Region")
plt.axis("off")
plt.show()
````

````python
gray_flat   = cv2.cvtColor(flat_img, cv2.COLOR_RGB2GRAY)
gray_edge   = cv2.cvtColor(edge_img, cv2.COLOR_RGB2GRAY)
gray_corner = cv2.cvtColor(corner_img, cv2.COLOR_RGB2GRAY)
````

````python
def compute_gradients_and_plot(gray, threshold=None):
    # Sobel kernels
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]])
    sobel_y = np.array([[-1, -2, -1],
                        [ 0,  0,  0],
                        [ 1,  2,  1]])
    # Compute gradients
    Ix = cv2.filter2D(gray.astype(np.float32), -1, sobel_x)
    Iy = cv2.filter2D(gray.astype(np.float32), -1, sobel_y)
    # Show Ix and Iy
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.title("Ix (Horizontal Gradient)")
    plt.imshow(Ix, cmap="gray")
    plt.axis("off")
    plt.subplot(1,2,2)
    plt.title("Iy (Vertical Gradient)")
    plt.imshow(Iy, cmap="gray")
    plt.axis("off")
    plt.show()
    # Flatten
    Ix_flat = Ix.flatten()
    Iy_flat = Iy.flatten()
    # Thresholding
    if threshold is not None:
        mag = np.sqrt(Ix_flat**2 + Iy_flat**2)
        mask = mag > threshold
        Ix_flat = Ix_flat[mask]
        Iy_flat = Iy_flat[mask]
    # Scatter plot
    plt.figure(figsize=(5,5))
    plt.scatter(Ix_flat, Iy_flat, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Scatter Plot of Gradients")
    plt.grid(True)
    plt.show()
    return Ix, Iy
````

````python
print("FLAT REGION")
Ix_flat, Iy_flat = compute_gradients_and_plot(gray_flat, threshold=None)
print("EDGE REGION")
Ix_edge, Iy_edge = compute_gradients_and_plot(gray_edge, threshold=5)
print("CORNER REGION")
Ix_corner, Iy_corner = compute_gradients_and_plot(gray_corner, threshold=5)
````

````python
#Sobel filters were used to compute horizontal and vertical image gradients Ix and Iy on grayscale image.
#Three regions were cropped from the same image: a flat sky region, a skyline edge region, and a building region containing corners.
#Scatter plots of Ix and Iy show distinct behaviors: flat regions cluster near the origin, edge regions exhibit variation primarily in one direction, and corner regions show significant spread in both directions.
#No threshold was applied for the flat region to preserve near-zero gradients, while a small threshold of 5 was used for edge and corner regions to suppress weak gradients caused by noise.
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def compute_R_map(Ix, Iy, window_size=7, k=0.04):
    h, w = Ix.shape
    R_map = np.zeros((h, w), dtype=np.float32)
    offset = window_size // 2
    for y in range(offset, h - offset):
        for x in range(offset, w - offset):
            Ix_win = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            Iy_win = Iy[y-offset:y+offset+1, x-offset:x+offset+1]
            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)
            M = np.array([[a, b],
                          [b, c]])
            lambda1, lambda2 = np.linalg.eigvals(M)
            R_map[y, x] = lambda1 * lambda2 - k * (lambda1 + lambda2)**2
    return R_map
````

````python
R_flat = compute_R_map(Ix_flat, Iy_flat)
R_edge = compute_R_map(Ix_edge, Iy_edge)
R_corner = compute_R_map(Ix_corner, Iy_corner)
````

````python
def normalize_for_display(R):
    R = R - R.min()
    if R.max() != 0:
        R = R / R.max()
    return R
R_flat_n   = normalize_for_display(R_flat)
R_edge_n   = normalize_for_display(R_edge)
R_corner_n = normalize_for_display(R_corner)
````

````python
# Grayscale image
gray_hough = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Canny edge detection
edges = cv2.Canny(gray_hough, threshold1=100, threshold2=200)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Map")
plt.axis("off")
plt.show()
````

````python
def hough_slope_intercept(edges, m_range=(-1,1), m_step=0.01, c_step=1):
    h, w = edges.shape
    m_values = np.arange(m_range[0], m_range[1], m_step)
    c_values = np.arange(-h, h, c_step)
    accumulator = np.zeros((len(m_values), len(c_values)), dtype=np.int32)
    ys, xs = np.where(edges > 0)
    for i, m in enumerate(m_values):
        for x, y in zip(xs, ys):
            c = y - m * x
            c_idx = int((c + h) / c_step)
            if 0 <= c_idx < len(c_values):
                accumulator[i, c_idx] += 1
    return accumulator, m_values, c_values
````

````python
acc_si, m_vals, c_vals = hough_slope_intercept(edges)
plt.imshow(acc_si, cmap="hot", aspect="auto")
plt.title("Accumulator (SlopeIntercept)")
plt.xlabel("c index")
plt.ylabel("m index")
plt.colorbar()
plt.show()
````

````python
img_lines_si = img.copy()
threshold = 0.9 * acc_si.max()
for i in range(acc_si.shape[0]):
    for j in range(acc_si.shape[1]):
        if acc_si[i, j] > threshold:
            m = m_vals[i]
            c = c_vals[j]
            x0, x1 = 0, img.shape[1]
            y0, y1 = int(m*x0 + c), int(m*x1 + c)
            cv2.line(img_lines_si, (x0, y0), (x1, y1), (255, 0, 0), 1)
plt.imshow(img_lines_si)
plt.title("Detected Lines (SlopeIntercept)")
plt.axis("off")
plt.show()
````

````python
def hough_rho_theta(edges, rho_res=1, theta_res=np.pi/180):
    h, w = edges.shape
    diag = int(np.sqrt(h*h + w*w))
    rhos = np.arange(-diag, diag, rho_res)
    thetas = np.arange(0, np.pi, theta_res)
    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.int32)
    ys, xs = np.where(edges > 0)
    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = x*np.cos(theta) + y*np.sin(theta)
            r_idx = int((rho + diag) / rho_res)
            if 0 <= r_idx < len(rhos):
                accumulator[r_idx, t_idx] += 1
    return accumulator, rhos, thetas
````

````python
acc_rt, rhos, thetas = hough_rho_theta(edges)
plt.imshow(acc_rt, cmap="hot", aspect="auto")
plt.title("Accumulator (RhoTheta)")
plt.xlabel("Theta index")
plt.ylabel("Rho index")
plt.colorbar()
plt.show()
````

````python
img_lines_rt = img.copy()
threshold = 0.9 * acc_rt.max()
for r_idx in range(acc_rt.shape[0]):
    for t_idx in range(acc_rt.shape[1]):
        if acc_rt[r_idx, t_idx] > threshold:
            rho = rhos[r_idx]
            theta = thetas[t_idx]
            a, b = np.cos(theta), np.sin(theta)
            x0, y0 = a*rho, b*rho
            x1, y1 = int(x0 + 1000*(-b)), int(y0 + 1000*(a))
            x2, y2 = int(x0 - 1000*(-b)), int(y0 - 1000*(a))
            cv2.line(img_lines_rt, (x1, y1), (x2, y2), (0, 255, 0), 1)
plt.imshow(img_lines_rt)
plt.title("Detected Lines (RhoTheta)")
plt.axis("off")
plt.show()
````

````python
#A Canny edge map was first computed to identify significant intensity transitions.
#The Hough Transform was implemented using both slopeintercept and rhotheta parameterizations.
#In the slopeintercept form, each edge pixel votes for possible line parameters, resulting in a large and slow accumulator.
#The rhotheta formulation provides a more compact and stable representation, allowing faster and more reliable line detection.
#Peaks in the accumulator correspond to strong line structures in the image, which were visualized by overlaying detected lines.
````
