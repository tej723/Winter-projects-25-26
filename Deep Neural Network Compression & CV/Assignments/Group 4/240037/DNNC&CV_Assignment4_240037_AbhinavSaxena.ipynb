To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Sobel Kernels
sobel_x = np.array([[1, 0, -1],
                    [2, 0, -2],
                    [1, 0, -1]])

sobel_y = np.array([[1, 2, 1],
                    [0, 0, 0],
                    [-1, -2, -1]])

def grayscale(img):
    """Convert RGB to grayscale"""
    if len(img.shape) == 3:
        return np.dot(img[...,:3], [0.2989, 0.5870, 0.1140])
    return img

def convolve2d(image, kernel):

    h, w = image.shape
    kh, kw = kernel.shape
    pad_h, pad_w = kh // 2, kw // 2
    # Pad image with zeros
    padded = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant', constant_values=0)
    out = np.zeros_like(image)

    # Convolve
    for i in range(h):
        for j in range(w):
            region = padded[i:i+kh, j:j+kw]
            out[i, j] = np.sum(region * kernel)
    return out

def compute_sobel(img, threshold=None):
    """Compute Ix, Iy using Sobel and plot results"""
    gray = grayscale(img)

    Ix = convolve2d(gray, sobel_x)
    Iy = convolve2d(gray, sobel_y)

    # Plot Ix
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix (Sobel X)")
    plt.axis('off')

    # Plot Iy
    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy (Sobel Y)")
    plt.axis('off')

    # Scatter plot Iy vs Ix
    plt.subplot(1,3,3)
    if threshold is not None:
        mask = (np.abs(Ix) > threshold) & (np.abs(Iy) > threshold)
        plt.scatter(Ix[mask], Iy[mask], s=1, c='blue')
        print(f"Applied threshold: {threshold}")
    else:
        plt.scatter(Ix.flatten(), Iy.flatten(), s=1, c='blue')
    plt.title("Scatter: Iy vs Ix")
    plt.xlabel("Ix")
    plt.ylabel("Iy")

    plt.tight_layout()
    plt.show()

    return Ix, Iy

# Example usage with the images you loaded
Ix_flat, Iy_flat = compute_sobel(flat, threshold=50)
Ix_edge, Iy_edge = compute_sobel(edge, threshold=50)
Ix_corner, Iy_corner = compute_sobel(corner, threshold=50)
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Assume these are your images in grayscale np.array format
images = [flat, edge,corner]
titles = ["Flat", "Edge", "Corner"]

# ---------------- Part A: Compute Gradients ----------------
def compute_gradients(img_gray):
    # Sobel gradients
    Ix = cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)
    return Ix, Iy

def plot_gradients_scatter(Iy, Ix, title="Gradient Scatter"):
    plt.figure(figsize=(5,5))
    plt.scatter(Ix.flatten(), Iy.flatten(), alpha=0.5, s=2)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(title)
    plt.grid(True)
    plt.show()

# ---------------- Part B: Harris R-map ----------------
def harris_response(img_gray, window_size=3, k=0.04):
    Ix, Iy = compute_gradients(img_gray)
    Ix2 = Ix**2
    Iy2 = Iy**2
    Ixy = Ix*Iy

    # Create empty R map
    R = np.zeros(img_gray.shape)
    offset = window_size // 2

    # Loop over pixels
    for i in range(offset, img_gray.shape[0]-offset):
        for j in range(offset, img_gray.shape[1]-offset):
            # Get local window
            Ix2_win = Ix2[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy2_win = Iy2[i-offset:i+offset+1, j-offset:j+offset+1]
            Ixy_win = Ixy[i-offset:i+offset+1, j-offset:j+offset+1]

            # Sum over window (second moments)
            a = Ix2_win.sum()
            b = Ixy_win.sum()
            c = Iy2_win.sum()

            # Harris R value
            detM = a*c - b**2
            traceM = a + c
            R[i,j] = detM - k*(traceM**2)
    return R

def plot_r_map(R, title="R Map"):
    plt.figure(figsize=(5,5))
    plt.imshow(R, cmap='gray')
    plt.title(title)
    plt.colorbar()
    plt.show()

# ---------------- Run for all images ----------------
for idx, img in enumerate(images):
    print(f"Processing {titles[idx]} Image")

    # Part A: Gradient scatter
    Ix, Iy = compute_gradients(img)
    plot_gradients_scatter(Iy, Ix, title=f"Gradient Scatter - {titles[idx]}")

    # Part B: Harris R map
    R = harris_response(img, window_size=3, k=0.04)
    plot_r_map(R, title=f"Harris R Map - {titles[idx]}")
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

#  Synthetic Line Image
img = np.zeros((200, 200), dtype=np.uint8)
cv2.line(img, (10,50), (190,50), 255, 2)   # horizontal
cv2.line(img, (50,10), (50,190), 255, 2)   # vertical
cv2.line(img, (10,10), (190,190), 255, 2)  # diagonal

plt.imshow(img, cmap='gray')
plt.title("Original Image with Lines")
plt.show()

# ---------------- Slope-Intercept Hough ----------------
def hough_slope_intercept(image, theta_res=1, rho_res=1, threshold=0.5):
    edges = cv2.Canny(image, 50, 150)

    # Prepare accumulator
    rows, cols = edges.shape
    accumulator = np.zeros((rows, cols), dtype=np.float32)

    # For each edge pixel, vote for lines
    y_idxs, x_idxs = np.nonzero(edges)

    for x, y in zip(x_idxs, y_idxs):
        for x2 in range(cols):
            if x2 != x:
                m = (y - y) / (x - x2 + 1e-5)
                b = y - m*x
                # Convert slope-intercept to integer index
                m_idx = int((m+5)*10)   # scale slope
                b_idx = int(b)
                if 0 <= m_idx < cols and 0 <= b_idx < rows:
                    accumulator[b_idx, m_idx] += 1

    # Threshold accumulator
    acc_max = accumulator.max()
    lines_idx = np.argwhere(accumulator > acc_max*threshold)

    output = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    for b_idx, m_idx in lines_idx:
        m = (m_idx/10 - 5)
        b = b_idx
        x1 = 0
        y1 = int(m*x1 + b)
        x2 = cols-1
        y2 = int(m*x2 + b)
        cv2.line(output, (x1,y1), (x2,y2), (0,0,255), 1)

    plt.figure(figsize=(6,6))
    plt.imshow(output)
    plt.title("Detected Lines - Slope Intercept Hough")
    plt.show()

# ---------------- Rho-Theta Hough ----------------
def hough_rho_theta(image):
    edges = cv2.Canny(image, 50, 150)
    output = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    lines = cv2.HoughLines(edges, 1, np.pi/180, 100)
    if lines is not None:
        for rho_theta in lines:
            rho, theta = rho_theta[0]
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))
            cv2.line(output, (x1,y1), (x2,y2), (0,0,255), 2)

    plt.figure(figsize=(6,6))
    plt.imshow(output)
    plt.title("Detected Lines - Rho-Theta Hough")
    plt.show()

hough_slope_intercept(img)   # Slope-Intercept (slow)
hough_rho_theta(img)         # Rho-Theta (fast)
````
