To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def create_synthetic_images():

    height, width = 200, 200

    # 1. Flat Image: Just a solid gray color
    flat_img = np.ones((height, width), dtype=np.uint8) * 128

    # 2. Edge Image: Vertical edge (Left half black, Right half white)
    edge_img = np.zeros((height, width), dtype=np.uint8)
    edge_img[:, 100:] = 255

    # 3. Corner Image: A white square in the middle of black background
    corner_img = np.zeros((height, width), dtype=np.uint8)
    corner_img[50:150, 50:150] = 255

    return [("Flat Region", flat_img), ("Edge Region", edge_img), ("Corner Region", corner_img)]

def analyze_and_plot_gradients(image, title, threshold=None):

    # Ensure image is grayscale
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 1. Calculate Gradients using Sobel
    # cv2.CV_64F is used to keep negative values (important for direction)
    ksize = 3
    gx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=ksize)
    gy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=ksize)

    # 2. Setup Plots
    fig, axes = plt.subplots(1, 4, figsize=(20, 5))
    fig.suptitle(f"Analysis of: {title}", fontsize=16)

    # Plot Original
    axes[0].imshow(image, cmap='gray')
    axes[0].set_title("Original Image")
    axes[0].axis('off')

    # Plot Ix (Gradient X)
    axes[1].imshow(gx, cmap='jet')
    axes[1].set_title("Gradient X (Ix)")
    axes[1].axis('off')

    # Plot Iy (Gradient Y)
    axes[2].imshow(gy, cmap='jet')
    axes[2].set_title("Gradient Y (Iy)")
    axes[2].axis('off')

    # 3. Scatter Plot (Iy vs Ix)
    # We flatten the arrays to treat pixels as individual data points
    ix_flat = gx.flatten()
    iy_flat = gy.flatten()

    # Optional Thresholding to remove noise (e.g., flat areas)
    if threshold is not None:
        magnitude = np.sqrt(ix_flat**2 + iy_flat**2)
        # Only keep indices where magnitude is greater than threshold
        indices = magnitude > threshold
        ix_plot = ix_flat[indices]
        iy_plot = iy_flat[indices]
        scatter_title = f"Scatter: Iy vs Ix (Thresh>{threshold})"
    else:
        ix_plot = ix_flat
        iy_plot = iy_flat
        scatter_title = "Scatter: Iy vs Ix (No Threshold)"

    axes[3].scatter(ix_plot, iy_plot, alpha=0.3, s=2, c='blue')
    axes[3].set_title(scatter_title)
    axes[3].set_xlabel("Ix")
    axes[3].set_ylabel("Iy")
    axes[3].grid(True, linestyle='--', alpha=0.6)

    # Add crosshair at 0,0
    axes[3].axhline(0, color='black', linewidth=1)
    axes[3].axvline(0, color='black', linewidth=1)

    plt.tight_layout()
    plt.show()

    return gx, gy

# --- Execution ---

# 1. Generate the images
images = create_synthetic_images()

# 2. Run the function for each image
results = {}

for title, img in images:
    # We use a small threshold (e.g., 10) to remove the pure flat areas
    # from the scatter plot so the relevant points are visible.
    Ix, Iy = analyze_and_plot_gradients(img, title, threshold=10)

    # Store for further questions if needed
    results[title] = {"Ix": Ix, "Iy": Iy}
````

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# --- 1. Generate Synthetic Images (No Base64 needed) ---
def create_synthetic_images():

    height, width = 200, 200

    # Flat Image: Solid gray
    flat_img = np.ones((height, width), dtype=np.float32) * 128

    # Edge Image: Vertical line (Step edge)
    edge_img = np.zeros((height, width), dtype=np.float32)
    edge_img[:, 100:] = 255  # Right half is white

    # Corner Image: White square in center
    corner_img = np.zeros((height, width), dtype=np.float32)
    corner_img[50:150, 50:150] = 255

    return [("Flat Image", flat_img), ("Edge Image", edge_img), ("Corner Image", corner_img)]

# --- 2. Harris Corner Response Function ---
def compute_harris_response(image, k=0.04, window_size=3):

    # Ensure image is float32
    gray = np.float32(image)

    # A. Calculate derivatives (Ix, Iy) using Sobel
    dx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    # B. Calculate product of derivatives
    Ixx = dx * dx
    Iyy = dy * dy
    Ixy = dx * dy

    # C. Compute Second Moment Matrix components (A, B, C)
    # Apply Gaussian window to sum the products over the window
    # standard sigma for 3x3 window is often around 1.0
    sigma = 1.0
    # ensure ksize is odd
    g_ksize = (window_size, window_size) if window_size % 2 == 1 else (window_size + 1, window_size + 1)

    A = cv2.GaussianBlur(Ixx, g_ksize, sigma) # Sum of Ix^2
    B = cv2.GaussianBlur(Iyy, g_ksize, sigma) # Sum of Iy^2
    C = cv2.GaussianBlur(Ixy, g_ksize, sigma) # Sum of Ix*Iy

    # D. Compute R (Corner Response)
    # Determinant = A*B - C^2
    # Trace = A + B
    det_M = (A * B) - (C**2)
    trace_M = A + B

    # Harris Response Equation: R = Det - k * (Trace)^2
    R_map = det_M - k * (trace_M**2)

    return R_map



# Generate the images
images_data = create_synthetic_images()

fig, axes = plt.subplots(len(images_data), 2, figsize=(12, 15))

for i, (title, img) in enumerate(images_data):
    # Calculate R
    R = compute_harris_response(img)

    # 1. Plot Original Image
    axes[i, 0].imshow(img, cmap='gray')
    axes[i, 0].set_title(f"Original: {title}")
    axes[i, 0].axis('off')

    # 2. Plot R-Map
    # Corners will be bright spots (Large positive R)
    # Edges will be dark/black (Large negative R)
    # Flat areas will be grey (R approx 0)
    im = axes[i, 1].imshow(R, cmap='jet') # 'jet' or 'viridis' highlights peaks well
    axes[i, 1].set_title(f"R-Map (Harris Response)")
    axes[i, 1].axis('off')

    # Optional: Add colorbar to see range of values
    fig.colorbar(im, ax=axes[i, 1], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

def hough_slope_intercept(image_path):
    # 1. Load and Preprocess
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        # Fallback synthetic image if load fails
        img = np.zeros((100, 100), dtype=np.uint8)
        cv2.line(img, (20, 20), (80, 80), 255, 1)

    # Resize
    scale_percent = min(150.0 / img.shape[0], 150.0 / img.shape[1])
    width = int(img.shape[1] * scale_percent)
    height = int(img.shape[0] * scale_percent)
    img_small = cv2.resize(img, (width, height))

    # 2. Edge Detection
    edges = cv2.Canny(img_small, 50, 150)

    # 3. Define Accumulator Space
    m_min, m_max = -5.0, 5.0
    m_step = 0.1
    m_axis = np.arange(m_min, m_max, m_step)

    c_min, c_max = -height * 1.5, height * 1.5
    c_step = 1.0
    c_axis = np.arange(c_min, c_max, c_step)

    accumulator = np.zeros((len(m_axis), len(c_axis)), dtype=int)

    # 4. Voting Procedure
    y_idxs, x_idxs = np.nonzero(edges)

    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for m_idx, m in enumerate(m_axis):
            c = y - (m * x)

            if c >= c_min and c < c_max:
                c_idx = int((c - c_min) / c_step)
                if 0 <= c_idx < len(c_axis):
                    accumulator[m_idx, c_idx] += 1

    # 5. Thresholding and Drawing
    acc_norm = (accumulator / np.max(accumulator)) * 100
    threshold_val = 90
    peaks = np.argwhere(acc_norm > threshold_val)

    result_img = cv2.cvtColor(img_small, cv2.COLOR_GRAY2BGR)

    for peak in peaks:
        m_idx, c_idx = peak
        m = m_axis[m_idx]
        c = c_axis[c_idx]

        y1 = int(c)
        y2 = int(m * width + c)

        cv2.line(result_img, (0, y1), (width, y2), (0, 0, 255), 1)

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(edges, cmap='gray'), plt.title('Canny Edges')
    plt.subplot(122), plt.imshow(result_img), plt.title('Detected Lines (y=mx+c)')
    plt.show()

# Run
hough_slope_intercept('my image.jpg')
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

def hough_rho_theta(image_path):
    # 1. Load and Preprocess
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    if img is None:
        img = np.zeros((100, 100), dtype=np.uint8)
        cv2.line(img, (20, 20), (80, 80), 255, 1)

    scale_percent = min(150.0 / img.shape[0], 150.0 / img.shape[1])
    width = int(img.shape[1] * scale_percent)
    height = int(img.shape[0] * scale_percent)
    img_small = cv2.resize(img, (width, height))

    # 2. Edge Detection
    edges = cv2.Canny(img_small, 50, 150)

    # 3. Define Accumulator Space
    diag_len = int(np.ceil(np.sqrt(width**2 + height**2)))
    rhos = np.arange(-diag_len, diag_len, 1)
    thetas = np.deg2rad(np.arange(-90, 90, 1))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=int)

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    # 4. Voting Procedure
    y_idxs, x_idxs = np.nonzero(edges)

    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for t_idx in range(len(thetas)):
            rho = round(x * cos_t[t_idx] + y * sin_t[t_idx])
            rho_idx = int(rho + diag_len)

            if 0 <= rho_idx < len(rhos):
                accumulator[rho_idx, t_idx] += 1

    # 5. Thresholding and Drawing
    acc_norm = (accumulator / np.max(accumulator)) * 100
    threshold_val = 90
    peaks = np.argwhere(acc_norm > threshold_val)

    result_img = cv2.cvtColor(img_small, cv2.COLOR_GRAY2BGR)

    for peak in peaks:
        rho_idx, theta_idx = peak
        rho = rhos[rho_idx]
        theta = thetas[theta_idx]

        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(result_img, (x1, y1), (x2, y2), (0, 255, 0), 1)

    # Display
    plt.figure(figsize=(10, 5))
    plt.subplot(121), plt.imshow(edges, cmap='gray'), plt.title('Canny Edges')
    plt.subplot(122), plt.imshow(result_img), plt.title('Detected Lines (Rho-Theta)')
    plt.show()

# Run
hough_rho_theta('my image.jpg')
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt


def create_ball_image(width, height, ball_color, bg_color, radius, pos):

    # Create background
    img = np.zeros((height, width, 3), dtype=np.uint8)
    img[:] = bg_color

    # Draw ball (Circle)
    # cv2.circle expects center as (x, y) and color as (R, G, B)
    cv2.circle(img, pos, radius, ball_color, -1)

    # Add some noise to make gradient detection interesting
    noise = np.random.randint(0, 50, (height, width, 3), dtype=np.uint8)
    img = cv2.add(img, noise)

    return img

# Generate "Image 1": White Ball on grey background
# Simulates the input for the stencil creation
ball1_img = create_ball_image(400, 300, (255, 255, 255), (50, 50, 50), 60, (200, 150))

# Generate "Image 2": Red Ball on Green Table
# Simulates the destination image
ball2_img = create_ball_image(400, 300, (200, 0, 0), (0, 100, 0), 60, (100, 150))



def get_ball_stencil(image, x1, y1, x2, y2):

    # 1. Crop
    crop = image[y1:y2, x1:x2]

    # 2. Grayscale & Blur
    gray = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # 3. Gradient / Edge Detection
    # Using Canny to find gradients
    edges = cv2.Canny(blurred, 30, 150)

    # Close gaps in the edges to ensure a full circle contour
    kernel = np.ones((3,3), np.uint8)
    closed_edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    # 4. Find Contours
    contours, _ = cv2.findContours(closed_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours:
        print("No contours found in box.")
        return crop

    # Select the biggest completed gradient (maximum area contour)
    max_contour = max(contours, key=cv2.contourArea)

    # 5. Create Mask & Stencil
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [max_contour], -1, 255, thickness=cv2.FILLED)

    # Convert RGB crop to RGBA
    b_channel, g_channel, r_channel = cv2.split(crop)
    alpha_channel = np.ones_like(b_channel) * 255
    rgba = cv2.merge((b_channel, g_channel, r_channel, alpha_channel))

    # Apply mask to alpha channel
    rgba[:, :, 3] = mask

    return rgba

def overlay_ball_on_table(background_img, stencil, center_x, center_y):

    bg = background_img.copy()
    # Ensure bg is RGBA for easy blending
    if bg.shape[2] == 3:
        bg = cv2.cvtColor(bg, cv2.COLOR_RGB2RGBA)

    h_stencil, w_stencil = stencil.shape[:2]
    h_bg, w_bg = bg.shape[:2]

    # Calculate Top-Left coordinates
    x_offset = center_x - (w_stencil // 2)
    y_offset = center_y - (h_stencil // 2)

    # Handle Clipping (if ball goes off screen)
    y1, y2 = max(0, y_offset), min(h_bg, y_offset + h_stencil)
    x1, x2 = max(0, x_offset), min(w_bg, x_offset + w_stencil)

    y1_s, y2_s = max(0, -y_offset), min(h_stencil, h_bg - y_offset)
    x1_s, x2_s = max(0, -x_offset), min(w_stencil, w_bg - x_offset)

    bg_slice = bg[y1:y2, x1:x2]
    stencil_slice = stencil[y1_s:y2_s, x1_s:x2_s]

    if bg_slice.size == 0 or stencil_slice.size == 0:
        return bg

    # Alpha Blending
    alpha_s = stencil_slice[:, :, 3] / 255.0

    for c in range(0, 3):
        bg_slice[:, :, c] = (alpha_s * stencil_slice[:, :, c] +
                             (1 - alpha_s) * bg_slice[:, :, c])

    bg_slice[:, :, 3] = 255 # Keep background fully opaque

    bg[y1:y2, x1:x2] = bg_slice
    return bg



# STEP 1: Define Box around White Ball (x1, y1, x2, y2)
# We know the ball is at (200, 150) with radius 60, so we make a box around it
box_coords = (130, 80, 270, 220)

# STEP 2: Create Stencil (Extract White Ball)
stencil = get_ball_stencil(ball1_img, *box_coords)

# STEP 3: Define Target Center (x3, y3)
target_coords = (300, 150)

# STEP 4: Overlay
result_img = overlay_ball_on_table(ball2_img, stencil, *target_coords)


plt.figure(figsize=(15, 5))

plt.subplot(1, 4, 1)
plt.imshow(ball1_img)
plt.title("Img 1: White Ball\n(Input)")
plt.axis("off")

debug_img = ball1_img.copy()
cv2.rectangle(debug_img, (box_coords[0], box_coords[1]), (box_coords[2], box_coords[3]), (255, 255, 0), 2)
plt.subplot(1, 4, 2)
plt.imshow(debug_img)
plt.title("Box Selection")
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(stencil)
plt.title("Extracted Stencil\n(Gradients)")
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(result_img)
plt.title("Img 2: Result\n(Overlay)")
plt.axis("off")

plt.tight_layout()
plt.show()
````
