To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# sobel kernels
sobel_x = [[-1, 0, 1],
      [-2, 0, 2],
      [-1, 0, 1]]

sobel_y = [[ 1,  2,  1],
      [ 0,  0,  0],
      [-1, -2, -1]]

def Ix_Iy(img):
    # convert image to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = gray.astype(float)

    H = gray.shape[0]
    W = gray.shape[1]

    Ix = np.zeros((H, W))
    Iy = np.zeros((H, W))

    # apply sobel using loops


    for i in range(1, H-1):
        for j in range(1, W-1):
            region = gray[i-1:i+2, j-1:j+2]
            Ix[i, j] = np.sum(region * sobel_x)
            Iy[i, j] = np.sum(region * sobel_y)


    # show Ix
    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix")
    plt.axis("off")


    # show Iy
    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy")
    plt.axis("off")


    # scatter plot Iy vs Ix
    plt.subplot(1,3,3)
    plt.scatter(Ix.flatten(), Iy.flatten(), s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix")
    plt.show()

    # return values for next parts
    return Ix,Iy



# for flat image
print("flat image")
Ix_Iy(flat)

# for edge image
print("edge image")
Ix_Iy(edge)

# for corner image
print("corner image")
Ix_Iy(corner)






````

````python
import numpy as np
import matplotlib.pyplot as plt

def R_map(img):
    Ix,Iy=Ix_Iy(img)
    H = Ix.shape[0]
    W= Ix.shape[1]

    R = np.zeros((H, W))

    k = 0.04   # constant
    win = 1    # 3x3 window

    for i in range(win, H-win):
        for j in range(win, W-win):

            # take small region
            ix = Ix[i-win:i+win+1, j-win:j+win+1]
            iy = Iy[i-win:i+win+1, j-win:j+win+1]

            # using * and sum
            a = np.sum(ix * ix)
            b = np.sum(ix * iy)
            c = np.sum(iy * iy)

            # trace and determinant
            trace = a + c
            det = a * c - b * b

            # eigenvalues
            value = trace*trace - 4*det
            if value < 0:
                value = 0

            lambda1 = (trace + np.sqrt(value)) / 2
            lambda2 = (trace - np.sqrt(value)) / 2

            # R value
            R[i,j] = lambda1 * lambda2 - k * (trace * trace)

    # show R map
    plt.imshow(R, cmap='gray')
    plt.title("R Map")
    plt.axis("off")
    plt.show()



 #for flat image
print("flat image")
R_map(flat)

# for edge image
print("edge image")
R_map(edge)

# for corner image
print("corner image")
R_map(corner)




````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def hough_transform(img):
    # convert to gray
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # canny edge detection
    edges = cv2.Canny(gray, 100, 200)

    H, W = edges.shape

    # limits for slope and intercept (not too large)
    slopes = np.linspace(-1, 1, 100)
    intercepts = np.linspace(-W, W, 200)

    # accumulator
    accumulator = np.zeros((len(slopes), len(intercepts)))

    # voting
    for y in range(H):
        for x in range(W):
            if edges[y,x] > 0:
                for i in range(len(slopes)):
                    m = slopes[i]
                    c = y - m * x

                    # check bounds
                    if c >= -W and c <= W:
                        j = np.argmin(np.abs(intercepts - c))
                        accumulator[i,j] += 1

    # normalize accumulator
    accumulator = accumulator / np.max(accumulator)

    # threshold (90 percent)
    threshold = 0.9
    print("Threshold used:", threshold)

    # draw detected lines
    img_out = img.copy()

    for i in range(len(slopes)):
        for j in range(len(intercepts)):
            if accumulator[i,j] > threshold:
                m = slopes[i]
                c = intercepts[j]

                x1 = 0
                y1 = int(m * x1 + c)
                x2 = W
                y2 = int(m * x2 + c)

                cv2.line(img_out, (x1,y1), (x2,y2), (0,0,255), 1)
                img_out=cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB)
    # show result
    plt.imshow(img_out)
    plt.title("Hough Lines (Slope-Intercept)")
    plt.axis("off")
    plt.show()



hough_transform(ball)
````

````python
# using rho theta
import numpy as np
import cv2
import matplotlib.pyplot as plt
def hough_rho_theta(img):
    # convert to gray
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # canny edge detection
    edges = cv2.Canny(gray, 100, 200)

    H, W = edges.shape
    diag = int(np.sqrt(H*H + W*W))

    rhos = np.arange(-diag, diag)
    thetas = np.deg2rad(np.arange(-90, 90))

    # accumulator
    accumulator = np.zeros((len(rhos), len(thetas)))

    # voting
    for y in range(H):
        for x in range(W):
            if edges[y,x] > 0:
                for i in range(len(thetas)):
                    rho = int(x*np.cos(thetas[i]) + y*np.sin(thetas[i])) + diag
                    accumulator[rho,i] += 1

    # normalize
    accumulator = accumulator / np.max(accumulator)

    # threshold
    threshold = 0.9
    print("Threshold used:", threshold)

    # draw lines
    img_out = img.copy()

    for r in range(len(rhos)):
        for i in range(len(thetas)):
            if accumulator[r,i] > threshold:
                rho = rhos[r]
                theta = thetas[i]

                a = np.cos(theta)
                b = np.sin(theta)

                x0 = a * rho
                y0 = b * rho

                x1 = int(x0 + 1000*(-b))
                y1 = int(y0 + 1000*(a))
                x2 = int(x0 - 1000*(-b))
                y2 = int(y0 - 1000*(a))

                cv2.line(img_out, (x1,y1), (x2,y2), (255,0,0), 1)
                img_out=cv2.cvtColor(img_out, cv2.COLOR_BGR2RGB)

    # show result
    plt.imshow(img_out)
    plt.title("Hough Lines (Rho-Theta)")
    plt.axis("off")
    plt.show()



hough_rho_theta(ball)



````
