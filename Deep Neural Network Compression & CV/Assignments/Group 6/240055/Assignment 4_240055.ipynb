To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def convolve2d(image, kernel):
    """
    Simple 2D convolution using numpy.
    """
    kernel = np.flipud(np.fliplr(kernel))
    output = np.zeros_like(image)


    pad_h = kernel.shape[0] // 2
    pad_w = kernel.shape[1] // 2
    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='constant')

    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            region = padded_image[y:y+kernel.shape[0], x:x+kernel.shape[1]]
            output[y, x] = (kernel * region).sum()

    return output

def part_a_gradients(image, title="Image"):
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    else:
        gray = image
    gray = gray.astype(np.float32) / 255.0
    Kx = np.array([[-1, 0, 1],
                   [-2, 0, 2],
                   [-1, 0, 1]], dtype=np.float32)

    Ky = np.array([[-1, -2, -1],
                   [ 0,  0,  0],
                   [ 1,  2,  1]], dtype=np.float32)
    Ix = cv2.filter2D(gray, -1, Kx)
    Iy = cv2.filter2D(gray, -1, Ky)
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title(f"{title} - Ix (Gradient X)")
    plt.imshow(Ix, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.title(f"{title} - Iy (Gradient Y)")
    plt.imshow(Iy, cmap='gray')
    plt.axis('off')
    plt.subplot(1, 3, 3)
    plt.title(f"{title} - Gradient Distribution")
    flat_Ix = Ix.flatten()
    flat_Iy = Iy.flatten()
    threshold = 0.1
    mask = (np.abs(flat_Ix) > threshold) | (np.abs(flat_Iy) > threshold)
    plt.scatter(flat_Ix[mask], flat_Iy[mask], alpha=0.1, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.grid(True)
    plt.show()
    return Ix, Iy
print("--- FLAT IMAGE ---")
Ix_f, Iy_f = part_a_gradients(flat, "Flat")

print("--- EDGE IMAGE ---")
Ix_e, Iy_e = part_a_gradients(edge, "Edge")

print("--- CORNER IMAGE ---")
Ix_c, Iy_c = part_a_gradients(corner, "Corner")
````

````python
def hough_slope_intercept(image, threshold_vote=100):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(image, 50, 150)
    height, width = edges.shape
    m_min, m_max = -5, 5
    m_step = 0.1
    m_values = np.arange(m_min, m_max, m_step)
    c_min, c_max = -height, width + height
    c_step = 1
    c_values = np.arange(c_min, c_max, c_step)
    accumulator = np.zeros((len(m_values), len(c_values)), dtype=int)
    y_idxs, x_idxs = np.nonzero(edges)
    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for m_idx, m in enumerate(m_values):
            c = y - m * x
            if c_min <= c < c_max:
                c_idx = int((c - c_min) / c_step)
                if 0 <= c_idx < len(c_values):
                    accumulator[m_idx, c_idx] += 1

    result_img = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    peaks = np.argwhere(accumulator > threshold_vote)

    for m_idx, c_idx in peaks:
        m = m_values[m_idx]
        c = c_values[c_idx]
        p1 = (0, int(c))
        p2 = (width, int(m*width + c))
        cv2.line(result_img, p1, p2, (0, 255, 0), 1)

    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    plt.title("Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.subplot(1,2,2)
    plt.title(f"Hough Slope-Intercept (Thresh={threshold_vote})")
    plt.imshow(result_img)
    plt.show()
hough_slope_intercept(edge, threshold_vote=60)
````

````python
def hough_rho_theta(image, threshold_vote=100):
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(image, 50, 150)

    height, width = edges.shape
    diag_len = int(np.sqrt(height**2 + width**2))
    rhos = np.arange(-diag_len, diag_len, 1)
    thetas = np.deg2rad(np.arange(0, 180, 1))
    accumulator = np.zeros((len(rhos), len(thetas)), dtype=int)

    y_idxs, x_idxs = np.nonzero(edges)
    print("Voting in Rho-Theta Accumulator...")
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for t_idx in range(len(thetas)):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            rho_idx = int(rho + diag_len)
            accumulator[rho_idx, t_idx] += 1
    result_img = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    peaks = np.argwhere(accumulator > threshold_vote)

    for rho_idx, t_idx in peaks:
        rho = rhos[rho_idx]
        theta = thetas[t_idx]

        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(result_img, (x1, y1), (x2, y2), (0, 0, 255), 1)

    plt.figure(figsize=(10,5))
    plt.title(f"Hough Rho-Theta (Thresh={threshold_vote})")
    plt.imshow(result_img)
    plt.show()

hough_rho_theta(edge, threshold_vote=60)
````
