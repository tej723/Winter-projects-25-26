To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def sobel_corner_analysis(img, threshold=30):


    # convert to float
    img = img.astype(np.float32)

   #here is the sobel kernel to go pixel by pixel for thr detction
    sobel_x=np.array([[-1, 0, 1],
                      [-2, 0, 2],
                      [-1, 0, 1]],dtype=np.float32)

    sobel_y=np.array([[-1, -2, -1],
                      [ 0,  0,  0],
                      [ 1,  2,  1]],dtype=np.float32)

    # gradients
    Ix = cv2.filter2D(img,-1,sobel_x)
    Iy = cv2.filter2D(img,-1,sobel_y)

    # gradient magnitude
    mag=np.sqrt(Ix**2+Iy**2)

    # threshold mask
    mask=mag>threshold
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix (Sobel X)")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy (Sobel Y)")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.scatter(Ix[mask], Iy[mask], s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix (threshold = {threshold})")
    plt.scatter(Ix.flatten(), Iy.flatten())

    plt.tight_layout()
    plt.show()

    return Ix, Iy
````

````python

def r_map(gray,size=5,k=0.04):
  gray=gray.astype(np.float32)
  Ix=cv2.Sobel(gray,cv2.CV_32F,1,0,ksize=3)
  Iy=cv2.Sobel(gray,cv2.CV_32F,0,1,ksize=3)
  Ix2=Ix*Ix
  Iy2=Iy*Iy
  Ixy=Ix*Iy
  a=cv2.boxFilter(Ix2,-1,(size,size))
  b=cv2.boxFilter(Iy2,-1,(size,size))
  c=cv2.boxFilter(Ixy,-1,(size,size))
  det=a*c-b*b
  trace=a+c
  R=det-k*(trace**2)
  return R
````

````python
def show_r_map(gray, R, title):
    plt.figure(figsize=(10,4))

    plt.subplot(1,2,1)
    plt.imshow(gray, cmap='gray')
    plt.title("Original Image")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(R, cmap='gray')
    plt.title(title)
    plt.axis("off")

    plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img_n = cv2.imread("your_image.jpg.png")   # change path if needed
if img_n is None:
    raise ValueError("image not found")

img_n = cv2.cvtColor(img_n, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img_n, cv2.COLOR_RGB2GRAY)


def hough_function1(img):
    # canny edge detection
    edges = cv2.Canny(img,100,200)
    h,w = edges.shape

    # slope and intercept ranges
    range_m = np.linspace(-10,10,400)
    range_c = np.linspace(-2*h,2*h,400)

    # accumulator
    accumulate = np.zeros((len(range_m),len(range_c)))

    # edge points
    y_pts,x_pts = np.where(edges>0)

    # voting
    for i in range(len(x_pts)):
        x = x_pts[i]
        y = y_pts[i]
        for m_id in range(len(range_m)):
            m = range_m[m_id]
            c = y - (m*x)
            if range_c[0] <= c <= range_c[-1]:
                c_idx = np.argmin(np.abs(range_c-c))
                accumulate[m_id,c_idx] += 1

    # thresholding
    thresh = np.max(accumulate)*0.6
    m_win,c_win = np.where(accumulate > thresh)

    # draw detected lines
    output_img = cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
    for i in range(len(m_win)):
        m = range_m[m_win[i]]
        c = range_c[c_win[i]]
        x1,x2 = 0,w
        y1 = int(m*x1 + c)
        y2 = int(m*x2 + c)
        cv2.line(output_img,(x1,y1),(x2,y2),(0,255,0),2)

    # show results
    plt.figure(figsize=(15,7))
    plt.subplot(1,2,1)
    plt.imshow(edges,cmap="gray")
    plt.title("canny edge method")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(output_img)
    plt.title("detected lines")
    plt.axis("off")
    plt.show()

    return accumulate


acc = hough_function1(gray_img)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img_n = cv2.imread("your_image.jpg.png")   # change path if needed
if img_n is None:
    raise ValueError("image not found")

img_n = cv2.cvtColor(img_n, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img_n, cv2.COLOR_RGB2GRAY)


def hough_rho_theta(img):
    # canny edge detection
    edges = cv2.Canny(img,100,200)
    h,w = edges.shape

    # rho and theta ranges
    diag = int(np.sqrt(h*h + w*w))
    rhos = np.linspace(-diag,diag,400)
    thetas = np.linspace(-np.pi/2,np.pi/2,400)

    # accumulator
    accumulate = np.zeros((len(rhos),len(thetas)))

    # edge points
    y_pts,x_pts = np.where(edges>0)

    # precompute trig
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    # voting
    for i in range(len(x_pts)):
        x = x_pts[i]
        y = y_pts[i]
        for t_id in range(len(thetas)):
            rho = x*cos_t[t_id] + y*sin_t[t_id]
            r_id = np.argmin(np.abs(rhos-rho))
            accumulate[r_id,t_id] += 1

    # thresholding
    thresh = np.max(accumulate)*0.6
    r_win,t_win = np.where(accumulate > thresh)

    # draw detected lines
    output_img = cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
    for i in range(len(r_win)):
        rho = rhos[r_win[i]]
        theta = thetas[t_win[i]]
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(output_img,(x1,y1),(x2,y2),(255,0,0),2)

    # show results
    plt.figure(figsize=(15,7))
    plt.subplot(1,2,1)
    plt.imshow(edges,cmap="gray")
    plt.title("canny edge method")
    plt.axis("off")

    plt.subplot(1,2,2)
    plt.imshow(output_img)
    plt.title("rho theta detected lines")
    plt.axis("off")
    plt.show()

    return accumulate

acc = hough_rho_theta(gray_img)
````
