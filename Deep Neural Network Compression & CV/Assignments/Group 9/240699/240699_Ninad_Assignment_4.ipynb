To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python

def compute_harris_manual_eigenvalues(image, k=0.04, window_size=3):
    height, width = image.shape
    
    #Gradients
    Ix = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    
    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy
    R_map = np.zeros((height, width))
    offset = window_size // 2

    for y in range(offset, height - offset):
        for x in range(offset, width - offset):
            
            #Windows
            window_Ixx = Ixx[y-offset : y+offset+1, x-offset : x+offset+1]
            window_Iyy = Iyy[y-offset : y+offset+1, x-offset : x+offset+1]
            window_Ixy = Ixy[y-offset : y+offset+1, x-offset : x+offset+1]
            
            #Summation of windows
            a = np.sum(window_Ixx) # Sum of Ix^2 
            b = np.sum(window_Ixy) # Sum of IxIy 
            c = np.sum(window_Iyy) # Sum of Iy^2 
            
            #Eigenvalues using the Expanded Formula
            # Formula: lambda = ( (a+c) +/- sqrt( (a-c)^2 + 4b^2 ) ) / 2
            trace_term = a + c
            # The part inside the square root
            sqrt_term = np.sqrt((a - c)**2 + 4 * (b**2))
            
            lambda1 = (trace_term + sqrt_term) / 2
            lambda2 = (trace_term - sqrt_term) / 2
            
            #Calculateing R using the Eigenvalues
            r = (lambda1 * lambda2) - k * ((lambda1 + lambda2)**2)
            R_map[y, x] = r
            
    return R_map



R_flat= compute_harris_manual_eigenvalues(flat_g, window_size=3)
R_edge= compute_harris_manual_eigenvalues(edge_g, window_size=3)
R_corner= compute_harris_manual_eigenvalues(corner_g, window_size=3)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(R_flat, cmap='gray')
plt.title("for flat image")
plt.axis('off')
plt.subplot(1,3,2)
plt.imshow(R_edge, cmap='gray')
plt.title("for edge image")
plt.axis('off')
plt.subplot(1,3,3)
plt.imshow(R_corner, cmap='gray')
plt.title("for corner image")
plt.axis('off')
plt.show()
````

````python

def manual_hough_slope_intercept(image, threshold_percent=0.9):
    gray_1 =  cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    rows, cols = gray_1.shape
    displ_img = image.copy()
    # Canny Edge Map
    edges = cv2.Canny(gray_1, 50, 150)

    # Accumulator Ranges
    m_min, m_max = -100.0, 100.0
    m_step = 0.1
    num_m = int(np.ceil((m_max - m_min) / m_step))

    # Intercept (c): c = y - mx. Range depends on image size and max slope.
    c_min = -cols * 5  # Allow for steep negative slopes #i saw this calculations online
    c_max = rows + cols * 5
    c_step = 2.0 # Step of 2 pixels for speed
    num_c = int(np.ceil((c_max - c_min) / c_step))

    #Initializing Accumulator
    accumulator = np.zeros((num_m, num_c), dtype=np.uint64)
    #Edge Pixels
    y_idxs, x_idxs = np.nonzero(edges)
    
    # Voting Loop
    for i in range(len(x_idxs)):
        x= x_idxs[i]
        y= y_idxs[i]

        for m_idx in range(num_m):
            m= m_min + m_idx * m_step
            
            # c = y - mx
            c = y - (m * x)

            # Check bounds
            if c_min<= c< c_max:
                c_idx= int((c - c_min) / c_step)
                if 0<= c_idx < num_c:
                    accumulator[m_idx, c_idx]+= 1

    

    # Thresholding (90%)
    max_vote= np.max(accumulator)
    vote_thresh= max_vote* threshold_percent
    # Extracting peaks
    m_bins, c_bins= np.where(accumulator > vote_thresh)

    # i searched for this calculations online as well as on Ai
    for i in range(len(m_bins)):
        m= m_min+ m_bins[i]* m_step
        c= c_min+c_bins[i]*c_step
        
        #start (x=0) and end (x=cols) points
        y1= int(c)
        y2= int(m * cols + c)
        cv2.line(displ_img, (0, y1), (cols, y2), (0, 0, 255), 2)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(displ_img, cv2.COLOR_BGR2RGB))
    plt.title(f"Detected Lines (m-c method)")
    plt.axis('off')
    plt.show()

manual_hough_slope_intercept(new_img)
````

````python
def hough_rho_theta_fast(image, threshold_votes=150):
    gray_2 = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    dis_img = image.copy()
    #Canny Edge Detection
    edges = cv2.Canny(gray_2, 50, 150)
    #OpenCV HoughLines
    #rho accuracy (1 pixel)
    # np.pi/180: theta accuracy (1 degree)
    # threshold_votes
    lines = cv2.HoughLines(edges, 1, np.pi/180, threshold_votes)

    
    for line in lines:
        rho, theta= line[0]
        # Converting polar to cartesian for drawing
        a= np.cos(theta)
        b= np.sin(theta)
        x0= a* rho
        y0= b* rho
            
        # Extending the line to be visible across the image
        x1= int(x0+ 2000* (-b))
        y1= int(y0+ 2000* (a))
        x2= int(x0- 2000* (-b))
        y2= int(y0- 2000* (a))
        cv2.line(dis_img, (x1, y1), (x2, y2), (0, 255, 0), 2)
    

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.axis('off')
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(dis_img, cv2.COLOR_BGR2RGB))
    plt.title("Detected Lines (Rho-Theta)")
    plt.axis('off')
    plt.show()

hough_rho_theta_fast(new_img)
#comparing time between two methods, the slope-intercept method took more than a minute for detection
#while the rho-theta took less than a second to process same image.
````
