To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def R_map_plot(Ix,Iy,window_size):

  row,col=Ix.shape
  pad=window_size//2 #This is so we dont choose centers whose window goes outside our image

  R_map=np.zeros((row,col),dtype=np.float32)

  for i in range(pad,row-pad):
    for j in range(pad,col-pad):

      a=0
      b=0
      c=0

      for k in range(-pad,pad+1):
        for l in range(-pad,pad+1):

          a+=Ix[i+k,j+l]*Ix[i+k,j+l]
          b+=Ix[i+k,j+l]*Iy[i+k,j+l]
          c+=Iy[i+k,j+l]*Iy[i+k,j+l]

      #Eigen value calculation done below:
      det=a*c-b*b
      trace=a+c

      R=det-0.04*(trace*trace)
      R_map[i,j]=R


  plt.figure()
  plt.imshow(R_map,cmap='gray')
  plt.axis('off')
  plt.show()

R_map_plot(flat_ix,flat_iy,3)
R_map_plot(edge_ix,edge_iy,3)
R_map_plot(corner_ix,corner_iy,3)
````

````python
#Transform using slope intercept form (m,b)

def hough_mb(img,threshold):
    img=np.array(img)
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    edge_map=cv2.Canny(gray_img,100,200)

    l,w=edge_map.shape

    m_min,m_max=-10,10 #For max and min slope
    b_min,b_max=-l,l #For max and min intercept

    num_m=1000
    num_b=1000
    m_vals=np.linspace(m_min,m_max,num_m)
    b_vals=np.linspace(b_min,b_max,num_b)

    accum=np.zeros((num_m,num_b),dtype=np.float32)

    #We don't want to waste time looping over non-edge pixels, hence these arrays will store only edge pixel locations
    row_idx,col_idx=np.where(edge_map>0)

    for idx in range(len(row_idx)):
        y=row_idx[idx]
        x=col_idx[idx]

        for m_idx in range(len(m_vals)):
            b=y-m_vals[m_idx]*x

            if b<b_min or b>b_max:
                continue

            b_idx=int((b-b_min)/(b_max-b_min)*(num_b-1))
            accum[m_idx,b_idx]+=1

    #normalize accum map
    accum=accum/np.max(accum)

    #Finding valid lines according to threshold
    lines=np.where(accum>threshold)

    #We will put edges onto copy of input image
    temp_img=img.copy()

    for i in range(len(lines[0])):
        m_idx=lines[0][i]
        b_idx=lines[1][i]

        m=m_vals[m_idx]
        b=b_vals[b_idx]

        x1=0
        y1=int(b)
        x2=w
        y2=int(m*w+b)

        cv2.line(temp_img,(x1,y1),(x2,y2),(0,0,255),1)

    plt.imshow(temp_img)
    plt.axis('off')

hough_mb(rubiks,0.5)
````

````python
#Code for r,theta implementation.
#Code is mostly similar to m,b implementation, with a few tweaks
#We can clearly see r,theta can find all edges in the image unlike m,b
def hough_r_theta(img,threshold):
    img=np.array(img)
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    edge_map=cv2.Canny(gray_img,100,200)

    l,w=edge_map.shape

    r_max=int(np.sqrt(l*l+w*w))
    r_min=-r_max

    theta_min=-np.pi/2
    theta_max=np.pi/2

    num_r=1000
    num_theta=1000

    r_vals=np.linspace(r_min,r_max,num_r)
    theta_vals=np.linspace(theta_min,theta_max,num_theta)

    accum=np.zeros((num_r,num_theta),dtype=np.float32)
    row_idx,col_idx=np.where(edge_map>0)

    for idx in range(len(row_idx)):
        y=row_idx[idx]
        x=col_idx[idx]

        for t_idx in range(len(theta_vals)):
            r=x*np.cos(theta_vals[t_idx])+y*np.sin(theta_vals[t_idx])

            if r<r_min or r>r_max:
                continue

            r_idx=int((r-r_min)/(r_max-r_min)*(num_r-1))
            accum[r_idx,t_idx]+=1

    #normalize accum map
    accum=accum/np.max(accum)

    #Finding valid lines according to threshold
    lines=np.where(accum>threshold)

    #We will put edges onto copy of input image
    temp_img=img.copy()

    for i in range(len(lines[0])):
        r_idx=lines[0][i]
        t_idx=lines[1][i]

        r=r_vals[r_idx]
        theta=theta_vals[t_idx]

        a=np.cos(theta)
        b=np.sin(theta)

        x0=a*r
        y0=b*r

        x1=int(x0+1000*(-b))
        y1=int(y0+1000*(a))
        x2=int(x0-1000*(-b))
        y2=int(y0-1000*(a))

        cv2.line(temp_img,(x1,y1),(x2,y2),(0,0,255),1)

    plt.imshow(temp_img)
    plt.axis('off')

hough_r_theta(rubiks,0.67)
````

````python
#Stencil creator using manually found coords code block

def create_stencil(ball,x1,y1,x2,y2):

    crop=ball[y1:y2,x1:x2]
    gray=cv2.cvtColor(crop,cv2.COLOR_BGR2GRAY)
    edge_map=cv2.Canny(gray,100,200)

    #using cv2 findcountour function documentation
    contours,_=cv2.findContours(edge_map,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

    max_val=contours[0]
    max_area=0

    #We need to find max area contour
    for cnt in contours:
        area=cv2.contourArea(cnt)
        if area>max_area:
            max_area=area
            max_val=cnt

    #Making binary mask to use as stencil
    mask=np.zeros(gray.shape,dtype=np.uint8)
    cv2.drawContours(mask,[max_val],-1,255,-1)

    length,width=mask.shape

    #Stencil will have 4 channels,  r g b and one alpha channel
    stencil=np.zeros((length,width,4),dtype=np.uint8)

    stencil[:,:,0]=crop[:,:,0]
    stencil[:,:,1]=crop[:,:,1]
    stencil[:,:,2]=crop[:,:,2]
    stencil[:,:,3]=mask

    return stencil

ball_stencil=create_stencil(ball,105,188,240,325)
````
