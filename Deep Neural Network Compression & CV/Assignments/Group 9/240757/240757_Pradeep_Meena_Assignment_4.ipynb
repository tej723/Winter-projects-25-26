To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def analyze_gradients(img_rgb, label):
    # Manual grayscale conversion
    gray = 0.299*img_rgb[:,:,0] + 0.587*img_rgb[:,:,1] + 0.114*img_rgb[:,:,2]

    # Sobel Kernels
    Sx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    Sy = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

    h, w = gray.shape
    Ix = np.zeros_like(gray)
    Iy = np.zeros_like(gray)

    # Manual Convolution with padding
    padded = np.pad(gray, 1, mode='edge')
    for r in range(h):
        for c in range(w):
            window = padded[r:r+3, c:c+3]
            Ix[r, c] = np.sum(window * Sx)
            Iy[r, c] = np.sum(window * Sy)

    # Plotting Gradients
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    axes[0].imshow(Ix, cmap='gray'); axes[0].set_title(f"Ix - {label}")
    axes[1].imshow(Iy, cmap='gray'); axes[1].set_title(f"Iy - {label}")
    plt.show()

    # Scatter Plot
    plt.figure(figsize=(5, 5))
    plt.scatter(Ix.flatten(), Iy.flatten(), alpha=0.3, s=2)
    plt.title(f"Scatter Plot (Iy vs Ix) for {label}")
    plt.xlabel("Ix"); plt.ylabel("Iy")
    plt.grid(True)
    plt.show()

    return Ix, Iy

# Execution
ix_f, iy_f = analyze_gradients(img_flat, "Flat")
ix_e, iy_e = analyze_gradients(img_edge, "Edge")
ix_c, iy_c = analyze_gradients(img_corner, "Corner")
````

````python
def get_harris_rmap(Ix, Iy, window_size=3, k=0.04):
    h, w = Ix.shape
    R_map = np.zeros((h, w))


    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy

    offset = window_size // 2
    for y in range(offset, h - offset):
        for x in range(offset, w - offset):
 # Sum of squared gradients in the local window
            Sxx = np.sum(Ixx[y-offset:y+offset+1, x-offset:x+offset+1])
            Syy = np.sum(Iyy[y-offset:y+offset+1, x-offset:x+offset+1])
            Sxy = np.sum(Ixy[y-offset:y+offset+1, x-offset:x+offset+1])

            # Determinant and Trace of M
            det_M = (Sxx * Syy) - (Sxy**2)
            trace_M = Sxx + Syy

            # Harris Response calculation
            R_map[y, x] = det_M - k * (trace_M**2)

    return R_map

# Plotting the Corner R-Map
corner_r = get_harris_rmap(ix_c, iy_c)
plt.imshow(corner_r > (0.1 * corner_r.max()), cmap='gray')
plt.title("Detected Corners (R-Map Thresholded)")
plt.show()
````

````python
def hough_transform(img_rgb):
    # Grayscale and basic edge detection
    gray = np.mean(img_rgb, axis=2)
    dy, dx = np.gradient(gray)
    mag = np.sqrt(dx**2 + dy**2)
    edges = mag > 50

    h, w = edges.shape
    diagonal = int(np.sqrt(h**2 + w**2))
    thetas = np.deg2rad(np.arange(-90, 90))

    # Accumulator: rows are Rho , cols are Theta
    accumulator = np.zeros((2 * diagonal, len(thetas)))

    y_points, x_points = np.where(edges)

    for i in range(len(x_points)):
        x, y = x_points[i], y_points[i]
        for t_idx, th in enumerate(thetas):
            rho = int(x * np.cos(th) + y * np.sin(th))
            accumulator[rho + diagonal, t_idx] += 1

    return accumulator

# Display result for the Multiple Corners image
acc_map = hough_transform(img_corner2)
plt.imshow(acc_map, aspect='auto', cmap='hot')
plt.title("Hough Accumulator (Polar)")
plt.xlabel("Theta Index"); plt.ylabel("Rho")
plt.show()
````

````python
def process_ball_transfer(src, dest, box_coords, target_center):
    x1, y1, x2, y2 = box_coords
    x3, y3 = target_center

    # 1. Crop ball and create alpha mask based on brightness
    ball_crop = src[y1:y2, x1:x2].astype(float)
    mask = (np.mean(ball_crop, axis=2) > 180).astype(float)

    # 2. Build 4-channel Stencil
    h, w = mask.shape
    stencil = np.zeros((h, w, 4))
    stencil[:,:,:3] = ball_crop / 255.0
    stencil[:,:,3] = mask

    # 3. Blending onto target
    output = dest.copy().astype(float) / 255.0
    half_h, half_w = h // 2, w // 2

    # Target region on the green table
    reg_y1, reg_y2 = y3 - half_h, y3 - half_h + h
    reg_x1, reg_x2 = x3 - half_w, x3 - half_w + w

    bg_region = output[reg_y1:reg_y2, reg_x1:reg_x2]
    alpha = stencil[:,:,3][..., None]

    # Blending: New = (Foreground * Alpha) + (Background * (1 - Alpha))
    blended = (stencil[:,:,:3] * alpha) + (bg_region * (1 - alpha))
    output[reg_y1:reg_y2, reg_x1:reg_x2] = blended

    plt.imshow(output)
    plt.title("Ball Placed on Green Table")
    plt.axis('off')
    plt.show()
````
