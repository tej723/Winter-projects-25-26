To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def rmap_function(img,k,w_s): #w_s is window_size
  img_f = img.astype(np.float32)
  sobelx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
  sobely = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
  #finding ix,iy
  ix = cv2.filter2D(img_f,-1,sobelx)
  iy = cv2.filter2D(img_f,-1,sobely)

  #finding the squared gradients
  ixx = ix**2
  iyy = iy**2
  ixy = ix*iy

  #height and width of the image
  h,w = img.shape
  r_map = np.zeros((h,w))
  off = w_s//2

  #iterating over the pixels
  for i in range(off,h-off):
    for j in range(off,w-off):
      #finding the window
      win_ixx = ixx[i-off:i+off+1,j-off:j]
      win_iyy = iyy[i-off:i+off+1,j-off:j]
      win_ixy = ixy[i-off:i+off+1,j-off:j]

      #finding a,b,c
      a = np.sum(win_ixx)
      b = np.sum(win_iyy)
      c = np.sum(win_ixy)

      tr = a+b #trace
      det = a*b - c**2 #determinant

      temp1 = np.sqrt((a-b)**2 + 4*(c**2))  #using the quadratic formula for finding the eigen values
      l1 = (tr+temp1)/2 #lembda1
      l2 = (tr-temp1)/2 #lembda2

      r = det - k*(tr**2)
      r_map[i,j] = r

  #plotting the obtained results
  plt.figure(figsize=(13,5))
  plt.subplot(1,2,1)
  plt.imshow(img,cmap='gray')
  plt.title('original greyscale img')
  plt.axis('off')

  plt.subplot(1,2,2)
  plt.imshow(r_map,cmap='gray')
  plt.title("R - Map")
  plt.axis('off')

  plt.show()
  return r_map

flat_rmap = rmap_function(gray_flat,0.06,5)
edge_rmap = rmap_function(gray_edge,0.06,5)
corner2_rmap = rmap_function(gray_corner2,0.06,5)



````

````python
gray_img = cv2.cvtColor(img_n, cv2.COLOR_RGB2GRAY)  #conversion to greyscale

def hough_function1(img):   #using the slope-intercept method
    edges = cv2.Canny(img,100,200) #using threshold range from 100 to 200
    h, w = edges.shape
    range_m = np.linspace(-10,10,1500)
    range_c = np.linspace(-h,h,1500)
    accumulate = np.zeros((len(range_m),len(range_c)))
    y_pts, x_pts = np.where(edges>0)

    for i in range(len(x_pts)):    #iterating
        x = x_pts[i]
        y = y_pts[i]
        for m_id in range(len(range_m)):
            m = range_m[m_id]
            c = y-(m*x)   #equation of line c=y-mx
            #checking the range of c
            if range_c[0]<=c<=range_c[-1]:
                #finding the idx in accumulator
                c_idx = np.argmin(np.abs(range_c-c))
                accumulate[m_id,c_idx] += 1

    thresh = 0.4*np.max(accumulate)  #thresholding
    m_win, c_win = np.where(accumulate>thresh)
    output_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)  #drawing the detected lines
    for i in range(len(m_win)):
        m = range_m[m_win[i]]
        c = range_c[c_win[i]]
        #calculating starting and end point of the line
        x1,x2 = 0,w
        y1 = int(m*x1+c)
        y2 = int(m*x2+c)
        #drawing the line
        cv2.line(output_img,(x1,y1),(x2,y2), (0,255,0),2)

    #showing the results
    plt.figure(figsize=(15,7))
    plt.subplot(1,2,1)
    plt.imshow(edges, cmap='gray')
    plt.title("canny edge method")

    plt.subplot(1,2,2)
    plt.imshow(output_img)
    plt.title("detected lines")

    plt.show()
    return accumulate

acc = hough_function1(gray_img)
#some verticle edges are unable to detect using this method
````

````python
def hough_function2(img):   #using the rho-theta method
    edges = cv2.Canny(img,50,150)
    h,w = edges.shape
    diag_len = int(np.sqrt(h**2+w**2))
    range_theta = np.linspace(0, np.pi, 180) #theta :0 to 180 deg
    range_rho = np.linspace(-diag_len, diag_len,6*diag_len)
    accumulate = np.zeros((len(range_rho),len(range_theta)))
    y_pts, x_pts = np.where(edges>0)
    for i in range(len(x_pts)):    #iterating through edge pts
        x = x_pts[i]
        y = y_pts[i]
        for t_id in range(len(range_theta)):
            theta = range_theta[t_id]
            #equation of line : rho=x*cos(theta)+y*sin(theta)
            rho = x*np.cos(theta)+y*np.sin(theta)
            #finding the idx in accumulator
            #since rho ranges from neg to pos, we match it to range_rho
            rho_idx = np.argmin(np.abs(range_rho-rho))
            accumulate[rho_idx,t_id]+=1

    thresh = np.max(accumulate)*0.4  #thresholding
    rho_win, theta_win = np.where(accumulate > thresh)
    output_img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB) #drawing the detected lines

    for i in range(len(rho_win)):
        rho = range_rho[rho_win[i]]
        theta = range_theta[theta_win[i]]
        #converting polar coordinates back to cartesian to draw the line
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        #moving 2000 pixels in both directions along the line to find start/end points
        x1 = int(x0 + 2000*(-b))
        y1 = int(y0 + 2000*a)
        x2 = int(x0 - 2000*(-b))
        y2 = int(y0 - 2000*a)
        cv2.line(output_img,(x1,y1),(x2,y2),(0,255,0),2)

    # showing the results
    plt.figure(figsize=(15,7))
    plt.subplot(1,2,1)
    plt.imshow(edges, cmap='gray')
    plt.title("canny edge method")

    plt.subplot(1,2,2)
    plt.imshow(output_img)
    plt.title("detected lines (rho-theta)")

    plt.show()
    return accumulate

acc = hough_function2(gray_img)
````
