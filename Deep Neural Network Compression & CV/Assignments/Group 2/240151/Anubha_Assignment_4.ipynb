To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Then, to embed the string into your Jupyter Notebook :

````python
import numpy as np
import matplotlib.pyplot as plt


def harris_r_map(gray_img, Ix, Iy, window_size=5, k=0.04):

    h, w = gray_img.shape
    offset = window_size // 2

    R_map = np.zeros((h, w), dtype=np.float32)

    for y in range(offset, h - offset):
        for x in range(offset, w - offset):

            # Window extraction
            Ix_win = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            Iy_win = Iy[y-offset:y+offset+1, x-offset:x+offset+1]

            # Second moment values
            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            # Eigenvalues (explicit computation)
            trace = a + c
            det = (a * c) - (b ** 2)

            lambda1 = (trace + np.sqrt((a - c)**2 + 4*b*b)) / 2
            lambda2 = (trace - np.sqrt((a - c)**2 + 4*b*b)) / 2

            # Harris response
            R = det - k * (trace ** 2)

            R_map[y, x] = R

    return R_map
def show_r_map(R_map, title="R Map"):
    plt.figure(figsize=(5, 5))
    plt.imshow(R_map, cmap='gray')
    plt.colorbar()
    plt.title(title)
    plt.axis('off')
    plt.show()
# Flat image
R_flat = harris_r_map(grayf, Ix_f, Iy_f)
show_r_map(R_flat, "Harris R Map - Flat Image")

# Edge image
R_edge = harris_r_map(graye, Ix_e, Iy_e)
show_r_map(R_edge, "Harris R Map - Edge Image")

# Corner image
R_corner = harris_r_map(grayc1, Ix_c1, Iy_c1)
show_r_map(R_corner, "Harris R Map - Corner Image")
````

````python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import base64

# --------------------------------------------------
# Save image from base64 string
# --------------------------------------------------
img_data_decoded = base64.b64decode(b64)
with open("human-eye.jpg", "wb") as f:
    f.write(img_data_decoded)

# --------------------------------------------------
# Read image (grayscale)
# --------------------------------------------------
img = cv2.imread("human-eye.jpg", cv2.IMREAD_GRAYSCALE)

if img is None:
    raise ValueError("Image could not be loaded")

# --------------------------------------------------
# Canny edge detection
# --------------------------------------------------
edges = cv2.Canny(img, 100, 200)

plt.figure(figsize=(5,5))
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Map")
plt.axis('off')
plt.show()

# --------------------------------------------------
# Manual Hough Transform (mc form)
# --------------------------------------------------
def hough_mc(edges, m_range=(-2, 2), m_bins=400, c_bins=400):
    """
    Manual Hough Transform using slope-intercept form
    """

    h, w = edges.shape

    m_vals = np.linspace(m_range[0], m_range[1], m_bins)
    c_vals = np.linspace(-h, h, c_bins)

    accumulator = np.zeros((m_bins, c_bins), dtype=np.uint64)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x

            if c_vals[0] <= c <= c_vals[-1]:
                ci = int((c - c_vals[0]) / (c_vals[-1] - c_vals[0]) * (c_bins - 1))
                accumulator[mi, ci] += 1

    return accumulator, m_vals, c_vals

# --------------------------------------------------
# Compute accumulator
# --------------------------------------------------
acc_mc, m_vals, c_vals = hough_mc(edges)

# --------------------------------------------------
# Normalize accumulator (safe)
# --------------------------------------------------
if acc_mc.max() == 0:
    acc_norm = acc_mc
else:
    acc_norm = acc_mc / acc_mc.max()

# --------------------------------------------------
# Threshold
# --------------------------------------------------
threshold = 0.9
lines = np.where(acc_norm > threshold)

# --------------------------------------------------
# Draw detected lines
# --------------------------------------------------
img_lines = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

for mi, ci in zip(lines[0], lines[1]):
    m = m_vals[mi]
    c = c_vals[ci]

    x1, x2 = 0, img.shape[1]
    y1 = int(m * x1 + c)
    y2 = int(m * x2 + c)

    if 0 <= y1 < img.shape[0] and 0 <= y2 < img.shape[0]:
        cv2.line(img_lines, (x1, y1), (x2, y2), (0, 0, 255), 1)

plt.figure(figsize=(6,6))
plt.imshow(img_lines)
plt.title("Detected Lines (mc Hough)")
plt.axis('off')
plt.show()
````

````python
def hough_rho_theta(edges, theta_bins=180, rho_bins=300):
    h, w = edges.shape
    diag = int(np.sqrt(h*h + w*w))

    thetas = np.linspace(-np.pi/2, np.pi/2, theta_bins)
    rhos = np.linspace(-diag, diag, rho_bins)

    accumulator = np.zeros((rho_bins, theta_bins), dtype=np.uint64)

    ys, xs = np.nonzero(edges)

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    for x, y in zip(xs, ys):
        for ti in range(theta_bins):
            rho = x * cos_t[ti] + y * sin_t[ti]
            ri = int((rho + diag) / (2 * diag) * (rho_bins - 1))
            accumulator[ri, ti] += 1

    return accumulator, rhos, thetas

acc_rt, rhos, thetas = hough_rho_theta(edges)

# Normalize
acc_rt_norm = acc_rt / acc_rt.max()

# Threshold
threshold = 0.9
lines = np.where(acc_rt_norm > threshold)

img_lines_rt = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

for ri, ti in zip(lines[0], lines[1]):
    rho = rhos[ri]
    theta = thetas[ti]

    a = np.cos(theta)
    b = np.sin(theta)

    x0 = a * rho
    y0 = b * rho

    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    cv2.line(img_lines_rt, (x1,y1), (x2,y2), (0,255,0), 1)

plt.figure(figsize=(6,6))
plt.imshow(img_lines_rt)
plt.title("Detected Lines ( Hough)")
plt.axis('off')
plt.show()
````
