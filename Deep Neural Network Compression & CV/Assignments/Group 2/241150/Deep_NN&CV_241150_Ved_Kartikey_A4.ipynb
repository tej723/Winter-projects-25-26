To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def harris_corner_detection(image, k=0.04, window_size=3):

    Ix = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy

    Sxx = cv2.boxFilter(Ixx, -1, (window_size, window_size), normalize=False)
    Syy = cv2.boxFilter(Iyy, -1, (window_size, window_size), normalize=False)
    Sxy = cv2.boxFilter(Ixy, -1, (window_size, window_size), normalize=False)

    det_M = (Sxx * Syy) - (Sxy**2)
    trace_M = Sxx + Syy

    discriminant = np.sqrt(trace_M**2 - 4*det_M)

    lambda1 = (trace_M + discriminant) / 2
    lambda2 = (trace_M - discriminant) / 2

    R_map = (lambda1 * lambda2) - k * (lambda1 + lambda2)**2

    return R_map, lambda1, lambda2

def display_r_map(image, title="Image"):
    R, l1, l2 = harris_corner_detection(image)
    R_norm = cv2.normalize(R, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)

    plt.figure(figsize=(10, 4))

    # Plot Original
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title(f"Original: {title}")
    plt.axis('off')

    # Plot R-Map
    plt.subplot(1, 2, 2)
    plt.imshow(R_norm, cmap='gray')
    plt.title(f"R-Map (Corners in White)\n{title}")
    plt.axis('off')

    plt.tight_layout()
    plt.show()


display_r_map(flat, title="Flat Region")
display_r_map(edge, title="Edge Region")
display_r_map(corner, title="Corner Region")
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hough_line_slope_intercept(image):
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image

    edges = cv2.Canny(gray, 50, 150)
    height, width = edges.shape

    m_min, m_max = -5.0, 5.0
    m_step = 0.1
    m_range = np.arange(m_min, m_max, m_step)

    c_max = int(np.sqrt(height**2 + width**2)) * 2
    c_min = -c_max

    num_m = len(m_range)
    num_c = c_max - c_min
    accumulator = np.zeros((num_c, num_m), dtype=np.uint64)
    y_idxs, x_idxs = np.nonzero(edges)

    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for m_idx, m in enumerate(m_range):
            c = y - (m * x)
            if c_min <= c < c_max:
                c_idx = int(c - c_min)
                accumulator[c_idx, m_idx] += 1

    acc_norm = cv2.normalize(accumulator.astype('float'), None, 0, 255, cv2.NORM_MINMAX)
    threshold = 150
    strong_lines = np.argwhere(acc_norm > threshold)

    output_img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    for c_idx, m_idx in strong_lines:
        m = m_range[m_idx]
        c = c_idx + c_min # Map index back to real 'c' value

        # To draw the line, we need two points (x1, y1) and (x2, y2)
        # We pick x=0 and x=width
        x1 = 0
        y1 = int(m * x1 + c)
        x2 = width
        y2 = int(m * x2 + c)

        cv2.line(output_img, (x1, y1), (x2, y2), (0, 0, 255), 1)

    # Display
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.subplot(1, 2, 2)
    plt.imshow(output_img)
    plt.title(f"Slope-Intercept Hough (Thresh={threshold})")
    plt.show()

sudoku = cv2.imread('/content/sudoku_image.jpg')
hough_line_slope_intercept(sudoku)
````

````python
def hough_line_rho_theta(image):

    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image

    edges = cv2.Canny(gray, 50, 150)
    height, width = edges.shape
    diag_len = int(np.ceil(np.sqrt(height**2 + width**2)))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)

    thetas = np.deg2rad(np.arange(0, 180, 1.0))
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)
    num_thetas = len(thetas)

    accumulator = np.zeros((2 * diag_len, num_thetas), dtype=np.uint64)
    y_idxs, x_idxs = np.nonzero(edges)

    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        for t_idx in range(num_thetas):
            # Calculate rho = x cos(theta) + y sin(theta)
            rho = round(x * cos_t[t_idx] + y * sin_t[t_idx])
            rho_idx = int(rho + diag_len)
            accumulator[rho_idx, t_idx] += 1

    acc_norm = cv2.normalize(accumulator.astype('float'), None, 0, 255, cv2.NORM_MINMAX)
    threshold = 100
    strong_lines = np.argwhere(acc_norm > threshold)

    # 4. Draw Lines
    output_img = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

    for rho_idx, t_idx in strong_lines:
        rho = rhos[rho_idx]
        theta = thetas[t_idx]

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(output_img, (x1, y1), (x2, y2), (0, 255, 0), 1)

    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.subplot(1, 2, 2)
    plt.imshow(output_img)
    plt.title("Rho-Theta Hough")
    plt.show()


hough_line_rho_theta(sudoku)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def create_ball_stencil(image, x1, y1, x2, y2):
    x1, y1 = max(0, x1), max(0, y1)
    x2, y2 = min(image.shape[1], x2), min(image.shape[0], y2)

    roi = image[y1:y2, x1:x2]
    if len(roi.shape) == 3:
        gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    else:
        gray = roi

    edges = cv2.Canny(gray, 30, 100)
    kernel = np.ones((5, 5), np.uint8)
    closed_edges = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel, iterations=2)
    contours, _ = cv2.findContours(closed_edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours:
        print("No contours found in the selected box!")
        return None

    largest_contour = max(contours, key=cv2.contourArea)
    mask = np.zeros_like(gray)
    cv2.drawContours(mask, [largest_contour], -1, 255, thickness=cv2.FILLED)

    mask = cv2.GaussianBlur(mask, (3, 3), 0)
    b, g, r = cv2.split(roi)
    rgba_stencil = cv2.merge([b, g, r, mask])

    return rgba_stencil

def overlay_stencil(background_img, stencil, center_x, center_y):

    output = background_img.copy()
    bg_h, bg_w = output.shape[:2]
    s_h, s_w = stencil.shape[:2]

    x_start = center_x - (s_w // 2)
    y_start = center_y - (s_h // 2)
    x_end = x_start + s_w
    y_end = y_start + s_h


    bg_x1 = max(0, x_start)
    bg_y1 = max(0, y_start)
    bg_x2 = min(bg_w, x_end)
    bg_y2 = min(bg_h, y_end)

    st_x1 = bg_x1 - x_start
    st_y1 = bg_y1 - y_start
    st_x2 = st_x1 + (bg_x2 - bg_x1)
    st_y2 = st_y1 + (bg_y2 - bg_y1)


    if bg_x2 <= bg_x1 or bg_y2 <= bg_y1:
        return output

    # Extract the valid region from stencil and background
    stencil_crop = stencil[st_y1:st_y2, st_x1:st_x2]
    bg_crop = output[bg_y1:bg_y2, bg_x1:bg_x2]

    alpha = stencil_crop[:, :, 3] / 255.0
    inv_alpha = 1.0 - alpha

    for c in range(3):
        bg_crop[:, :, c] = (alpha * stencil_crop[:, :, c] +
                            inv_alpha * bg_crop[:, :, c])

    # Put the blended region back into the output image
    output[bg_y1:bg_y2, bg_x1:bg_x2] = bg_crop

    return output

img1 = ball
img2 = ball2
x1, y1 = 100, 175
x2, y2 = 250, 325

print("Extracting white ball...")
ball_stencil = create_ball_stencil(img1, x1, y1, x2, y2)

if ball_stencil is not None:

    h, w, _ = img2.shape
    target_x = int(w * 0.75)
    target_y = int(h * 0.6)

    print(f"Overlaying ball at ({target_x}, {target_y})...")
    final_result = overlay_stencil(img2, ball_stencil, target_x, target_y)

    plt.figure(figsize=(15, 6))

    plt.subplot(1, 3, 1)
    plt.imshow(ball_stencil[:, :, 3], cmap='gray')
    plt.title("Extracted Stencil (Alpha)")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(cv2.cvtColor(final_result, cv2.COLOR_BGR2RGB))
    plt.title("Final Overlay")
    plt.axis('off')

    plt.tight_layout()
    plt.show()
````
