To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# with thresholding

def scatter_plot_threshold(img, threshold):
    Ix = cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 3)
    Iy = cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 3)

    mag = np.sqrt(Ix**2 + Iy**2)
    mask = mag > threshold

    Ix_filt = Ix[mask]
    Iy_filt = Iy[mask]

    plt.figure(figsize = (10,10))
    plt.scatter(Ix_filt.ravel(),Iy_filt.ravel(),alpha = 0.5)
    plt.title("Scatterplot for Iy vs Ix")
    plt.xlabel("Ix")
    plt.axhline(0, color = "black", lw = 0.7)
    plt.axvline(0, color = "black", lw = 0.7)
    plt.ylabel("Iy")
    plt.show()

    return Ix_filt, Iy_filt

Ix1,Iy1 = scatter_plot_threshold(flat_gray, 20)
print(f"The value for Ix is \n {Ix1} \n and that for Iy is \n {Iy1}")
````

````python
def harris_r_map (img, k, windowsize):
  H, W = img.shape[0],img.shape[1]
  Ix = cv2.Sobel(img,cv2.CV_64F,1,0,windowsize)
  Iy = cv2.Sobel(img,cv2.CV_64F,0,1,windowsize)
  Ixx = Ix*Ix
  Iyy = Iy*Iy
  Ixy = Ix*Iy

  a = cv2.boxFilter(Ixx, -1, (windowsize, windowsize), normalize=False)
  c = cv2.boxFilter(Iyy, -1, (windowsize, windowsize), normalize=False)
  b = cv2.boxFilter(Ixy, -1, (windowsize, windowsize), normalize=False)

  lambda_1 = ((a+c)+np.sqrt((a-c)**2 + 4*b**2))/2
  lambda_2 = ((a+c)-np.sqrt((a-c)**2 + 4*b**2))/2

  R = lambda_1*lambda_2 - k*(lambda_1 + lambda_2)**2
  plt.imshow(R, cmap = "gray")
  plt.title("R map")
  plt.show()
  return R

harris_r_map(flat_gray, 0.04,3)
````

````python
corner2_gray = cv2.cvtColor(corner2,cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(corner2_gray,50,150)

def hough_transform_intercept(img, m_range, c_range,step):
  H,W = img.shape[0],img.shape[1]
  acc = np.zeros((step,step))
  m_vals = np.linspace(m_range[0],m_range[1],step)
  c_vals = np.linspace(c_range[0],c_range[1],step)

  y_idx,x_idx = np.where(img == 255)

  for i in range(len(x_idx)):
    x, y = x_idx[i], y_idx[i]
    for m_idx, m in enumerate(m_vals):
      c = y - m*x
      if (c_range[0] <= c <= c_range[1]):
        c_res = (c_range[1] - c_range[0]) / (step - 1)
        c_idx = int((c - c_range[0]) / c_res)
        acc[m_idx,c_idx] += 1
  return acc

accumulator = hough_transform_intercept(edges, (-10,10),(-500,500),100)
acc_norm = accumulator/np.max(accumulator)
mark = np.max(acc_norm)
t = 90
t_acc = (t/100) * mark
idx = np.where(acc_norm >= t_acc)

new = np.zeros_like(acc_norm)
new[idx] = 255

plt.imshow(new, cmap = "gray")
plt.title("Accumulator Map for Slope-Intercept method")
plt.axis("off")
````

````python
def hough_transform_rho(img,theta_range,step):
  H,W = img.shape[0],img.shape[1]

  acc = np.zeros((step,step))
  diag_len = np.sqrt(H**2 + W**2)
  theta_vals = np.linspace(theta_range[0],theta_range[1],step)
  rho_vals = np.linspace(-diag_len,diag_len,step)

  cos_t = np.cos(theta_vals)
  sin_t = np.sin(theta_vals)
  rho_res = (2*diag_len) / (step - 1)
  y_idx,x_idx = np.where(img == 255)

  for i in range(len(x_idx)):
    x, y = x_idx[i], y_idx[i]
    for theta_idx in range(len(theta_vals)):
        rho = x*cos_t[theta_idx] + y*sin_t[theta_idx]
        if -diag_len <= rho <= diag_len :
          rho_idx = int((rho + diag_len) / rho_res)
          acc[rho_idx,theta_idx] += 1
  return acc

accumulator = hough_transform_rho(edges,(-np.pi,np.pi),100)
acc_norm = accumulator/np.max(accumulator)
mark = np.max(acc_norm)
t = 90
t_acc = (t/100) * mark
idx = np.where(acc_norm >= t_acc)

new = np.zeros_like(acc_norm)
new[idx] = 255

plt.imshow(new, cmap = "gray")
plt.title("Accumulator Map for Rho-Theta method")
plt.axis("off")
````

````python
def create_stencil(img, x1, y1, x2, y2):
  region = img[y1:y2,x1:x2]
  region_gray = cv2.cvtColor(region,cv2.COLOR_RGB2GRAY)

  Ix = cv2.Sobel(region_gray,cv2.CV_64F,1,0,ksize = 3)
  Iy = cv2.Sobel(region_gray,cv2.CV_64F,0,1,ksize = 3)

  mag = np.sqrt(Ix**2 + Iy**2)

  thresh_val = 50
  mask = (mag > thresh_val).astype(np.uint8) * 255
  contours,_= cv2.findContours(mask, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)

  if not contours:
      return None

  biggest_contour = contours[0]
  max_area = 0
  for i in contours:
      area = cv2.contourArea(i)
      if area > max_area:
          max_area = area
          biggest_contour = i

  final_mask = np.zeros_like(region_gray)
  cv2.fillPoly(final_mask, [biggest_contour], 255)

  r, g, b = cv2.split(region)
  stencil = cv2.merge([r, g, b, final_mask])

  return stencil

stencil = create_stencil(ball, 0,0,417,626)
plt.imshow(stencil)
plt.axis("off")
````

````python
def overlay_stencil(stencil, background, x3, y3):
  h, w = stencil.shape[0],stencil.shape[1]
  H, W = background.shape[0],background.shape[1]

  top = y3 - h // 2
  left = x3 - w // 2
  bottom = top + h
  right = left + w

  bg_top = np.maximum(top, 0)
  bg_left = np.maximum(left, 0)
  bg_bottom = np.minimum(bottom, H)
  bg_right = np.minimum(right, W)

  st_top = bg_top - top
  st_left = bg_left - left
  st_bottom = st_top + (bg_bottom - bg_top)
  st_right = st_left + (bg_right - bg_left)

  stencil = stencil[st_top:st_bottom, st_left:st_right]

  alpha_s = stencil[:,:,3]/255.0
  alpha_l = 1.0 - alpha_s

  for c in range(0,3):
    background[bg_top:bg_bottom, bg_left:bg_right,c] = (alpha_s * stencil[:,:,c] + alpha_l * background[bg_top:bg_bottom, bg_left:bg_right,c])
  return background

img2 = overlay_stencil(stencil, ball2, 400, 220)
plt.imshow(img2)
plt.axis("off")
````
