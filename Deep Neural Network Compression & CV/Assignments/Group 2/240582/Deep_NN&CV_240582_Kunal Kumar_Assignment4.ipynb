To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :

````python
#C
#RHO-THETA
def hough_rho_theta(image, gray_image):
    edges=cv2.Canny(gray_image, 50, 150)
    H, W=edges.shape
    theta_deg=1.0
    thetas=np.deg2rad(np.arange(0, 180, theta_deg))

    diag_len = int(np.ceil(np.sqrt(H*H+W*W)))
    rhos = np.arange(-diag_len, diag_len, 1)
    acc=np.zeros((len(rhos), len(thetas)), dtype=np.int32)
    cos_t=np.cos(thetas)
    sin_t=np.sin(thetas)

    y_idxs, x_idxs=np.nonzero(edges)

    print(f"Rho-Theta: Processing {len(y_idxs)} edge pixels...")

    for i in range(len(y_idxs)):
        x=x_idxs[i]
        y=y_idxs[i]
        for t_idx in range(len(thetas)):
            rho=x*cos_t[t_idx]+y*sin_t[t_idx]
            rho_idx=int(rho+diag_len)

            if 0<=rho_idx<len(rhos):
                acc[rho_idx,t_idx]+=1

    acc_max=np.max(acc)
    if acc_max==0:
        print("No lines")
        return image

    acc_norm=(acc/acc_max)*100
    threshold_val=60
    rho_indices,theta_indices=np.where(acc_norm>threshold_val)
    output_img=image.copy()
    print(f"Found {len(rho_indices)} lines.")

    for i in range(len(rho_indices)):
        rho=rhos[rho_indices[i]]
        theta=thetas[theta_indices[i]]
        a=np.cos(theta)
        b=np.sin(theta)
        x0=a*rho
        y0=b*rho
        scale=2000
        x1=int(x0+scale*(-b))
        y1=int(y0+scale*(a))
        x2=int(x0-scale*(-b))
        y2=int(y0-scale*(a))
        cv2.line(output_img,(x1,y1),(x2,y2),(0,255,0),2)

    plt.figure(figsize=(15, 5))
    plt.subplot(1,3,1)
    plt.imshow(edges,cmap='gray')
    plt.title("Canny Edges")

    plt.subplot(1,3,2)
    plt.imshow(acc,cmap='jet', aspect='auto')
    plt.title("Accumulator (Rho vs Theta)")
    plt.xlabel("Theta")
    plt.ylabel("Rho")

    plt.subplot(1,3,3)
    plt.imshow(output_img)
    plt.title(f"Detected Lines (Rho-Theta)\nThresh: {threshold_val}%")

    plt.show()

hough_rho_theta(rgb_img, gray_img)
````

````python
#Slope-intercept
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hough_slope_intercept(image, gray_image):
    edges=cv2.Canny(gray_image,50,150)
    H,W=edges.shape
    m_min,m_max=-5.0, 5.0
    m_step=0.05
    m_values=np.arange(m_min,m_max,m_step)
    c_min, c_max=-int(H*2),int(H+W*2)
    c_step=1.0
    c_values=np.arange(c_min,c_max,c_step)
    accumulator=np.zeros((len(m_values), len(c_values)), dtype=np.int32)
    y_idxs,x_idxs=np.nonzero(edges)

    print(f"Slope-Intercept: Processing {len(y_idxs)} edge pixels...")
    for i in range(len(y_idxs)):
        x=x_idxs[i]
        y=y_idxs[i]
        for m_idx, m in enumerate(m_values):
            c=y-(m*x)
            if c_min<=c<c_max:
                c_idx=int((c-c_min)/c_step)
                if 0<=c_idx<len(c_values):
                    accumulator[m_idx,c_idx] += 1
    acc_max=np.max(accumulator)
    if acc_max == 0:
        print("No lines detected!")
        return image
    acc_norm=(accumulator/acc_max)*100
    threshold_val=60

    m_indices,c_indices=np.where(acc_norm>threshold_val)
    output_img=image.copy()
    print(f"Found {len(m_indices)} lines.")

    for i in range(len(m_indices)):
        m=m_values[m_indices[i]]
        c=c_values[c_indices[i]]
        p1_x=0
        p1_y=int(c)
        p2_x=W
        p2_y=int(m*W+c)
        cv2.line(output_img,(p1_x,p1_y),(p2_x,p2_y),(255,0,0),2)
    plt.figure(figsize=(15, 5))

    plt.subplot(1,3,1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")

    plt.subplot(1,3,2)
    plt.imshow(accumulator, cmap='jet', aspect='auto')
    plt.title("Accumulator (m vs c)")
    plt.xlabel("Intercept (c)")
    plt.ylabel("Slope (m)")

    plt.subplot(1,3,3)
    plt.imshow(output_img)
    plt.title(f"Detected Lines (m-c)\nThresh: {threshold_val}%")

    plt.show()

hough_slope_intercept(rgb_img, gray_img)
````
