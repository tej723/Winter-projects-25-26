To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def compute_r_map(Ix, Iy, k=0.04):
    # Compute second moments
    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy

    # Apply a 3x3 window (Gaussian or Box filter)
    Sxx = cv2.GaussianBlur(Ixx, (3, 3), 0)
    Syy = cv2.GaussianBlur(Iyy, (3, 3), 0)
    Sxy = cv2.GaussianBlur(Ixy, (3, 3), 0)

    # Compute R: det(M) - k * (trace(M))^2
    det = (Sxx * Syy) - (Sxy**2)
    trace = Sxx + Syy
    R = det - k * (trace**2)

    # Normalize for display
    R_norm = cv2.normalize(R, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    return R_norm

r_flat = compute_r_map(ix_f, iy_f)
r_edge = compute_r_map(ix_e, iy_e)
r_corner = compute_r_map(ix_c, iy_c)

print("R-Maps (Corners highlighted in white):")
show_step(r_flat, "R Map - Flat")
show_step(r_edge, "R Map - Edge")
show_step(r_corner, "R Map - Corner")
````

````python
def hough_slope_intercept(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)
    show_step(edges, "Canny Edge Map")

    h, w = edges.shape
    # m from -5 to 5, c from -h to h (simplified)
    accumulator = np.zeros((100, 200))

    y_coords, x_coords = np.where(edges > 0)
    for x, y in zip(x_coords, y_coords):
        for m_idx, m in enumerate(np.linspace(-5, 5, 100)):
            c = y - m * x
            c_idx = int(c + h) # Offset to keep index positive
            if 0 <= c_idx < 200:
                accumulator[m_idx, c_idx] += 1

    # Normalize and Threshold (90th percentile)
    acc_norm = accumulator / np.max(accumulator)
    thresh = 0.9
    m_found, c_found = np.where(acc_norm > thresh)

    output = img.copy()
    for m_i, c_i in zip(m_found, c_found):
        m = np.linspace(-5, 5, 100)[m_i]
        c = c_i - h
        # Draw line: y = mx + c
        x1, x2 = 0, w
        y1, y2 = int(m*x1 + c), int(m*x2 + c)
        cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 2)

    show_step(output, "Hough Lines (Slope-Intercept)")

hough_slope_intercept(img_edge)
````
