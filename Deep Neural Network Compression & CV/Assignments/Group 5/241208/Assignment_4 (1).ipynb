To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def convolve2d(image, kernel):

    image_h, image_w = image.shape
    kernel_h, kernel_w = kernel.shape

    pad_h = kernel_h // 2
    pad_w = kernel_w // 2

    padded_image = np.pad(image, ((pad_h, pad_h), (pad_w, pad_w)), mode='edge')

    output = np.zeros_like(image, dtype=float)

    for y in range(image_h):
        for x in range(image_w):
            output[y, x] = np.sum(padded_image[y:y+kernel_h, x:x+kernel_w] * kernel)
    return output
````

````python
def harris_r_map(img, window_size=3, k=0.04):

    # Convert to grayscale (same as sobel_analysis)
    if img.ndim == 3:
        gray_img = np.dot(img[..., :3], [0.2989, 0.5870, 0.1140])
    else:
        gray_img = img

    gray_img = gray_img.astype(np.float64)

    # Gradients using manual Sobel
    Ix = convolve2d(gray_img, sobel_x)
    Iy = convolve2d(gray_img, sobel_y)

    h, w = gray_img.shape
    offset = window_size // 2

    R_map = np.zeros((h, w))
    lambda1_map = np.zeros((h, w))
    lambda2_map = np.zeros((h, w))

    for y in range(offset, h - offset):
        for x in range(offset, w - offset):

            # Windowed gradients
            Ix_w = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            Iy_w = Iy[y-offset:y+offset+1, x-offset:x+offset+1]

            # Second moment matrix components
            a = np.sum(Ix_w ** 2)
            b = np.sum(Iy_w ** 2)
            c = np.sum(Ix_w * Iy_w)

            # Eigenvalues of structure tensor
            trace = a + b
            det = a * b - c**2
            temp = np.sqrt((a - b)**2 + 4*c**2)

            lambda1 = (trace + temp) / 2
            lambda2 = (trace - temp) / 2

            # Harris response
            R = det - k * (trace ** 2)

            R_map[y, x] = R
            lambda1_map[y, x] = lambda1
            lambda2_map[y, x] = lambda2

    return R_map, lambda1_map, lambda2_map
````

````python
def show_r_map(R_map, title="R Map"):
    # Normalize to [0, 255]
    R_min = np.min(R_map)
    R_max = np.max(R_map)

    if R_max > R_min:
        R_norm = (R_map - R_min) / (R_max - R_min)
    else:
        R_norm = np.zeros_like(R_map)

    R_norm = (R_norm * 255).astype(np.uint8)

    plt.figure(figsize=(5,5))
    plt.imshow(R_norm, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
````

````python
def hough_mb(gray,
             m_min=-2, m_max=2, m_bins=200,
             b_min=-500, b_max=500, b_bins=200):

    edges = cv2.Canny(gray, 100, 200)

    h, w = gray.shape
    m_vals = np.linspace(m_min, m_max, m_bins)
    b_vals = np.linspace(b_min, b_max, b_bins)

    accumulator = np.zeros((m_bins, b_bins))

    # Looping through edge pixels
    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for i, m in enumerate(m_vals):
            b = y - m * x
            if b_min <= b <= b_max:
                j = int((b - b_min) / (b_max - b_min) * (b_bins - 1))
                accumulator[i, j] += 1

    return accumulator, m_vals, b_vals, edges
````

````python
acc, m_vals, b_vals, edges = hough_mb(gray)

acc_norm = acc / acc.max()

threshold = 0.9
lines = np.where(acc_norm > threshold)
````

````python

img_mb = gray.copy()
for i, j in zip(lines[0], lines[1]):
    m = m_vals[i]
    b = b_vals[j]

    x1, x2 = 0, gray.shape[1]
    y1 = int(m * x1 + b)
    y2 = int(m * x2 + b)

    if 0 <= y1 < gray.shape[0] and 0 <= y2 < gray.shape[0]:
        cv2.line(img_mb, (x1, y1), (x2, y2), (0,0,255), 1)

plt.imshow(cv2.cvtColor(img_mb, cv2.COLOR_BGR2RGB))
plt.title("Slope-Intercept Hough Lines")
plt.axis("off")
plt.show()
````

````python
def hough_rho_theta(gray):

    edges = cv2.Canny(gray, 100, 200)
    h, w = gray.shape

    thetas = np.deg2rad(np.arange(-90, 90))
    diag_len = int(np.sqrt(h*h + w*w))
    rhos = np.linspace(-diag_len, diag_len, 2*diag_len)

    accumulator = np.zeros((len(rhos), len(thetas)))

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for t_idx, theta in enumerate(thetas):
            rho = int(x * np.cos(theta) + y * np.sin(theta)) + diag_len
            accumulator[rho, t_idx] += 1

    return accumulator, rhos, thetas
````

````python
acc_rt, rhos, thetas = hough_rho_theta(gray)

acc_norm = acc_rt / acc_rt.max()
threshold = 0.5   # lower works better here

lines = np.where(acc_norm > threshold)



img_rt = gray.copy()

for rho_idx, theta_idx in zip(lines[0], lines[1]):
    rho = rhos[rho_idx]
    theta = thetas[theta_idx]

    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho

    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    cv2.line(img_rt, (x1, y1), (x2, y2), (0,255,0), 1)

plt.imshow(cv2.cvtColor(img_rt, cv2.COLOR_BGR2RGB))
plt.title("Rho-Theta Hough Lines")
plt.axis("off")
plt.show()
````
