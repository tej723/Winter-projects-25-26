To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# defining sobel kernels

Sx = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

Sy = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])
````

````python
# writing convolution function

def convolve(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape
    pad = kh // 2

    output = np.zeros((h, w))
    padded = np.pad(image, pad, mode='constant')

    for i in range(h):
        for j in range(w):
            patch = padded[i:i+kh, j:j+kw]
            output[i, j] = np.sum(patch * kernel)

    return output
````

````python
# compute Ix and Iy

Ix_flat = convolve(gray_flat, Sx)
Iy_flat = convolve(gray_flat, Sy)

Ix_edge = convolve(gray_edge, Sx)
Iy_edge = convolve(gray_edge, Sy)

Ix_corner = convolve(gray_corner, Sx)
Iy_corner = convolve(gray_corner, Sy)
````

````python
def analyze_image(img, name=""):
    gray = to_grayscale(img)
    Ix = convolve(gray, Sx)
    Iy = convolve(gray, Sy)

    plt.imshow(Ix, cmap='gray')
    plt.title(f"Ix - {name}")
    plt.axis("off")
    plt.show()

    plt.imshow(Iy, cmap='gray')
    plt.title(f"Iy - {name}")
    plt.axis("off")
    plt.show()

    plt.scatter(Ix.flatten(), Iy.flatten(), s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix - {name}")
    plt.show()

    return Ix, Iy
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
# Your PART B codes follow here. You can create more cells under this. Write clean and commented code.
def compute_R_map(Ix, Iy, window_size=3, k=0.04):
    h, w = Ix.shape
    offset = window_size // 2
    R_map = np.zeros((h, w))

    for i in range(offset, h - offset):
        for j in range(offset, w - offset):
            Ix_win = Ix[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy_win = Iy[i-offset:i+offset+1, j-offset:j+offset+1]

            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            det = a * c - b ** 2
            trace = a + c

            R_map[i, j] = det - k * (trace ** 2)

    return R_map
````

````python
Ix, Iy = analyze_image(corner,"corner")
R = compute_R_map(Ix, Iy)

plt.imshow(R, cmap='gray')
plt.title("R Map - Corner")
plt.axis("off")
plt.show()

Ix, Iy = analyze_image(edge, "edge")
R = compute_R_map(Ix, Iy)

plt.imshow(R, cmap='gray')
plt.title("R Map - edge")
plt.axis("off")
plt.show()

Ix, Iy = analyze_image(flat, "Flat")
R = compute_R_map(Ix, Iy)

plt.imshow(R, cmap='gray')
plt.title("R Map - flat")
plt.axis("off")
plt.show()

````

Using the gradient images Ix and Iy obtained in Part A, second moment values a, b and c were computed over a local window around each pixel. These values form the structure matrix, whose eigenvalues represent intensity variations in orthogonal directions. R was then calculated for every pixel. Flat regions produce low R values, edges produce negative or small values, while corners result in large positive R values. The resulting R-map clearly highlights corner regions in white.

````python
# Your PART C codes follow here. You can create more cells under this. Write clean and commented code.
import cv2
edges = cv2.Canny(edge.astype(np.uint8), 50, 150)
m_vals = np.linspace(-2, 2, 400)
c_vals = np.linspace(-500, 500, 400)

acc = np.zeros((len(m_vals), len(c_vals)))
ys, xs = np.where(edges > 0)



for x, y in zip(xs, ys):
    for i, m in enumerate(m_vals):
        c = y - m * x
        j = np.argmin(np.abs(c_vals - c))
        if 0 <= j < len(c_vals):
            acc[i, j] += 1

acc_norm = acc / acc.max()
lines = np.where(acc_norm > 0.9)
img_lines = edge.copy()



for i, j in zip(lines[0], lines[1]):
    m = m_vals[i]
    c = c_vals[j]

    x1, x2 = 0, edge.shape[1]
    y1 = int(m * x1 + c)
    y2 = int(m * x2 + c)

    cv2.line(img_lines, (x1, y1), (x2, y2), (255,0,0), 1)
````

````python
def extract_ball_stencil(img, x1, y1, x2, y2):
    crop = img[y1:y2, x1:x2]

    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)

    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    largest = max(contours, key=cv2.contourArea)

    mask = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(mask, [largest], -1, 255, -1)

    rgba = np.dstack((crop, mask))
    return rgba


````

````python
def overlay_stencil(bg, stencil, x3, y3):
    h, w, _ = stencil.shape
    x1 = x3 - w // 2
    y1 = y3 - h // 2

    for i in range(h):
        for j in range(w):
            if stencil[i, j, 3] > 0:  # alpha channel
                bg[y1 + i, x1 + j] = stencil[i, j, :3]

    return bg
````
