To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def compute_harris_r_map(img, k=0.04, window_size=3):

    Kx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    Ky = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    Ix = cv2.filter2D(img.astype(float), -1, Kx)
    Iy = cv2.filter2D(img.astype(float), -1, Ky)


    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix*Iy


    Sxx = cv2.boxFilter(Ixx, -1, (window_size, window_size))
    Syy = cv2.boxFilter(Iyy, -1, (window_size, window_size))
    Sxy = cv2.boxFilter(Ixy, -1, (window_size, window_size))


    det_M = (Sxx * Syy) - (Sxy**2)
    trace_M = Sxx + Syy
    R = det_M - k * (trace_M**2)


    R_norm = cv2.normalize(R, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1); plt.imshow(img, cmap='gray'); plt.title('Original')
    plt.subplot(1, 2, 2); plt.imshow(R_norm, cmap='jet'); plt.title('R-Map (Corners in Red/White)')
    plt.show()

    return R


r_map = compute_harris_r_map(corner_img)
````

````python


def hough_lines_slope_intercept(img, threshold_percent=0.5):
    # 1. Edge Detection
    edges = cv2.Canny(img, 100, 200)
    h, w = img.shape


    m_range = np.linspace(-10, 10, 200)
    c_range = np.linspace(-h, h, 200)

    accumulator = np.zeros((len(m_range), len(c_range)))

    # Get coordinates of all edge pixels
    y_points, x_points = np.nonzero(edges)

    # 2. Voting procedure
    for i in range(len(x_points)):
        x, y = x_points[i], y_points[i]
        for m_idx, m in enumerate(m_range):
            # c = y - mx
            c = y - m * x

            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                accumulator[m_idx, c_idx] += 1

    # 3. Normalization and Thresholding
    acc_max = np.max(accumulator)
    acc_norm = (accumulator / acc_max) if acc_max > 0 else accumulator

    # Find peaks above threshold
    peaks = np.argwhere(acc_norm > threshold_percent)

    # Visualization
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(acc_norm, cmap='hot', extent=[c_range[0], c_range[-1], m_range[-1], m_range[0]])
    plt.title("Slope-Intercept Accumulator")

    plt.subplot(1, 2, 2)
    plt.imshow(img, cmap='gray')
    for m_idx, c_idx in peaks:
        m, c = m_range[m_idx], c_range[c_idx]
        # Draw the line: y = mx + c
        x_vals = np.array([0, w])
        y_vals = m * x_vals + c
        plt.plot(x_vals, y_vals, color='red', linewidth=1)
    plt.xlim(0, w); plt.ylim(h, 0)
    plt.title("Detected Lines (m, c)")
    plt.show()

def hough_lines_rho_theta(img, threshold_percent=0.6):
    edges = cv2.Canny(img, 100, 200)
    h, w = img.shape

    # Diagonal length is the max possible rho
    diag_len = np.ceil(np.sqrt(h**2 + w**2))
    rhos = np.linspace(-diag_len, diag_len, int(2 * diag_len))
    thetas = np.deg2rad(np.arange(-90, 90))

    accumulator = np.zeros((len(rhos), len(thetas)))

    y_points, x_points = np.nonzero(edges)

    # Pre-compute cos and sin for speed
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    # Voting
    for i in range(len(x_points)):
        x, y = x_points[i], y_points[i]
        for t_idx in range(len(thetas)):
            # rho = x*cos(theta) + y*sin(theta)
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            rho_idx = np.argmin(np.abs(rhos - rho))
            accumulator[rho_idx, t_idx] += 1

    # Normalize and find peaks
    acc_norm = accumulator / np.max(accumulator)
    peaks = np.argwhere(acc_norm > threshold_percent)

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(acc_norm, cmap='magma', extent=[-90, 90, rhos[-1], rhos[0]])
    plt.title("Rho-Theta Accumulator")

    plt.subplot(1, 2, 2)
    plt.imshow(img, cmap='gray')
    for r_idx, t_idx in peaks:
        rho = rhos[r_idx]
        theta = thetas[t_idx]
        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a * rho, b * rho
        # Find two points far away to draw the line
        pt1 = (int(x0 + 1000 * (-b)), int(y0 + 1000 * (a)))
        pt2 = (int(x0 - 1000 * (-b)), int(y0 - 1000 * (a)))
        plt.plot([pt1[0], pt2[0]], [pt1[1], pt2[1]], color='cyan', linewidth=1)
    plt.xlim(0, w); plt.ylim(h, 0)
    plt.title("Detected Lines (rho, theta)")
    plt.show()

# Run it on your loaded 'edge' or 'corner' image
hough_lines_rho_theta(corner_img)
````

````python
def hough_circles_manual(img, min_r, max_r, step_r=2):

    edges = cv2.Canny(img, 100, 200)
    h, w = img.shape


    Kx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    Ky = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
    Ix = cv2.filter2D(img.astype(float), -1, Kx)
    Iy = cv2.filter2D(img.astype(float), -1, Ky)


    radii = np.arange(min_r, max_r, step_r)
    accumulator = np.zeros((h, w, len(radii)))

    edge_pixels = np.argwhere(edges > 0)

    # Voting process
    for y, x in edge_pixels:

        theta = np.arctan2(Iy[y, x], Ix[y, x])

        for r_idx, r in enumerate(radii):

            for sign in [-1, 1]:
                a = int(x + sign * r * np.cos(theta))
                b = int(y + sign * r * np.sin(theta))

                if 0 <= a < w and 0 <= b < h:
                    accumulator[b, a, r_idx] += 1


    y_center, x_center, r_idx_best = np.unravel_index(np.argmax(accumulator), accumulator.shape)
    best_r = radii[r_idx_best]


    output = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    cv2.circle(output, (x_center, y_center), best_r, (0, 255, 0), 3)
    cv2.circle(output, (x_center, y_center), 2, (0, 0, 255), 5)

    plt.figure(figsize=(8, 8))
    plt.imshow(output)
    plt.title(f"Detected Circle: Center({x_center},{y_center}), Radius {best_r}")
    plt.axis("off")
    plt.show()


hough_circles_manual(corner_img, min_r=20, max_r=100)
````
