To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2

def ix_iy_info(img):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY).astype(np.float32)
    sobel_x = np.array([[-1, 0, 1],[-2, 0, 2],[-1, 0, 1]], dtype=np.float32)
    sobel_y = np.array([[-1, -2, -1],[ 0,  0,  0],[ 1,  2,  1]], dtype=np.float32)
    Ix = np.zeros_like(gray)
    Iy = np.zeros_like(gray)

    # Finding the convolutions
    for i in range(1, gray.shape[0] - 1):
        for j in range(1, gray.shape[1] - 1):
            region = gray[i-1:i+2, j-1:j+2]
            Ix[i, j] = np.sum(region * sobel_x)
            Iy[i, j] = np.sum(region * sobel_y)

    # For plotting the scatter plot with threshold for small gradients
    mask = (Ix**2 + Iy**2) > 50 # Threshold
    plt.figure(figsize=(5,5))
    plt.scatter(Ix[mask], Iy[mask], s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix Scatter")
    plt.grid(True)
    plt.show()

    return Ix, Iy


ix_iy_info(flat)
ix_iy_info(edge)
ix_iy_info(corner)
ix_iy_info(corner2)
````

````python
def find_R_map(img, pad = 3, k=0.04):
    Ix, Iy = ix_iy_info(img)
    h, w = Ix.shape
    R = np.zeros((h, w), dtype=np.float32)

    for i in range(pad, h - pad):
        for j in range(pad, w - pad):
            Ix_1 = Ix[i-pad:i+pad+1, j-pad:j+pad+1]
            Iy_1 = Iy[i-pad:i+pad+1, j-pad:j+pad+1]

            a = np.sum(Ix_1**2)
            b = np.sum(Ix_1*Iy_1)
            c = np.sum(Iy_1**2)

            det = a*c - b**2
            trace = a + c
            R[i, j] = det - k * (trace ** 2)

    return R

def show_R_map(R, title):
    R_1 = R.copy()
    R_1[R_1 < 0] = 0    # This is for omitting the negative values (if any)
    if R_1.max() > 0 :
      R_1 = R_1 / R_1.max()  # This is for converting the R-map values to [0,1] so that it can be displayed in an image
    R_1[R_1 > 0.005] = 1.0 # This is the threshold for making the corners extra visible, reached this number after trying many numbers as thresholds
    plt.figure(figsize=(5,5))
    plt.imshow(R_1, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Finding and plotting the R-maps for the 'flat', 'edge' and the 'corner' images
R_flat = find_R_map(flat)
R_edge = find_R_map(edge)
R_corner = find_R_map(corner)
R_corner2 = find_R_map(corner2)
show_R_map(R_flat, "R-map (Flat)")
show_R_map(R_edge, "R-map (Edge)")
show_R_map(R_corner, "R-map (Corner)")
show_R_map(R_corner2, "R-map (Corner2)")
````

````python
# Hough-Transform through Slope Intercept method

def hough_slope_intercept(img):
    m_range = (-2, 2) # Selected range of slope
    c_range = (-200, 200)  # Selected range of intercept
    m_bins= 200 # For Descretization
    c_bins = 200 # For Descretization
    threshold = 0.9
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(gray, 100, 200)

    h, w = gray.shape
    m_val = np.linspace(m_range[0], m_range[1], m_bins) # Creating the m linespace
    c_val = np.linspace(c_range[0], c_range[1], c_bins) # Creating the c linespace

    accumulator = np.zeros((m_bins, c_bins), dtype=np.uint64)

    ys, xs = np.where(edges > 0)  # Permitting the edge coordinates only

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_val):
            c = y - m*x
            ci = int((c - c_range[0]) / (c_range[1]-c_range[0]) * c_bins)
            if 0 <= ci < c_bins:
                accumulator[mi, ci] += 1

    accumulator_normalised = accumulator / accumulator.max()
    lines = np.where(accumulator_normalised > threshold)

    img_lines = img.copy()
    for mi, ci in zip(*lines):
        m = m_val[mi]
        c = c_val[ci]
        x1, x2 = 0, w
        y1 = int(m*x1 + c)
        y2 = int(m*x2 + c)
        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img_lines, (x1,y1), (x2,y2), (0,255,0), 1)

    return img_lines, accumulator

out_img, acc = hough_slope_intercept(edge)

plt.figure(figsize=(6,6))
plt.imshow(out_img)
plt.title("SlopeIntercept Hough Lines")
plt.axis('off')
plt.show()
````

````python
def create_ball_stencil(img, x1, y1, x2, y2):
    roi = img[y1:y2, x1:x2].copy() # Region of Interest
    gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    # Converting to grayscale
    gray = cv2.GaussianBlur(gray, (9,9), 1.5)

    # Hough Circle Tranform for detecting the circle
    circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT, dp=1.2, minDist=gray.shape[0]//2, param1=100, param2=30, minRadius=10, maxRadius=min(gray.shape)//2)
    cx, cy, r = circles[0][0].astype(int) # Gives the strongest circle's centre and radii

    # Creating a circular alpha mask
    h, w = gray.shape
    yy, xx = np.ogrid[:h, :w] # ogrid is better than meshgrid for saving memory
    mask = (xx - cx)**2 + (yy - cy)**2 <= r*r
    alpha = np.zeros((h, w), dtype=np.uint8)
    alpha[mask] = 255 # This means fully transparent pixels

    # Making the stencil RGBA
    rgba = np.dstack((roi, alpha))

    # PLotting the circle
    plt.imshow(rgba)
    plt.title("RGBA Ball Stencil (Hough Circle)")
    plt.axis("off")
    plt.show()

    return rgba


stencil = create_ball_stencil(ball, x1=100, y1=170, x2=290, y2=330) # Got these coordinates by trial and error
````
