To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analysis(img, thresh=30):

  #Converting to grayscale
  #I'm using cv2.cvtColor method here instead of weighted method for ease and quickness

  if (img.ndim==3):
    gray_scale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  else:
    gray_scale = img.copy()

  gray_scale = gray_scale.astype(np.float32)

  #Declaring Sobel kernels

  S_x = [[-1,0,1],
        [-2,0,2],
        [-1,0,1]
        ]

  S_y = [[-1,-2,-1],
        [0,0,0],
        [1,2,1]
        ]

  S_x = np.array(S_x, dtype = np.float32)
  S_y = np.array(S_y, dtype = np.float32)

  #Computing Gradients again using cv2.filter instead of custom convolve function like in previous assignment mainly for efficiency

  I_x = cv2.filter2D(gray_scale, -1, S_x)
  I_y = cv2.filter2D(gray_scale, -1, S_y)

  #Computing intensity magnitude

  I = np.sqrt(I_x**2 + I_y**2)

  #Adding thresholding by creating custom mask

  mask = I > thresh
  I_x_thresh = I_x[mask]
  I_y_thresh = I_y[mask]

  #Plotting

  sns.set_theme(style="whitegrid", context="talk")

  plt.figure(figsize=(20, 6))

  plt.subplot(1, 3, 1)
  sns.heatmap(
      I_x,
      cmap="gray",
      cbar=True,
      square=False
  )
  plt.title("Horizontal Gradient (Ix)")
  plt.axis("off")

  plt.subplot(1, 3, 2)
  sns.heatmap(
      I_y,
      cmap="gray",
      cbar=True,
      square=False
  )
  plt.title("Vertical Gradient (Iy)")
  plt.axis("off")

  plt.subplot(1, 3, 3)
  sns.scatterplot(
      x=I_x_thresh,
      y=I_y_thresh,
      s=10,
      alpha=0.5,
      edgecolor=None
  )

  plt.xlabel("Ix")
  plt.ylabel("Iy")
  plt.title("Ix vs Iy Scatter Plot")

  plt.tight_layout()
  plt.show()

  return I_x, I_y
````

````python
#Defining Gradient Calculating function

def calculate_grads(img):

  #Converting to grayscale
  #I'm using cv2.cvtColor method here instead of weighted method for ease and quickness

  if (img.ndim==3):
    gray_scale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  else:
    gray_scale = img.copy()

  gray_scale = gray_scale.astype(np.float32)

  #Declaring Sobel kernels

  S_x = [[-1,0,1],
        [-2,0,2],
        [-1,0,1]
        ]

  S_y = [[-1,-2,-1],
        [0,0,0],
        [1,2,1]
        ]

  S_x = np.array(S_x, dtype = np.float32)
  S_y = np.array(S_y, dtype = np.float32)

  #Computing Gradients again using cv2.filter instead of custom convolve function like in previous assignment mainly for efficiency

  I_x = cv2.filter2D(gray_scale, -1, S_x)
  I_y = cv2.filter2D(gray_scale, -1, S_y)

  return I_x, I_y
````

````python
from types import LambdaType
#Defining R map computing function

def harris_R_map(img, window_size = 5, k= 0.04):

  #Getting gradients

  Ix, Iy = calculate_grads(img)

  #Calculating relevant terms

  Ix_sq = Ix**2
  Iy_sq = Iy**2
  Ixy = Ix*Iy

  #Padding

  pad = window_size//2
  h,w = Ix.shape

  #Initialising zero matrix

  R_map = np.zeros_like(img, dtype=np.float32)

  #Looping window over image

  for y in range(pad, h-pad):
    for x in range(pad, w-pad):

      #Calculating matrix elements a,b and c

      a = np.sum(Ix_sq[y-pad:y+pad+1, x-pad:x+pad+1])
      b= np.sum(Ixy[y-pad:y+pad+1, x-pad:x+pad+1])
      c = np.sum(Iy_sq[y-pad:y+pad+1, x-pad:x+pad+1])

      #Defining matrix

      matrix = [[a,b],
                [b,c]
                ]

      matrix = np.array(matrix)

      #Calculating eigen values

      lamda_1, lambda_2 = np.linalg.eigvals(matrix)

      #Calculating Harris respnse R

      R = (lamda_1*lambda_2) - k*(lamda_1+lambda_2)**2
      R_map[y,x] = R

  return R_map


````

````python
#Creating show function

def show_R_map(R_map, title="Harris R Map"):
    R_norm = R_map.copy()
    R_norm[R_norm < 0] = 0
    R_norm = R_norm / R_norm.max()

    plt.figure(figsize=(6,6))
    plt.imshow(R_norm, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()
````

````python
def hough_trans_SI(img,
                   lower_hysteresis = 100,
                   upper_histeresis = 200,
                   m_range = (-50,50),
                   c_range = (-500,500),
                   m_bins = 200,
                   c_bins = 400,
                   acc_threshold = 0.9
                   ):

  #Converting image to array

  img = np.array(img, dtype = np.uint8)

  #Converting image to grayscale

  if (img.ndim==3):
    gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

  else:
    gray_img = img.copy()

  #Applying gaussian blur before canny to get better results

  blur = cv2.GaussianBlur(gray_img, (5,5), sigmaX=3, sigmaY=3)

  #Using Canny Edge Detection

  canny = cv2.Canny(blur,lower_hysteresis, upper_histeresis)

  #Making accumalator with finite values

  height, width = canny.shape
  m_vals = np.linspace(m_range[0], m_range[1], m_bins)
  c_vals = np.linspace(c_range[0], c_range[1], c_bins)

  accumulator = np.zeros((m_bins, c_bins), dtype = np.uint32)

  row_idx, col_idx = np.nonzero(canny)

  #Voting Algorythm

  #Looping through each pixel using zip which pairs elements from x and y non zero pixels in canny

  for x,y in zip(col_idx, row_idx):

    #Looping through each possible value of m

    for mi,m in enumerate(m_vals):
      c= y - m*x

      #Mapping c to proper column index

      ci = int((c-c_range[0])/(c_range[1] - c_range[0]) * (c_bins-1))

      #Only adding ci in appropriate range

      if (0<=ci<c_bins):  accumulator[mi,ci] +=1

  #Normalising Accumulator

  accumulator_norm = accumulator / accumulator.max()

  #Drawing detected lines

  output_img = img.copy()

  for mi, ci in np.argwhere(accumulator_norm >= acc_threshold):

    m = m_vals[mi]
    c = c_vals[ci]

    #Getting another point

    x1, x2 = 0, width

    y1 = int(c)
    y2 = int(m*x2 + c)

    #Passing these points to cv2.line to get a red line that spans the whole width of the image

    cv2.line(output_img, (x1,y1), (x2,y2), (0,0,255), 1)

  #Displaying

  plt.figure(figsize=(10,10))
  plt.imshow(output_img, cmap="gray")
  plt.title("Slope-intercept Hough Lines")
  plt.show()

  return output_img, accumulator
````

````python
#Using rho theta method

def hough_trans_RT(img,
                   lower_hysteresis = 100,
                   upper_hysteresis = 200,
                   theta_range = (0,180),
                   theta_resolution = np.deg2rad(1),
                   rho_resolution = 1,
                   acc_threshold = 0.7
                   ):

  #Converting image to array

  img_array = np.array(img, dtype = np.uint8)

  #Converting image to grayscale

  if (img_array.ndim==3):

    gray_img = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)

  else:
    gray_img = img_array.copy()

  #Applying gaussian blur before canny to get better results

  blur = cv2.GaussianBlur(gray_img, (5,5), sigmaX=2, sigmaY=2)


  #Finding canny edge map

  canny = cv2.Canny(blur, lower_hysteresis, upper_hysteresis)

  #Making rho theta accumulator

  height, width = canny.shape
  rho_max = int(np.sqrt(height**2 + width**2))
  rho_vals = np.arange((-1)*rho_max, rho_max+1, rho_resolution)
  theta_vals = np.arange(0,np.pi, theta_resolution)

  accumulator = np.zeros((len(rho_vals), len(theta_vals)), dtype = np.uint32)

  row_idx, col_idx = np.nonzero(canny)

  #Voting
  for x, y in zip(col_idx, row_idx):

    for ti, theta in enumerate(theta_vals):
      rho = int(x*np.cos(theta) + y*np.sin(theta))

      #Mapping rho to proper row
      ri = int(np.round((rho+rho_max)/rho_resolution))  #We divide by bin values spp we know in which bin it belongs and round to the nearest integer

      #Checking annd inserting values

      if (0<=ri<len(rho_vals)): accumulator[ri,ti] +=1

  #Normalising accumulator

  accumulator_norm = accumulator / accumulator.max()

  #Drawing detected lines

  output_img = img_array.copy()

  for ri, ti in np.argwhere(accumulator_norm>=acc_threshold):
    rho = rho_vals[ri]
    theta = theta_vals[ti]

    #Getting points on line using normal form

    a, b = np.cos(theta), np.sin(theta)
    x0, y0 = a*rho, b*rho

    #Creating points in oppsite direction

    x1 = int(x0 + width*(-b))
    y1 = int(y0 + width*(a))
    x2 = int(x0 - width*(-b))
    y2 = int(y0 - width*a)

    cv2.line(output_img, (x1,y1), (x2,y2), (0,0,255), 1)

  #Displaying

  plt.figure(figsize=(10,10))
  plt.imshow(output_img, cmap="gray")
  plt.title("Rho-theta Hough Lines")
  plt.show()

  return output_img, accumulator

````

````python
def stencil(img, t_left_pt, b_right_pt):

  #Converting to numpy array

  image_array = np.array(img, dtype = np.uint8)

  #Extracting region of interest

  x1,y1 = t_left_pt
  x2,y2 = b_right_pt

  roi = image_array[y1:y2, x1:x2]

  #Converting to grayscale

  gray_scale = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

  #Applying Gaussian Blur

  gray_scale = cv2.GaussianBlur(gray_scale, (3,3), sigmaX=2, sigmaY=2)

  #Calculating Gradients

  G_x = cv2.Sobel(gray_scale, cv2.CV_64F, 1, 0, ksize=3)
  G_y = cv2.Sobel(gray_scale, cv2.CV_64F, 0, 1, ksize=3)
  G = np.sqrt(G_x**2 + G_y**2)

  #Normalising and thresholding

  G = cv2.normalize(G, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
  _, thresh = cv2.threshold(G, 100, 255, cv2.THRESH_BINARY)

  #Finding contours using cv2.findContours. We can also build this function ourselves but cv2's implementation is highly optimised

  contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  #Choosing the larget contour areawise

  largest_contour = max(contours, key=cv2.contourArea)

  #Creating A mask using formula
  # Iout = alhpa*foreground + (1-alpha)*background

  alpha = np.zeros(gray_scale.shape, dtype=np.uint8)
  cv2.drawContours(alpha, [largest_contour], -1, 255, thickness = -1)

  #Building stencil by stakcing along depth

  rgba = np.dstack((roi,alpha))

  return rgba
````

````python
def see_through(background, stencil_rgba, center):

  #Separating background and foreground

  bg = background.copy()
  stencil = stencil_rgba.copy()

  height, width = stencil.shape[:2]

  #Shifting towards bottom left

  x3 , y3 = center

  x0 = int(x3 - width/2)
  y0 = int(y3 - height/2)

  #Extracting A channel

  alpha = stencil[...,3]/255.0

  #Extending dimensions of alpha to blend

  alpha =alpha[...,None]

  foreground = stencil[...,:3]

  #Slicing background to exactly where we place stencil

  bg_crop = bg[y0:y0+height, x0:x0+width]

  #Applying alpha blending

  blended = alpha*foreground + (1-alpha)*bg_crop
  blended = blended.astype(np.uint8)

  #Placing background at appropriate position

  background[y0:y0+height, x0:x0+width] = blended

  return background
````
