To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# defining convolve
def convolve(img, kernel):

    x, y = kernel.shape
    img_with_padding = np.pad(
        img,((x//2, x//2), (y//2, y//2)),mode='reflect'
    )

    output = np.zeros(img.shape)
    a, b = img.shape

    for i in range(a):
        for j in range(b):
            output[i, j] = np.sum(
                img_with_padding[i:i+x, j:j+y] * kernel
            )

    return output
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python


def R_map_from_gradients(img, window_size=3, k=0.04):

    # Convert to grayscale
    b= img[:,:,0]
    g= img[:,:,1]
    r= img[:,:,2]
    gray=(0.299*r + 0.587*g +0.114*b).astype(np.uint8)

    # Sobel kernels
    sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    sobel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])

    # Gradients (g_x, g_y)
    g_x = convolve(gray, sobel_x)
    g_y = convolve(gray, sobel_y)

    # Second moment components
    g_x2 = g_x ** 2
    g_y2 = g_y ** 2
    g_xy = g_x * g_y

    # Forming the window
    window = np.ones((window_size, window_size))

    # finding a, b , c matrices
    a = convolve(g_x2, window)
    b = convolve(g_xy, window)
    c = convolve(g_y2, window)

    # Harris corner response in terms of a,b,c (we can also write it in terms of lambdas)

    R = (a * c- b ** 2) - k * (a + c) ** 2

    R[R < 0] = 0


    return R
````

````python
#defining non maximal suppression
def non_max_suppression(R, window_size):
    h, w = R.shape
    offset = window_size // 2
    R_nms = np.zeros_like(R)

#if we have window size as k then the valid center pixels are (offset,.... h-offt-1)
# where offset = k//2

    for y in range(offset, h - offset):
        for x in range(offset, w - offset):
            window = R[y-offset:y+offset+1, x-offset:x+offset+1]
            if R[y, x] == np.max(window):
                R_nms[y, x] = R[y, x]

    return R_nms
````

````python
img_5 = cv2.imread("Screenshot 2025-12-19 003116.png")
R = R_map_from_gradients(img_5)
img_color = img_5.copy()

#using non maximal suppression
R_nms = non_max_suppression(R,3)

#defining corners by using thresholding
T = 0.01 * R_nms.max()    #keeping corners that are at least 1% as strong as the strongest corner
corners = R_nms > T

#overlapping the corners on the original marvel logo
ys, xs = np.where(corners)

for (x, y) in zip(xs, ys):
    cv2.circle(img_color, (x, y), radius=2, color=(255, 0, 0), thickness=-1)

#normalising r_nor
R_min = R.min()
R_max = R.max()

R_nor = (R - R_min) / (R_max - R_min)
R_nor = (R_nor * 255).astype(np.uint8)

plt.figure(figsize=(15,6))
plt.subplot(1, 2, 1)
plt.imshow(R_nor, cmap='gray')
plt.title("Harris Response R")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Detected Corners")
plt.axis('off')

plt.show()
````

````python
img = cv2.imread("Screenshot 2025-12-19 040849.png")
img_bgr=img[ :, :,::-1]
b= img_bgr[:,:,0]
g= img_bgr[:,:,1]
r= img_bgr[:,:,2]
gray=(0.299*r + 0.587*g +0.114*b).astype(np.uint8)

edges = cv2.Canny(gray,90, 200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Map")
plt.axis("off")
plt.show()

h, w = edges.shape

m_min, m_max = 15,-15
c_min, c_max = -h, h

num_m = 400
num_c = 400

m_vals = np.linspace(m_min, m_max, num_m)
c_vals = np.linspace(c_min, c_max, num_c)

accumulator = np.zeros((num_m, num_c), dtype=np.int32)

ys, xs = np.where(edges > 0)

for x, y in zip(xs, ys):
    for i_m, m in enumerate(m_vals):
        c = y - m * x
        if c_min <= c <= c_max:
            i_c = int((c - c_min) / (c_max - c_min) * (num_c - 1))
            accumulator[i_m, i_c] += 1

acc_norm = accumulator / accumulator.max()
threshold = 0.9
lines_mc = np.where(acc_norm >= threshold)

img_lines = img.copy()

for i_m, i_c in zip(lines_mc[0], lines_mc[1]):
    m = m_vals[i_m]
    c = c_vals[i_c]

    x1, x2 = 0, w
    y1 = int(m * x1 + c)
    y2 = int(m * x2 + c)

    cv2.line(img_lines, (x1, y1), (x2, y2), (255, 0, 0), 1)

plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(img_lines, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.title("SlopeIntercept Hough")
plt.show()
````
