To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python

def Ellipse(Ix, Iy):
    # Find a,b,c using 2nd moments
    a = np.sum(Ix**2)
    b = 2*np.sum(Ix * Iy)
    c = np.sum(Iy**2)

    # Calculate lambda1 and lambda2
    lambda1 = (a + c - np.sqrt((a - c)**2 + b**2)) / 2
    lambda2 = (a + c + np.sqrt((a - c)**2 + b**2)) / 2
    return lambda1, lambda2

def R(img_gray,k, T):

  # Obtaining Gradients
  Ix=cv2.Sobel(img_gray, cv2.CV_64F, 1, 0, ksize=3)
  Iy=cv2.Sobel(img_gray, cv2.CV_64F, 0, 1, ksize=3)
  h, w = Ix.shape
  R = np.zeros((h, w))

  for i in range(1, h-1):
    for j in range(1, w-1):

      Ix_w = Ix[i-1:i+2, j-1:j+2].ravel()
      Iy_w = Iy[i-1:i+2, j-1:j+2].ravel()

      lambda1, lambda2 = Ellipse(Ix_w, Iy_w)

      # Obtaining R value
      R[i, j] = lambda1*lambda2-k*(lambda1+lambda2)**2


  return (R > T).astype(np.uint8)*255


R_flat=R(flat_gray,0.04,0)
R_edge=R(edge_gray,0.04,0)
R_corner=R(corner_gray,0.04,0)

# Displaying images
plt.figure(figsize=(24,6))
# Flat
plt.subplot(1,6,1)
plt.imshow(flat_gray,cmap='gray')
plt.axis("off")
plt.title("Flat Image")

plt.subplot(1, 6, 2)
plt.imshow(R_flat,cmap='gray')
plt.axis("off")
plt.title("Flat R Map")


# Edge
plt.subplot(1,6,3)
plt.imshow(edge_gray,cmap='gray')
plt.axis("off")
plt.title("Edge Image")

plt.subplot(1, 6, 4)
plt.imshow(R_edge,cmap='gray')
plt.axis("off")
plt.title("Edge R Map")


# Corner
plt.subplot(1,6,5)
plt.imshow(corner_gray,cmap='gray')
plt.axis("off")
plt.title("Corner Image")

plt.subplot(1, 6, 6)
plt.imshow(R_corner,cmap='gray')
plt.axis("off")
plt.title("Corner R Map")

plt.show()







````

````python

def draw_lines_m_c(img_gray, T):
    h, w = img_gray.shape

    #  Canny edge detection
    edges = cv2.Canny(img_gray, 100, 200)


    ys, xs = np.where(edges > 0)

    #  Define slope (m) and intercept (c) ranges
    m_values = np.linspace(-5, 5, 1500)
    c_values= np.linspace(-h, h,1500 )
    Nm = len(m_values)
    Nc = len(c_values)

    #  Accumulator
    accumulator = np.zeros((len(m_values), len(c_values)), dtype=np.float32)

    d = c_values[1] - c_values[0]

    #  Voting
    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_values):
            c = y - m * x
            ci = int((c - c_values[0]) // d)
            if 0 <= ci < len(c_values):
                accumulator[mi, ci] += 1

    #  Normalize accumulator
    if accumulator.max() > 0:
        accumulator /= accumulator.max()

    #  Threshold accumulator
    indices = np.where(accumulator > T)

    detected_m = m_values[indices[0]]
    detected_c = c_values[indices[1]]

    # Drawing Lines
    for m, c in zip(detected_m, detected_c):
        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img, (x1, y1), (x2, y2), (0, 0, 255), 1)

    #  Display results
    plt.figure(figsize=(15,5))

    plt.subplot(1,3,1)
    plt.title("Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Accumulator")
    plt.imshow(accumulator, cmap='gray',)
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Slope-Intercept parametrization")
    plt.imshow(img)
    plt.axis("off")

    plt.show()
draw_lines_m_c(img_gray,0.9)
````

````python
# Rho-Theta Parameter Space
def lines_rho_theta(img_gray, T):
    h, w = img_gray.shape

    #  Canny edge detection
    edges = cv2.Canny(img_gray, 100, 200)


    ys,xs=np.where(edges>0)

    #3 rho, theta
    r=np.sqrt(h**2+w**2)
    rho_values=np.linspace(0,r,num=500)
    theta_values=np.linspace(0,np.pi,num=900)
    Nr=len(rho_values)
    Nt=len(theta_values)

    #4. Accumulator
    accumulator=np.zeros((Nr,Nt),dtype=np.float32)

    d=rho_values[1]-rho_values[0]


    #5. Voting
    for x,y in zip(xs,ys):
      for theta_i, theta in enumerate(theta_values):
        rho=-x*np.sin(theta)+y*np.cos(theta)
        rho_i=int((rho-rho_values[0])//d)
        if 0<=rho_i<Nr:
          accumulator[rho_i,theta_i]+=1

    #6. Normalize
    if accumulator.max()>0:
      accumulator/=accumulator.max()

    #7. Threshold
    indices=np.where(accumulator>T)

    detected_rho=rho_values[indices[0]]
    detected_theta=theta_values[indices[1]]

    for rho,theta in zip(detected_rho,detected_theta):
      a=np.cos(theta)
      b=np.sin(theta)
      x0=a*rho
      y0=b*rho
      x1=int(x0-1000*(b))
      y1=int(y0+1000*(a))
      x2=int(x0+1000*(b))
      y2=int(y0-1000*(a))
      if 0 <= y1 < h and 0 <= y2 < h:
        cv2.line(img,(x1,y1),(x2,y2),(0,0,255),1)

    plt.figure(figsize=(15,5))
    #  Edge-Map
    plt.subplot(1,3,1)
    plt.title("Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    # Accumulator
    plt.subplot(1,3,2)
    plt.title("Accumulator")
    plt.imshow(accumulator, cmap='gray',)
    plt.axis("off")

    # Hough Lines
    plt.subplot(1,3,3)
    plt.title("Rho-Theta")
    plt.imshow(img)
    plt.axis("off")

    plt.show()

lines_rho_theta(img_gray,0.9)
````

````python
def overlap_stencil(ball, stencil, cx, cy):
    #Normalize
    ball = ball.astype(float) / 255.0
    out = ball.copy()

    h, w = stencil.shape[:2]

    x0 = cx - w // 2
    y0 = cy - h // 2

    for i in range(h):
        for j in range(w):
            y = y0 + i
            x = x0 + j
            if 0 <= y < out.shape[0] and 0 <= x < out.shape[1]:
                a = stencil[i, j, 3]
                out[y, x] = (
                    a * stencil[i, j, :3] +
                    (1 - a) * out[y, x]
                )

    return (out * 255).astype(np.uint8)
````
