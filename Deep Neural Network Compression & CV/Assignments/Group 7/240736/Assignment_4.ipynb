To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Sobel kernels
sobel_x = np.array([
    [-1,  0,  1],
    [-2,  0,  2],
    [-1,  0,  1]
])

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
])
````

````python
def convolve_gray(img, kernel):
    img = img.astype(float)
    kernel = np.flipud(np.fliplr(kernel))

    H, W = img.shape
    kH, kW = kernel.shape
    pad_h = kH // 2
    pad_w = kW // 2

    padded = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode="reflect")
    out = np.zeros((H, W))

    for i in range(H):
        for j in range(W):
            region = padded[i:i+kH, j:j+kW]
            out[i, j] = np.sum(region * kernel)

    return out
````

````python
def sobel_and_scatter(gray_img, title, threshold=None):
    # Sobel gradients
    Ix = convolve_gray(gray_img, sobel_x)
    Iy = convolve_gray(gray_img, sobel_y)

    # Optional thresholding
    if threshold is not None:
        mask = np.sqrt(Ix**2 + Iy**2) > threshold
        Ix_plot = Ix[mask]
        Iy_plot = Iy[mask]
    else:
        Ix_plot = Ix.flatten()
        Iy_plot = Iy.flatten()

    # Plot gradients
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap="gray")
    plt.title(f"{title} - Ix")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap="gray")
    plt.title(f"{title} - Iy")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.scatter(Ix_plot, Iy_plot, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix Scatter Plot")

    plt.show()

    return Ix, Iy
````

````python
def harris_r_map(Ix, Iy, window_size=5, k=0.04):
    """
    Computes Harris R-map from Ix and Iy
    Ix, Iy: gradient images (same shape)
    window_size: size of local window (odd)
    k: Harris constant
    """
    assert window_size % 2 == 1, "Window size must be odd"

    H, W = Ix.shape
    pad = window_size // 2

    # Pad gradients
    Ix_p = np.pad(Ix, ((pad, pad), (pad, pad)), mode="reflect")
    Iy_p = np.pad(Iy, ((pad, pad), (pad, pad)), mode="reflect")

    R = np.zeros((H, W))

    for i in range(H):
        for j in range(W):
            # Local window
            Ix_w = Ix_p[i:i+window_size, j:j+window_size]
            Iy_w = Iy_p[i:i+window_size, j:j+window_size]

            # Second moments
            a = np.sum(Ix_w ** 2)
            b = np.sum(Ix_w * Iy_w)
            c = np.sum(Iy_w ** 2)

            # Harris response
            R[i, j] = (a * c - b ** 2) - k * (a + c) ** 2

    return R
````

````python
def show_r_map(gray_small, title, window_size=5, k=0.04):
    # Gradients
    Ix = convolve_gray(gray_small, sobel_x)
    Iy = convolve_gray(gray_small, sobel_y)

    # Harris R-map
    R = harris_r_map(Ix, Iy, window_size, k)

    # Normalize for display
    R_norm = (R - R.min()) / (R.max() - R.min() + 1e-8)

    plt.figure(figsize=(6,6))
    plt.imshow(R_norm, cmap="gray")
    plt.title(f"{title} - Harris R Map")
    plt.axis("off")
    plt.show()

    return R
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("download(8).jpeg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Canny edge map
edges = cv2.Canny(gray, 100, 200)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Map")
plt.axis("off")

plt.show()
````

````python
def hough_mc(edges, m_range=(-2, 2), m_steps=400, c_steps=400):
    H, W = edges.shape

    m_vals = np.linspace(m_range[0], m_range[1], m_steps)
    c_vals = np.linspace(-H, H, c_steps)

    acc = np.zeros((m_steps, c_steps))

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int((c + H) * (c_steps / (2 * H)))
            if 0 <= ci < c_steps:
                acc[mi, ci] += 1

    return acc, m_vals, c_vals
````

````python
acc_mc, m_vals, c_vals = hough_mc(edges)

# Normalize accumulator
acc_norm = acc_mc / acc_mc.max()

# Threshold (90% of max)
threshold = 0.9
lines = np.argwhere(acc_norm > threshold)

img_lines = img_rgb.copy()

for mi, ci in lines:
    m = m_vals[mi]
    c = c_vals[ci]

    # Draw line y = mx + c
    x0, x1 = 0, img.shape[1]
    y0 = int(c)
    y1 = int(m * x1 + c)

    cv2.line(img_lines, (x0, y0), (x1, y1), (255, 0, 0), 2)

plt.figure(figsize=(6,6))
plt.imshow(img_lines)
plt.title("SlopeIntercept Hough Lines")
plt.axis("off")
plt.show()
````

````python
def hough_rho_theta(edges, rho_res=1, theta_res=1):
    H, W = edges.shape
    diag = int(np.sqrt(H*H + W*W))

    rhos = np.arange(-diag, diag, rho_res)
    thetas = np.deg2rad(np.arange(-90, 90, theta_res))

    acc = np.zeros((len(rhos), len(thetas)))

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int((rho + diag) / rho_res)
            if 0 <= ri < len(rhos):
                acc[ri, ti] += 1

    return acc, rhos, thetas
````

````python
acc_rt, rhos, thetas = hough_rho_theta(edges)

acc_norm = acc_rt / acc_rt.max()
threshold = 0.5

img_lines_rt = img_rgb.copy()
lines = np.argwhere(acc_norm > threshold)

for ri, ti in lines:
    rho = rhos[ri]
    theta = thetas[ti]

    a = np.cos(theta)
    b = np.sin(theta)

    x0 = a * rho
    y0 = b * rho

    pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
    pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))

    cv2.line(img_lines_rt, pt1, pt2, (0, 255, 0), 2)

plt.figure(figsize=(6,6))
plt.imshow(img_lines_rt)
plt.title("RhoTheta Hough Lines")
plt.axis("off")
plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def create_ball_stencil(img, x1, y1, x2, y2):
    crop = img[y1:y2, x1:x2].copy()

    gray = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)

    # Gradients
    gx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    gy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
    grad = np.sqrt(gx**2 + gy**2)

    # Threshold gradient magnitude
    thresh = np.mean(grad)
    binary = (grad > thresh).astype(np.uint8) * 255

    # Find contours
    contours, _ = cv2.findContours(
        binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE
    )

    # Largest contour  ball
    largest = max(contours, key=cv2.contourArea)

    mask = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(mask, [largest], -1, 255, -1)

    # Create RGBA
    alpha = mask / 255.0
    rgba = np.zeros((crop.shape[0], crop.shape[1], 4), dtype=float)
    rgba[..., :3] = crop / 255.0
    rgba[..., 3] = alpha

    return rgba
````

````python
def overlay_stencil(bg, stencil, cx, cy):
    bg = bg.astype(float) / 255.0
    out = bg.copy()

    h, w = stencil.shape[:2]
    x0 = cx - w // 2
    y0 = cy - h // 2

    for i in range(h):
        for j in range(w):
            y = y0 + i
            x = x0 + j
            if 0 <= y < out.shape[0] and 0 <= x < out.shape[1]:
                a = stencil[i, j, 3]
                out[y, x] = (
                    a * stencil[i, j, :3] +
                    (1 - a) * out[y, x]
                )

    return (out * 255).astype(np.uint8)
````
