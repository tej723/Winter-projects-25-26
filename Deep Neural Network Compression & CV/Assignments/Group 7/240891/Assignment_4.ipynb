To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def rgb2gray(img):
    return (0.3*img[:,:,0] + 0.6*img[:,:,1] + 0.1*img[:,:,2]).astype(np.float32)

g1 = rgb2gray(img1)
g2 = rgb2gray(img2)
g3 = rgb2gray(img3)
````

````python
def my_convolve(img, kernel):
    k = kernel.shape[0]//2
    padded = np.pad(img, k, mode="reflect")
    out = np.zeros_like(img)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            out[i,j] = np.sum(padded[i:i+2*k+1, j:j+2*k+1] * kernel)

    return out
````

````python
def sobel_scatter(gray, title, thresh):
    Kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    Ky = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])

    Ix = my_convolve(gray, Kx)
    Iy = my_convolve(gray, Ky)

    mag = np.sqrt(Ix**2 + Iy**2)
    mask = mag > thresh

    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap="gray"); plt.title("Ix"); plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap="gray"); plt.title("Iy"); plt.axis("off")

    plt.subplot(1,3,3)
    plt.scatter(Ix[mask], Iy[mask], s=1)
    plt.title("Iy vs Ix")

    plt.suptitle(title)
    plt.show()

    return Ix, Iy
````

````python
Ix1, Iy1 = sobel_scatter(g1, "Flat Image", 10)
Ix2, Iy2 = sobel_scatter(g2, "Edge Image", 20)
Ix3, Iy3 = sobel_scatter(g3, "Corner Image", 30)
````

````python
def harris_R(Ix, Iy, win=5, k=0.05):
    pad = win//2
    Ix2, Iy2, Ixy = Ix*Ix, Iy*Iy, Ix*Iy

    Ix2 = np.pad(Ix2, pad, mode="reflect")
    Iy2 = np.pad(Iy2, pad, mode="reflect")
    Ixy = np.pad(Ixy, pad, mode="reflect")

    R = np.zeros(Ix.shape)

    for i in range(R.shape[0]):
        for j in range(R.shape[1]):
            a = np.sum(Ix2[i:i+win, j:j+win])
            b = np.sum(Ixy[i:i+win, j:j+win])
            c = np.sum(Iy2[i:i+win, j:j+win])

            det = a*c - b*b
            trace = a + c
            R[i,j] = det - k*(trace**2)

    return R
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def extract_ball_rgba(image, p1, p2):
    # Crop region
    roi = image[p1[1]:p2[1], p1[0]:p2[0]].copy()

    # Convert to grayscale
    gray_roi = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)

    # Compute gradients separately
    grad_x = cv2.Sobel(gray_roi, cv2.CV_32F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(gray_roi, cv2.CV_32F, 0, 1, ksize=3)

    magnitude = np.abs(grad_x) + np.abs(grad_y)

    # Adaptive threshold (percentile instead of mean)
    t = np.percentile(magnitude, 60)
    edge_map = (magnitude > t).astype(np.uint8) * 255

    # Extract contours
    cnts, _ = cv2.findContours(edge_map, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Choose contour with max area
    ball_cnt = max(cnts, key=lambda c: cv2.contourArea(c))

    # Create mask
    mask = np.zeros(gray_roi.shape, dtype=np.uint8)
    cv2.fillPoly(mask, [ball_cnt], 255)

    # Build RGBA stencil
    alpha = mask.astype(float) / 255.0
    rgba = np.zeros((roi.shape[0], roi.shape[1], 4), dtype=float)
    rgba[..., :3] = roi.astype(float) / 255.0
    rgba[..., 3] = alpha

    return rgba
````

````python
def paste_rgba(background, stencil, center):
    bg = background.astype(float) / 255.0
    result = bg.copy()

    sh, sw = stencil.shape[:2]
    cx, cy = center

    start_x = cx - sw // 2
    start_y = cy - sh // 2

    for r in range(sh):
        for c in range(sw):
            y = start_y + r
            x = start_x + c

            if 0 <= y < result.shape[0] and 0 <= x < result.shape[1]:
                alpha = stencil[r, c, 3]
                if alpha > 0:
                    result[y, x] = (
                        stencil[r, c, :3] * alpha +
                        result[y, x] * (1 - alpha)
                    )

    return (result * 255).astype(np.uint8)
````
