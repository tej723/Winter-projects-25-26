To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Custom function for convolution
def convolve(img, kernel):
  out_img = np.zeros(img.shape)

  # Getting shapes for iteration and padding
  img_h, img_w = img.shape[:2]
  k_h, k_w = kernel.shape[:2]

  # Padding
  pad_h = k_h // 2
  pad_w = k_w // 2
  pad_img = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode = 'reflect')

  # Convolution (* just does element wise multiplication)
  for i in range(img_h):
    for j in range(img_w):
      out_img[i,j] = np.sum(pad_img[i:i+k_h, j:j+k_w] * kernel) # np.sum() computes convolution sum at a pixel

  return out_img
````

````python
# Sobel X
sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
# Sobel Y
sobel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
````

````python
# Takes RGB Image and returns Ix, Iy
def compute_Ix_Iy(img):
  # Conversion to Grayscale
  img_gray = get_grayscale(img)
  # Convolution of sobel kernels to get Ix, Iy
  Ix = convolve(img_gray, sobel_x)
  Iy = convolve(img_gray, sobel_y)
  return Ix, Iy
````

````python
# Takes RGB Image and plots its corresponding Ix and Iy
def plot(img, title):
  # Above defined function to compute Ix and Iy
  Ix, Iy = compute_Ix_Iy(img)

  # Display
  plt.figure(figsize=(12,5))

  plt.subplot(1,2,1)
  plt.imshow(Ix, cmap = "gray")
  plt.title("Ix")
  plt.axis("off")

  plt.subplot(1,2,2)
  plt.imshow(Iy, cmap = "gray")
  plt.title("Iy")
  plt.axis("off")

  plt.suptitle(title, fontsize=20)  # Super Title
  plt.show()
````

````python
# Thresholding to suppress weak gradient response mostly due to noise
def threshold_Ix_Iy(img, threshold):
  Ix, Iy = compute_Ix_Iy(img) # Already defined above
  # Threshold
  mask = np.abs(Ix) + np.abs(Iy) > threshold
  Ix_thres = Ix * mask
  Iy_thres = Iy * mask

  return Ix_thres, Iy_thres
````

````python
# Custom Finction for scatter plot
def scatter_plot(img, threshold = 30):  # Takes corresponding RGB Image as input
  Ix_thres, Iy_thres = threshold_Ix_Iy(img, threshold)  # Using threshold of 30

  # Ix and Iy are 2D array but scatter expects 1D, so using .flatten()
  Ix_thres = Ix_thres.flatten()
  Iy_thres = Iy_thres.flatten()

  # Display
  plt.scatter(Ix_thres, Iy_thres)
  plt.xlabel("Ix")
  plt.ylabel("Iy")
  plt.show()
````

````python
# Custom function which takes Ix, Iy and window size to compute a,b and c (the second moments)
def second_moment(Ix, Iy, window_size = 3):
  # Computing Ix^2, Iy^2, Ix * Iy
  Ix2 = Ix * Ix
  Iy2 = Iy * Iy
  IxIy = Ix * Iy

  # Padding
  pad = window_size // 2
  Ix2_pad = np.pad(Ix2, pad, mode = 'reflect')
  Iy2_pad = np.pad(Iy2, pad, mode = 'reflect')
  IxIy_pad = np.pad(IxIy, pad, mode = 'reflect')

  # Making a,b and c
  a = np.zeros_like(Ix)
  b = np.zeros_like(Ix)
  c = np.zeros_like(Ix)

  # Computing the second moments
  for i in range(Ix.shape[0]):
    for j in range(Ix.shape[1]):
      a[i,j] = np.sum(Ix2_pad[i:i+window_size, j:j+window_size])
      b[i,j] = np.sum(IxIy_pad[i:i+window_size, j:j+window_size])
      c[i,j] = np.sum(Iy2_pad[i:i+window_size, j:j+window_size])

  return a,b,c
````

````python
# Custom function which takes the second moments and return corresponding eigenvalues
def eigenvalues(a,b,c):
  # Using the formulas given in pre-session videos
  second_term = np.sqrt((b**2)+((a-c)**2))
  first_term = a + c

  lambda1 = (first_term + second_term) / 2
  lambda2 = (first_term - second_term) / 2

  return lambda1, lambda2
````

````python
# Custom Function to compute R_map
def R_map(img, window_size, k):
  # Using above defined functions
  Ix, Iy = compute_Ix_Iy(img)
  a,b,c = second_moment(Ix, Iy, window_size)
  lambda1, lambda2 = eigenvalues(a,b,c)

  # Using the formula R = det - k * (trace)^2
  R = (lambda1 * lambda2) - k * (lambda1 + lambda2) ** 2

  return R
````

````python
# Custom function to show the R_map
def show_R_map(img, title, window_size = 3, k = 0.04):  # k ranges from 0.04 to 0.08
  R = R_map(img, window_size, k)  # Already defined above

  # Normalization for display
  R_norm = (R - R.min()) / (R.max() - R.min() + 1e-8)  # Adding small term in denominator to avoid division by zero

  # Display
  plt.figure(figsize=(12,6))
  plt.imshow(R, cmap = 'gray')
  plt.title(title)
  plt.axis("off")
  plt.show()
````

````python
# Custom function to get Canny Edge Map
def canny(img):
  gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
  edge = cv2.Canny(gray, 100, 200)  # lower threshold = 100, upper threshold = 200 according to our image
  return edge
````

````python
# Custom function to get m_values, c_values and accumulator
def slope_intercept(edges):
  # Get shapes
  h,w = edges.shape

  # Getting the m_values (slope) using below range
  m_min, m_max, m_step = -2, 2, 0.02
  m_vals = np.arange(m_min, m_max, m_step)

  # Getting the c_values (intercept) using below range
  c_min, c_max, c_step = -h, h, 2
  c_vals = np.arange(c_min, c_max, c_step)

  # Initialising accumulator
  acc = np.zeros((len(m_vals), len(c_vals)))

  # Edges is a binary edge image from canny edge map
  ys, xs = np.nonzero(edges)  # Returns indices of all non-zero (edge) pixels

  # Loop for all edge pixels
  for y,x in zip(ys,xs):
    # For each pixel trying all possible m_values
    for i,m in enumerate(m_vals):
      # Using y = m * x + c
      c = y - m * x
      # COnverting c_values into accumulator index
      j = int((c-c_min) / c_step)
      # If index lies inside accumulator bounds
      if 0 <= j < len(c_vals):
        # Vote for line parameterized by m,c
        acc[i,j] += 1

  return acc, m_vals, c_vals
````

````python
# Custom Function that takes RGB Image and threshold and returns Output Image with lines detected using Hough Transform
def hough_transform_mc(img, threshold = 0.9):
  edges = canny(img)  # Already defined above
  h,w = edges.shape
  acc, m_vals, c_vals = slope_intercept(edges)  # Already defined
  acc_norm = acc / acc.max()  # Normalization
  lines = np.argwhere(acc_norm > threshold) # Thresholding

  # Creating a copy of Image to create output image
  output = img.copy()
  # Drawing lines
  for i,j in lines:
    m = m_vals[i]
    c = c_vals[j]
    x1, y1 = 0 ,int(c)
    x2, y2 = w, int(m * w + c)
    if 0 <= y1 < h or 0 <= y2 < h:
      cv2.line(output, (x1,y1), (x2,y2), (255,0,0), 6)  # Red Color line with thickness = 6

  return output
````

````python
# Custom Function to do Hough Transform using Rho-Theta method
def hough_rho_theta(img, threshold=0.5):
  edges = canny(img)  # Already defined
  h, w = edges.shape

  # Parameter space
  diag_len = int(np.sqrt(h*h + w*w))
  rhos = np.arange(-diag_len, diag_len + 1, 1)
  thetas = np.deg2rad(np.arange(-90, 90, 1))  # Conversion to radians for cos, sin

  # Accumulator
  acc = np.zeros((len(rhos), len(thetas)), dtype=np.uint64)

  cos_t = np.cos(thetas)
  sin_t = np.sin(thetas)

  ys, xs = np.nonzero(edges)

  # Voting
  for y, x in zip(ys, xs):
    for t_idx in range(len(thetas)):
      # Using formula x * cos(theta) - y * sin(theta) + rho = 0
      rho = y * cos_t[t_idx] - x * sin_t[t_idx]
      rho_idx = int(round(rho)) + diag_len

      if 0 <= rho_idx < len(rhos):
        acc[rho_idx, t_idx] += 1

  acc_norm = acc / acc.max()  # Normalization
  lines = np.argwhere(acc_norm > threshold) # Thresholding

  output = img.copy()

  # Drawing lines
  for rho_idx, theta_idx in lines:
    rho = rhos[rho_idx]
    theta = thetas[theta_idx]

    a, b = np.cos(theta), np.sin(theta)
    # Compute point (x0,y0) on line using the normal form
    x0 = -b * rho
    y0 = a * rho
    # Extend line to span across image by generating two far-points along the line direction
    x1 = int(x0 + 1000 * (a))
    y1 = int(y0 + 1000 * (b))
    x2 = int(x0 - 1000 * (a))
    y2 = int(y0 - 1000 * (b))
    # Drawing detected line on output
    cv2.line(output, (x1, y1), (x2, y2), (255, 0, 0), 6)

  return output
````

````python
# Custom function to create RGBA stencil
def create_ball_stencil(img, x1, y1, x2, y2):
  # Crop the region of the ball
  roi = img[y1:y2, x1:x2]
  # Getting edge map
  edges = canny(roi)  # Already defined (can also use Sobel as was done in Assignment 3)

  # Finding contours from binary edges image
  contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  # Select the largest contour i.e. one with maximum area
  largest_contour = max(contours, key = cv2.contourArea)

  # Creating binary mask from largest contour
  mask = np.zeros(roi.shape[:2])
  cv2.drawContours(mask, [largest_contour], -1, 255, thickness=-1)

  # Creating RGBA stencil
  rgba = cv2.cvtColor(roi, cv2.COLOR_RGB2RGBA)
  rgba[:, :, 3] = mask   # Alpha channel

  return rgba
````

````python
# Since ball stencil has blue reflection from blue pool,
# this custom function convert it to green reflection for green pool using hue rotation
def hue_rotation(img_rgba):
  # Split alpha channel
  rgb = img_rgba[:, :, :3]
  alpha = img_rgba[:, :, 3]

  # Conversion to HSV
  hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)

  h, s, v = cv2.split(hsv)

  # Blue color range in Hue channel
  blue_mask = (h >= 100) & (h <= 130)

  h[blue_mask] -= 60  # Getting blue converted to green

  # Recombine
  hsv_mod = cv2.merge([h, s, v])
  rgb_mod = cv2.cvtColor(hsv_mod, cv2.COLOR_HSV2RGB)

  # Reattach alpha
  rgba = np.dstack([rgb_mod, alpha])

  return rgba
````

````python
# Custom Function to overlay stencil on backgound with (x3,y3) as center point
def overlay_stencil(background, stencil_rgba, x3, y3):
  bg = background.copy()
  h, w = stencil_rgba.shape[:2]

  # Compute top-left corner so that (x3, y3) is center
  x1 = int(x3 - w // 2)
  y1 = int(y3 - h // 2)

  # Similarly, computing bottom-right corner
  x2 = x1 + w
  y2 = y1 + h

  # Clipping to background bounds
  H, W = bg.shape[:2]
  sx1, sy1 = max(0, x1), max(0, y1)
  sx2, sy2 = min(W, x2), min(H, y2)

  # Corresponding stencil region
  st_x1 = sx1 - x1
  st_y1 = sy1 - y1
  st_x2 = st_x1 + (sx2 - sx1)
  st_y2 = st_y1 + (sy2 - sy1)

  # Extract regions
  stencil_crop = stencil_rgba[st_y1:st_y2, st_x1:st_x2]
  bg_crop = bg[sy1:sy2, sx1:sx2]

  # Separating alpha channel and Normalization
  alpha = stencil_crop[:, :, 3] / 255.0
  alpha = alpha[:, :, None]

  # Alpha blending
  bg[sy1:sy2, sx1:sx2] = (alpha * stencil_crop[:, :, :3] + (1 - alpha) * bg_crop)
  bg = bg.astype(np.uint8)

  return bg
````
