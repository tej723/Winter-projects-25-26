````python
def gray_histogram(gray_img):             # grayscale histogram using numpy
  histogram=np.zeros(256)                 #0-255
  for i in range (gray_img.shape[0]):
    for j in range (gray_img.shape[1]):
      histogram[gray_img[i,j]]+=1
  return histogram


h= gray_histogram (gray)
plt.plot(h)
plt.title('Grayscale Histogram')
plt.show()

````

````python
def rgb_histogram(rgb_img):
   hist_r=np.zeros(256)
   hist_g=np.zeros(256)
   hist_b=np.zeros(256)
   r , g , b = rgb_img[:,:,0], rgb_img[:,:,1], rgb_img[:,:,2]
   for i in range (rgb_img.shape[0]):
    for j in range (rgb_img.shape[1]):
      hist_r[r[i,j]]+=1
      hist_g[g[i,j]]+=1
      hist_b[b[i,j]]+=1
   return hist_r,hist_g ,hist_b

plt.figure(figsize=(6,6))
plt.plot( rgb_histogram(rgb)[0], 'r', color='red')
plt.plot( rgb_histogram(rgb)[1], 'g' ,color='green')
plt.plot( rgb_histogram(rgb)[2], 'b',color='blue')

plt.title('RGB Histogram')
plt.legend()
plt.show()
````

````python
def oops_too_much_senpai(img):
    arr = img.astype(np.float32)
    if arr.max() <= 1:
        arr = arr * 255.0  #to convert everthing to floatt 0-255 for saferside

    gray = 0.299 * arr[:,:,0] + 0.587 * arr[:,:,1] + 0.114 * arr[:,:,2]     #grayscale


    s = float(0.9)
    arr = gray[..., None] * (1.0 - s) + arr * s  #mixing grayscale and orginal image to produce low saturation img

    contrast = float(0.7)
    arr = 128.0 + contrast * (arr - 128.0)
    arr = np.clip(arr, 0, 255)


    y = float(0.2)
                                   # yellow = red + green so i reduced blue and increased red and green
    arr[:,:,0] += y*20
    arr[:,:,1] += y*15
    arr[:,:,2] -= y*15


    fade = np.array([240, 235, 220], dtype=np.float32)

    f = float(0.2)
    arr = arr * (1.0 - f) + fade * f


    h, w = arr.shape[:2]                  # for dark corners
    y, x = np.indices((h, w))
    cx, cy = w/2, h/2
    distance = np.sqrt((x - cx)**2 + (y - cy)**2)
    distance = distance / distance.max()
    vignette = 1 - (distance ** 1) * 0.5
    arr = arr * vignette[:,:, None]


    return np.clip(arr, 0, 255).astype(np.uint8)
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
def white_balance(img_rgb):
  rgb=img_rgb.astype(np.float32)
  r,g,b=rgb[:,:,0],rgb[:,:,1],rgb[:,:,2]
  V=np.max(rgb,axis=-1)
  idx=np.unravel_index(V.argmax(),V.shape)
  brightest_pixel=rgb[idx]
  r_w,g_w,b_w=brightest_pixel[0],brightest_pixel[1],brightest_pixel[2]

  r_s = 255/r_w
  g_s = 255/g_w
  b_s=  255/b_w
  r_after = r*r_s
  g_after = g*g_s
  b_after =  b*b_s


  rgb=np.stack((r_after,g_after,b_after),axis=-1)
  white_balance_img=np.clip(rgb, 0, 255).astype(np.uint8)

  return white_balance_img
````
