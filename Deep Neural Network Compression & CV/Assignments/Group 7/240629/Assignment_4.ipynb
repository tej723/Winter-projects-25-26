To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    pad_h = kh // 2
    pad_w = kw // 2

    padded_image = np.pad(image, ((pad_h, pad_h),
                                  (pad_w, pad_w)),
                                  mode='constant')

    output = np.zeros((h, w))

    for i in range(h):
        for j in range(w):
            region = padded_image[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)

    return output
````

````python
def corner_analysis(image, threshold):

    # Sobel gradients
    Ix = convolve2d(image, Gx)
    Iy = convolve2d(image, Gy)

    # Plot Ix
    plt.figure()
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix")
    plt.axis('off')
    plt.show()

    # Plot Iy
    plt.figure()
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy")
    plt.axis('off')
    plt.show()

    # Scatter plot Iy vs Ix
    x = Ix.flatten()
    y = Iy.flatten()

    mask = (np.abs(x) > threshold) | (np.abs(y) > threshold)

    plt.figure()
    plt.scatter(x[mask], y[mask], s=5)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix (threshold = {threshold})")
    plt.show()

    return Ix, Iy
````

````python
size = 100

# Flat image
flat_img = np.ones((size, size)) * 128

# Edge image (vertical edge)
edge_img = np.zeros((size, size))
edge_img[:, size//2:] = 255

# Corner image
corner_img = np.zeros((size, size))
corner_img[size//2:, size//2:] = 255
````

````python
threshold = 50

Ix_flat, Iy_flat = corner_analysis(flat_img,threshold)
Ix_edge, Iy_edge = corner_analysis(edge_img,threshold)
Ix_corner, Iy_corner = corner_analysis(corner_img, threshold)
````

````python
def second_moment_matrix(Ix, Iy, window_size):

    half = window_size // 2

    a = np.zeros_like(Ix)
    b = np.zeros_like(Ix)
    c = np.zeros_like(Ix)

    for i in range(half, Ix.shape[0] - half):
        for j in range(half, Ix.shape[1] - half):

            Wx = Ix[i-half:i+half+1, j-half:j+half+1]
            Wy = Iy[i-half:i+half+1, j-half:j+half+1]

            a[i, j] = np.sum(Wx ** 2)
            b[i, j] = np.sum(Wx * Wy)
            c[i, j] = np.sum(Wy ** 2)

    return a, b, c
````

````python
def eigenvalues(a, b, c):

    trace = a + c
    det_term = np.sqrt((a - c)**2 + 4 * b**2)

    lambda1 = (trace + det_term) / 2
    lambda2 = (trace - det_term) / 2

    return lambda1, lambda2
````

````python
def harris_response(a, b, c, k=0.04):

    det_M = a * c - b**2
    trace_M = a + c

    R = det_M - k * (trace_M ** 2)

    return R
````

````python
def compute_R_map(image, window_size=3, k=0.04):

    Ix = convolve2d(image, Gx)
    Iy = convolve2d(image, Gy)

    a, b, c = second_moment_matrix(Ix, Iy, window_size)

    R = harris_response(a, b, c, k)

    return R
````

````python
def show_R_map(R, title):

    plt.figure()
    plt.imshow(R, cmap='gray')
    plt.title(title)
    plt.colorbar()
    plt.axis('off')
    plt.show()
````

````python
#METHOD 1: SLOPEINTERCEPT HOUGH
m_min, m_max = -2, 2
c_min, c_max = -300, 300

m_bins = 200
c_bins = 200

m_vals = np.linspace(m_min, m_max, m_bins)
c_vals = np.linspace(c_min, c_max, c_bins)

accumulator = np.zeros((m_bins, c_bins))
````

````python
ys, xs = np.where(edges > 0)

for x, y in zip(xs, ys):
    for mi, m in enumerate(m_vals):
        c = y - m * x

        if c_min <= c <= c_max:
            ci = int((c - c_min) * (c_bins - 1) / (c_max - c_min))
            accumulator[mi, ci] += 1
````

````python
acc_norm = accumulator / accumulator.max()

threshold = 0.9   # 90 percent
lines = np.where(acc_norm > threshold)
````

````python
img_lines = img.copy()

for mi, ci in zip(lines[0], lines[1]):
    m = m_vals[mi]
    c = c_vals[ci]

    x1, x2 = 0, img.shape[1]
    y1 = int(m * x1 + c)
    y2 = int(m * x2 + c)

    cv2.line(img_lines, (x1, y1), (x2, y2), (0,0,255), 1)

plt.imshow(cv2.cvtColor(img_lines, cv2.COLOR_BGR2RGB))
plt.title("Slope-Intercept Hough Lines")
plt.axis('off')
plt.show()
````

````python
def create_ball_stencil(image, x1, y1, x2, y2):

    # Crop region containing ball
    crop = image[y1:y2, x1:x2]

    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)

    # Sobel gradients
    gx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    gy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    mag = np.sqrt(gx**2 + gy**2)
    mag = np.uint8(255 * mag / mag.max())

    # Threshold gradients
    _, thresh = cv2.threshold(mag, 50, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL,
                                   cv2.CHAIN_APPROX_SIMPLE)

    # Select largest contour (white ball)
    largest_contour = max(contours, key=cv2.contourArea)

    # Create alpha mask
    alpha = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(alpha, [largest_contour], -1, 255, -1)

    # Create RGBA image
    b, g, r = cv2.split(crop)
    rgba = cv2.merge((b, g, r, alpha))

    return rgba
````

````python
def overlay_stencil(background, stencil, x3, y3):

    bg = background.copy()

    h, w = stencil.shape[:2]

    # Convert center (x3,y3) to top-left
    x_start = int(x3 - w // 2)
    y_start = int(y3 - h // 2)

    # Boundary check
    if x_start < 0 or y_start < 0 or \
       x_start + w > bg.shape[1] or \
       y_start + h > bg.shape[0]:
        return bg

    # Split RGBA
    b, g, r, a = cv2.split(stencil)
    alpha = a.astype(float) / 255.0

    for c, channel in enumerate([b, g, r]):
        bg[y_start:y_start+h, x_start:x_start+w, c] = (
            alpha * channel +
            (1 - alpha) * bg[y_start:y_start+h, x_start:x_start+w, c]
        )

    return bg
````
