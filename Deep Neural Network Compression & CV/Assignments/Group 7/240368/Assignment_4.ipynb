To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Compute eigenvalues from second-moment matrix
def second_moment_eigenvalues(ix_vals, iy_vals):
    m_xx = np.sum(ix_vals ** 2)
    m_xy = np.sum(ix_vals * iy_vals) * 2
    m_yy = np.sum(iy_vals ** 2)

    trace = m_xx + m_yy
    det_term = np.sqrt((m_xx - m_yy) ** 2 + m_xy ** 2)

    eig_small = (trace - det_term) / 2
    eig_large = (trace + det_term) / 2

    return eig_small, eig_large
````

````python
# Compute Harris response map
def compute_harris_response(gray_img, alpha, thresh):

    grad_x = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=3)

    height, width = gray_img.shape
    response_map = np.zeros((height, width))

    for row in range(1, height - 1):
        for col in range(1, width - 1):

            ix_patch = grad_x[row-1:row+2, col-1:col+2].flatten()
            iy_patch = grad_y[row-1:row+2, col-1:col+2].flatten()

            l1, l2 = second_moment_eigenvalues(ix_patch, iy_patch)

            response_map[row, col] = (l1 * l2) - alpha * (l1 + l2) ** 2

    return (response_map > thresh).astype(np.uint8) * 255


harris_flat   = compute_harris_response(flat_img_gray,   0.04, 0)
harris_edge   = compute_harris_response(edge_img_gray,   0.04, 0)
harris_corner = compute_harris_response(corner_img_gray, 0.04, 0)
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img2 = cv2.imread('img2.jpg')

gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

# Thresholds 50 and 150 are standard for clear grid lines
canny_edges = cv2.Canny(gray, 50, 150)

# Displaying the canny Edge Map
plt.figure(figsize=(6, 6))
plt.imshow(canny_edges, cmap='gray')
plt.title("Canny Edge Map ")
plt.axis('off')
plt.show()
````

````python
import numpy as np

def run_slope_intercept(image, edges):
    height, width = edges.shape
    output = image.copy()

    # Accumulator Space (m, c)
    # m: Slope. c: Intercept.
    m_range = np.linspace(-10, 10, 300)
    c_range = np.linspace(-height, height + width, 400)
    accumulator = np.zeros((len(m_range), len(c_range)))


    y_idxs, x_idxs = np.nonzero(edges)
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for m_idx, m in enumerate(m_range):
            # Calculate c = y - mx
            c = y - m * x

            # Check if c falls within our defined bins
            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                accumulator[m_idx, c_idx] += 1

    # The maxima represent the most likely lines in the image


    threshold = 0.9 * np.max(accumulator)
    m_peaks, c_peaks = np.where(accumulator > threshold)



    #Drawing the Lines on Output Image
    for i in range(len(m_peaks)):
        m = m_range[m_peaks[i]]
        c = c_range[c_peaks[i]]

        x1, x2 = 0, width
        y1, y2 = int(m * x1 + c), int(m * x2 + c)

        #  grid lines in green
        cv2.line(output, (x1, y1), (x2, y2), (0, 255, 0), 2)

    return output



result = run_slope_intercept(img2, canny_edges)

plt.figure(figsize=(6, 6))
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.title("Slope-Intercept Result")
plt.axis('off')
plt.show()
````

````python
def solve_rho_theta(image_name):
    # Grayscale image
    img = cv2.imread(image_name)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Canny edge map
    edges = cv2.Canny(gray, 50, 150)
    height, width = edges.shape
    diag_len = int(np.ceil(np.sqrt(width**2 + height**2)))

    # Define Accumulator (rho, theta)
    thetas = np.deg2rad(np.arange(0, 180))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)
    accumulator = np.zeros((len(rhos), len(thetas)))

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)


    y_idxs, x_idxs = np.nonzero(edges)
    for i in range(len(x_idxs)):
        x, y = x_idxs[i], y_idxs[i]
        for t_idx in range(len(thetas)):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            rho_idx = np.argmin(np.abs(rhos - rho))
            accumulator[rho_idx, t_idx] += 1

    # Thresholding (90% threshold)
    threshold = 0.9 * np.max(accumulator)
    rho_peaks, theta_peaks = np.where(accumulator > threshold)

    output = img.copy()
    for i in range(len(rho_peaks)):
        r = rhos[rho_peaks[i]]
        t = thetas[theta_peaks[i]]
        a, b = np.cos(t), np.sin(t)
        x0, y0 = a*r, b*r
        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))
        cv2.line(output, (x1, y1), (x2, y2), (0, 0, 255), 2)

    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.title("Rho-Theta Result (Faster)")
    plt.show()

solve_rho_theta('img2.jpg')
````

The Slope-Intercept ($m, c$) method fails to detect vertical lines because the slope becomes infinite, which is why only the horizontal bars are visible in your green result 1.

The Rho-Theta ($\rho, \theta$) method is much more accurate and robust because it uses polar coordinates to handle all angles equally, successfully capturing the full Sudoku grid2. It is significantly faster and more memory-efficient because its search space is finite and bounded by the image size.
