To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def harris_response_map(img, window_size=3, k=0.04):

    if len(img.shape) == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    img = img.astype(np.float32)

    Ix = cv2.filter2D(img, -1, sobel_x)
    Iy = cv2.filter2D(img, -1, sobel_y)

    H, W = img.shape
    offset = window_size // 2

    R_map = np.zeros((H, W), dtype=np.float32)

    for y in range(offset, H - offset):
        for x in range(offset, W - offset):

            Ix_win = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            Iy_win = Iy[y-offset:y+offset+1, x-offset:x+offset+1]

            a = np.sum(Ix_win ** 2)
            b = np.sum(Ix_win * Iy_win)
            c = np.sum(Iy_win ** 2)

            det = a * c - b ** 2
            trace = a + c

            R_map[y, x] = det - k * (trace ** 2)

    return R_map


def show_rmap(R_map, title='R-map'):
    plt.figure(figsize=(5,5))
    plt.imshow(R_map, cmap='gray')
    plt.colorbar()
    plt.title(title)
    plt.axis('off')
    plt.show()


R_flat = harris_response_map(flat)
show_rmap(R_flat, "R-map (Flat Image)")

R_edge = harris_response_map(edge)
show_rmap(R_edge, "R-map (Edge Image)")

R_corner = harris_response_map(corner)
show_rmap(R_corner, "R-map (Corner Image)")
````

````python
def hough_mc(img,
             m_min=-2, m_max=2, m_bins=180,
             c_bins=180,
             acc_thresh=0.9):

    # Ensure grayscale
    if len(img.shape) == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    else:
        gray = img.copy()

    edges = cv2.Canny(gray, 100, 200)

    H, W = gray.shape

    m_vals = np.linspace(m_min, m_max, m_bins)
    c_vals = np.linspace(-H, H, c_bins)

    accumulator = np.zeros((m_bins, c_bins), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    # Voting
    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x

            c_idx = int((c + H) * (c_bins / (2 * H)))
            if 0 <= c_idx < c_bins:
                accumulator[mi, c_idx] += 1

    # Normalize accumulator
    accumulator /= accumulator.max()

    # Threshold
    line_indices = np.where(accumulator >= acc_thresh)

    return line_indices, m_vals, c_vals, edges

def draw_mc_lines(img, lines, m_vals, c_vals):

    if len(img.shape) == 2:
        out = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    else:
        out = img.copy()

    H, W = img.shape[:2]

    for m_idx, c_idx in zip(lines[0], lines[1]):
        m = m_vals[m_idx]
        c = c_vals[c_idx]

        x1, x2 = 0, W
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        cv2.line(out, (x1, y1), (x2, y2), (0, 0, 255), 1)

    return out

    img=edge


lines_mc, m_vals, c_vals, edges_mc = hough_mc(img, acc_thresh=0.9)
result_mc = draw_mc_lines(img, lines_mc, m_vals, c_vals)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(edges_mc, cmap='gray')
plt.title("Canny Edges")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result_mc)
plt.title("SlopeIntercept Hough")
plt.axis("off")
plt.show()

def hough_rho_theta(img,
                    theta_bins=180,
                    rho_bins=200,
                    acc_thresh=0.6):

    # Ensure grayscale
    if len(img.shape) == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        gray = img.copy()

    edges = cv2.Canny(gray, 100, 200)

    H, W = gray.shape
    diag = int(np.sqrt(H**2 + W**2))

    thetas = np.linspace(0, np.pi, theta_bins)
    rhos = np.linspace(-diag, diag, rho_bins)

    accumulator = np.zeros((rho_bins, theta_bins), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    # Voting
    for x, y in zip(xs, ys):
        for t_idx in range(theta_bins):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]
            rho_idx = int((rho + diag) * rho_bins / (2 * diag))
            if 0 <= rho_idx < rho_bins:
                accumulator[rho_idx, t_idx] += 1

    # Normalize
    accumulator /= accumulator.max()

    lines = np.where(accumulator >= acc_thresh)

    return lines, rhos, thetas, edges
def draw_rho_theta_lines(img, lines, rhos, thetas):

    if len(img.shape) == 2:
        out = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
    else:
        out = img.copy()

    for rho_idx, theta_idx in zip(lines[0], lines[1]):
        rho = rhos[rho_idx]
        theta = thetas[theta_idx]

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = int(a * rho)
        y0 = int(b * rho)

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(out, (x1, y1), (x2, y2), (0, 255, 0), 1)

    return out
lines_rt, rhos, thetas, edges_rt = hough_rho_theta(img, acc_thresh=0.6)
result_rt = draw_rho_theta_lines(img, lines_rt, rhos, thetas)

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.imshow(edges_rt, cmap='gray')
plt.title("Canny Edges")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(result_rt)
plt.title("RhoTheta Hough (Fast)")
plt.axis("off")
plt.show()

````
