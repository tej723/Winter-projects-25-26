````python
def convolve2d(image, kernel):
    h, w = image.shape
    kh, kw = kernel.shape

    pad_h = kh // 2
    pad_w = kw // 2

    padded_image = np.pad(image, ((pad_h, pad_h),
                                  (pad_w, pad_w)),
                                  mode='constant')

    output = np.zeros((h, w))

    for i in range(h):
        for j in range(w):
            region = padded_image[i:i+kh, j:j+kw]
            output[i, j] = np.sum(region * kernel)

    return output
````

````python
def corner_analysis(image, threshold):

    Ix = convolve2d(image, Gx)
    Iy = convolve2d(image, Gy)

    plt.figure()
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix")
    plt.axis('off')
    plt.show()

    plt.figure()
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy")
    plt.axis('off')
    plt.show()

    x = Ix.flatten()
    y = Iy.flatten()

    mask = (np.abs(x) > threshold) | (np.abs(y) > threshold)

    plt.figure()
    plt.scatter(x[mask], y[mask], s=5)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title(f"Iy vs Ix (threshold = {threshold})")
    plt.show()

    return Ix, Iy
````

````python
size = 100

flat_img = np.ones((size, size)) * 128

edge_img = np.zeros((size, size))
edge_img[:, size//2:] = 255

corner_img = np.zeros((size, size))
corner_img[size//2:, size//2:] = 255
````

Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def compute_part_b_rmap(base64_str, window_size=3, k=0.04):
    img_data = base64.b64decode(base64_str)
    img = Image.open(io.BytesIO(img_data)).convert('L')
    img_np = np.array(img).astype(np.float32)

    Ix = cv2.Sobel(img_np, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(img_np, cv2.CV_64F, 0, 1, ksize=3)

    Ixx = Ix**2
    Iyy = Iy**2
    Ixy = Ix * Iy

    kernel = np.ones((window_size, window_size))
    A = cv2.filter2D(Ixx, -1, kernel)
    B = cv2.filter2D(Iyy, -1, kernel)
    C = cv2.filter2D(Ixy, -1, kernel)

    trace = A + B
    det = (A * B) - (C**2)

    root_term = np.sqrt(np.maximum(0, (trace/2)**2 - det))
    lambda_1 = (trace/2) + root_term
    lambda_2 = (trace/2) - root_term

    R_map = det - k * (trace**2)

    return img_np, R_map, lambda_1, lambda_2

def display_results(base64_str, label):
    img, r_map, l1, l2 = compute_part_b_rmap(base64_str)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(img, cmap='gray')
    plt.title(f"{label} Image")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(r_map, cmap='gray')
    plt.title("R-Map (Corner Response)")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    threshold = 0.05 * r_map.max()
    corner_mask = np.zeros_like(r_map)
    corner_mask[r_map > threshold] = 255
    plt.imshow(corner_mask, cmap='hot')
    plt.title("Final Detected Corners")
    plt.axis('off')

    plt.show()

display_results(base64_image_f, "Flat")
display_results(base64_image_e, "Edge")
display_results(base64_image_c1, "Corner")
````

````python
import numpy as np
import cv2
import base64
import io
from PIL import Image
import matplotlib.pyplot as plt

def robust_decode(base64_str):

    if "," in str(base64_str):
        base64_str = str(base64_str).split(",")[1]
    img_data = base64.b64decode(base64_str)
    nparr = np.frombuffer(img_data, np.uint8)
    img = cv2.imdecode(nparr, cv2.IMREAD_COLOR)
    return img

def solve_part_c_hough(base64_str):
    img = robust_decode(base64_str)
    if img is None:
        print("Could not decode the image. Please check the base64 string.")
        return

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    edges = cv2.Canny(gray, 50, 150, apertureSize=3)

    m_range = np.linspace(-5, 5, 100)
    c_range = np.linspace(-gray.shape[0], gray.shape[0], 200)
    accumulator = np.zeros((len(m_range), len(c_range)))

    y_coords, x_coords = np.where(edges > 0)
    for x, y in zip(x_coords, y_coords):
        for m_idx, m in enumerate(m_range):
            c = y - m * x
            if c_range[0] <= c <= c_range[-1]:
                c_idx = np.argmin(np.abs(c_range - c))
                accumulator[m_idx, c_idx] += 1

    thresh = np.percentile(accumulator, 99.9)
    m_indices, c_indices = np.where(accumulator > thresh)

    img_slope_intercept = img_rgb.copy()
    for mi, ci in zip(m_indices, c_indices):
        m, c = m_range[mi], c_range[ci]
        x1, x2 = 0, img.shape[1]
        y1, y2 = int(m*x1 + c), int(m*x2 + c)
        cv2.line(img_slope_intercept, (x1, y1), (x2, y2), (255, 0, 0), 2)

    lines = cv2.HoughLines(edges, 1, np.pi/180, 150)
    img_rho_theta = img_rgb.copy()
    if lines is not None:
        for line in lines:
            rho, theta = line[0]
            a, b = np.cos(theta), np.sin(theta)
            x0, y0 = a*rho, b*rho
            x1, y1 = int(x0 + 1000*(-b)), int(y0 + 1000*(a))
            x2, y2 = int(x0 - 1000*(-b)), int(y0 - 1000*(a))
            cv2.line(img_rho_theta, (x1, y1), (x2, y2), (0, 255, 0), 2)

    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(img_slope_intercept)
    plt.title("Hough: Slope-Intercept")
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(img_rho_theta)
    plt.title("Hough: Rho-Theta")
    plt.axis('off')

    plt.tight_layout()
    plt.show()

solve_part_c_hough(base64_image)
````
