To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def compute_and_show_R_map(img,title):
    gx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
    gy=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
    ixx=gx**2 # squaring x gradients for structure tensor
    iyy=gy**2 # squaring y gradients for structure tensor
    ixy=gx*gy # multiplying x and y for off diagonal elements
    window_size=(5,5)
    sigma=1
    sxx=cv2.GaussianBlur(ixx,window_size,sigma) # summing neighbors using gaussian blur
    syy=cv2.GaussianBlur(iyy,window_size,sigma)
    sxy=cv2.GaussianBlur(ixy,window_size,sigma)
    det=(sxx*syy)-(sxy**2) # calculating determinant of matrix
    trace=sxx+syy # calculating trace of matrix
    k=0.04 # constant value empirically chosen for harris
    r_score=det-k*(trace**2) # computing final harris response score
    r_norm=cv2.normalize(r_score,None,0,255,cv2.NORM_MINMAX,cv2.CV_8U) # normalizing huge values to 0-255 for display
    plt.imshow(r_norm,cmap='gray')
    plt.title(title)
    plt.axis('off')
    return r_score
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
compute_and_show_R_map(flat,"R-Map: Flat")
plt.subplot(1,3,2)
compute_and_show_R_map(edge,"R-Map: Edge")
plt.subplot(1,3,3)
compute_and_show_R_map(corner,"R-Map: Corner")
plt.show()
````

````python
def hough_slope_intercept(image):
    edges=cv2.Canny(image,50,150) # finding binary edges for voting
    h,w=image.shape
    m_step=0.1
    m_min,m_max=-5,5 # limiting slope range to avoid infinite loops(it will run for large times if a cap is not put as m goes to inf)
    m_len=int((m_max-m_min)/m_step)+1
    c_min,c_max=-h,2*h
    c_len=int(c_max-c_min)+1
    accumulator=np.zeros((m_len,c_len),dtype=np.uint64) # creating voting grid
    y_idxs,x_idxs=np.nonzero(edges) # getting coordinates of all edge pixels
    for i in range(len(x_idxs)):
        x,y=x_idxs[i],y_idxs[i]
        for m_idx in range(m_len): # checking every possible slope
            m=m_min+m_idx*m_step
            c=y-m*x # calculating intercept c using line equation
            if c_min<=c<c_max:
                c_idx=int(c-c_min)
                accumulator[m_idx,c_idx]+=1 # voting for this m and c pair
    result=cv2.cvtColor(image,cv2.COLOR_GRAY2RGB) # converting to RGB for display
    threshold=0.5*np.max(accumulator) # picking lines with enough votes
    lines=np.argwhere(accumulator>threshold)
    for m_idx,c_idx in lines:
        m=m_min+m_idx*m_step
        c=c_min+c_idx
        pt1=(0,int(c))
        pt2=(w,int(m*w+c))
        cv2.line(result,pt1,pt2,(255,0,0),1) # drawing red lines (255,0,0) for matplotlib
    return result
output_mc=hough_slope_intercept(corner)
plt.imshow(output_mc)
plt.title("Hough: Slope-Intercept (m, c)")
plt.show()
````

````python
def hough_rho_theta(image):
    if len(image.shape)==3:
        image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    edges=cv2.Canny(image,50,150) # finding binary edges for voting
    h,w=image.shape
    diag_len=int(np.sqrt(h**2+w**2)) # calculating diagonal to set max rho
    rhos=2*diag_len
    thetas=180
    accumulator=np.zeros((rhos,thetas),dtype=np.uint64) # creating voting grid
    y_idxs,x_idxs=np.nonzero(edges)
    cos_t=np.cos(np.deg2rad(np.arange(thetas))) # precalculating trig values for speed
    sin_t=np.sin(np.deg2rad(np.arange(thetas)))
    for i in range(len(x_idxs)):
        x,y=x_idxs[i],y_idxs[i]
        for t in range(thetas): # checking every possible angle
            rho=int(x*cos_t[t]+y*sin_t[t])+diag_len # calculating rho using polar equation
            accumulator[rho,t]+=1 # voting for this rho and theta
    result=cv2.cvtColor(image,cv2.COLOR_GRAY2BGR)
    threshold=0.5*np.max(accumulator) # selecting strong lines
    lines=np.argwhere(accumulator>threshold)
    for r_idx,t_idx in lines:
        rho=r_idx-diag_len # removing offset to get real rho
        theta=np.deg2rad(t_idx)
        a=np.cos(theta)
        b=np.sin(theta)
        x0=a*rho
        y0=b*rho
        x1=int(x0+1000*(-b)) # converting polar to cartesian for drawing
        y1=int(y0+1000*(a))
        x2=int(x0-1000*(-b))
        y2=int(y0-1000*(a))
        cv2.line(result,(x1,y1),(x2,y2),(0,255,0),1)
    return result
output_rt=hough_rho_theta(corner)
plt.imshow(output_rt)
plt.title("Hough: Rho-Theta")
plt.show()
````
