To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Part A: Corner detection intuition with Sobel gradients (Ix, Iy)
# This cell computes Ix and Iy using Sobel kernels on grayscale images
# It also stores Ix_corner2 and Iy_corner2 for Part B

import numpy as np
import cv2
import matplotlib.pyplot as plt

import torch
import torch.nn.functional as F

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

def to_gray_np(img):
    # Converts RGB uint8 image to float32 grayscale using the weighted method.
    if img.ndim == 2:
        return img.astype(np.float32)
    r = img[..., 0].astype(np.float32)
    g = img[..., 1].astype(np.float32)
    b = img[..., 2].astype(np.float32)
    return 0.299 * r + 0.587 * g + 0.114 * b

def normalize01_np(x):
    # Normalizes a NumPy array to [0,1] for visualization
    x = x.astype(np.float32)
    mn = float(x.min())
    mx = float(x.max())
    return (x - mn) / (mx - mn + 1e-8)

def to_torch_gray(img_rgb):
    # Returns grayscale tensor (H,W) float32 on GPU
    g = to_gray_np(img_rgb)
    return torch.from_numpy(g).to(device=device, dtype=torch.float32)

def conv2d_torch(gray_hw, kernel_kk):
    # 2D convolution on GPU using reflect padding and torch.conv2d
    k = int(kernel_kk.shape[0])
    pad = k // 2
    x = gray_hw[None, None, ...]
    w = kernel_kk[None, None, ...]
    x = F.pad(x, (pad, pad, pad, pad), mode="reflect")
    y = F.conv2d(x, w)
    return y[0, 0, ...]

SOBEL_X = torch.tensor([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=torch.float32, device=device)

SOBEL_Y = torch.tensor([[-1, -2, -1],
                        [ 0,  0,  0],
                        [ 1,  2,  1]], dtype=torch.float32, device=device)

def sobel_IxIy(img_rgb):
    # Computes Ix and Iy (H,W) float32 on GPU
    g = to_torch_gray(img_rgb)
    Ix = conv2d_torch(g, SOBEL_X)
    Iy = conv2d_torch(g, SOBEL_Y)
    return Ix, Iy

def show_gray(img, title):
    # Displays a 2D array as grayscale
    plt.figure(figsize=(5, 5))
    plt.imshow(img, cmap="gray")
    plt.title(title)
    plt.axis("off")
    plt.show()

def show_rgb(img, title):
    # Displays an RGB image
    plt.figure(figsize=(5, 5))
    plt.imshow(img)
    plt.title(title)
    plt.axis("off")
    plt.show()

# Computes gradients for the three reference images
Ix_flat, Iy_flat = sobel_IxIy(flat)
Ix_edge, Iy_edge = sobel_IxIy(edge)
Ix_corner, Iy_corner = sobel_IxIy(corner)

# Uses corner2 as the main example for later parts
Ix_corner2, Iy_corner2 = sobel_IxIy(corner2)

# Visualizes flat example
show_rgb(flat, "Flat image")
show_gray(normalize01_np(Ix_flat.detach().cpu().numpy()), "Flat: Ix")
show_gray(normalize01_np(Iy_flat.detach().cpu().numpy()), "Flat: Iy")

# Visualizes edge example
show_rgb(edge, "Edge image")
show_gray(normalize01_np(Ix_edge.detach().cpu().numpy()), "Edge: Ix")
show_gray(normalize01_np(Iy_edge.detach().cpu().numpy()), "Edge: Iy")

# Visualizes corner example
show_rgb(corner, "Corner image")
show_gray(normalize01_np(Ix_corner.detach().cpu().numpy()), "Corner: Ix")
show_gray(normalize01_np(Iy_corner.detach().cpu().numpy()), "Corner: Iy")

# Shows magnitude for corner2 to confirm strong gradients in both directions
mag_corner2 = torch.sqrt(Ix_corner2 * Ix_corner2 + Iy_corner2 * Iy_corner2)
show_rgb(corner2, "Corner2 image")
show_gray(normalize01_np(mag_corner2.detach().cpu().numpy()), "Corner2: Gradient magnitude")

# Interpretation of Ix and Iy
# Flat regions have Ix and Iy near zero
# Edges have strong change in one direction so one of Ix or Iy dominates locally
# Corners have strong change in both directions so both Ix and Iy are strong
````

````python
# Part B: Ellipse intuition via structure tensor and Harris R map
# This cell builds the structure tensor from Ix and Iy and computes the Harris response map R

import numpy as np
import cv2
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

def to_gray01(img):
    # This converts RGB uint8 or float images to grayscale float32 in [0,1] (normalised after applying weighted method)
    if img.ndim == 2:
        g = img.astype(np.float32)
        if g.max() > 1.5:
            g = g / 255.0
        return g
    x = img.astype(np.float32)
    if x.max() > 1.5:
        x = x / 255.0
    r, g, b = x[..., 0], x[..., 1], x[..., 2]
    return (0.299 * r + 0.587 * g + 0.114 * b).astype(np.float32)

def sobel_torch(gray01_t):
    # This computes Sobel gradients on GPU
    kx = torch.tensor([[-1, 0, 1],
                       [-2, 0, 2],
                       [-1, 0, 1]], dtype=torch.float32, device=gray01_t.device).view(1, 1, 3, 3)
    ky = torch.tensor([[-1, -2, -1],
                       [ 0,  0,  0],
                       [ 1,  2,  1]], dtype=torch.float32, device=gray01_t.device).view(1, 1, 3, 3)
    x = gray01_t.unsqueeze(0).unsqueeze(0)
    x = F.pad(x, (1, 1, 1, 1), mode="replicate")
    Ix = F.conv2d(x, kx)[0, 0]
    Iy = F.conv2d(x, ky)[0, 0]
    return Ix, Iy

def gaussian_kernel(ksize=7, sigma=1.5, device="cpu"):
    # This builds a normalized 2D Gaussian kernel for smoothing the structure tensor
    ax = torch.arange(-(ksize // 2), ksize // 2 + 1, device=device, dtype=torch.float32)
    xx, yy = torch.meshgrid(ax, ax, indexing="ij")
    K = torch.exp(-(xx * xx + yy * yy) / (2.0 * sigma * sigma))
    K = K / (K.sum() + 1e-8)
    return K.view(1, 1, ksize, ksize)

def smooth_torch(x2d, K):
    # This applies Gaussian smoothing via conv2d
    x = x2d.unsqueeze(0).unsqueeze(0)
    p = K.shape[-1] // 2
    x = F.pad(x, (p, p, p, p), mode="replicate")
    return F.conv2d(x, K)[0, 0]

def harris_R(gray01_t, k=0.04, win=7, sigma=1.5):
    # This computes Harris response R = det(M) - k * trace(M)^2
    # I have used the standard value of k (around 0.04), but we can adjust it to check the variations in corner and edge detection.
    # Larger k suppresses edges more but may reduce corner response.
    Ix, Iy = sobel_torch(gray01_t)
    Ixx = Ix * Ix
    Iyy = Iy * Iy
    Ixy = Ix * Iy
    K = gaussian_kernel(ksize=win, sigma=sigma, device=gray01_t.device)
    Sxx = smooth_torch(Ixx, K)
    Syy = smooth_torch(Iyy, K)
    Sxy = smooth_torch(Ixy, K)
    det = Sxx * Syy - Sxy * Sxy
    trace = Sxx + Syy
    R = det - k * (trace * trace)
    return R

def nms_harris(R, radius=6, thresh_rel=0.01, topk=200):
    # This extracts corners using max-pool NMS on positive R values
    Rp = torch.clamp(R, min=0.0)
    t = thresh_rel * float(Rp.max().item() + 1e-12)
    x = Rp.unsqueeze(0).unsqueeze(0)
    m = F.max_pool2d(x, kernel_size=2 * radius + 1, stride=1, padding=radius)[0, 0]
    keep = (Rp == m) & (Rp > t)
    ys, xs = torch.where(keep)
    if ys.numel() == 0:
        return np.zeros((0, 2), dtype=np.int32), t
    scores = Rp[ys, xs]
    idx = torch.argsort(scores, descending=True)
    if topk is not None:
        idx = idx[:topk]
    pts = torch.stack([xs[idx], ys[idx]], dim=1).detach().cpu().numpy().astype(np.int32)
    return pts, t

corner2_gray = to_gray01(corner2)

g_t = torch.from_numpy(corner2_gray).to(device=device, dtype=torch.float32)
R_t = harris_R(g_t, k=0.04, win=7, sigma=1.5)

R_np = R_t.detach().cpu().numpy()
print("R stats:", "min", float(R_np.min()), "max", float(R_np.max()), "p99", float(np.percentile(R_np, 99)), "p99.9", float(np.percentile(R_np, 99.9)))

# This visualizes the required outputs
R_vis = np.log1p(np.maximum(R_np, 0.0))
R_vis = (R_vis - R_vis.min()) / (R_vis.max() - R_vis.min() + 1e-8)

pts, thr = nms_harris(R_t, radius=6, thresh_rel=0.01, topk=200)
print("N corners:", len(pts), "threshold:", thr)

corner2_rgb = corner2.copy()
if corner2_rgb.ndim == 2:
    corner2_rgb = np.stack([corner2_rgb] * 3, axis=-1)
if corner2_rgb.dtype != np.uint8:
    corner2_rgb = np.clip(corner2_rgb, 0, 255).astype(np.uint8)

overlay = corner2_rgb.copy()
for x, y in pts:
    cv2.circle(overlay, (int(x), int(y)), 3, (255, 0, 0), 1)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.imshow(R_vis, cmap="gray")
plt.title("Harris R map (log1p(ReLU(R)))")
plt.axis("off")
plt.subplot(1, 2, 2)
plt.imshow(overlay)
plt.title("Detected corners overlay")
plt.axis("off")
plt.show()

# Using OpenCV cornerHarris for comparison
g8 = (corner2_gray * 255.0).astype(np.uint8)
cvR = cv2.cornerHarris(g8, blockSize=2, ksize=3, k=0.04)
cvR_vis = np.log1p(np.maximum(cvR, 0.0))
cvR_vis = (cvR_vis - cvR_vis.min()) / (cvR_vis.max() - cvR_vis.min() + 1e-8)
plt.figure(figsize=(5, 5))
plt.imshow(cvR_vis, cmap="gray")
plt.title("OpenCV Harris (log1p(ReLU))")
plt.axis("off")
plt.show()
````

````python
# Part C: Hough Transform for line detection
# This cell compares slope-intercept voting with rho-theta voting
# The slope-intercept method becomes unstable for near-vertical lines

import numpy as np
import cv2
import matplotlib.pyplot as plt
import time
import math

import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Device: {device}")

def edges_for_hough(img_rgb, low=80, high=160):
    # Extracts edges using Canny on grayscale
    g = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY) if img_rgb.ndim == 3 else img_rgb
    g8 = np.clip(g, 0, 255).astype(np.uint8)
    return cv2.Canny(g8, low, high)

def hough_slope_intercept(edge_map, m_min=-3.0, m_max=3.0, num_m=720, c_res=1.0):
    # Hough voting in (m,c) space using GPU vectorization
    ys, xs = np.where(edge_map > 0)
    if ys.size == 0:
        return None
    H, W = edge_map.shape[:2]
    x = torch.from_numpy(xs.astype(np.float32)).to(device)
    y = torch.from_numpy(ys.astype(np.float32)).to(device)
    m_vals = torch.linspace(m_min, m_max, steps=num_m, device=device)

    Cmax = float(H + abs(m_max) * W)
    c_min = -Cmax
    c_max = Cmax
    num_c = int(math.floor((c_max - c_min) / c_res)) + 1

    c = y[:, None] - m_vals[None, :] * x[:, None]
    c_idx = torch.round((c - c_min) / c_res).to(torch.int64)
    t_idx = torch.arange(num_m, device=device, dtype=torch.int64)[None, :].expand_as(c_idx)

    valid = (c_idx >= 0) & (c_idx < num_c)
    flat = c_idx[valid] * num_m + t_idx[valid]

    acc = torch.bincount(flat, minlength=num_c * num_m).reshape(num_c, num_m)
    return acc, m_vals, c_min, c_res

def hough_rho_theta(edge_map, num_theta=720, rho_res=1.0):
    # Hough voting in (rho,theta) space using GPU vectorization
    ys, xs = np.where(edge_map > 0)
    if ys.size == 0:
        return None
    H, W = edge_map.shape[:2]
    x = torch.from_numpy(xs.astype(np.float32)).to(device)
    y = torch.from_numpy(ys.astype(np.float32)).to(device)

    thetas = torch.linspace(-math.pi / 2, math.pi / 2, steps=num_theta, device=device)
    cos_t = torch.cos(thetas)
    sin_t = torch.sin(thetas)

    rho_max = float(math.hypot(H, W))
    rho_min = -rho_max
    num_rho = int(math.floor((2.0 * rho_max) / rho_res)) + 1

    rho = x[:, None] * cos_t[None, :] + y[:, None] * sin_t[None, :]
    rho_idx = torch.round((rho - rho_min) / rho_res).to(torch.int64)
    t_idx = torch.arange(num_theta, device=device, dtype=torch.int64)[None, :].expand_as(rho_idx)

    valid = (rho_idx >= 0) & (rho_idx < num_rho)
    flat = rho_idx[valid] * num_theta + t_idx[valid]

    acc = torch.bincount(flat, minlength=num_rho * num_theta).reshape(num_rho, num_theta)
    return acc, thetas, rho_min, rho_res

def topk_peaks(acc, k=10, min_votes=1):
    # Extracts top accumulator peaks above min_votes
    flat = acc.flatten()
    vals, idxs = torch.topk(flat, k=min(k, flat.numel()))
    peaks = []
    for v, idx in zip(vals.tolist(), idxs.tolist()):
        if v < min_votes:
            continue
        peaks.append((idx, v))
    return peaks

def draw_rho_theta_lines(img_rgb, peaks, thetas, rho_min, rho_res, num_theta):
    # Draws detected rho-theta lines on an RGB image
    out = img_rgb.copy()
    H, W = out.shape[:2]
    for flat_idx, votes in peaks:
        rho_i = flat_idx // num_theta
        th_i = flat_idx % num_theta
        rho = rho_min + rho_i * rho_res
        theta = float(thetas[th_i].detach().cpu().item())
        a = math.cos(theta)
        b = math.sin(theta)
        if abs(b) > 1e-6:
            x0, x1 = 0, W - 1
            y0 = (rho - x0 * a) / b
            y1 = (rho - x1 * a) / b
            p0 = (int(x0), int(round(y0)))
            p1 = (int(x1), int(round(y1)))
        else:
            x = rho / (a + 1e-8)
            p0 = (int(round(x)), 0)
            p1 = (int(round(x)), H - 1)
        cv2.line(out, p0, p1, (255, 0, 0), 2)
    return out

def draw_mc_lines(img_rgb, peaks, m_vals, c_min, c_res, num_m):
    # Draws detected slope-intercept lines on an RGB image
    out = img_rgb.copy()
    H, W = out.shape[:2]
    for flat_idx, votes in peaks:
        c_i = flat_idx // num_m
        m_i = flat_idx % num_m
        m = float(m_vals[m_i].detach().cpu().item())
        c = float(c_min + c_i * c_res)
        x0, x1 = 0, W - 1
        y0 = m * x0 + c
        y1 = m * x1 + c
        p0 = (int(x0), int(round(y0)))
        p1 = (int(x1), int(round(y1)))
        cv2.line(out, p0, p1, (0, 255, 0), 2)
    return out

img_demo = corner2
edges = edges_for_hough(img_demo, low=80, high=160)

plt.figure(figsize=(6, 6))
plt.imshow(edges, cmap="gray")
plt.title("Edge map for Hough")
plt.axis("off")
plt.show()

if device.type == "cuda":
    torch.cuda.synchronize()
t0 = time.time()
mc = hough_slope_intercept(edges, m_min=-3.0, m_max=3.0, num_m=720, c_res=1.0)
if device.type == "cuda":
    torch.cuda.synchronize()
t1 = time.time()

if device.type == "cuda":
    torch.cuda.synchronize()
t2 = time.time()
rt = hough_rho_theta(edges, num_theta=720, rho_res=1.0)
if device.type == "cuda":
    torch.cuda.synchronize()
t3 = time.time()

print(f"Slope-intercept Hough time: {t1 - t0:.3f} s")
print(f"Rho-theta Hough time:      {t3 - t2:.3f} s")

if mc is not None:
    acc_mc, m_vals, c_min, c_res = mc
    mv_mc = int(max(20, 0.55 * float(acc_mc.max().detach().cpu().item())))
    peaks_mc = topk_peaks(acc_mc, k=10, min_votes=mv_mc)
    out_mc = draw_mc_lines(img_demo, peaks_mc, m_vals, c_min, c_res, num_m=acc_mc.shape[1])
    plt.figure(figsize=(6, 6))
    plt.imshow(out_mc)
    plt.title("Top lines from slope-intercept Hough (green)")
    plt.axis("off")
    plt.show()

if rt is not None:
    acc_rt, thetas, rho_min, rho_res = rt
    mv_rt = int(max(20, 0.55 * float(acc_rt.max().detach().cpu().item())))
    peaks_rt = topk_peaks(acc_rt, k=10, min_votes=mv_rt)
    out_rt = draw_rho_theta_lines(img_demo, peaks_rt, thetas, rho_min, rho_res, num_theta=acc_rt.shape[1])
    plt.figure(figsize=(6, 6))
    plt.imshow(out_rt)
    plt.title("Top lines from rho-theta Hough (red)")
    plt.axis("off")
    plt.show()
````

````python
# Bonus Part D: Build an RGBA stencil from the ball and composite it onto the pool image
# This cell segments the ball using thresholding and contours, creates a soft alpha matte, and blends on GPU for faster computation.

import numpy as np
import cv2
import torch
import matplotlib.pyplot as plt

ball_rgb = ball.copy()
pool_rgb = ball2.copy()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

def segment_ball_circle_mask(ball_rgb):
    # This segments the white ball with HSV and returns a clean filled-circle mask
    hsv = cv2.cvtColor(ball_rgb, cv2.COLOR_RGB2HSV)
    lower = np.array([0, 0, 120], dtype=np.uint8)
    upper = np.array([179, 100, 255], dtype=np.uint8)
    m0 = cv2.inRange(hsv, lower, upper)
    k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    m0 = cv2.morphologyEx(m0, cv2.MORPH_OPEN, k, iterations=1)
    m0 = cv2.morphologyEx(m0, cv2.MORPH_CLOSE, k, iterations=2)
    num, labels, stats, _ = cv2.connectedComponentsWithStats((m0 > 0).astype(np.uint8), connectivity=8)
    if num <= 1:
        return None
    largest = 1 + int(np.argmax(stats[1:, cv2.CC_STAT_AREA]))
    m1 = (labels == largest).astype(np.uint8) * 255
    cnts, _ = cv2.findContours(m1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not cnts:
        return None
    cnt = max(cnts, key=cv2.contourArea)
    (xc, yc), r = cv2.minEnclosingCircle(cnt)
    xc, yc, r = int(round(xc)), int(round(yc)), int(round(r))
    if r < 6:
        return None
    mask = np.zeros_like(m1)
    cv2.circle(mask, (xc, yc), r, 255, -1)
    return mask, (xc, yc, r)

def soft_alpha_from_mask(mask, r):
    # This creates a soft alpha matte with distance feathering
    m = (mask > 0).astype(np.uint8)
    din = cv2.distanceTransform(m, cv2.DIST_L2, 5)
    dout = cv2.distanceTransform(1 - m, cv2.DIST_L2, 5)
    signed = din - dout
    feather = max(2, int(0.06 * r))
    alpha = np.clip((signed + feather) / (2.0 * feather), 0.0, 1.0).astype(np.float32)
    alpha = np.clip(alpha ** 1.05, 0.0, 1.0).astype(np.float32)
    return alpha

def crop_circle(img, alpha, xc, yc, r):
    # This crops around the detected ball for a tight stencil
    pad = max(6, int(0.12 * r))
    y0 = max(0, yc - r - pad)
    y1 = min(img.shape[0], yc + r + pad)
    x0 = max(0, xc - r - pad)
    x1 = min(img.shape[1], xc + r + pad)
    return img[y0:y1, x0:x1], alpha[y0:y1, x0:x1]

def estimate_bg_color(ball_crop, alpha_crop):
    # This estimates the blue background color from near-transparent pixels
    ring = alpha_crop < 0.05
    if np.any(ring):
        return ball_crop[ring].astype(np.float32).mean(axis=0)
    h, w = alpha_crop.shape
    border = np.zeros_like(alpha_crop, dtype=bool)
    border[:3, :] = True
    border[-3:, :] = True
    border[:, :3] = True
    border[:, -3:] = True
    return ball_crop[border].astype(np.float32).mean(axis=0)

def inverse_composite_defringe(ball_crop, alpha_crop, bg_rgb):
    # This removes the blue rim by inverse compositing and edge defringing
    a = alpha_crop[..., None].astype(np.float32)
    a_safe = np.clip(a, 1e-4, 1.0)
    C = ball_crop.astype(np.float32)
    B = bg_rgb[None, None, :].astype(np.float32)
    F = (C - (1.0 - a) * B) / a_safe
    F = np.clip(F, 0.0, 255.0).astype(np.uint8)

    edge = (alpha_crop > 0.05) & (alpha_crop < 0.90)
    if np.any(edge):
        hsv = cv2.cvtColor(F, cv2.COLOR_RGB2HSV).astype(np.float32)
        core = alpha_crop > 0.90
        if np.any(core):
            core_rgb = F[core].astype(np.float32)
            core_mean = core_rgb.mean(axis=0)
            core_hsv = cv2.cvtColor(core_mean[None, None, :].astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)[0, 0]
            hsv[..., 0][edge] = core_hsv[0]
            hsv[..., 1][edge] = hsv[..., 1][edge] * 0.55
        F = cv2.cvtColor(np.clip(hsv, 0, 255).astype(np.uint8), cv2.COLOR_HSV2RGB)

    return F

def neutralize_white_balance(fg_rgb, alpha_crop):
    # This removes residual blue tint using the ball core as reference
    core = alpha_crop > 0.90
    if not np.any(core):
        return fg_rgb
    m = fg_rgb[core].astype(np.float32).mean(axis=0)
    t = float(m.mean() + 1e-6)
    s = t / (m + 1e-6)
    out = np.clip(fg_rgb.astype(np.float32) * s[None, None, :], 0.0, 255.0).astype(np.uint8)
    return out

def detect_red_ball(pool_rgb):
    # This finds the red ball and its radius for placement and shadow direction
    hsv = cv2.cvtColor(pool_rgb, cv2.COLOR_RGB2HSV)
    m1 = cv2.inRange(hsv, np.array([0, 70, 50], np.uint8), np.array([10, 255, 255], np.uint8))
    m2 = cv2.inRange(hsv, np.array([170, 70, 50], np.uint8), np.array([179, 255, 255], np.uint8))
    mask = cv2.bitwise_or(m1, m2)
    k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, k, iterations=1)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k, iterations=2)
    cnts, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not cnts:
        return None
    cnt = max(cnts, key=cv2.contourArea)
    (xc, yc), r = cv2.minEnclosingCircle(cnt)
    return int(round(xc)), int(round(yc)), int(round(r))

def estimate_shadow_direction(pool_rgb, red_info):
    # This estimates light direction by locating the darkest point near the red ball as shadow cue
    if red_info is None:
        return np.array([1.0, 0.25], dtype=np.float32)
    rx, ry, rr = red_info
    g = cv2.cvtColor(pool_rgb, cv2.COLOR_RGB2GRAY).astype(np.float32) / 255.0
    y0 = int(np.clip(ry, 0, g.shape[0] - 1))
    y1 = int(np.clip(ry + int(2.2 * rr), 0, g.shape[0]))
    x0 = int(np.clip(rx, 0, g.shape[1] - 1))
    x1 = int(np.clip(rx + int(3.0 * rr), 0, g.shape[1]))
    if y1 <= y0 + 2 or x1 <= x0 + 2:
        return np.array([1.0, 0.25], dtype=np.float32)
    patch = g[y0:y1, x0:x1]
    py, px = np.unravel_index(np.argmin(patch), patch.shape)
    dx = float((x0 + px) - rx)
    dy = float((y0 + py) - ry)
    v = np.array([dx, dy], dtype=np.float32)
    n = float(np.linalg.norm(v) + 1e-6)
    v = v / n
    if v[0] < 0.2:
        v[0] = 0.2
    if v[1] < -0.1:
        v[1] = -0.1
    v = v / float(np.linalg.norm(v) + 1e-6)
    return v

def rotated_anisotropic_blur(sh, angle_deg, sigma_long, sigma_short):
    # This blurs in a rotated frame so the shadow elongates along the estimated direction
    h, w = sh.shape
    M = cv2.getRotationMatrix2D((w * 0.5, h * 0.5), -angle_deg, 1.0)
    sh_r = cv2.warpAffine(sh, M, (w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=0)
    sh_r = cv2.GaussianBlur(sh_r, (0, 0), sigmaX=float(sigma_long), sigmaY=float(sigma_short))
    Minv = cv2.getRotationMatrix2D((w * 0.5, h * 0.5), angle_deg, 1.0)
    sh_b = cv2.warpAffine(sh_r, Minv, (w, h), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=0)
    return np.clip(sh_b, 0.0, 1.0).astype(np.float32)

def build_shadows(alpha_resized, dir_vec, target_d):
    # This creates both a contact shadow and a cast shadow consistent with the scene lighting
    dx = float(dir_vec[0])
    dy = float(dir_vec[1])
    angle = float(np.degrees(np.arctan2(dy, dx)))

    base = (alpha_resized ** 1.8).astype(np.float32)
    sigma_long = 0.20 * target_d
    sigma_short = 0.06 * target_d
    cast = rotated_anisotropic_blur(base, angle, sigma_long, sigma_short)

    cast_shift_x = int(round(0.32 * target_d * dx))
    cast_shift_y = int(round(0.22 * target_d * dy))
    M = np.float32([[1, 0, cast_shift_x], [0, 1, cast_shift_y]])
    cast = cv2.warpAffine(cast, M, (target_d, target_d), flags=cv2.INTER_LINEAR, borderMode=cv2.BORDER_CONSTANT, borderValue=0)

    y = np.linspace(-1.0, 1.0, target_d, dtype=np.float32)
    x = np.linspace(-1.0, 1.0, target_d, dtype=np.float32)
    xx, yy = np.meshgrid(x, y)
    contact = np.exp(-(((xx / 0.55) ** 2) + ((yy / 0.23) ** 2)) * 3.4).astype(np.float32)
    contact = cv2.GaussianBlur(contact, (0, 0), sigmaX=0.06 * target_d, sigmaY=0.04 * target_d)
    contact = contact * (alpha_resized ** 1.2).astype(np.float32)
    contact = np.clip(contact, 0.0, 1.0)

    return cast, contact

def match_ball_to_scene(ball_resized, alpha_resized, pool_rgb, ty0, tx0):
    # This matches brightness and adds subtle green bounce light from the felt
    H, W = pool_rgb.shape[:2]
    d = ball_resized.shape[0]
    y0 = max(0, ty0 - int(0.15 * d))
    y1 = min(H, ty0 + d + int(0.15 * d))
    x0 = max(0, tx0 - int(0.15 * d))
    x1 = min(W, tx0 + d + int(0.15 * d))

    patch = pool_rgb[y0:y1, x0:x1].astype(np.float32)
    bg = patch.reshape(-1, 3).mean(axis=0)

    core = alpha_resized > 0.90
    if np.any(core):
        core_rgb = ball_resized[core].astype(np.float32)
        core_lum = (0.299 * core_rgb[:, 0] + 0.587 * core_rgb[:, 1] + 0.114 * core_rgb[:, 2]).mean()
    else:
        core_lum = (0.299 * ball_resized[..., 0] + 0.587 * ball_resized[..., 1] + 0.114 * ball_resized[..., 2]).mean()

    bg_lum = (0.299 * bg[0] + 0.587 * bg[1] + 0.114 * bg[2])
    target_lum = 0.90 * core_lum + 0.10 * bg_lum
    gain = float(np.clip(target_lum / (core_lum + 1e-6), 0.75, 1.05))

    out = np.clip(ball_resized.astype(np.float32) * gain, 0.0, 255.0)

    y = np.linspace(0.0, 1.0, d, dtype=np.float32)[:, None]
    bounce = (y ** 2) * 0.10
    out = out + bounce[..., None] * (bg[None, None, :] - out) * alpha_resized[..., None]

    return np.clip(out, 0.0, 255.0).astype(np.uint8)

seg = segment_ball_circle_mask(ball_rgb)

mask, (xc, yc, r) = seg
alpha_full = soft_alpha_from_mask(mask, r)
ball_crop, alpha_crop = crop_circle(ball_rgb, alpha_full, xc, yc, r)

bg_rgb = estimate_bg_color(ball_crop, alpha_crop)
fg = inverse_composite_defringe(ball_crop, alpha_crop, bg_rgb)
fg = neutralize_white_balance(fg, alpha_crop)

stencil_rgb = (fg.astype(np.float32) * alpha_crop[..., None]).astype(np.uint8)

pool_h, pool_w = pool_rgb.shape[:2]
target_d = int(round(0.12 * pool_h))
target_d = max(18, target_d)

ball_resized = cv2.resize(fg, (target_d, target_d), interpolation=cv2.INTER_AREA)
alpha_resized = cv2.resize(alpha_crop, (target_d, target_d), interpolation=cv2.INTER_AREA)
alpha_resized = np.clip(alpha_resized, 0.0, 1.0).astype(np.float32)

red_info = detect_red_ball(pool_rgb)
dir_vec = estimate_shadow_direction(pool_rgb, red_info)

if red_info is not None:
    rx, ry, rr = red_info
    ground_y = ry + rr
    cy = int(np.clip(ground_y - target_d * 0.52, target_d // 2, pool_h - target_d // 2))
    cx = int(np.clip(rx + rr + int(0.95 * target_d), target_d // 2, pool_w - target_d // 2))
else:
    cy = int(round(0.62 * pool_h))
    cx = int(round(0.62 * pool_w))

ty0 = int(np.clip(cy - target_d // 2, 0, pool_h - target_d))
tx0 = int(np.clip(cx - target_d // 2, 0, pool_w - target_d))

ball_resized = match_ball_to_scene(ball_resized, alpha_resized, pool_rgb, ty0, tx0)

cast_shadow_roi, contact_shadow_roi = build_shadows(alpha_resized, dir_vec, target_d)

shadow_canvas = np.zeros((pool_h, pool_w), dtype=np.float32)
contact_canvas = np.zeros((pool_h, pool_w), dtype=np.float32)
shadow_canvas[ty0:ty0 + target_d, tx0:tx0 + target_d] = cast_shadow_roi
contact_canvas[ty0:ty0 + target_d, tx0:tx0 + target_d] = contact_shadow_roi

pool_t = torch.from_numpy(pool_rgb).to(device=device, dtype=torch.float32) / 255.0
ball_t = torch.from_numpy(ball_resized).to(device=device, dtype=torch.float32) / 255.0
alpha_t = torch.from_numpy(alpha_resized).to(device=device, dtype=torch.float32).unsqueeze(-1)
shadow_t = torch.from_numpy(shadow_canvas).to(device=device, dtype=torch.float32).unsqueeze(-1)
contact_t = torch.from_numpy(contact_canvas).to(device=device, dtype=torch.float32).unsqueeze(-1)

out_t = pool_t.clone()

shadow_strength = 0.46
contact_strength = 0.30
out_t = out_t * (1.0 - shadow_strength * shadow_t)
out_t = out_t * (1.0 - contact_strength * contact_t)

roi = out_t[ty0:ty0 + target_d, tx0:tx0 + target_d]
roi = ball_t * alpha_t + roi * (1.0 - alpha_t)
out_t[ty0:ty0 + target_d, tx0:tx0 + target_d] = roi

composite_result = (out_t.clamp(0.0, 1.0) * 255.0).byte().cpu().numpy()

plt.figure(figsize=(16, 8))
plt.subplot(2, 3, 1); plt.imshow(ball_rgb); plt.title("Ball image"); plt.axis("off")
plt.subplot(2, 3, 2); plt.imshow(mask, cmap="gray"); plt.title("Binary mask"); plt.axis("off")
plt.subplot(2, 3, 3); plt.imshow(alpha_full, cmap="gray"); plt.title("Alpha matte"); plt.axis("off")
plt.subplot(2, 3, 4); plt.imshow(pool_rgb); plt.title("Pool image"); plt.axis("off")
plt.subplot(2, 3, 5); plt.imshow(stencil_rgb); plt.title("Stencil RGB"); plt.axis("off")
plt.subplot(2, 3, 6); plt.imshow(composite_result); plt.title("Composite result"); plt.axis("off")
plt.show()
````
