To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def rmaps(img,win=3,k=0.04):
  # no thresholding here
  Ix=convolve(img,kx)
  Iy=convolve(img,ky)
  h,w=img.shape
  pad=win//2

  R=np.zeros((h,w),dtype=np.float32)
  for i in range(pad,h-pad):
    for j in range(pad,w-pad):
        # Local window
        Ix_w = Ix[i-pad:i+pad+1, j-pad:j+pad+1]
        Iy_w = Iy[i-pad:i+pad+1, j-pad:j+pad+1]

        # Second moments (a, b, c)
        a = np.sum(Ix_w ** 2)
        b = np.sum(Ix_w * Iy_w)
        c = np.sum(Iy_w ** 2)

        # lmbda1 and lmbda2
        temp = np.sqrt((a - c)**2 + 4*b*b)
        lambda1 = (a + c + temp) / 2
        lambda2 = (a + c - temp) / 2

        # Harris response
        R[i, j] = lambda1 * lambda2 - k * (lambda1 + lambda2)**2

  return R
````

````python
def hough_lines(img, mbins=200, cbins=200):
    img_g = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    h, w = img_g.shape

    # find canny edge map
    blur=cv2.GaussianBlur(img_g,(5,5),0) # blur to remove noise
    edges = cv2.Canny(blur, 50, 150)
    # plt.imshow(edges,cmap='gray')


    m_vals = np.linspace(-5, 5, mbins)   # m in [-5,5]
    c_vals = np.linspace(-h, h, cbins)   # c in [-h,h]

    acc = np.zeros((mbins, cbins), dtype=np.int32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int(np.round((c + h) * (cbins - 1) / (2 * h)))

            if 0 <= ci < cbins:
                acc[mi, ci] += 1

    if acc.max() == 0:
        plt.imshow(img)
        plt.axis("off")
        plt.show()
        return

    acc_norm = acc / acc.max()
    t = 0.9
    peaks = np.argwhere(acc_norm >= t)

    # get the lines
    lines = []
    for mi, ci in peaks:
        lines.append((m_vals[mi], c_vals[ci]))

    # draw lines (convert to BGR first)
    img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

    for m, c in lines:
        x0 = 0
        y0 = int(c)

        x1 = w - 1
        y1 = int(m * x1 + c)

        if 0 <= y0 < h and 0 <= y1 < h:
            cv2.line(img_bgr, (x0, y0), (x1, y1), (255, 0, 0), 2)

    # show result
    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
hough_lines(img_rgb)
# we got only horizontal line since accumulator is not large enough to accomodate large m values
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hough_lines_improved(img, rho_res=1, theta_res=np.pi/180, threshold=100):
    img_g = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(img_g, 50, 150)

    h, w = img_g.shape
    # max possible rho is the diagonal of the image
    diag_len = int(np.sqrt(h**2 + w**2))

    # thetas: 0 to 180 degrees
    # rhos: -diag_len to +diag_len
    thetas = np.arange(0, np.pi, theta_res)
    rhos = np.arange(-diag_len, diag_len, rho_res)

    # accumulator
    acc = np.zeros((len(rhos), len(thetas)), dtype=np.int32)

    # pre-calculate cos and sin
    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)

    ys, xs = np.nonzero(edges) # get edge points

    # voting
    for x, y in zip(xs, ys):
        for ti in range(len(thetas)):
            rho = x * cos_t[ti] + y * sin_t[ti]
            # map rho to an index in the accumulator
            # (rho + diag_len) shifts it to be 0 or positive
            ri = int(round(rho + diag_len) / rho_res)
            if 0 <= ri < len(rhos):
                acc[ri, ti] += 1

    # finding Peaks
    peaks = np.argwhere(acc > threshold)

    # drawing Lines
    img_bgr = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    for ri, ti in peaks:
        rho = rhos[ri]
        theta = thetas[ti]

        # convert polar back to cartesian for drawing
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        # two points far apart to draw the full line
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_bgr, (x1, y1), (x2, y2), (255, 0, 0), 2)

    plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()

hough_lines_improved(img_rgb, threshold=120)
````
