To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
sobel_x = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
], dtype=np.float32)

sobel_y = np.array([
    [-1, -2, -1],
    [ 0,  0,  0],
    [ 1,  2,  1]
], dtype=np.float32)
````

````python
def convolve(img, kernel):

    img_height, img_width = img.shape
    kernel_height, kernel_width = kernel.shape

    pad_height = kernel_height // 2
    pad_width = kernel_width // 2

    padded_image = np.pad(img, ((pad_height, pad_height), (pad_width, pad_width)), mode='reflect')

    output = np.zeros_like(img)

    for i in range(img_height):
        for j in range(img_width):
            region = padded_image[i:i+kernel_height, j:j+kernel_width]
            output[i, j] = np.sum(region * kernel)
    
    return output
````

````python
def analyze_gradients(img_gray, threshold=None):
    Ix = convolve(img_gray, sobel_x)
    Iy = convolve(img_gray, sobel_y)

    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.title("Ix (Sobel X)")
    plt.imshow(Ix, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Iy (Sobel Y)")
    plt.imshow(Iy, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("Input Image")
    plt.imshow(img_gray, cmap='gray')
    plt.axis("off")
    plt.show()

    Ix_flat = Ix.flatten()
    Iy_flat = Iy.flatten()

    if threshold is not None:
        mask = (np.abs(Ix_flat) > threshold) | (np.abs(Iy_flat) > threshold)
        Ix_flat = Ix_flat[mask]
        Iy_flat = Iy_flat[mask]

    plt.figure(figsize=(5,5))
    plt.scatter(Ix_flat, Iy_flat, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Scatter Plot: Iy vs Ix")
    plt.grid(True)
    plt.show()

    return Ix, Iy
````

````python
def compute_r_map(img_gray, window_size=10, k=0.01):
    Ix = convolve(img_gray, sobel_x)
    Iy = convolve(img_gray, sobel_y)

    pad = window_size // 2
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    Ix2_p = np.pad(Ix2, pad, mode='reflect')
    Iy2_p = np.pad(Iy2, pad, mode='reflect')
    Ixy_p = np.pad(Ixy, pad, mode='reflect')

    R_map = np.zeros_like(img_gray, dtype=np.float32)

    for i in range(img_gray.shape[0]):
        for j in range(img_gray.shape[1]):
            wx = Ix2_p[i:i+window_size, j:j+window_size]
            wy = Iy2_p[i:i+window_size, j:j+window_size]
            wxy = Ixy_p[i:i+window_size, j:j+window_size]

            a = np.sum(wx)
            b = np.sum(wxy)
            c = np.sum(wy)

            trace = a + c
            det = a * c - b * b

            R_map[i, j] = det - k * (trace ** 2)

    return R_map
````

````python
import cv2
def canny_edges(img_gray, low=100, high=200):
    img_u8 = img_gray.astype(np.uint8)
    edges = cv2.Canny(img_u8, low, high)
    return edges
````

````python
def hough_mc(edges, m_range, c_range, num_m=200, num_c=200):
    H, W = edges.shape

    ms = np.linspace(m_range[0], m_range[1], num_m)
    cs = np.linspace(c_range[0], c_range[1], num_c)

    acc = np.zeros((num_m, num_c), dtype=np.int32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for i, m in enumerate(ms):
            c = y - m * x
            j = int((c - c_range[0]) / (c_range[1] - c_range[0]) * num_c)
            if 0 <= j < num_c:
                acc[i, j] += 1

    return acc, ms, cs
````

````python
def draw_mc_lines(img, acc, ms, cs, threshold_ratio=0.9):
    acc_norm = acc / acc.max()
    lines = np.argwhere(acc_norm > threshold_ratio)

    img_color = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_GRAY2BGR)

    H, W = img.shape

    for i, j in lines:
        m = ms[i]
        c = cs[j]
        x0, x1 = 0, W
        y0 = int(c)
        y1 = int(m * x1 + c)
        cv2.line(img_color, (x0, y0), (x1, y1), (0,0,255), 1)

    return img_color
````

````python
def hough_rho_theta(edges, num_rho=300, num_theta=180):
    H, W = edges.shape
    diag = int(np.sqrt(H*H + W*W))

    rhos = np.linspace(-diag, diag, num_rho)
    thetas = np.linspace(-np.pi/2, np.pi/2, num_theta)

    acc = np.zeros((num_rho, num_theta), dtype=np.int32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for j, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            i = int((rho + diag) / (2 * diag) * num_rho)
            if 0 <= i < num_rho:
                acc[i, j] += 1

    return acc, rhos, thetas
````

````python
def draw_rho_theta_lines(img, acc, rhos, thetas, threshold_ratio=0.9):
    acc_norm = acc / acc.max()
    lines = np.argwhere(acc_norm > threshold_ratio)

    img_color = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_GRAY2BGR)
    H, W = img.shape

    for i, j in lines:
        rho = rhos[i]
        theta = thetas[j]

        a, b = np.cos(theta), np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))

        cv2.line(img_color, pt1, pt2, (255,0,0), 1)

    return img_color
````

````python
def extract_ball_stencil(img, x1, y1, x2, y2):
    crop = img[y1:y2, x1:x2].astype(np.float32)
    
    gray = crop.mean(axis=2)

    sobel_x = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], np.float32)
    sobel_y = np.array([[-1,-2,-1],[0,0,0],[1,2,1]], np.float32)

    gx = convolve(gray, sobel_x)
    gy = convolve(gray, sobel_y)
    grad = np.sqrt(gx**2 + gy**2)

    grad = grad / grad.max()
    edge_mask = grad > 0.3

    h, w = edge_mask.shape
    cy, cx = h//2, w//2

    Y, X = np.ogrid[:h, :w]
    dist = np.sqrt((X-cx)**2 + (Y-cy)**2)
    radius = np.mean(dist[edge_mask])

    alpha = (dist <= radius).astype(np.float32)

    rgba = np.zeros((h, w, 4), dtype=np.float32)
    rgba[:, :, :3] = crop[:, :, :3] / 255.0
    rgba[:, :, 3] = alpha

    return rgba
````

````python
def overlay_stencil(bg_img, stencil_rgba, x3, y3):
    bg = bg_img.astype(np.float32) / 255.0
    fg = stencil_rgba

    h, w = fg.shape[:2]
    y1 = y3 - h//2
    x1 = x3 - w//2

    out = bg.copy()

    for i in range(h):
        for j in range(w):
            yi, xi = y1 + i, x1 + j
            if 0 <= yi < bg.shape[0] and 0 <= xi < bg.shape[1]:
                a = fg[i,j,3]
                out[yi,xi,:3] = (
                    a * fg[i,j,:3] +
                    (1-a) * out[yi,xi,:3]
                )

    return (out * 255).astype(np.uint8)
````
