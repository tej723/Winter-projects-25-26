````python
def compute_r_map_explicit(img, window_size=3, k=0.05):
    Ix, Iy = sobel_gradients(img)

    h, w = img.shape
    pad = window_size // 2

    # Pad gradients for windowing
    Ix_pad = np.pad(Ix, pad, mode='constant')
    Iy_pad = np.pad(Iy, pad, mode='constant')

    R_map = np.zeros((h, w), dtype=np.float32)

    # Looping over each pixel
    for i in range(h):
        for j in range(w):
            Ix_win = Ix_pad[i:i+window_size, j:j+window_size]
            Iy_win = Iy_pad[i:i+window_size, j:j+window_size]

            a = np.sum(Ix_win**2)
            b = np.sum(Iy_win**2)
            c = np.sum(Ix_win * Iy_win)

            # Eigenvalues
            trace = a + b
            det = a*b - c**2
            lambda1 = trace/2 + np.sqrt((a-b)**2/4 + c**2)
            lambda2 = trace/2 - np.sqrt((a-b)**2/4 + c**2)

            R_map[i, j] = det - k*trace**2

    return R_map
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

````python
def hough_lines_mc(image,
                   m_min=-1.5, m_max=1.5, m_steps=300,
                   c_min=-400, c_max=400, c_steps=400,
                   acc_thresh=0.9):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Canny Edge Detection
    edges = cv2.Canny(gray, 100, 200)

    h, w = edges.shape

    m_vals = np.linspace(m_min, m_max, m_steps)
    c_vals = np.linspace(c_min, c_max, c_steps)

    accumulator = np.zeros((m_steps, c_steps), dtype=np.float32)

    # Voting
    ys, xs = np.where(edges > 0)
    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int((c - c_min) / (c_max - c_min) * c_steps)
            if 0 <= ci < c_steps:
                accumulator[mi, ci] += 1

    # Normalize accumulator
    accumulator /= accumulator.max()

    # Threshold
    peaks = np.where(accumulator > acc_thresh)

    # Draw detected lines
    output = image.copy()
    for mi, ci in zip(peaks[0], peaks[1]):
        m = m_vals[mi]
        c = c_vals[ci]

        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(output, (x1, y1), (x2, y2), (0, 0, 255), 1)

    # Display results
    plt.figure(figsize=(15,4))

    plt.subplot(1,3,1)
    plt.title("Canny Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Accumulator (mc)")
    plt.imshow(accumulator, cmap='hot')
    plt.colorbar()

    plt.subplot(1,3,3)
    plt.title("Detected Lines (mc)")
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.axis("off")

    plt.show()

    return output
````

````python
def hough_lines_rho_theta(image,
                          rho_res=1,
                          theta_res=1,
                          acc_thresh=0.6):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 100, 200)

    h, w = edges.shape
    diag_len = int(np.sqrt(h*h + w*w))

    rhos = np.arange(-diag_len, diag_len, rho_res)
    thetas = np.deg2rad(np.arange(-90, 90, theta_res))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    # Voting
    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x*np.cos(theta) + y*np.sin(theta)
            ri = int((rho + diag_len) / rho_res)
            if 0 <= ri < len(rhos):
                accumulator[ri, ti] += 1

    # Normalize
    accumulator /= accumulator.max()

    # Threshold
    peaks = np.where(accumulator > acc_thresh)

    output = image.copy()
    for ri, ti in zip(peaks[0], peaks[1]):
        rho = rhos[ri]
        theta = thetas[ti]

        a, b = np.cos(theta), np.sin(theta)
        x0, y0 = a*rho, b*rho

        x1 = int(x0 + 1000*(-b))
        y1 = int(y0 + 1000*(a))
        x2 = int(x0 - 1000*(-b))
        y2 = int(y0 - 1000*(a))

        cv2.line(output, (x1, y1), (x2, y2), (0,255,0), 1)

    # Display
    plt.figure(figsize=(15,4))

    plt.subplot(1,3,1)
    plt.title("Canny Edge Map")
    plt.imshow(edges, cmap='gray')
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Accumulator ()")
    plt.imshow(accumulator, cmap='hot', aspect='auto')
    plt.colorbar()

    plt.subplot(1,3,3)
    plt.title("Detected Lines ()")
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.axis("off")

    plt.show()

    return output
````

````python
def create_ball_stencil(image, x1, y1, x2, y2):
    # Crop region containing ball
    crop = image[y1:y2, x1:x2]

    # Convert to grayscale
    gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)

    # Edge detection
    edges = cv2.Canny(gray, 100, 200)

    # Find contours from edges
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Select largest contour (maximum area)
    largest_contour = max(contours, key=cv2.contourArea)

    # Create mask
    mask = np.zeros(gray.shape, dtype=np.uint8)
    cv2.drawContours(mask, [largest_contour], -1, 255, thickness=-1)

    # Create RGBA stencil
    rgba = np.zeros((crop.shape[0], crop.shape[1], 4), dtype=np.uint8)
    rgba[..., :3] = crop
    rgba[..., 3] = mask

    # Display intermediate results
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.title("Cropped Ball")
    plt.imshow(crop)
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.title("Gradient / Edges")
    plt.imshow(edges, cmap="gray")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.title("RGBA Stencil (Alpha Mask)")
    plt.imshow(mask, cmap="gray")
    plt.axis("off")

    plt.show()

    return rgba
````

````python
def overlay_rgba_stencil(image2, stencil, x3, y3):
    """
    image2 : background image (H, W, 3)
    stencil : RGBA image (h, w, 4)
    (x3, y3) : center location for stencil
    """

    output = image2.copy()

    h, w = stencil.shape[:2]

    # Compute top-left corner from center
    x1 = int(x3 - w // 2)
    y1 = int(y3 - h // 2)
    x2 = x1 + w
    y2 = y1 + h

    # Boundary check
    if x1 < 0 or y1 < 0 or x2 > output.shape[1] or y2 > output.shape[0]:
        raise ValueError("Stencil placement goes outside image")

    # Extract ROI
    roi = output[y1:y2, x1:x2]

    # Separate RGB and Alpha
    stencil_rgb = stencil[:, :, :3].astype(np.float32)
    alpha = stencil[:, :, 3].astype(np.float32) / 255.0
    alpha = alpha[:, :, np.newaxis]

    # Alpha blending
    blended = alpha * stencil_rgb + (1 - alpha) * roi
    blended = blended.astype(np.uint8)

    # Replace ROI
    output[y1:y2, x1:x2] = blended

    # Display result
    plt.figure(figsize=(6,4))
    plt.title("Ball Placed on Pool Table")
    plt.imshow(output)
    plt.axis("off")
    plt.show()

    return output
````
