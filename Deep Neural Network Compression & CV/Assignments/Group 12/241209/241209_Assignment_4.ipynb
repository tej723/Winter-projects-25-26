To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def convolve(img_gray, kernel):
    # Using only ksize instead of height and width because in the assignment, we only need ksize x ksize kernals
    ksize = kernel.shape[0]
    pad = ksize // 2

    # Padding with mode reflect
    img_padded = np.pad(img_gray, pad_width=pad, mode='reflect')

    # Image after convolution
    img_convolve = np.zeros_like(img_gray, dtype=np.float64)

    # Convolution
    for i in range(img_gray.shape[0]):
        for j in range(img_gray.shape[1]):
            patch = img_padded[i:i+ksize, j:j+ksize]
            img_convolve[i, j] = np.sum(patch * kernel)

    return img_convolve
````

````python
#Computing Ix and Iy from Sobel Kernel
def Sobel(img_gray):
  sobel_x = np.array([[ -1, 0, 1],[ -2, 0, 2],[ -1, 0, 1]])

  sobel_y = np.array([[ -1, -2, -1],[  0,  0,  0],[  1,  2,  1]])

  # Gradient in the x-direction
  Ix = convolve(img_gray,sobel_x)
  # Gradient in the y-direction
  Iy = convolve(img_gray,sobel_y)

  return Ix, Iy

# Plotting Scatter Plot of Ix v/s Iy
def Scatter_Plot(img_gray):
  Ix, Iy = Sobel(img_gray)
  ix = Ix.flatten()
  iy = Iy.flatten()

  plt.figure(figsize =(4,4))
  plt.scatter(ix, iy, s=1)
  plt.xlabel("Ix")
  plt.ylabel("Iy")
  plt.title("Scatter Plot Ix v/s Iy")
  plt.show()

Scatter_Plot(flat_gray)
Scatter_Plot(edge_gray)
Scatter_Plot(corner_gray)
````

````python
def Compute_R_map(img_gray,Window_size, k):
    Ix, Iy = Sobel(img_gray)
    h, w = Ix.shape
    pad = Window_size // 2

    R_map = np.zeros_like(img_gray)

    for i in range(pad, h - pad):
        for j in range(pad, w - pad):

            # Gradient Ix and Iy in one window
            Ix_w = Ix[i-pad:i+pad+1, j-pad:j+pad+1]
            Iy_w = Iy[i-pad:i+pad+1, j-pad:j+pad+1]

            a = np.sum(Ix_w ** 2)
            b = np.sum(Ix_w * Iy_w)
            c = np.sum(Iy_w ** 2)

            lambda1 = (a + c + np.sqrt((a - c)**2 + 4*b**2)) / 2
            lambda2 = (a + c - np.sqrt((a - c)**2 + 4*b**2)) / 2

            # Harris Corner Response Function
            R_map[i, j] = lambda1 * lambda2 - k * (lambda1 + lambda2)**2 # 0.04 <= k <= 0.06

    # Normalise R_map in (0,1)
    R_map = R_map/(R_map.max() + 1e-8)
    R_map = np.clip(R_map,0,1)

    return R_map

plt.figure(figsize = (12,4))
plt.subplot(1,3,1)
plt.imshow(Compute_R_map(flat_gray, 5, 0.04), cmap='gray')
plt.title("R Map of Flat Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(Compute_R_map(edge_gray,5,0.04), cmap = 'gray')
plt.title("R Map of Edge Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(Compute_R_map(corner_gray,5,0.04), cmap = 'gray')
plt.title("R Map of Corner Image")
plt.axis("off")
plt.show()
````

Hough Transform using slope intercept method

````python
def hough_slope_intercept(img):

    # Canny edge map
    edges = cv2.Canny(img, 100, 200)
    h, w = img.shape

    # slope (m) and intercept (c) ranges
    m_min, m_max, m_step = -2, 2, 0.01
    c_min, c_max, c_step = -500, 500, 1

    m_vals = np.arange(m_min, m_max, m_step)
    c_vals = np.arange(c_min, c_max, c_step)

    accumulator = np.zeros((len(m_vals), len(c_vals)), dtype=np.float32)

    # Voting Algorithm
    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int((c - c_min) / c_step)
            if 0 <= ci < len(c_vals):
                accumulator[mi, ci] += 1

    # Normalize accumulator
    acc_norm = accumulator / (accumulator.max() + 1e-8)

    # Applying threshold in accumulator
    threshold = 0.9
    line_indices = np.where(acc_norm > threshold)

    # Draw detected lines
    img_lines = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    for mi, ci in zip(line_indices[0], line_indices[1]):
        m = m_vals[mi]
        c = c_vals[ci]

        x1, x2 = 0, w
        y1 = int(m * x1 + c)
        y2 = int(m * x2 + c)

        if 0 <= y1 < h and 0 <= y2 < h:
            cv2.line(img_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

    return edges, acc_norm, img_lines

edges, acc, result = hough_slope_intercept(corner2_gray)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.title("Edges")
plt.imshow(edges, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Accumulator")
plt.imshow(acc, cmap='hot')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Detected Lines")
plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
````

Hough Transform using Rho Theta method

````python
def hough_rho_theta(img):

    # Canny edge map
    edges = cv2.Canny(img, 100, 200)
    h, w = img.shape

    # theta and rho ranges
    thetas = np.deg2rad(np.arange(-90, 90, 1))
    diag_len = int(np.sqrt(h*h + w*w))
    rhos = np.arange(-diag_len, diag_len, 1)

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    # Voting algorithm
    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = int(x * np.cos(theta) + y * np.sin(theta))
            ri = rho + diag_len
            accumulator[ri, ti] += 1

    # Normalize accumulator
    acc_norm = accumulator / (accumulator.max() + 1e-8)

    # Applying Threshold
    threshold = 0.75
    line_indices = np.where(acc_norm > threshold)

    # Draw detected lines
    img_lines = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    for ri, ti in zip(line_indices[0], line_indices[1]):
        rho = rhos[ri]
        theta = thetas[ti]

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

    return edges, acc_norm, img_lines

edges_rt, acc_rt, result_rt = hough_rho_theta(corner2_gray)

plt.figure(figsize=(15,4))

plt.subplot(1,3,1)
plt.title("Edges")
plt.imshow(edges_rt, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("Accumulator ()")
plt.imshow(acc_rt, cmap='hot')
plt.axis('off')

plt.subplot(1,3,3)
plt.title("Detected Lines")
plt.imshow(cv2.cvtColor(result_rt, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
````
