To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sobel(image):

    img_float = np.float32(image)

    Ix = cv2.Sobel(img_float, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(img_float, cv2.CV_64F, 0, 1, ksize=3)


    ix_flat = Ix.flatten()
    iy_flat = Iy.flatten()

    plt.figure(figsize=(16, 4))


    plt.subplot(1, 4, 1)
    plt.imshow(image, cmap='gray')
    plt.title("original")
    plt.axis('off')


    plt.subplot(1, 4, 2)
    plt.imshow(np.abs(Ix), cmap='gray')
    plt.title("Ix (Horizontal Gradient)")
    plt.axis('off')


    plt.subplot(1, 4, 3)
    plt.imshow(np.abs(Iy), cmap='gray')
    plt.title("Iy (Vertical Gradient)")
    plt.axis('off')


    plt.subplot(1, 4, 4)
    plt.scatter(ix_flat, iy_flat, alpha=0.1, s=1, c='blue')
    plt.title("Scatter: Iy vs Ix")
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.grid(True)

    limit = max(np.max(np.abs(ix_flat)), np.max(np.abs(iy_flat))) + 10
    plt.xlim(-limit, limit)
    plt.ylim(-limit, limit)

    plt.tight_layout()
    plt.show()

    return Ix, Iy
````

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def hr(image, Ix, Iy, k=0.04):

    Ixx = Ix ** 2
    Iyy = Iy ** 2
    Ixy = Ix * Iy


    window_size = 5
    a = cv2.GaussianBlur(Ixx, (window_size, window_size), 1)
    c = cv2.GaussianBlur(Iyy, (window_size, window_size), 1)
    b = cv2.GaussianBlur(Ixy, (window_size, window_size), 1)


    det = (a * c) - (b ** 2)
    trace = a + c


    R = det - k * (trace ** 2)

    return R

````

````python
def r_map(image, R):
    plt.figure(figsize=(10, 4))


    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title("original")
    plt.axis('off')


    plt.subplot(1, 2, 2)
    plt.imshow(R, cmap='gray')
    plt.title(f"R-Map (Response)\nWhite = Corner")
    plt.colorbar()
    plt.axis('off')

    plt.tight_layout()
    plt.show()
````

````python
def hough(image):

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) if len(image.shape) == 3 else image
    edges = cv2.Canny(gray, 50, 150)
    height, width = edges.shape


    diag_len = int(np.sqrt(height**2 + width**2))
    rhos = np.linspace(-diag_len, diag_len, diag_len * 2)


    thetas = np.deg2rad(np.arange(-90, 90, 1))


    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.uint64)


    cos_t = np.cos(thetas)
    sin_t = np.sin(thetas)


    y_idxs, x_idxs = np.nonzero(edges)




    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]


        rho_vals = x * cos_t + y * sin_t


        for t_idx in range(len(thetas)):
            rho_val = rho_vals[t_idx]
            rho_idx = int(rho_val + diag_len)
            accumulator[rho_idx, t_idx] += 1


    threshold = 0.9 * np.max(accumulator)
    rho_idxs, theta_idxs = np.where(accumulator > threshold)


    result = image.copy()
    if len(result.shape) == 2: result = cv2.cvtColor(result, cv2.COLOR_GRAY2BGR)

    for i in range(len(rho_idxs)):
        rho = rhos[rho_idxs[i]]
        theta = thetas[theta_idxs[i]]

        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho


        p1 = (int(x0 + 1000 * (-b)), int(y0 + 1000 * (a)))
        p2 = (int(x0 - 1000 * (-b)), int(y0 - 1000 * (a)))

        cv2.line(result, p1, p2, (0, 255, 0), 1)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")
    plt.subplot(1, 2, 2)
    plt.imshow(result)
    plt.title(f"Rho-Theta Output\n(Threshold: {int(threshold)})")
    plt.show()
````

````python


def create_ball_stencil(image, top_left, bottom_right):

    x1, y1 = top_left
    x2, y2 = bottom_right


    roi = image[y1:y2, x1:x2]
    if roi.size == 0:
        print("Error: Empty ROI selected.")
        return None


    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)


    blurred = cv2.GaussianBlur(gray, (5, 5), 0)


    edges = cv2.Canny(blurred, 50, 150)


    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours:
        print("No contour found in the box!")
        return None


    max_contour = max(contours, key=cv2.contourArea)


    mask = np.zeros_like(gray)

    kernel = np.ones((3, 3), np.uint8)


    mask = cv2.erode(mask, kernel, iterations=1)


    cv2.drawContours(mask, [max_contour], -1, 255, thickness=cv2.FILLED)


    b, g, r = cv2.split(roi)


    rgba = cv2.merge([b, g, r, mask])

    return rgba
````

````python

def overlay_stencil(background, stencil, center_x, center_y):

    bg_h, bg_w = background.shape[:2]
    s_h, s_w = stencil.shape[:2]


    x_offset = center_x - (s_w // 2)
    y_offset = center_y - (s_h // 2)


    output = background.copy()


    y1, y2 = max(0, y_offset), min(bg_h, y_offset + s_h)
    x1, x2 = max(0, x_offset), min(bg_w, x_offset + s_w)


    sy1, sy2 = max(0, -y_offset), min(s_h, s_h - (y_offset + s_h - bg_h))
    sx1, sx2 = max(0, -x_offset), min(s_w, s_w - (x_offset + s_w - bg_w))


    if y1 >= y2 or x1 >= x2 or sy1 >= sy2 or sx1 >= sx2:
        return output

    roi_bg = output[y1:y2, x1:x2]


    roi_stencil = stencil[sy1:sy2, sx1:sx2]


    stencil_bgr = roi_stencil[:, :, :3]
    alpha_mask = roi_stencil[:, :, 3] / 255.0


    alpha_inv = 1.0 - alpha_mask

    for c in range(3):
        roi_bg[:, :, c] = (stencil_bgr[:, :, c] * alpha_mask +
                           roi_bg[:, :, c] * alpha_inv)


    output[y1:y2, x1:x2] = roi_bg

    return output
````
