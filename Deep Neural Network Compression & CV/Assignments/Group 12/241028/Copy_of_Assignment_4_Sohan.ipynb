To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
# --- Helpers (Required specifically for Part B inputs) ---
def generate_flat():
    return np.ones((100, 100), dtype=np.uint8) * 128
def generate_edge():
    img = np.zeros((100, 100), dtype=np.uint8)
    img[:, 50:] = 255
    return img
def generate_corner():
    img = np.zeros((100, 100), dtype=np.uint8)
    img[30:70, 30:70] = 255
    return img
def get_gradients(image):
    """Calculates Ix and Iy without plotting (Helper for Part B)."""
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    Ix = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    Iy = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    return Ix, Iy
# --- Part B: Ellipse Fitting & R Maps (Main Solution) ---
def calculate_r_map(image, window_size=5, k=0.04):
    """
    Calculates the R-map (Harris corner response) w/ window iteration.
    Display the R map as requested.
    """
    # 1. Get Gradients (Ix, Iy)
    Ix, Iy = get_gradients(image)

    height, width = image.shape
    r_map = np.zeros((height, width), dtype=np.float64)
    offset = window_size // 2

    # 2. Loop over pixels to find R values
    # We loop excluding the border covered by the window
    for y in range(offset, height - offset):
        for x in range(offset, width - offset):
            # Extract window
            window_Ix = Ix[y-offset:y+offset+1, x-offset:x+offset+1]
            window_Iy = Iy[y-offset:y+offset+1, x-offset:x+offset+1]

            # Compute moments (Elements of Matrix M)
            a = np.sum(window_Ix ** 2)
            b = np.sum(window_Ix * window_Iy)
            c = np.sum(window_Iy ** 2)

            # Construct Second Moment Matrix M
            # M = [[a, b],
            #      [b, c]]

            # Calculate Eigenvalues
            # Using manual quadratic formula for 2x2 eigenvalues or numpy
            # lambda = (tr(M) +/- sqrt(tr(M)^2 - 4*det(M))) / 2
            # Here using numpy's linear algebra solver for clarity
            M = np.array([[a, b], [b, c]])
            w = np.linalg.eigvalsh(M)
            l1, l2 = w[0], w[1]

            # Calculate R
            # R = det(M) - k(trace(M))^2
            # R = (l1 * l2) - k * (l1 + l2)^2
            R = (l1 * l2) - k * ((l1 + l2) ** 2)

            r_map[y, x] = R
    return r_map
# --- Execution & Display ---
def run_part_b():
    images = [
        ("Flat Image", generate_flat()),
        ("Edge Image", generate_edge()),
        ("Corner Image", generate_corner())
    ]
    plt.figure(figsize=(15, 5))
    plt.suptitle("Part B: R-Map Analysis", fontsize=16)
    for i, (name, img) in enumerate(images):
        print(f"Calculating R-map for {name}...")
        r_map_val = calculate_r_map(img)

        # Display
        plt.subplot(1, 3, i+1)
        plt.imshow(r_map_val, cmap='gray')
        plt.title(f'{name}\nR Map')
        plt.axis('off')
    plt.tight_layout()
    plt.show()
if __name__ == "__main__":
    run_part_b()
````

````python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import math
# --- 1. Image Generation ---
def generate_line_image():
    """Generates an image with a diagonal line and a horizontal line."""
    img = np.zeros((100, 100), dtype=np.uint8)
    # Diagonal line
    for i in range(20, 80):
        img[i, i] = 255
    # Horizontal line
    img[80, 20:80] = 255
    return img
# --- 2. Rho-Theta Hough Transform ---
def hough_rho_theta(image, edge_map):
    """
    Implements Hough Transform using rho = x*cos(theta) + y*sin(theta).
    Can detect all lines including vertical ones.
    """
    height, width = image.shape
    diagonal = int(np.sqrt(height**2 + width**2))

    # Define Parameter Limits
    # Theta: 0 to 180 degrees
    theta_step = 1.0 # degree
    thetas_deg = np.arange(0, 180, theta_step)
    thetas_rad = np.deg2rad(thetas_deg)

    # Rho: -diagonal to +diagonal
    rho_min, rho_max = -diagonal, diagonal
    rho_len = int(rho_max - rho_min) + 1

    # Accumulator: (rho_bins, theta_bins)
    accumulator = np.zeros((rho_len, len(thetas_rad)), dtype=np.int32)

    print("Voting in (rho, theta) accumulator...")

    # Get all edge pixel coordinates
    y_idxs, x_idxs = np.nonzero(edge_map)

    # Precompute cos and sin to optimize inner loop speed slightly
    cos_t = np.cos(thetas_rad)
    sin_t = np.sin(thetas_rad)

    # Voting Loop
    for i in range(len(x_idxs)):
        x = x_idxs[i]
        y = y_idxs[i]

        # Calculate rho for every theta
        # rho = x * cos(theta) + y * sin(theta)
        for t_idx in range(len(thetas_rad)):
            rho = x * cos_t[t_idx] + y * sin_t[t_idx]

            # Map rho to index
            rho_idx = int(rho - rho_min)

            accumulator[rho_idx, t_idx] += 1

    return accumulator, thetas_rad, rho_min
def draw_lines_rho_theta(image, accumulator, thetas_rad, rho_min, threshold_ratio=0.7):
    """Drawing lines based on peaks in (rho, theta) accumulator."""
    output_img = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    height, width = image.shape

    threshold = np.max(accumulator) * threshold_ratio

    # Find peaks where votes > threshold
    rho_idxs, theta_idxs = np.where(accumulator >= threshold)

    print(f"Found {len(rho_idxs)} lines (Threshold: {threshold}).")

    for i in range(len(rho_idxs)):
        rho = rho_min + rho_idxs[i] # reconstruct rho
        theta = thetas_rad[theta_idxs[i]]

        a = np.cos(theta)
        b = np.sin(theta)

        x0 = a * rho
        y0 = b * rho

        # Find two points on valid line extended far out
        # x = x0 - 1000 * sin(theta)
        # y = y0 + 1000 * cos(theta)
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))

        cv2.line(output_img, pt1, pt2, (0, 0, 255), 1)

    return output_img
# --- Main Execution ---
if __name__ == "__main__":
    # Setup
    img = generate_line_image()
    edges = cv2.Canny(img, 100, 200)

    # Run Algorithm
    print("--- Rho-Theta Method ---")
    acc, thetas_rad, rho_min = hough_rho_theta(img, edges)
    result_img = draw_lines_rho_theta(img, acc, thetas_rad, rho_min, threshold_ratio=0.6)

    # Display
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 4, 1)
    plt.imshow(img, cmap='gray')
    plt.title("Original Image")

    plt.subplot(1, 4, 2)
    plt.imshow(edges, cmap='gray')
    plt.title("Canny Edges")

    plt.subplot(1, 4, 3)
    plt.imshow(acc, cmap='jet', aspect='auto')
    plt.title("Hough Accumulator")

    plt.subplot(1, 4, 4)
    plt.imshow(result_img)
    plt.title("Detected Lines (Red)")
    plt.show()
````

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def create_ball_stencil(image, box_coords):
    """
    Extracts a stencil (RGBA) of the object inside the box coordinates.
    Finds the largest gradient contour (the ball) to create the alpha mask.

    Args:
        image: Source image (BGR)
        box_coords: Tuple (x1, y1, x2, y2) defining the crop box

    Returns:
        rgba_stencil: 4-channel image (The cutout ball)
    """
    x1, y1, x2, y2 = box_coords

    # 0. Debug Checks
    if image is None:
        print("Error: Image not found!")
        return None

    # Ensure coordinates are within bounds
    h, w = image.shape[:2]
    x1, x2 = max(0, x1), min(w, x2)
    y1, y2 = max(0, y1), min(h, y2)

    print(f"DEBUG: Processing Box ({x1}:{x2}, {y1}:{y2}) on Image {w}x{h}")

    # 1. Crop the ROI
    roi = image[y1:y2, x1:x2].copy()

    if roi.size == 0:
        print("Error: Cropped ROI is empty. Check your box coordinates!")
        return None

    # 2. Find Gradients / Edges
    gray_roi = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)

    # Gaussian blur to reduce noise
    blurred = cv2.GaussianBlur(gray_roi, (5, 5), 0)

    # --- FIX: Dynamic Canny Thresholding ---
    # Using the median pixel intensity to robustly find edges
    v = np.median(blurred)
    sigma = 0.33
    lower = int(max(0, (1.0 - sigma) * v))
    upper = int(min(255, (1.0 + sigma) * v))

    print(f"DEBUG: Auto-Canny Thresholds: Lower={lower}, Upper={upper}")

    edges = cv2.Canny(blurred, lower, upper)

    # 3. Find Contours
    # RETR_EXTERNAL to get the outer boundary
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours:
        print("Warning: No contours found in box!")
        print("Tip: If the ball is very faint, try manually setting lower/upper in cv2.Canny")
        # Fallback: Show the edge map to help debug
        plt.imshow(edges, cmap='gray')
        plt.title("Debug: No Contours Found - Edge Map")
        plt.show()
        return None

    # 4. Select Biggest Completed Gradient (Max Area Contour)
    max_contour = max(contours, key=cv2.contourArea)
    print(f"DEBUG: Found contour with area {cv2.contourArea(max_contour)}")

    # 5. Create Alpha Mask
    mask = np.zeros_like(gray_roi)
    # Draw filed contour in white
    cv2.drawContours(mask, [max_contour], -1, (255), thickness=cv2.FILLED)

    # Optional: Smooth the mask edges
    mask = cv2.GaussianBlur(mask, (3, 3), 0)

    # 6. Create RGBA Stencil
    # Split BGR channels
    b, g, r = cv2.split(roi)

    # Merge with mask as Alpha channel
    rgba_stencil = cv2.merge([b, g, r, mask])

    return rgba_stencil
````

````python
def overlay_stencil_on_image(background_img, stencil, center_xy):
    """
    Overlays the RGBA stencil onto the background image at the specified center.
    """
    if background_img is None or stencil is None:
        print("Error: Invalid inputs for overlay.")
        return background_img
    bg_h, bg_w = background_img.shape[:2]
    s_h, s_w = stencil.shape[:2]

    cx, cy = center_xy

    # Top-left coordinates for stencil
    tl_x = cx - s_w // 2
    tl_y = cy - s_h // 2

    # Bottom-right coordinates
    br_x = tl_x + s_w
    br_y = tl_y + s_h

    # Background ROI
    bg_x1 = max(0, tl_x)
    bg_y1 = max(0, tl_y)
    bg_x2 = min(bg_w, br_x)
    bg_y2 = min(bg_h, br_y)

    # Stencil ROI
    st_x1 = bg_x1 - tl_x
    st_y1 = bg_y1 - tl_y
    st_x2 = st_x1 + (bg_x2 - bg_x1)
    st_y2 = st_y1 + (bg_y2 - bg_y1)

    if bg_x2 <= bg_x1 or bg_y2 <= bg_y1:
        return background_img

    bg_roi = background_img[bg_y1:bg_y2, bg_x1:bg_x2]
    stencil_roi = stencil[st_y1:st_y2, st_x1:st_x2]

    st_b, st_g, st_r, st_a = cv2.split(stencil_roi)
    st_rgb = cv2.merge([st_b, st_g, st_r])

    # Alpha Blending
    alpha = st_a.astype(float) / 255.0
    alpha = cv2.merge([alpha, alpha, alpha])

    foreground = st_rgb.astype(float)
    background = bg_roi.astype(float)

    out_roi = (foreground * alpha) + (background * (1.0 - alpha))
    out_roi = out_roi.astype(np.uint8)

    result = background_img.copy()
    result[bg_y1:bg_y2, bg_x1:bg_x2] = out_roi

    return result
# --- REVISED USAGE BLOCK ---
# We assume 'ball' and 'ball2' are already defined in your notebook
# If not, please load them before running this!
if __name__ == "__main__":

    # 1. Check for Image Variables
    # In Colab, simple variables are global. We'll check local/global scope.
    # Note: When pasting, just ensure 'ball' and 'ball2' exist.

    # For this script to be safe, I will explicitly ask you to assign them:
    # REPLACE THE VARIABLES BELOW WITH YOUR ACTUAL IMAGE VARIABLES
    image_source = locals().get('ball')  # or just use: image_source = ball
    image_target = locals().get('ball2') # or just use: image_target = ball2

    if image_source is None:
        print("Wait! 'ball' variable not found.")
        print("Please set: image_source = ball  (or whatever your variable is named)")
        # Using a dummy check just to prevent crash if run blindly
        image_source = np.zeros((100,100,3), dtype=np.uint8) + 50 # Small gray image

    if image_target is None:
         # Using a dummy check
         image_target = np.zeros((100,100,3), dtype=np.uint8)

    # 2. Extract Stencil
    # Adjust 'box' to surround your white ball
    # Format: (x_min, y_min, x_max, y_max)
    box = (150, 250, 350, 450) # <-- CHANGE THIS if needed

    print("\n--- Running Part D ---")
    stencil = create_ball_stencil(image_source, box)

    if stencil is not None:
        plt.figure()
        plt.imshow(cv2.cvtColor(stencil, cv2.COLOR_BGRA2RGBA))
        plt.title("Step 1: Extracted Stencil")
        plt.show()

        # 3. Overlay
        center_x = 450
        center_y = 250

        final_result = overlay_stencil_on_image(image_target, stencil, (center_x, center_y))

        plt.figure(figsize=(10, 6))
        plt.imshow(cv2.cvtColor(final_result, cv2.COLOR_BGR2RGB))
        plt.title("Step 2: Final Result")
        plt.show()
    else:
        print("Stencil creation failed. Check box coordinates.")
````
