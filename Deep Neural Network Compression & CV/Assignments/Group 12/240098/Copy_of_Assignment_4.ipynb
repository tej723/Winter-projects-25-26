To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
# Your PART A codes follow here. You can create more cells under this. Write clean and commented code.

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
def load_gray_image(path):
    img = Image.open(path).convert("L")   # grayscale
    return np.array(img, dtype=np.float32)
def sobel_corner_analysis(img_gray, title="", threshold=None):
    """
    Input:
        img_gray : grayscale image (float32)
        title    : image name for plots
        threshold: optional threshold on gradient magnitude

    Output:
        Ix, Iy : Sobel gradients
    """

    # Sobel kernels
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[ 1,  2,  1],
                        [ 0,  0,  0],
                        [-1, -2, -1]], dtype=np.float32)

    # Compute gradients
    Ix = cv2.filter2D(img_gray, -1, sobel_x)
    Iy = cv2.filter2D(img_gray, -1, sobel_y)

    # Optional thresholding
    if threshold is not None:
        mag = np.sqrt(Ix**2 + Iy**2)
        mask = mag > threshold
        Ix_plot = Ix[mask]
        Iy_plot = Iy[mask]
    else:
        Ix_plot = Ix.flatten()
        Iy_plot = Iy.flatten()

    # Plot Ix and Iy
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title(f"{title} - Ix")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title(f"{title} - Iy")
    plt.axis("off")

    # Scatter plot Iy vs Ix
    plt.subplot(1,3,3)
    plt.scatter(Ix_plot, Iy_plot, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix Scatter")

    plt.tight_layout()
    plt.show()

    if threshold is not None:
        print(f"Threshold applied on gradient magnitude: {threshold}")

    return Ix, Iy
````

````python
# Your PART B codes follow here. You can create more cells under this. Write clean and commented code.

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
def load_gray_image(path):
    img = Image.open(path).convert("L")
    return np.array(img, dtype=np.float32)
def compute_sobel_gradients(img_gray):
    sobel_x = np.array([[-1, 0, 1],
                        [-2, 0, 2],
                        [-1, 0, 1]], dtype=np.float32)

    sobel_y = np.array([[ 1,  2,  1],
                        [ 0,  0,  0],
                        [-1, -2, -1]], dtype=np.float32)

    Ix = cv2.filter2D(img_gray, -1, sobel_x)
    Iy = cv2.filter2D(img_gray, -1, sobel_y)

    return Ix, Iy
````

````python
def compute_r_map(img_gray, window_size=5, k=0.04):
    """
    Returns:
        R_map : 2D array where each pixel has its Harris R value
    """

    Ix, Iy = compute_sobel_gradients(img_gray)

    h, w = img_gray.shape
    offset = window_size // 2

    R_map = np.zeros((h, w), dtype=np.float32)

    for i in range(offset, h - offset):
        for j in range(offset, w - offset):

            # Window
            Ix_win = Ix[i-offset:i+offset+1, j-offset:j+offset+1]
            Iy_win = Iy[i-offset:i+offset+1, j-offset:j+offset+1]

            # Second moments
            a = np.sum(Ix_win ** 2)
            b = np.sum(Iy_win ** 2)
            c = np.sum(Ix_win * Iy_win)

            # Harris response
            R = (a * b - c ** 2) - k * (a + b) ** 2
            R_map[i, j] = R

    return R_map

def show_r_map(R_map, title="R Map"):
    plt.figure(figsize=(5,5))
    plt.imshow(R_map, cmap='gray')
    plt.colorbar()
    plt.title(title)
    plt.axis("off")
    plt.show()
````

````python
# Load images
img1 = load_gray_image("Virat.jpg")
img2 = load_gray_image("vk2.jpg")
img3 = load_gray_image("vk3.jpg")

# Compute R maps
R1 = compute_r_map(img1, window_size=5, k=0.04)
R2 = compute_r_map(img2, window_size=5, k=0.04)
R3 = compute_r_map(img3, window_size=5, k=0.04)

# Display
show_r_map(R1, "Virat - R Map")
show_r_map(R2, "vk2 - R Map")
show_r_map(R3, "vk3 - R Map")
````

````python
# Your PART C codes follow here. You can create more cells under this. Write clean and commented code.

import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
def load_gray_image(path):
    img = Image.open(path).convert("L")
    return np.array(img, dtype=np.uint8)
def get_canny_edges(img_gray, low=100, high=200):
    edges = cv2.Canny(img_gray, low, high)
    return edges
````

````python
def hough_line_mc(edges, m_range, c_range):
    """
    m_range : array of slopes
    c_range : array of intercepts
    """

    accumulator = np.zeros((len(m_range), len(c_range)), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    # Calculate the step size for c_range once
    c_step_size = c_range[1] - c_range[0]

    for x, y in zip(xs, ys):
        for i, m in enumerate(m_range):
            c = y - m * x

            # Check if c is within the c_range with a small epsilon for floating point comparison
            if (c_range[0] - 1e-9) <= c <= (c_range[-1] + 1e-9):
                # Calculate the index j based on the value c and the step size
                j = int(round((c - c_range[0]) / c_step_size))

                # Ensure j is within the valid bounds [0, len(c_range)-1]
                j = np.clip(j, 0, len(c_range) - 1)

                accumulator[i, j] += 1

    return accumulator

def draw_lines_mc(img, m_range, c_range, accumulator, threshold):
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    acc_norm = accumulator / accumulator.max()

    idxs = np.where(acc_norm > threshold)

    for i, j in zip(idxs[0], idxs[1]):
        m = m_range[i]
        c = c_range[j]

        x1, x2 = 0, img.shape[1]
        y1 = int(c)
        y2 = int(m * x2 + c)

        cv2.line(img_color, (x1, y1), (x2, y2), (0,255,0), 1)

    return img_color

img = load_gray_image("Virat.jpg")
edges = get_canny_edges(img)

m_range = np.linspace(-1, 1, 200)
c_range = np.linspace(-img.shape[0], img.shape[0], 300)

acc_mc = hough_line_mc(edges, m_range, c_range)

result_mc = draw_lines_mc(img, m_range, c_range, acc_mc, threshold=0.9)

plt.figure(figsize=(6,6))
plt.imshow(result_mc)
plt.title("Hough Lines (m, c)")
plt.axis("off")
plt.show()

#Slopeintercept Hough is computationally expensive since every edge pixel votes for all slopes. The accumulator becomes large and runtime increases significantly.
````

````python
def hough_line_rho_theta(edges, rho_res=1, theta_res=1):
    h, w = edges.shape
    diag_len = int(np.sqrt(h*h + w*w))

    rhos = np.arange(-diag_len, diag_len, rho_res)
    thetas = np.deg2rad(np.arange(-90, 90, theta_res))

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    ys, xs = np.where(edges > 0)

    for x, y in zip(xs, ys):
        for j, theta in enumerate(thetas):
            rho = int(x * np.cos(theta) + y * np.sin(theta))
            i = np.where(rhos == rho)[0]
            if len(i) > 0:
                accumulator[i[0], j] += 1

    return accumulator, rhos, thetas

def draw_lines_rho_theta(img, accumulator, rhos, thetas, threshold):
    img_color = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

    acc_norm = accumulator / accumulator.max()
    idxs = np.where(acc_norm > threshold)

    for i, j in zip(idxs[0], idxs[1]):
        rho = rhos[i]
        theta = thetas[j]

        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho

        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        cv2.line(img_color, (x1,y1), (x2,y2), (255,0,0), 1)

    return img_color

acc_rt, rhos, thetas = hough_line_rho_theta(edges)

result_rt = draw_lines_rho_theta(img, acc_rt, rhos, thetas, threshold=0.9)

plt.figure(figsize=(6,6))
plt.imshow(result_rt)
plt.title("Hough Lines (rho-theta)")
plt.axis("off")
plt.show()

#The slopeintercept Hough transform is slow due to voting over a large parameter space.
#The  formulation is significantly faster and more stable, since vertical lines are handled naturally and parameter bounds are compact.
#Hence, practical Hough implementations prefer the  representation.
````

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def create_ball_stencil(img, x1, y1, x2, y2, grad_thresh=20):
    """
    img : RGB image (numpy)
    (x1,y1) top-left, (x2,y2) bottom-right bounding box
    returns RGBA stencil
    """

    # Crop ROI
    roi = img[y1:y2, x1:x2].copy()

    # Convert to grayscale
    gray = np.mean(roi, axis=2)

    # Simple gradients
    gx = np.abs(np.gradient(gray, axis=1))
    gy = np.abs(np.gradient(gray, axis=0))
    grad_mag = np.sqrt(gx**2 + gy**2)

    # Binary mask from gradient
    mask = grad_mag > grad_thresh

    # Fill interior (rough heuristic)
    mask = mask.astype(np.uint8)
    for i in range(mask.shape[0]):
        row = mask[i]
        if np.any(row):
            first = np.argmax(row)
            last = len(row) - np.argmax(row[::-1]) - 1
            mask[i, first:last] = 1

    # Create RGBA
    h, w, _ = roi.shape
    rgba = np.zeros((h, w, 4), dtype=np.uint8)
    rgba[..., :3] = roi
    rgba[..., 3] = mask * 255

    return rgba
````

````python
def overlay_stencil(bg_img, stencil, center_x, center_y):
    """
    bg_img : RGB image
    stencil : RGBA image
    (center_x, center_y) is stencil center placement
    """

    output = bg_img.copy()
    h, w, _ = stencil.shape

    x_start = center_x - w // 2
    y_start = center_y - h // 2

    for i in range(h):
        for j in range(w):
            if stencil[i, j, 3] > 0:
                y = y_start + i
                x = x_start + j
                if 0 <= y < output.shape[0] and 0 <= x < output.shape[1]:
                    output[y, x] = stencil[i, j, :3]

    return output
````
