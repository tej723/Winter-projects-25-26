To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

````python
def convolve(img, kernel):
    """
    img    : 2D numpy array (grayscale image)
    kernel : 2D numpy array (odd-sized)
    return : convolved image
    """
    H, W = img.shape
    kH, kW = kernel.shape
    pad_h = kH // 2
    pad_w = kW // 2

    # Pad image using reflect mode
    padded = np.pad(img, ((pad_h, pad_h), (pad_w, pad_w)), mode='reflect')

    # Output image
    out = np.zeros_like(img, dtype=np.float32)

    # Convolution
    for i in range(H):
        for j in range(W):
            region = padded[i:i+kH, j:j+kW]   # neighborhood
            out[i, j] = np.sum(region * kernel)

    return out
````

````python
def analyze_gradients(img_gray, threshold=None):
    """
    img_gray : 2D grayscale image (float or uint8)
    threshold: optional magnitude threshold
    returns  : Ix, Iy
    """

    # Ensure float
    img_gray = img_gray.astype(np.float32)

    # Compute gradients
    Ix = convolve(img_gray, sobel_x)
    Iy = convolve(img_gray, sobel_y)

    # Optional thresholding
    if threshold is not None:
        mag = np.sqrt(Ix**2 + Iy**2)
        mask = mag > threshold

        Ix_plot = Ix[mask]
        Iy_plot = Iy[mask]
    else:
        Ix_plot = Ix.flatten()
        Iy_plot = Iy.flatten()

    # ---- Plotting ----
    plt.figure(figsize=(12,4))

    plt.subplot(1,3,1)
    plt.imshow(Ix, cmap='gray')
    plt.title("Ix (Gx)")
    plt.axis("off")

    plt.subplot(1,3,2)
    plt.imshow(Iy, cmap='gray')
    plt.title("Iy (Gy)")
    plt.axis("off")

    plt.subplot(1,3,3)
    plt.scatter(Ix_plot, Iy_plot, s=1)
    plt.xlabel("Ix")
    plt.ylabel("Iy")
    plt.title("Iy vs Ix Scatter Plot")

    plt.show()

    return Ix, Iy
````

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Now, display these R maps. Notice how the corners are clearly visible in white.

````python
# Your PART B codes follow here. You can create more cells under this. Write clean and commented code.

def harris_r_map(img_gray, window_size=5, k=0.04):
    """
    img_gray   : 2D grayscale image
    window_size: size of local window (odd)
    k          : Harris constant
    return     : R-map (same size as image)
    """

    img_gray = img_gray.astype(np.float32)

    # Step 1: Gradients
    Ix = convolve(img_gray, sobel_x)
    Iy = convolve(img_gray, sobel_y)

    # Step 2: Products of gradients
    Ix2 = Ix * Ix
    Iy2 = Iy * Iy
    Ixy = Ix * Iy

    # Padding for window operation
    pad = window_size // 2
    Ix2_p = np.pad(Ix2, pad, mode='reflect')
    Iy2_p = np.pad(Iy2, pad, mode='reflect')
    Ixy_p = np.pad(Ixy, pad, mode='reflect')

    H, W = img_gray.shape
    R_map = np.zeros((H, W), dtype=np.float32)

    # Step 3: Slide window
    for i in range(H):
        for j in range(W):
            window_Ix2 = Ix2_p[i:i+window_size, j:j+window_size]
            window_Iy2 = Iy2_p[i:i+window_size, j:j+window_size]
            window_Ixy = Ixy_p[i:i+window_size, j:j+window_size]

            # Second moments
            a = np.sum(window_Ix2)
            b = np.sum(window_Iy2)
            c = np.sum(window_Ixy)

            # Harris response
            det = a * b - c * c
            trace = a + b
            R = det - k * (trace ** 2)

            R_map[i, j] = R

    return R_map
````

````python
def hough_slope_intercept(edges, m_range=(-2, 2), m_steps=200):
    H, W = edges.shape

    # Discretize slope values
    m_vals = np.linspace(m_range[0], m_range[1], m_steps)

    # Intercept range based on image size
    c_min, c_max = -H, H
    c_steps = 2 * H
    c_vals = np.linspace(c_min, c_max, c_steps)

    accumulator = np.zeros((len(m_vals), len(c_vals)), dtype=np.float32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x
            ci = int((c - c_min) / (c_max - c_min) * (c_steps - 1))
            if 0 <= ci < c_steps:
                accumulator[mi, ci] += 1

    return accumulator, m_vals, c_vals
````

````python
acc_mc, m_vals, c_vals = hough_slope_intercept(edges)

acc_norm = acc_mc / acc_mc.max()

threshold = 0.9   # 90% of max votes
lines = np.where(acc_norm > threshold)
````

````python
def hough_rho_theta(edges, theta_steps=180):
    H, W = edges.shape
    diag = int(np.sqrt(H**2 + W**2))

    rhos = np.linspace(-diag, diag, 2*diag)
    thetas = np.linspace(0, np.pi, theta_steps)

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int(rho + diag)
            accumulator[ri, ti] += 1

    return accumulator, rhos, thetas
````

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def create_ball_stencil(img, x1, y1, x2, y2, grad_thresh=20):
    """
    img : RGB image (numpy)
    (x1,y1) top-left, (x2,y2) bottom-right bounding box
    returns RGBA stencil
    """

    # Crop ROI
    roi = img[y1:y2, x1:x2].copy()

    # Convert to grayscale
    gray = np.mean(roi, axis=2)

    # Simple gradients
    gx = np.abs(np.gradient(gray, axis=1))
    gy = np.abs(np.gradient(gray, axis=0))
    grad_mag = np.sqrt(gx**2 + gy**2)

    # Binary mask from gradient
    mask = grad_mag > grad_thresh

    # Fill interior (rough heuristic)
    mask = mask.astype(np.uint8)
    for i in range(mask.shape[0]):
        row = mask[i]
        if np.any(row):
            first = np.argmax(row)
            last = len(row) - np.argmax(row[::-1]) - 1
            mask[i, first:last] = 1

    # Create RGBA
    h, w, _ = roi.shape
    rgba = np.zeros((h, w, 4), dtype=np.uint8)
    rgba[..., :3] = roi
    rgba[..., 3] = mask * 255

    return rgba
````

````python
def overlay_stencil(bg_img, stencil, center_x, center_y):
    """
    bg_img : RGB image
    stencil : RGBA image
    (center_x, center_y) is stencil center placement
    """

    output = bg_img.copy()
    h, w, _ = stencil.shape

    x_start = center_x - w // 2
    y_start = center_y - h // 2

    for i in range(h):
        for j in range(w):
            if stencil[i, j, 3] > 0:
                y = y_start + i
                x = x_start + j
                if 0 <= y < output.shape[0] and 0 <= x < output.shape[1]:
                    output[y, x] = stencil[i, j, :3]

    return output
````
