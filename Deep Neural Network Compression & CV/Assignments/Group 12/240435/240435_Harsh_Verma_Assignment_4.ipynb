To preserve an image in Google Colab without a continuous runtime or using Google Drive, you can use base64 string encoding of images ( its really simple )

First, begin by normally uploading the image into the contents folders as you normally do in colab or fetch it using Drive or any API.

To get the base64 string :
```
import base64
with open("your_image.jpg", "rb") as img_file:
    print(base64.b64encode(img_file.read()).decode())
```

Note that the base64 string is a very long string, click on the three dot icon to the left of the current code cell and click on <code> Copy Cell Output </code> to easily copy it to your clipboard.

Then, to embed the string into your Jupyter Notebook :
```
from PIL import Image
import io
import base64
import matplotlib.pyplot as plt
import numpy as np

# PASTE BASE64 STRING HERE
base64_image = "iVBORw0K..."  # long base64 string

img_data = base64.b64decode(base64_image)
img = Image.open(io.BytesIO(img_data))

plt.imshow(img)
plt.axis("off")
```

### Part B - Ellipse Fitting and R Maps for Corner Detection

Now, use the above functions ( basically Ix and Iy ) and the formulas in the pre-session videos to find values of a,b and c ( the second moments ) and calculate the eigenvalues lambda1 and lambda2. Then use those to create an R map for the entire image. Display the R map. Do it for the 3 images of part A. So, you might want to create a function that takes the image, loops over the pixels with a window, fits the ellipse over the window, calculates the lambdas, and value of R, essentially finding R values for all the pixels and returning an R-map ( a 2d image with only one channel where the value of a pixel is the r-value of that pixel in the original grayscale image ). Now, display these R maps. Notice how the corners are clearly visible in white.

````python
def compute_harris_response(img, k=0.04, window_size=3):
    if len(img.shape) == 3:
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    else:
        gray = img

    gray = np.float32(gray)

    dx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

    Ixx = dx * dx
    Iyy = dy * dy
    Ixy = dx * dy

    gauss = cv2.getGaussianKernel(window_size, -1)
    kernel = gauss @ gauss.T

    Sxx = cv2.filter2D(Ixx, -1, kernel)
    Syy = cv2.filter2D(Iyy, -1, kernel)
    Sxy = cv2.filter2D(Ixy, -1, kernel)

    # Det(M) = Sxx * Syy - Sxy^2
    # Trace(M) = Sxx + Syy
    det_M = (Sxx * Syy) - (Sxy ** 2)
    trace_M = Sxx + Syy

    R = det_M - k * (trace_M ** 2)

    return R
````

````python
def display_r_map(img, title="Image"):
    R = compute_harris_response(img)

    R_norm = cv2.normalize(R, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    plt.figure(figsize=(6, 6))
    plt.imshow(R_norm, cmap='gray')
    plt.title(f"{title}: Harris R-Map")
    plt.axis('off')
    plt.show()
````

````python
def hough_slope_intercept(edges, m_range=(-2, 2), m_steps=200):
    H, W = edges.shape
    m_vals = np.linspace(m_range[0], m_range[1], m_steps)

    c_min, c_max = -H, H
    c_steps = 2 * H
    c_vals = np.linspace(c_min, c_max, c_steps)

    accumulator = np.zeros((len(m_vals), len(c_vals)), dtype=np.float32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for mi, m in enumerate(m_vals):
            c = y - m * x

            ci = int((c - c_min) / (c_max - c_min) * (c_steps - 1))

            if 0 <= ci < c_steps:
                accumulator[mi, ci] += 1

    return accumulator, m_vals, c_vals
````

````python
acc_mc, m_vals, c_vals = hough_slope_intercept(edges)
````

````python
# Normalize and Threshold
acc_norm = acc_mc / acc_mc.max()
threshold = 0.9
lines = np.where(acc_norm > threshold)
````

Rho-Theta Method

````python
def hough_rho_theta(edges, theta_steps=180):
    H, W = edges.shape
    diag = int(np.sqrt(H**2 + W**2))

    # Rho: -diag to +diag
    rhos = np.linspace(-diag, diag, 2*diag)
    # Theta: 0 to 180 degrees
    thetas = np.linspace(0, np.pi, theta_steps)

    accumulator = np.zeros((len(rhos), len(thetas)), dtype=np.float32)

    ys, xs = np.nonzero(edges)

    for x, y in zip(xs, ys):
        for ti, theta in enumerate(thetas):
            rho = x * np.cos(theta) + y * np.sin(theta)
            ri = int(rho + diag)

            if 0 <= ri < len(rhos):
                accumulator[ri, ti] += 1

    return accumulator, rhos, thetas
````

````python
acc_rt, rhos, thetas = hough_rho_theta(edges)
````

````python
acc_rt_norm = acc_rt / acc_rt.max()
lines = np.where(acc_rt_norm > 0.5)
````

````python
img_lines_rt = corner2.copy()
for ri, ti in zip(lines[0], lines[1]):
    rho = rhos[ri]
    theta = thetas[ti]

    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho
    y0 = b * rho

    x1 = int(x0 + 1000 * (-b))
    y1 = int(y0 + 1000 * (a))
    x2 = int(x0 - 1000 * (-b))
    y2 = int(y0 - 1000 * (a))

    cv2.line(img_lines_rt, (x1,y1), (x2,y2), (0, 255, 0), 2)
````

````python
def create_stencil_simple(img, bbox):
    x1, y1, x2, y2 = bbox
    crop = img[y1:y2, x1:x2]

    # Gradients and mag Sobel X, Y
    gray = cv2.cvtColor(crop, cv2.COLOR_RGB2GRAY)
    grad = cv2.magnitude(cv2.Sobel(gray, cv2.CV_64F, 1, 0), cv2.Sobel(gray, cv2.CV_64F, 0, 1))

    # 2. Contour, gradient Threshold to find edges
    _, mask = cv2.threshold(grad, 50, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if not contours: return None

    #RGBA
    alpha = np.zeros_like(gray)
    cv2.drawContours(alpha, [max(contours, key=cv2.contourArea)], -1, 255, thickness=-1)

    return np.dstack([crop, alpha])
````

````python
def place_stencil(stencil, target, center):
    if stencil is None: return target

    h, w = stencil.shape[:2]
    tl_x, tl_y = center[0] - w//2, center[1] - h//2
    br_x, br_y = tl_x + w, tl_y + h

    # Extracting bg
    bg_roi = target[tl_y:br_y, tl_x:br_x].astype(float)

    # Extracting RGB,A from Stencil
    fg_rgb = stencil[..., :3].astype(float)
    alpha = stencil[..., 3].astype(float) / 255.0
    alpha = np.dstack([alpha] * 3)

    blended = (fg_rgb * alpha) + (bg_roi * (1.0 - alpha))

    result = target.copy()
    result[tl_y:br_y, tl_x:br_x] = blended.astype(np.uint8)
    return result
````

````python
def composite_ball_hough(img_source, img_target, target_pos):
    # Detecting Circle using Hough
    gray = cv2.cvtColor(img_source, cv2.COLOR_RGB2GRAY)
    gray_blur = cv2.medianBlur(gray, 5)

    circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, 1, 50,
                               param1=50, param2=25, minRadius=10, maxRadius=60)

    if circles is None:
        print("Error: No ball detected!")
        return img_target

    circles = np.uint16(np.around(circles))
    cx, cy, r = circles[0, :][0]

    mask = np.zeros_like(gray)
    cv2.circle(mask, (cx, cy), r, 255, -1)

    x1, y1 = cx - r, cy - r
    x2, y2 = cx + r, cy + r

    ball_roi = img_source[y1:y2, x1:x2]
    mask_roi = mask[y1:y2, x1:x2]
    mask_inv_roi = cv2.bitwise_not(mask_roi)

    tx, ty = target_pos
    tl_x = tx - r; tl_y = ty - r
    br_x = tl_x + (2*r); br_y = tl_y + (2*r)

    # Compositing
    target_roi = img_target[tl_y:br_y, tl_x:br_x]

    bg = cv2.bitwise_and(target_roi, target_roi, mask=mask_inv_roi)
    fg = cv2.bitwise_and(ball_roi, ball_roi, mask=mask_roi)
    combined = cv2.add(bg, fg)

    result = img_target.copy()
    result[tl_y:br_y, tl_x:br_x] = combined

    return result
````
